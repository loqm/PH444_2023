##Transformation properties of EM fields and sources


Before we head into the meat of the topic, let us review some fundamental mathematical tools that will be useful later, the transformation properties of EM fields and sources under rotations, spatial reflections and time reversal. It is assumed that the idea of space and time coordinate transformations is familiar to the reader, and we proceed to discuss the well-defined transformation properties of various electromagnetic quantities under these operations. A brief recap of terms that will be used is given below. Physical quantities are classified as rotational tensors of various ranks depending on how they transform under rotations. 

1. A vector is a quantity whose components transform according to the basic orthogonal transformation law $x'_{\alpha} = \sum_{\beta} a_{\alpha\beta} x_{\beta}, \sum_{\alpha} a_{\alpha\beta}a_{\alpha\gamma} = \delta_{\beta\gamma}$. These are tensors of rank one.

2. A scalar is a quantity which equals the scalar products of vectors, such as $\mathbf{x_1} \cdot \mathbf{x_2}$ and are invariant under rotations. These are tensors of rank zero.

3. A tensor is a group of quantities that transform according to $B'_{\alpha\beta} = \sum_{\gamma,\delta} a_{\alpha\gamma}a_{\beta\delta} B_{\gamma\delta}$. These are tensors of rank two.

4. A pseudoscalar is a quantity that behaves like a scalar except that it changes sign under a spatial inversion.

5. A pseudovector is a quantity that is defined as a function of some vectors or other geometric function that resembles a vector and behaves like a vector in many situations but changes sign under a spatial inversion.


###Transformation properties under different symmetries


####Rotations

Electromagnetic fields and sources are invariant under rotations. This means that if an observer rotates their coordinate system, the fields and sources will transform in a specific way so that they appear the same in the new coordinate system as they did in the original.

$$x_\alpha^{\prime}=\sum_\beta a_{\alpha \beta} x_\beta$$

$$\sum_\alpha a_{\alpha \beta} a_{\alpha \gamma}=\delta_{\beta \gamma}$$

$$B_{\alpha \beta}^{\prime}=\sum_{\gamma, \delta} a_{\alpha \gamma} a_{\beta \delta} B_{\gamma \delta}$$

$$\phi^{\prime}\left(\mathbf{x}_i^{\prime}\right)=\phi\left(\mathbf{x}_i\right)$$

####Spatial reflections

Electric fields are odd under spatial reflections, which means that they change sign when the coordinate system is reflected through a plane. Magnetic fields, on the other hand, are even under spatial reflections, which means that they remain the same when the coordinate system is reflected.

$$V_\alpha^{\prime}\left(\mathbf{x}_i^{\prime}\right)=\sum_\beta a_{\alpha \beta} V_\beta\left(\mathbf{x}_i\right)$$

$$A_\alpha=\sum_{\beta \gamma} \epsilon_{\alpha \beta \gamma} B_\beta C_\gamma$$

$$A_\alpha^{\prime}=\operatorname{det}(a) \sum_\beta a_{\alpha \beta} A_\beta$$

####Time reversal

Electromagnetic fields are even under time reversal, which means that they remain the same if the direction of time is reversed. However, sources are odd under time reversal, which means that they change sign if the direction of time is reversed.


$$\frac{d \mathbf{p}}{d t}=-\nabla U(\mathbf{x})$$

$$\boldsymbol{\nabla} \times \mathbf{E}+\frac{\partial \mathbf{B}}{\partial t}=0$$

$$\frac{1}{\mu_0} \boldsymbol{\nabla} \times \mathbf{B}-\epsilon_0 \frac{\partial \mathbf{E}}{\partial t}=\mathbf{J}$$

$$\mathbf{E} \times \mathbf{B}_0, \quad \frac{\partial \mathbf{E}}{\partial t} \times \mathbf{B}_0, \quad \frac{\partial^2 \mathbf{E}}{\partial t^2} \times \mathbf{B}_0, \cdots$$

$$\frac{1}{\epsilon_0} \mathbf{P}=\chi_0 \mathbf{E}+\chi_1 \frac{\partial \mathbf{E}}{\partial t} \times \mathbf{B}_0+\chi_2\left(\mathbf{B}_0 \cdot \mathbf{B}_0\right) \mathbf{E}+\chi_3\left(\mathbf{E} \cdot \mathbf{B}_0\right) \mathbf{B}_0+\cdots$$

$$\frac{1}{\epsilon_0} \mathbf{P}=\chi_0 \mathbf{E}+\chi_1 \frac{\partial \mathbf{E}}{\partial t} \times \mathbf{B}_0+\chi_2^{\prime}\left(\mathbf{B}_0 \cdot \mathbf{B}_0\right) \frac{\partial^2 \mathbf{E}}{\partial t^2}+\chi_3^{\prime}\left(\frac{\partial^2 \mathbf{E}}{\partial t^2} \cdot \mathbf{B}_0\right) \mathbf{B}_0$$



It is experimentally validated that electric charge is invariant under Galilean and Lorentz transformations and is a scalar under rotations. We can assume that it is a scalar under spatial inversion and time reversal too, but that assumption then follows through into physically measurable quantities like force.

Considering charge to be a true scalar under all three transformations, charge density $\rho$ is also a true scalar. As electric field is force per unit charge, we conclude that $\mathbf{E}$ is a polar vector, even under time reversal. This is corroborated by the Maxwell equation $\nabla \cdot \mathbf{E} = \rho/ϵ_0$ as both sides must transform similarly. 

The first term in the Maxwell equation representing Faraday's law,

$$\nabla \times \mathbf{E} + \frac{∂\mathbf{B}}{∂t} = 0 \tag{A1}$$

transforms as a pseudovector under rotations and spatial inversion, and is even under time reversal. Thus, the magnetic induction $\mathbf{B}$ is also a pseudovector, odd under time reversal.

The left hand side of the Ampere-Maxwell equation,

$$\frac{1}{\mu_0}\nabla \times \mathbf{B} - ϵ_0 \frac{\partial \mathbf{E}}{\partial t} = \mathbf{J} \tag{A2}$$
transforms as a polar vector, odd under time reversal. Thus, the current density $\mathbf{J}$ is a polar vector, odd under time reversal. 

We have seen that the microscopic fields and sources have well-defined transformation properties under rotations, spatial inversion and time reversal. From the derivation of the macroscopic Maxwell equations, one can observe that $\mathbf{E, P, D}$ all transform in the same as $\mathbf{B,M,H}$.

Now, we are ready to tackle the question of magnetic monopoles, equipped with an understanding of the various transformation properties of electromagnetic quantities.

## Introduction

A magnetic monopole is a hypothetical particle with a magnetic charge, which is analogous to the electric charge of a point particle.

The absence of magnetic monopoles has been a long-standing puzzle in physics. Maxwell's equations, which describe the behavior of electric and magnetic fields, predict the existence of magnetic monopoles, but they have never been observed experimentally.

Jackson discusses the theoretical implications of the existence of magnetic monopoles, such as the modification of the Coulomb law and the quantization of electric charge. He also explains the difficulties in detecting magnetic monopoles, which arise from the extremely weak magnetic fields that would be produced by a monopole.

Jackson presents some theoretical models for magnetic monopoles, such as the Dirac monopole, which predicts the existence of a magnetic monopole with a quantized magnetic charge. However, the Dirac monopole has not been observed experimentally, and the search for magnetic monopoles is still ongoing.

Overall, the question of magnetic monopoles remains an open problem in physics, and their hypothetical existence would have significant implications for our understanding of fundamental physical laws.

##Dirac's Quantization Condition


###Discrete Nature of Charge

Dirac considered the quantum mechanical nature of electrons in the presence of a magnetic monopole. For consistency, he required that a quantization condition be adhered to:

$$
\frac{e g}{4 \pi \hbar}=\frac{\alpha g}{Z_0 e}=\frac{n}{2} \quad(n=0, \pm 1, \pm 2, \ldots \tag{C1}
$$

This implies that the discreteness of charge is due to the existence of magentic monopoles! 



### Dirac Monopoles

If we reverse the same logic, we can infer from our observations about the discrete nature of electric charge, that monopoles do exist with an equivalent "magnetic fine structure" constant. 

**Fine Structure Constant:** A fundamental physical constant which quantifies the strength of the electromagnetic interaction between elementary charged particles. 

$$
\frac{g^2}{4 \pi \mu_0 \hbar c}=\frac{n^2}{4}\left(\frac{4 \pi \epsilon_0 \hbar c}{e^2}\right) \simeq \frac{137}{4} n^2 \tag{C2}
$$ 

These are known as Dirac monopoles, and it is evident from the expression that the magnitude of their coupling strength is extremely high.

##Dirac's Quantization Argument (1931)
We imagine the magnetic monopole to be either as a particle at the end of a line of dipoles or at the end of a tightly wound solenoid, stretching to infinity. This is depicted in the figure below. 



image.png



Both these configurations are equivalent and can be described by the vector potential of a magnetic dipole

\begin{equation}
A(x) = \frac{m \times (x-x')}{|x-x'|^3}, \tag{D1}
\end{equation}

where $x$ is the field point, $x'$ is the source point and $m$ is the magnetic dipole moment.

The line of dipoles is referred to as Dirac's string, and is a matter of significant controversy which will be discussed later in this report. The string is formed by infinitesimal magnetic dipole moments $dm$ located at $x'$ whose vector potential is $dA(x)=-dm \times ∇(1/|x-x'|)$, where we use the relation that $∇(1/|x-x'|) = -(x-x')/|x-x'|^3$. If we attribute $dm = gdl'$, where $g$ represents the magnetic charge and $dl'$ is a line element, the total vector potential for a string or solenoid on the curve $L$ reads

$$A_L = -g \int_L dl' \times ∇\left(\frac{1}{|x-x'|}\right). \tag{D2}$$

Using the result $∇ \times (dl'/|x-x'|) = -dl' \times \nabla(1/|x-x'|)$, we write $\text{(D2)}$ as 
$$A_L = g∇ \times \int_L \frac{dl'}{|x-x'|}. \tag{D3}$$
Observe that this potential formulation is already in the Coulomb gauge: $∇.A_L = 0$ as the divergence of a curl is always 0.

###Monopoles and Strings
After some vector calculus, we can show that the curl of this potential gives 

$$ \nabla \times A_L = \frac{g}{R^2}\hat{R} + 4\pi g\int_L \delta(x-x') dl', \tag{D4} $$

where $\delta(x-x')$ is the Dirac delta function, $R = |x-x'|$ and $\mathbf{\hat{R}} = (x-x')/R$. It is instructive to write $\text{(D4)}$ as 

$$B_{mon} = \nabla \times A_L - B_{string}, \tag{D5}$$

where $B_{mon} = \frac{g}{R^2}\mathbf{\hat{R}}$ is the field of the magnetic monopole $g$ located at the point $x'$ and $B_{string} = 4 \pi g \int_L \delta(x-x')dl'$ is a singular magnetic field contribution along the curve $L$. We can observe the divergence of $B_{mon}$ to be 

$$ \nabla \cdot B_{mon} = \nabla \cdot \left(\frac{g}{R^2} \mathbf{\hat{R}} \right) = 4 \pi g \delta(x-x'), \tag{D6}$$

where we have used $\nabla \cdot (\mathbf{\hat{R}}/R^2) = 4 \pi \delta(x-x')$. In an analogous manner, the divergence of $B_{string}$ is

\begin{align}
   \nabla \cdot B_{string} &= \nabla \cdot \left( 4 \pi\, g \int_L \delta(x-x') dl' \right) \\
  &= - 4 \pi\, g \int_L \nabla' \delta(x-x') \cdot dl' \\
  &= - 4 \pi\, g\, \delta(x-x'), \tag{D7}
  \end{align}

  where we have used $\nabla \delta(x-x') = \nabla' \delta(x-x')$. 
Using $\text{(D6)}$ and $\text{(D7)}$, we can verify the expected result that the divergence of $\text{(D4)}$ is $0$ i.e., $\nabla \cdot (\nabla \times A_L) = 0$. We can use Gauss theorem to express this in an equivalent way in terms of the fluxes of the fields $B_{mon}$ and $B_{string}$, which cancel.

$$\oint_S B_{mon} \cdot da = \int_V \nabla \cdot B_{mon}\,d^3 x = 4\pi\, g, \tag{D8}$$
$$\oint_S B_{string} \cdot da = \int_V \nabla \cdot B_{string}\,d^3 x = -4\pi\, g, \tag{D9}$$

where $da$ and $d^3 x$ denote the differential surface and volume elements.

### An illustrative example

Consider the case in which the string lays along the negative z-axis and the magnetic monopole is at the origin. In this case $dl' = dz' \hat{z}$, and the corresponding potential is 

$$A_L = g \nabla \times \hat{z} \int_{-\infty}^0 \frac{dz'}{|x-z'\hat{z}|}. \tag{D10}$$

After calculating the curl for this example we obtain

$$ \nabla \times A_L = \frac{g}{r^2}\hat{r} + 4 \pi g \delta(x) \delta(y) \Theta(-z)\hat{z}, \tag{D11}$$

where now $r = |x|$, $\hat{r} = x/r$, and $\Theta(z)$ is the step function which is undefined at $z=0$ but it is defined as $\Theta(z) = 0$ if $z < 0$ and $\Theta(z) = 1$ if $z>0$. The second term in the $\text{(D11)}$ captures the highly singular character of the magnetic field associated with the string.

### Dirac's solution
In his original (1931) paper, Dirac had proposed the same example and had written his solution for the vector potential in spherical coordinates $A_L = (g/r) \tan(\theta/2) \hat{\phi}$ and noted that this potential gave rise to the radial field $g \hat{r}/r^2$.

Dirac observed: "This solution is valid at all points except along the line $\theta = \pi$, where [$A_L$] become infinite." Dirac's solution is equivalent to

$$A_L = g \frac{1-\cos\theta}{r\sin\theta}\hat{\phi}. \tag{D12}$$

This expression can be determined by completing the integration specified in $\text{(D10)}$. Clearly, the curl of $\text{(D12)}$ under the constraint that $\sin \theta \neq 0$ gives only the field of the magnetic monopole $\nabla \times A_L = g \hat{r}/r^2 = B_{mon}$. The singularity originated by $\sin \theta = 0$ is avoided in such a differentiation process.

### Dirac Veto

If an electric charge is interacting with the potential given earlier, then it is essentially interacting with both a magnetic monopole and a magnetized string. Dirac's argument stressed that the interaction must only be with the magnetic monopole and thus, the charge $q$ must never "see" the singular field $B_{string}$ as defined earlier after $\text{(D5)}$. For this reason, he postulated that the wave function vanish along the string. This postulate has been intensely debated over the decades by eminent theorists and experimentalists and Dirac's peers criticized this requirement extensively as it would mean that the string does not exist at all. This postulate is commonly known as the "Dirac veto" whose alternative form states that any interaction of the electric charge with the string is forbidden. Dirac's own words on the matter were: "You must have the monopoles and the electric charges occupying distinct regions of space. The strings, which come out from the monopoles, can be drawn anywhere subject to the conditino that they must not pass through a region where there is electric charge present."



### Gauge Invariance

The next step of Dirac's argument is to show that the magnetic field of the monopole $\text{(D5)}$ is independent of the location of the string. To show this, we consider two different strings $L'$ and $L$ with their respective vector potentials $A_{L'}$ and $A_L$. If these potentials are equivalent, it will imply the equivalence of their respective strings thus indicating that the location of the string is irrelevant. 

image.png


We can obtain the difference in the potentials from $\text{(D3)}$ by taking the integration along the closed curve $C = L' - L$ around the area $S$ as shown in the figure above. The result is

\begin{equation}
A_{L'} - A_L = g\,\nabla \times \oint_C \frac{dl'}{|x-x'|} = \nabla(g \Omega_C), \tag{D13}
\end{equation}

where $\Omega_C$ is the solid angle function subtended by the curve $C$. Observe that $A_{L'}$ and $A_L$ are connected by the gradient of a function, and the similarity to the gauge transformation $A' = A + \nabla Λ$ where $Λ$ is a gauge function. Thus, without any loss of generality, we can write $A' ≡ A_{L'}$, $A ≡A_L$ and $ Λ=gΩ_C$. Notice that $A_{L'}$ and $A_L$ are in the Coulomb gauge: $∇ \cdot A_{L'}=0$ and $\nabla \cdot A_L =0$. However, this does not hinder these potentials from being connected by another gauge transformation as long as the gauge function is constrained by $\nabla^2 Λ=0$. This can be verified by taking the divergence of the difference of potentials and obtaining $\nabla^2 Λ=0$ to indicate that the potentials $A_{L'}$ and $A_L$ are connected by a restrictive gauge transformation.



### Single-valuedness of the wave function

Something quite noteworthy in the above argument is that different string positions correspond to different choices of gauge, or a change in string from $L$ to $L'$ is equivalent to a gauge transformation from $A_L$ to $A_{L'}$ with the gauge function $Λ=g\Omega_C$. With this identification of the gauge function, we observe that the associated phase transformation of the wave function $Ψ'= \mathrm{e}^{iq\lambda/(ħc)}\psi$ takes the form $Ψ'= \mathrm{e}^{iqg\Omega_C/(ħc)}\psi$. 

The crucial point of the argument is that the solid angle $\Omega_C$ undergoes a discontinuous variation of $4\pi$ as the observation point (or the charge $q$) crosses the surface $S$. This makes the gauge function $Λ = g\Omega_C$ multi-valued which implies that $\mathrm{e}^{iqg\Omega_C}$ is also multi-valued, i.e., $\mathrm{e}^{iqg\Omega_C} \neq \mathrm{e}^{iqg(\Omega_C+4\pi)}$. 

Thus when $q$ crosses $S$, the transformed wave function of the charge $q$ will be multi-valued, unless we impose the condition $\mathrm{e}^{i4\pi qg/\hbar c} = 1$. This condition and $\mathrm{e}^{i 2 \pi n} = 1 $ with integer $n$ directly imply $4 \pi q g/(\hbar c) = 2 \pi n$ which gives us the Dirac quantization condition

$$qg=\frac{n\hbar c}{2}. \tag{D14}$$

To conclude, the field of the monopole does not depend on the location of the string. In order to assemble the Dirac quantization condition, gauge invariance and the single-valuedness of the wave function are essential.


### Alternate interpretation

The above derivation emphasised the idea that the location of the Dirac string is irrelevant. However, one can pose the same argument on the basis that the string is unobservable. 

Consider the value $\Omega_1$ corresponding to one side of the surface $S$ and the value $\Omega_2$ corresponding to the other side. They are related by $\Omega_1 = \Omega_2 + 4 \pi$. It follows naturally that $\mathrm{e}^{iqg\Omega_1 / (\hbar c)} = \mathrm{e}^{iqg(\Omega_2 + 4 \pi) / (\hbar c)}$. 

This implies that the wave function of the charge $q$ differs by the quantity $\mathrm{e}^{i4 \pi qg / (\hbar c)}$ which would make the Dirac string observable as the charge crosses the surface, unless we again impose the condition $\mathrm{e}^{i4 \pi qg / (\hbar c)} = 1$, which is satisfied if $qg = n \hbar c/2$ holds. In conclusion, we must impose the Dirac condition for the unobservability of the string. 

## Induction Detection of Monopoles

One proposed method for detecting magnetic monopoles is using induction detectors. These detectors consist of a coil of wire, typically made of a superconducting material, that is cooled to very low temperatures. When a magnetic field changes in the vicinity of the coil, it induces a current in the coil, which can be detected and measured.

To detect magnetic monopoles using induction detectors, a magnetic monopole would have to pass through the coil, creating a magnetic field that would induce a current in the coil. The magnitude and direction of the induced current would provide information about the magnetic charge and direction of motion of the monopole.

An ideal monopole detector in theory is a closed loop of superconducting wire because it emits a clear signal whenever a monopole passes through the loop, no matter how slowly. We can use the integrated Maxwell equation to determine how a monopole passing through will affect the loop:

$$\oint_{\Gamma} \mathbf{E} \cdot \mathrm{d} \mathbf{r}=-\frac{\mathrm{d} \Phi}{\mathrm{d} t}-\frac{\mathrm{d} Q_{\mathrm{m}}}{\mathrm{d} t}$$


The integrated Maxwell equation, which has been appropriately modified to account for the magnetic monopole cutrent, may be used. In this instance, $\mathrm{d} Q_m / \mathrm{d} t$ is the monopole current flowing through the surface $S_{\Gamma}$ , and $\Phi$ is the magnetic flux through the surface  $S_{\Gamma}$ enclosed by the closed path $\Gamma$. Applying Equation to a path $\Gamma$ entirely contained in the superconducting wire, where $\mathbf{E}=0$, and integrating over time, we discover that the relationship between the total magnetic charge  $\Delta Q_{\mathrm{m}}$ and the change $\Delta \Phi$ in the magnetic flux linking the loop is:

$$\Delta \Phi=-\Delta Q_{\mathrm{m}}$$        

For example, if a magnetic monopole with the Dirac charge $g_{\mathrm{D}}$ passes through the loop, the flux changes by two quantized flux units. This factor of two results from the fact that a Cooper pair has an electric charge of 2e.

The fact that the magnetic field cannot penetrate the superconducting wire makes the above conclusion obvious. This causes the magnetic field lines coming from the monopole to be swept back as it approaches the wire, break off, and form closed loops around the wire, as seen in Figure below. Where the magnetic field disappears, it is allowed for the field lines to break and rejoin.

We can observe that a monopole travelling through the loop causes an abrupt shift in the magnetic flux connecting the loop and a corresponding change in the level of the dc current in the wire, with a rise time of the order of the loop's radius divided by the monopole's velocity. 

If the loop is appropriately shielded from other fluctuating magnetic disturbances, the shift can be easily observed using a SQUID (superconducting quantum interference device) magnetometer for a sufficiently tiny loop (less than 10 cm in diameter).

The magnetic shielding requirement and signal-to-noise issues are currently limiting the sizes of superconducting loop detectors, but monopole searches have been carried out with such detectors, and an experimental flux limit $F \lesssim 2 \times 10^{-11} \mathrm{~cm}^{-2} \mathrm{~s}^{-1} \mathrm{sr}^{-1}$ has been obtained with a 90% confidence level. 

One candidate event with a magnitude consistent with the magnetic charge of Dirac has been seen, but there are questions over whether it should be associated with a magnetic monopole or another particle.

Propose a semi-classical derivation for Dirac's quantization condition $(qg = n\hbar c/2)$ by considering an electric charge $q$ moving with a speed $\dot{\mathbf{x}}$ in the field of a monopole $g$. This configuration is illustrated below.

---


 The charge $q$ experiences the Lorentz force
$$
\frac{d \mathbf{p}}{d t}=q\left(\frac{\dot{\mathbf{x}}}{c} \times \mathbf{B}\right)
$$
where $\mathbf{p}=m \dot{\mathbf{x}}$ is the mechanical momentum associated to the charge $q$. 

The field of the monopole is spherically symmetric. This implies that the total angular momentum of the system is conserved. 

We first compute the torque experienced by the moving particle. 

$$
\frac{d(\mathbf{x} \times \mathbf{p})}{d t}=\frac{q}{c}(\mathbf{x} \times(\dot{\mathbf{x}} \times \mathbf{B}))=\frac{q g}{c}\left(\frac{\mathbf{x} \times(\dot{\mathbf{x}} \times \mathbf{x})}{r^3}\right)=\frac{d}{d t}\left(\frac{q g}{c} \hat{\mathbf{r}}\right)
$$
**Note:**
$$
\frac{\mathbf{x} \times(\dot{\mathbf{x}} \times \mathbf{x})}{r^3}=\frac{d \hat{\mathbf{r}}}{d t}
$$

However, the mechnical angular momentum is not conserved. Therefore, the condition of conservation of total angular momentum is written as, 

$$
\mathbf{J}=\mathbf{x} \times \mathbf{p}-\frac{q g}{c} \hat{\mathbf{r}}
$$

We can observe that the radial component of this angular momentum is constant $$\mathbf{J} \cdot \hat{\mathbf{r}}=-q g / c$$. According to quantum theory, angular momentum must be integral multiples of a fundamental unit, which is $\frac{ħ}{2}$. So we equate the classic value $q g / c$ to $\frac{n}{2} ħ$. Therefore, we can quantise the radial component of the angular momentum, $J_r=q g / c=n \hbar / 2$ (the minus sign is absorbed by $n$) and this yields the Dirac condition $q g=n \hbar c / 2$




## Conclusion

The existence of magnetic monopoles is a very general consequence of the 
unification of the fundamental interactions. But it is one thing to say that monopoles must exist, and quite another to say that we have a reasonable chance of observing one. If monopoles are as heavy as we expect, there is no hope of producing monopoles in any foreseen accelerator. Our best hope is to observe a monopole in cosmic rays. But since no process occurring in the present universe is sufficiently energetic to produce monopoles, any monopoles around today must have been produced in the very early universe, when higher energies were available. Thus, the abundance of magnetic monopoles is a cosmological issue. 

If a magnetic monopole is ever discovered, it will be a momentous 
occasion, with many fascinating implications. For one thing, that there are any monopoles at all would be evidence that the universe was once 
extremely hot. And severe constraints would be placed on our attempts at 
cosmological model building, for the observed monopole abundance would 
have to be explained by any realistic cosmological scenario.

Detection of a monopole would also confirm a very fundamental 
prediction of grand unification. The mass of the monopole, if it could be measured, would reveal the basic symmetry-breaking scale at which 
electrodynamics becomes truly united with the other particle interactions.

But even if nobody ever sees a magnetic monopole, there is surely much 
to be gained by studying the theory of monopoles. Many marvelous 
insights into gauge theory and quantum field theory have been derived 
from this study, and can be perused in Preskill.
