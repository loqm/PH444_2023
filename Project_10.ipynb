# **2. Time Reversal in Electrodynamics:**

#### **A. Time-Reversal Definition and Loschmidt’s Paradox:**
- A physical process is called time-reversal invariant or reversible if its evolution in the backward direction (think about the reversed video of the original process) is also a realistic process. That is, this evolution also satisfies the dynamic equations.
- An equivalent definition of a time-reversal invariant process was formulated by Casimir “if a system of particles and fields moves in a certain way during the period $0 <t<t_0$ and if at the moment $t_0$, we would invert all velocities, currents, magnetic fields, and so on, and then the system retraces its steps: at the time $2t_0$, particle coordinates and fields are what they were at $t = 0$, and at a time $2t_0 − t$, the situation is what it was at $t$”.

image.png


- If the equations describing the process do not change under substitution $t → −t$, the process is called time-reversal invariant. 
- $T\{r(t)\} = r(−t)$, while the corresponding velocities are related by $T\{v(t)\} = −v(−t)$.
- Furthermore, the time-reversal operation does not change the sign of differential dt (the time arrow direction remains the same), while it flips the sign of the infinitesimal displacement $dr(t)$.
- All the physical laws, with the only exception of those corresponding to weak interactions, are governed by the equations that are symmetric under time reversal.
- Our experience is also supported by the second law of thermodynamics expressed as $\frac{dS}{dt} ≥ 0$, which says that any closed physical system cannot evolve from a disordered to a more ordered state.

    **Loschmidt’s Paradox:** 
    <br>
    
- The behavior of particles in a system is the same regardless of the direction of time, then the observed irreversibility of macroscopic phenomena, such as diffusion and viscosity, is paradoxical.

image.png


- Consider a closed box with gas and assume that, at $t = 0$, all its molecules occupy only a small region in the corner of the box.
- The red arrows indicate the velocities of the molecules. After sometime t0, the molecules will spread somewhat uniform inside the box.
- If we choose the position of the molecules as in figure and launch them with the same velocities but in the opposite directions, then, after time period $t_0$, the molecules would come back exactly to the initial ordered state.
- In other words, the process reversal would be possible if we were able to ensure the microscopic initial conditions (position and velocity) for each molecule.
- In practice, assigning correct initial conditions even for a few molecules is a complicated engineering problem; therefore, for systems with large numbers of molecules.
- Even by assigning wrong velocity the final state of the system will be drastically different.
- The probability that, by satisfying the macroscopic initial conditions, we also ensure, by chance, that the microscopic ones are proportional to $n−N$ , where $N$ is the number of molecules in the closed box and $n$ is the number of possible states of each molecule.

#### **B. Time-Reversal Symmetry of Maxwell’s Equations:**
- It’s straightforward that the electric charge $q$, charge density $\rho$, and coordinate $r$ do not change under time reversal.
- Under this assumption, speed $v$ and electric current $j$ are odd with respect to time reversal.
- The force is proportional to the acceleration $a$ and, therefore, is symmetric.
- The electric field $\tilde{\mathbf{e}}$, being proportional to the force acting on a unit charge, is time-reversal even.
- From the formula for the Lorentz force $\tilde{\mathbf{F}}=q \tilde{\mathbf{e}}+q \tilde{\mathbf{v}} \times \tilde{\mathbf{b}}$, one can deduce that magnetic field $\tilde{\mathbf{b}}$ must be time-reversal odd (multiplication of speed and magnetic field must be time-reversal even).

image.png

- We see that under these assumptions, the microscopic Maxwell’s equations are invariant under time reversal.

#### **C.Time Reversal of Material Relations :**
- Let us consider time reversal of an arbitrary wave process in a stationary dielectric material. The dielectric is assumed to be isotropic, possibly nonuniform

$$\tilde{\mathbf{P}}(t)=\varepsilon_0 \int_{-\infty}^t \tilde{\chi}\left(t-t^{\prime}\right) \tilde{\mathbf{E}}\left(t^{\prime}\right) d t^{\prime}$$

- Where $P$ is the electric dipole moment per unit volume. An integral is from $\infty → −t$  is because to account for the history of the electric field acting on the material.

$$\tilde{\mathbf{D}}(t)=\int_0^{\infty} \varepsilon_0 \tilde{\varepsilon}(\tau) \tilde{\mathbf{E}}(t-\tau) d \tau$$

- After the standard replacement, the displacement vector can be written as the above convolution integral
- Using the Fourier transform, which yields the well known material relations for an isotropic dielectric without spatial dispersion.
$$
\mathbf{D}(\omega)=\varepsilon_0 \varepsilon(\omega) \mathbf{E}(\omega), \quad \mathbf{B}(\omega)=\mu_0 \mathbf{H}(\omega)
$$
where
$$
\varepsilon(\omega)=\int_0^{\infty} \tilde{\varepsilon}(\tau) e^{-j \omega \tau} d \tau
$$
- By applying the Fourier transform to both sides of Maxwell’s equations and substituting the above $D(\omega)$ and $B(\omega)$, we obtain their frequency-domain version:

$$\nabla \times \mathbf{E}(\omega, \mathbf{r})=-j \omega \mu_0 \mathbf{H}(\omega, \mathbf{r}), \quad \nabla \cdot[\varepsilon(\omega, \mathbf{r}) \mathbf{E}(\omega, \mathbf{r})]=0$$

$$\nabla \times \mathbf{H}(\omega, \mathbf{r})=j \omega \varepsilon_0 \varepsilon(\omega, \mathbf{r}) \mathbf{E}(\omega, \mathbf{r}), \quad \nabla \cdot \mathbf{H}(\omega, \mathbf{r})=0$$

- Assuming free currents and charge density as zero, with small algebraic calculations we get:

$$\nabla \times\left[\frac{1}{\varepsilon(\omega, \mathbf{r})} \nabla \times \mathbf{H}(\omega, \mathbf{r})\right]=\frac{\omega^2}{c^2} \mathbf{H}(\omega, \mathbf{r})$$

- Taking fourier transform of the magnetic field $H$ in frequency domain we get:

$$\mathbf{H}(\omega, \mathbf{r})=1 /(2 \pi) \int_{-\infty}^{+\infty} \tilde{\mathbf{H}}(t, \mathbf{r}) e^{-j \omega t} d t$$

- Under time reversal, the field in the time domain transforms as $T\{{\tilde{\mathbf{H}}(t, r)}\} = −\tilde{\mathbf{H}}(−t, r)$.

$$
\begin{gathered}
T\{\mathbf{H}(\omega, \mathbf{r})\}=1 /(2 \pi) \int_{-\infty}^{+\infty}-\tilde{\mathbf{H}}(-t, \mathbf{r}) e^{-j \omega t} d t \\
=-1 /(2 \pi) \int_{-\infty}^{+\infty} \tilde{\mathbf{H}}(t, \mathbf{r}) e^{j \omega t} d t . \\
T\{\mathbf{H}(\omega, \mathbf{r})\}=-\mathbf{H}(-\omega, \mathbf{r})=-\mathbf{H}^*(\omega, \mathbf{r}) .
\end{gathered}
$$

- For time-even fields, time reversal results in complex conjugate without the sign flip, that is, for electric field

$$T\{\mathbf{E}(\omega, \mathbf{r})\}=\mathbf{E}(-\omega, \mathbf{r})=\mathbf{E}^*(\omega, \mathbf{r})$$

- By applying time reversal to both sides of (5), we conclude that the time-reversal symmetry of field equations (here, including material relations) dictates the following rule for time reversal of the complex permittivity:

$$T\{\varepsilon(\omega)\}=\varepsilon^*(\omega)$$

- The above expression implies that under time reversal, lossy media become active and vice versa.
- If the direct process was lossy, in the reversed process, the phonons of the dielectric lattice will oscillate in such a way that their energy will be transformed back into the energy of electromagnetic waves.
- Naturally, this exact process reversal is impossible in practice due to the vast number of microscopic conditions to be satisfied.
- Then, the dielectric permittivity will remain lossy, and the reversed process will be different from the original one.

#### **D. Time Reversal of Wave Propagation in a Dielectric Material:**

- Let the original wave propagation process in a dielectric material with some complex permittivity $ε(\omega, r) = ε − jε$ be described by wave equation written as :

$$\nabla \times\left[\frac{1}{\varepsilon(\omega, \mathbf{r})} \nabla \times \mathbf{H}_{\text {orig }}(\omega, \mathbf{r})\right]=\frac{\omega^2}{c^2} \mathbf{H}_{\text {orig }}(\omega, \mathbf{r})$$

- Mathematically, it means that the time-reversed version of the dielectric is described by the complex conjugate of the original permittivity $ε^∗(\omega, r)$.
- Thus, the original and time-reversed (microscopically) waves propagating in a dielectric material have the same waveform but opposite propagation directions.

#### **E. Time Reversal of Wave Propagation in a Magneto–Optical Material :**

- Magneto–optical materials are materials biased by external or internal static (sometimes, quasi-static) magnetic field, which we denote $H_0$.
- In a magnetized free-electron plasma, due to electron cyclotron orbiting, the permittivity is described by a second-rank tensor.

$$\overline{\bar{\varepsilon}}\left(\omega, \mathbf{H}_0\right)=\left(\begin{array}{ccc}
\varepsilon_s(\omega) & j \varepsilon_a\left(\omega, \mathbf{H}_0\right) & 0 \\
-j \varepsilon_a\left(\omega, \mathbf{H}_0\right) & \varepsilon_s(\omega) & 0 \\
0 & 0 & \varepsilon_z(\omega)
\end{array}\right)$$

- Where $ε_s$, $ε_a$, and $ε_z$ are (real-valued in the lossless case) functions, and the external magnetic field is applied along the z-axis. Permittivity component εa is a linear function of $H_0$.
- The time-reversed version of the magneto–optical material is described by the complex conjugate of its permittivity, with an additional flip of the magnetic bias field $H_0$.

$$T\{\overline{\bar{\varepsilon}}\left(\omega, \mathbf{H}_0\right)\}=\overline{\bar{\varepsilon}}^*\left(\omega,-\mathbf{H}_0\right)$$

- This practically means changing loss to gain and reversing the direction of the bias field.
- The wave equation for magneto–optical media has the form:

$$\nabla \times\left[\overline{\bar{\varepsilon}}^{-1}\left(\omega, \mathbf{H}_0\right) \cdot\left[\nabla \times \mathbf{H}_{\text {orig }}\left(\omega, \mathbf{H}_0\right)\right]\right]=\frac{\omega^2}{c^2} \mathbf{H}_{\text {orig }}\left(\omega, \mathbf{H}_0\right)$$

- By applying complex conjugate to both sides of the above equation with additional flipping the sign of $H_0$ and taking into account 

$$T\left\{\mathbf{H}\left(\omega, \mathbf{H}_0\right)\right\}=-\mathbf{H}^*\left(\omega,-\mathbf{H}_0\right)$$

- We see that the wave solution in the time-reversed magneto– optical material corresponds to $T\{H_{\text {orig} }\}$ (the same waveform but in the opposite direction).
- The perfect inversion of the magneto-optical material is because of the definition we have taken for time- reversal. (microscopically reversing everything).

## IV) RECIPROCITY AND NONRECIPROCITY

In the previous sections,the concepts of time reversal and restricted time reversal have been discussed along with their applicability to several example materials. In this section we will show that the concept of reciprocity is closely related to that of restricted time reversal. For time-invariant systems (whose properties do not change with time), pointwise reciprocity holds if the restricted time reversal does not change the system, and vice versa. However, while the time-inversion concept is intrinsically theoretical and implies process inversion with correct macroscopic initial conditions, the reciprocity principle can be easily applied to real systems and requires only interchanging of the source and detector locations.

### A. Onsager Reciprocal Relations

All processes governed by time symmetric physical laws are time reversal symmetric on the microscopic level. In 1931, Onsager, using this microscopic reversibility, derived relations for lossy linear structures (where the processes are irreversible). These relations, sometimes referred to as the *"fourth law of thermodynamics"* due to their universality, can be applied to an enormous variety of physical phenomena since they are derived using only 4 basic assumptions: *microscopic reversibility* (holds even in lossy systems)*, linearity, causality and thermodynamic quasi-equilibrium*. In this section, we will go through the derivation of the Onsager Reciprocal relations and their applications to different phenomena.

According to quantum statistical mechanics, any system in a state of equilibrium undergoes fluctuations in its macroscopic parameters. For example, let us consider a polar dielectric without externally applied fields at equilibrium. For an arbitrary macroscopic region A, the electric polarization fluctuates around a mean value of zero (dielectric in neutral and no external electric field is applied). The polarization fluctuations $\tilde{P_A}$ at region A are different at each moment from the polarization fluctuations $\tilde{P_B}$ at reach B (here, we consider polarization along arbitrary directions). But the fluctuations in region A and B are not independent due to electrostatic interactions of polar molecules. For a single molecule, all orientations are equally likely but when there are 2 or more molecules, all orientations of the particles aren't equally likely (the orientation in which potential energy of interaction is minimum will have highest probability). This correlation of polarization fluctuations at different locations is conveniently characterized by the correlation function $\langle{\tilde{P_A}\tilde{P_B}}\rangle$ which implies, basically, averaging with respect to probabilities of various values of
$\tilde{P_A}$ and $\tilde{P_B}$. This correlation function can be verified by considering the case where $\tilde{P_A}$ and $\tilde{P_B}$ have arbitrary values independently. The correlation function in this case will be $\langle{\tilde{P_A}\tilde{P_B}}\rangle$ = 0 because for any given $\tilde{P}_{A,i}$, $\tilde{P_B}$ can be positive and negative with equal probability leading to a net zero summation.
In addition to the spatial correlation of fluctuations, we can analogously define temporal correlations. Moreover, correlation can be between fluctuations of different macroscopic quantities, for example, electric polarization and displacement of heat:
$\langle{\tilde{P_A}\tilde{\gamma}_B(t+\tau)}\rangle$ $\neq$ 0 (here $\tilde{\gamma}_B(t+\tau)$ defines the deviation from equilibrium along a given direction in B and $\tau$ is the time delay between the two fluctuations. Onsager recognized the fact that due to microscopic reversibility, some specific polarization $\tilde{P}$ = $\tilde{P_0}$, followed $\tau$ later by some specific heat displacement $\tilde{\gamma}$ = $\tilde{\gamma_0}$ must occur as often as the displacement $\tilde{\gamma}$ = $\tilde{\gamma_0}$ followed $\tau$ later by polarization $\tilde{P}$ = $\tilde{P_0}$.

$$\langle{\tilde{P_A(t)}\tilde{\gamma}_B(t+\tau)}\rangle = \langle{\tilde{P_A(t+\tau)}\tilde{\gamma}_B(t)}\rangle$$

The same equation can be written for fluctuations of the general macroscopic quantities $\tilde{x_i}$ and $\tilde{x_k}$ as:

$$\langle\tilde{x_i}(t)\tilde{x_k}(t+\tau)\rangle=\sigma\langle\tilde{x_i}(t+\tau)\tilde{x_k}(t)\rangle$$


where $\sigma$ = 1 if quantities $\tilde{x_i}$ and $\tilde{x_k}$ have the same time-reversal symmetry and $\sigma$ = -1 if they have opposite symmetry. We can write the same equation in frequency domain as:

$$x_ix_k = \sigma x_kx_i$$

where $x_ix_k = \int_{-\infty}^{\infty}\langle\tilde{x_i}(t+\tau)\tilde{x_k}(t)\rangle e^{(-j\omega\tau)} d\tau$

The above relations indicate the constraints on fluctuations of physical quantities in the equilibrium imposed by the microscopic time reversibility. Next, we need to determine what constraints are imposed by microscopic reversibility on stationary processes under small external perturbations. Stationary processes are processes wherein the system can be considered near thermodynamic equilibrium, that is, in quasi-equilibrium; there are no net macroscopic flows of energy at each moment of time. In the presence of an external perturbation, physical quantity $\tilde{x}_i(t)$ acquires some nonzero mean value $\overline{\tilde{x}_i}$(t) apart from the fluctuations:

$$\overline{\tilde{x}}_i(t)=\int_0^{\infty} \tilde{\alpha}_{i k}(\tau) \tilde{f}_k(t-\tau) d \tau$$

In this relation, $\tilde{\alpha}_{i k}$ is the generalized susceptibility tensor that relates the response of the system $\overline{\tilde{x}}_i(t)$ to the generalized forces $\tilde{f}_k(t)$. Here we note that the lower limit of the integral is 0 and not -$\infty$ due to causality. The above equation is applicable to all linear causal perturbation processes. 
The relation between fluctuations and perturbation processes is given by the fluctuation-dissipation theorem:
$$x_i x_k=\frac{-j \hbar}{2}\left(\alpha_{k i}^*-\alpha_{i k}\right) \operatorname{coth} \frac{\hbar \omega}{2 k_{\mathrm{B}} T}$$

The theorem states that thermal fluctuations of some macroscopic quantity in a system in thermal equilibrium have the same nature as the dissipation processes related to this quantity in the system in thermal quasi-equilibrium. Applied to electric polarization, the theorem implies that the intensity of the polarization fluctuations in the material is proportional to the imaginary part of its permittivity that is responsible for the dissipation of energy in the material. Thus, if there is a process accompanied by energy dissipation into heat, there should exist a reversed process, which converts heat into thermal fluctuations. Substituting the above equation in $x_ix_k = \sigma x_kx_i$, we get

$$\alpha_{k i}^*-\alpha_{i k}=\sigma\left(\alpha_{i k}^*-\alpha_{k i}\right)$$

This equation, along with Kramers-Kronig formulae, leads to

$$\alpha_{k i}(\omega) = \sigma\alpha_{i k}(\omega)$$

For every combination of i and k indices, $\sigma$ will either be +1 or -1 depending on whether $\tilde{x}_i$ and $\tilde{x}_k$ have the same symmetry under time reversal or not. These relations are referred to as the Onsager reciprocal relations. They impose fundamental restrictions on the generalized susceptibility tensors of arbitrary nature. If the relations are satisfied, the system is called reciprocal. When they do not hold, it is said that the system is nonreciprocal.
Casimir later pointed out that although nonreciprocal systems (systems with external time-odd bias) such as the magnetic field, are not constrained by the above relations, there is another relation that they must obey. This relation is as follows:

$$\alpha_{k i}\left(\omega, \mathbf{H}0\right)=\sigma \alpha{i k}\left(\omega,-\mathbf{H}_0\right)$$

Here, for the sake of compactness, we denote all the bias parameters as a single time-odd parameter, the magnetic field vector $\mathbf{H}_0$. The relation above is called the Onsager-Casimir relation and is a more general version of the Onsager reciprocal relation. They cannot be used to determine whether a system is reciprocal or nonreciprocal since they hold for either of these cases. These relations can be applied to a variety of irreversible physical processes of different nature: Acoustic, electromagnetic and so on. We will now consider 2 examples of  the application of the Onsager reciprocal relations to electromagnetic processes.

The first example is based on radiation from electromagnetic sources. Here, we assume that the sources are represented by an electric current density distribution $\mathbf{J(r)}$ with the dimensions of A/$m^2$ in a general nonhomogeneous and anisotropic medium. The electric field radiated by the sources (in the frequency domain) is given by the volume integral equation:

$$\mathbf{E}(\mathbf{r})=\int_V \bar{\bar{G}}\left(\mathbf{r}, \mathbf{r}^{\prime}\right) \cdot \mathbf{J}\left(\mathbf{r}^{\prime}\right) d V^{\prime}$$

where $\bar{\bar{G}}\left(\mathbf{r}, \mathbf{r}^{\prime}\right)$ is dyadic Green's function. For example, for an isotropic homogeneous medium with relative permeability $\mu$, the Green's function has the following form:

$$\bar{\bar{G}}\left(\mathbf{r}, \mathbf{r}^{\prime}\right)=\bar{\bar{G}}\left(\mathbf{r}^{\prime}, \mathbf{r}\right)=-j \omega \mu \mu_0\left[\overline{\bar{I}}+\frac{1}{k^2} \nabla \nabla\right] \frac{e^{-j k\left|\mathbf{r}-\mathbf{r}^{\prime}\right|}}{4 \pi\left|\mathbf{r}-\mathbf{r}^{\prime}\right|}$$


Green’s function describes how strong is an elementary electric field $d\mathbf{E(r)}$ at point *r* created by an elementary single point source $\mathbf{J}\left(\mathbf{r}^{\prime}\right) d V^{\prime}$ at point *r'*

$$d \mathbf{E}(\mathbf{r})=\bar{\bar{G}}\left(\mathbf{r}, \mathbf{r}^{\prime}\right) \cdot \mathbf{J}\left(\mathbf{r}^{\prime}\right) d V^{\prime}$$

The above relation implies the linear (electric field is a linear function of the current
density) and causal (electric field is the response function of the system depending on the current radiation in the
past only) process. Taking into account the macroscopic reversibility of the process and assuming that it evolves in thermodynamic quasi-equilibrium, we see that the Green's function satisfies all the conditions of a generalized susceptibility. Let us assume that there are no externally applied fields, in which case  $\mathbf{H}_0=0$. Applying the Onsager reciprocal relations. Applying the Onsager reciprocal relations for two infinitesimal current sources positioned at *r* and **r'**, we obtain:
$$\bar{\bar{G}}\left(\mathbf{r}, \mathbf{r}^{\prime}\right)=\bar{\bar{G^T}}\left(\mathbf{r}^{\prime}, \mathbf{r}\right)$$

In the derivations, the response functions $\tilde{x}_i$ and $\tilde{x}_k$ were replaced by $d \mathbf{E}(\mathbf{r})=\bar{\bar{G}}\left(\mathbf{r}, \mathbf{r}^{\prime}\right) \cdot \mathbf{J}\left(\mathbf{r}^{\prime}\right) d V^{\prime}$ and $d \mathbf{E}(\mathbf{r'})=\bar{\bar{G}}\left(\mathbf{r'}, \mathbf{r}\right) \cdot \mathbf{J}\left(\mathbf{r}\right) d V$ respectively. We have use the fact that $\sigma$ is equal to +1 since all response quantities are electric field components, which are time-even under time reversal. The dyadic Green's function satisfies the reciprocity relation. This fact implies that the process of radiation from electromagnetic sources in a homogeneous medium described by scalar permittivity $\epsilon$ and $\mu$ is reciprocal.

We will now see what kind of symmetry on the sources and their fields is imposed by the relation for dyadic Green’s function.  In order to do that, we consider the simplest electromagnetic system consisting of two sources *$J_A$* and *$J_B$* whose locations are described by vectors *$r_A$* and *$r_B$* (*$r_A$* and *$r_B$* define a manifold of vectors that indicate directions to all possible point sources in A and B), respectively. Similar considerations can be made for a system of 3 and more sources. Let us assume that the sources are located at different positions and have arbitrary orientations in the xz plane (the current densities have only the x and z components). 

$$\begin{aligned} & \left(\begin{array}{l}d E_{A, x} \\ d E_{A, z}\end{array}\right)=\left(\begin{array}{ll}G_{x x} & G_{x z} \\ G_{z x} & G_{z z}\end{array}\right) \cdot\left(\begin{array}{l}J_{A, x} d V_A \\ J_{A, z} d V_A\end{array}\right) \\ & \left(\begin{array}{ll}
d E_{B, x} \\ d E_{B, z}\end{array}\right)=\left(\begin{array}{ll}G_{x x} & G_{x z} \\ G_{z x} & G_{z z}\end{array}\right) \cdot\left(\begin{array}{l}J_{B, x} d V_B \\ J_{B, z} d V_B\end{array}\right)\end{aligned}$$

where $d\textbf{E}_A$ and $d\textbf{E}_B$ correspond to the electric fields at positions *$r_A$* and *$r_B$* created by sources *$J_A$* $dV_A$ and *$J_A$* $dV_B$, respectively. We can combime the two matrix equations and rewrite it as:

$$\begin{aligned} & \left(\begin{array}{ll}G_{x x} & G_{x z} \\ G_{z x} & G_{z z}\end{array}\right) = \left(\begin{array}{ll}d E_{A, x} & d E_{B, x} \\ d E_{A, z} & d E_{B, z}\end{array}\right) \cdot\left(\begin{array}{ll}J_{A, x} d V_A & J_{B, x} d V_B \\ J_{A, z} d V_A & J_{B, z} d V_B\end{array}\right)^{-1}\end{aligned}$$

For reciprocal systems, we require $G_{x z} = G_{z x}$. Upon equating their expressions, we get:

$$\begin{aligned} d E_{B, x} J_{A, x} d V_A + & d E_{B, z} J_{A, z} d V_A = d E_{A, x} J_{B, x} d V_B+d E_{A, z} J_{B, z} d V_B .\end{aligned}$$

The same derivations can be repeated for the other two other scenarios when the current densities have only the y and z components and only the x and y components. Summing up all the component equations derived from the 3 scenarios and vectorizing them, we get:

$$d \mathbf{E}_B \cdot \mathbf{J}_A d V_A=d \mathbf{E}_A \cdot \mathbf{J}_B d V_B$$

Integrating over the volume occupied by current source A for a fixed point *$r_B$* yields:

$$\int_{V_A} d \mathbf{E}_B \cdot \mathbf{J}_A d V_A=\mathbf{E}_A \cdot \mathbf{J}_B d V_B$$

Now, integrating over the volume occupied by current source b for a fixed point *$r_A$* yields:

$$\int_{V_A} \mathbf{E}B \cdot \mathbf{J}_A d V_A=\int{V_B} \mathbf{E}_A \cdot \mathbf{J}_B d V_B$$


The obtained relation for two electromagnetic sources and their fields in an anisotropic nonhomogeneous medium is a reciprocity condition that is called the Lorentz reciprocity relation. Interestingly, it was discovered by Lorentz in 1896, long before the Onsager reciprocal relations, which we used for our derivation, were known.
Likewise, we can derive a similar relation that will hold for radiation in both reciprocal and nonreciprocal media.
Applying the Onsager–Casimir relation in our geometry with two sources at *$r_A$* and *$r_B$* , we obtain

$$\bar{\bar{G}}\left(\mathbf{r}_A, \mathbf{r}_B, \mathbf{H}_0\right)=\bar{\bar{G^T}}\left(\mathbf{r}_B, \mathbf{r}_A,-\mathbf{H}_0\right)$$

Here, $-\textbf{H}_0$ implies that the corresponding quantity should be considered in the medium with all the
external bias fields reversed. We can now find the final relation connecting the currents with the electric fields in a different and more generalized way as:

$$\begin{aligned} d \mathbf{E}_B\left(\mathbf{r}_A, \mathbf{H}_0\right) & =\overline{\bar{G}}\left(\mathbf{r}_A, \mathbf{r}_B, \mathbf{H}_0\right) \cdot \mathbf{J}_B\left(\mathbf{r}_B\right) d V_B \\ d \mathbf{E}_A\left(\mathbf{r}_B,-\mathbf{H}_0\right) & =\overline{\bar{G}}\left(\mathbf{r}_B, \mathbf{r}_A,-\mathbf{H}_0\right) \cdot \mathbf{J}_A\left(\mathbf{r}_A\right) d V_A\end{aligned}$$


$\begin{aligned} & {\left[\overline{\bar{G}}\left(\mathbf{r}_A, \mathbf{r}_B, \mathbf{H}_0\right) \cdot \mathbf{J}_B\left(\mathbf{r}_B\right)\right]^T \mathbf{J}_A\left(\mathbf{r}_A\right) d V_A d V_B} =\mathbf{J}_B^T\left(\mathbf{r}_B\right) \overline{\bar{G}}^T\left(\mathbf{r}_A, \mathbf{r}_B, \mathbf{H}_0\right) \mathbf{J}_A\left(\mathbf{r}_A\right) d V_A d V_B \\ & =\mathbf{J}_B\left(\mathbf{r}_B\right) \cdot\left[\overline{\bar{G}}\left(\mathbf{r}_B, \mathbf{r}_A,-\mathbf{H}_0\right) \mathbf{J}_A\left(\mathbf{r}_A\right) d V_A\right] d V_B = d \mathbf{E}_A\left(\mathbf{r}_B,-\mathbf{H}_0\right) \cdot \mathbf{J}_B\left(\mathbf{r}_B\right) d V_B .\end{aligned}$

Upon integtrating over $V_A$ and $V_B$, we get

$$\int_{V_A} \mathbf{E}B\left(\mathbf{H}_0\right) \cdot \mathbf{J}_A d V_A=\int{V_B} \mathbf{E}_A\left(-\mathbf{H}_0\right) \cdot \mathbf{J}_B d V_B$$


This relation is referred to as the Onsager–Casimir theorem that is applied to both reciprocal and nonreciprocal linear time-invariant (LTI) systems. Naturally, for reciprocal media (with $\textbf{H}_0$ = 0), the above relation simplifies
to the Lorentz reciprocity relation.

$$\overline{\bar{S}}_{\text {gyrator }}=\left[\begin{array}{cc}0 & 1 \\ -1 & 0\end{array}\right]$$
A gyrator is considered as a fundamental nonreciprocal element that in combination with four other reciprocal elements, that is a resistor, capacitor, inductor, and ideal transformer, completes the set of building blocks needed for constructing an arbitrarily complex linear passive network.
For example, another nonreciprocal and lossless device is a circulator, a three-port device where the signal can flow between ports 1 → 2 → 3, but not in the opposite direction. The scattering matrix of an ideal circulator can be expressed as
$$\overline{\bar{S}}_{\text {circulator }}=\left[\begin{array}{lll}0 & 0 & 1 \\ 1 & 0 & 0 \\ 0 & 1 & 0\end{array}\right]$$

A three-port circulator can be constructed using the basic nonreciprocal building block, the gyrator, and two quater- wave transmission lines.
These scattering matrices that characterize these two examples are unitary, meaning that they are lossless systems.

*Non reciprocal and lossy systems*:Finally, there are devices whose matrices are not symmetric nor unitary. One of the most important devices fulfilling these properties is the isolator:
$$\overline{\bar{S}}_{\text {isolator }}=\left[\begin{array}{ll}
0 & 0 \\
1 & 0
\end{array}\right]$$

This two-port device allows transmission in one direction, but both transmission and reflection are forbidden in the opposite direction. Importantly, lossless isolators cannot exist: a two-port network described by the above scattering matrix is matched at both ports, meaning that the wave falling on the isolated port cannot be reflected back and must be absorbed inside the isolator.



### D. Different Routes for Breaking Reciprocity
Now, we will discuss the necessary condition to sustain reciprocity and how to break it-
1) time-reversal symmetry of microscopic equations; 
2) linear response;
3) causal response;
4) thermodynamic quasi-equilibrium.

The last condition should be discussed separately. All the previous formulations were supported by the assumption of thermodynamic quasi-equilibrium or, in other words, assumption that the system is in a stable and stationary state reached after interactions with its surroundings for enough long time (so-called linear or the Onsager region ). In this state, there are no net macroscopic flows of thermal energy. Particularly, in electromagnetic theory, this regime is achieved when the perturbations produced by the applied fields are slow enough to ensure that the particles equilibrate to the surrounding particles.

In order to achieve nonreciprocity in a system, at least one of the mentioned conditions must be made invalid (however, it is not a sufficient condition). Thus, we can list several possible routes toward breaking reciprocity. The first condition of time-reversal symmetry of micro- scopic equations can be violated by introducing to the system a time-odd external force/parameter H0. In this case, relation (29) does not hold anymore αki(ω,H0) = σαik(ω,H0), and the system may exhibit nonreciprocal response. Possible time-odd external parameters include, but not limited to, the following:

1) external magnetic fields, for example, applied to plasma or ferrite (see detailed discussion in Section V-A);
2) exchange interaction force, for example, in antiferro- magnets;
3) linear velocity using linearly moving structures or linear space–time modulation (see detailed discussion in Section VII);
4) angular velocity (rotating objects or space–time mod- ulation emulating rotation).

there can be a special discussion on breaking the odd time parameter to break the symmetry.The second condition of linear response can be nat- urally broken using nonlinear systems. However, as it will be discussed in Section VI, the nonlinearity route for breaking reciprocity is not universal and has its own limitations. The causality assumption does not apply to active systems, meaning that reciprocity can be broken in systems comprising amplifiers or parametric amplifiers. The use of systems far from equilibrium also appears possible for achieving nonreciprocity. It is known that the fluctuation–dissipation theorem is violated in nonequilib- rium glassy systems (systems that slowly approach their equilibrium state). In such systems, the Onsager reciprocal relations do not necessarily hold.














## V. Non Reciprocity in linear time invariant(LTI) media.

### Nonreciprocal effects using LTI materials

Faraday discovered the first nonreciprocal effect in 1845. He observed rotation of a linearly polarised light propagating through a rod of lead borate glass placed in a static magnetic field. Changing the direction of the field or incident light beam, the sense of rotation was reversed. We now give a classical description of Faraday rotation.

#### Faraday Rotation

Screenshot 2023-04-02 195449.png

Consider a free electron in the lattice of a magneto optical material. We apply an incident light beam, which applies a force to the electron. Consider only the force from the electric field of the light beam since the Lorentz force due to light beam is quite small ($\approx \frac{v}{c}E$). Let us assume the light is right circularly polarized and the bias is $B_0\textbf{z_0}$, toward the light source.

Equation of motion:
$$m\frac{d^2\tilde{\textbf{r}}}{dt^2} = -k_a\tilde{\textbf{r}}-\Gamma\frac{d\tilde{\textbf{r}}}{dt} + e\tilde{\textbf{E_i}} + eB_0\frac{d\tilde{\textbf{r}}}{dt}\times\textbf{z_0}$$
with $k_a$ being the coefficient for the Coulomb force, $\Gamma$ being the coefficient for the dissipative force. If we illuminate with LCP, the direction of the Lorentz force will be opposite, leading to a different effective cyclotron radii and hence microscopic polarizabilities. Taking the Fourier transform, and resolving along $x_0$ and $y_0$

$$(m\omega_0^2 - m\omega^2 +j\omega\Gamma)x(\omega) - j\omega eB_0y(\omega) = eE_{ix}$$
$$(m\omega_0^2 - m\omega^2 +j\omega\Gamma)y(\omega) + j\omega eB_0x(\omega) = eE_{iy}$$

Assuming circularly polarized light $E_{iy} = \pm jE_{ix} = \frac{E_i}{\sqrt{2}}$(+ for RCP and - for LCP). The electron also undergoes circular motion with $y(\omega) = \pm jx(\omega) = \frac{r(\omega)}{\sqrt{2}}$. The elctric polarizability for the ion-electron pair $\alpha_{\pm} = \frac{er}{E_i\epsilon_0}$ is given by

$$\alpha_{\pm} = \frac{\frac{e^2}{\epsilon}}{(m\omega_0^2-m\omega^2 + j\omega\Gamma)\pm e\omega B_0}$$
and so the refractive index is,
$$n_{\pm} = \sqrt{1+N_e\alpha{\pm}(\omega)} = n'_{\pm} - j\kappa_{\pm}(\omega)$$.

The Faraday rotation angle; $\theta_F(\omega) = \frac{\omega L}{2c}[n'_{RCP}-n'_{LCP}]$

It is now easy to demonstrate that $\theta_F$ flips sign for opposite propagation of light. With the opposite propagation, the magnetic field with respect to the wave vector reverses, which means that $n^{op}_{LCP} = n_{RCP}$ and vice versa, leading to a flip in the sign of the faraday rotation angle.

#### Kerr effect
By analogy with polarization rotation of transmitted linearly polarized light through a sample of a magneto–optical material, one should also expect polarization conversion for light reflected from such material. Here we consider a special case of the Kerr effect, the polar magneto-optical Kerr effect.

A normally incident light beam is linearly polarized along the $x$-axis, and the external static magnetic flux density $B_0z_0$ is along the $+z$-direction.
It is convenient to decompose the incident beam into a combination of LCP and RCP light, that is, $E_i\textbf{x}_0 = \frac{E_i}{2}(\textbf{x}_0 + j\textbf{y}_0)+\frac{E_i}{2}(\textbf{x}_0 - j\textbf{y}_0)$.
We can now use the Fresnal formula for the reflection coefficient and arrive at an expression for the reflected field:

$$E_r = \frac{1-n_{RCP}}{1+n_{RCP}}\frac{E_i}{2}(\textbf{x}_0 + j\textbf{y}_0)+\frac{1-n_{LCP}}{1+n_{LCP}}\frac{E_i}{2}(\textbf{x}_0 - j\textbf{y}_0)$$

Due to magnetization by $B_0$, the reflected light gets orthogonally polarized about the $y$ direction. The imaginary part of $\frac{E_{ry}}{E_{rx}} = j\frac{n_{RCP}-n_{LCP}}{n_{RCP}n_{LCP}-1}$ is the Kerr ellipticity, while the real part is the Kerr rotation. We can solve for the permittivity tensor from the refractive index and get the following expressions for the diagonal and off-diagonal terms

$\begin{aligned} \varepsilon_{x x} & =\varepsilon_{y y}=1+N_e e \frac{x(\omega)}{\varepsilon_0 E_{i x}(\omega)} \\ & =1+\Delta\left(m \omega_0^2-m \omega^2+j \omega \Gamma\right) \\ \varepsilon_{y x} & =-\varepsilon_{x y}=N_e e \frac{y(\omega)}{\varepsilon_0 E_{i x}(\omega)}=-j \Delta \omega e B_0 \\ \Delta & =\frac{N_e e^2 / \varepsilon_0}{\left(m \omega_0^2-m \omega^2+j \omega \Gamma\right)^2-\left(\omega e B_0\right)^2}\end{aligned}$

Using these expressions gives us $\frac{E_{ry}}{E_{rx}} = \frac{\epsilon_{yx}}{\sqrt{\epsilon_{xx}}(\epsilon_{xx}-1)}$. For lossless media, the ratio is purely imaginary implying elliptical polarization.

If we carry out the same mathematics for light polarized in the $y$ direction, we will see that the ratio comes out to be negative and inverted, that is, 

$$\frac{E_{ry}}{E_{rx}} = -(\frac{\epsilon_{yx}}{\sqrt{\epsilon_{xx}}(\epsilon_{xx}-1)})^{-1}$$

This implies a change of phase by $\pi$ radians, which shows nonreciprocal response of the medium. Fig 12 shows this effect for clarity.

Screenshot 2023-04-02 195715.png

As a conclusion to this discussion on the Kerr effect, note that opposite effects can exist, meaning that external light interacting with a medium leads to its magnetization. Such effects are called the inverse Kerr and the inverse Faraday effects. There exis other nonreciprocal effects as well, which will not be discussed here. Examples are Cotton-Mouton or Voigt effects. In this, light passes perpendicularly through a magnetic field applied to a vapour or a liquid. Magnetic linear birefringence takes place in such a setup.  

### C. General classification scheme based on space-time symmetries

Now we study and classify LTI materials on the basis of space and time symmetries. To do this, we look at how the response functions depend on the frequency and the external biasing, which we will represent using $\textbf{Q}$. The electromagnetic system under consideration must obey space and time inversion symmetries. We now write the material relations in general form
$$D = \epsilon(\omega,\textbf{Q})\cdot E + \xi(\omega,\textbf{Q})\cdot H$$
$$B = \zeta(\omega,\textbf{Q})\cdot E + \mu(\omega,\textbf{Q})\cdot H$$
Now we do two things:
1. Split the material tensors into a time-even and a time-odd part as follows
$$\epsilon_{TE}(\omega,\textbf{Q}_{TE}) = \epsilon_1(\omega)+\epsilon_2(\omega)\textbf{Q}_{TE}$$
$$\epsilon_{TO}(\omega,\textbf{Q}_{TO}) = \epsilon_3\textbf{Q}_TO$$
2. Apply time reversal operator to the material relations
3. Apply complex conjugation to the material relations
4. Compare coefficients

We get the following 8 relations
$$\epsilon_{TE}(\omega,\textbf{Q}_{TE})=\epsilon_{TE}^*(\omega,\textbf{Q}_{TE})$$
$$\epsilon_{TO}(\omega,\textbf{Q}_{TO})=-\epsilon_{TO}^*(\omega,\textbf{Q}_{TO})$$
$$\xi_{TE}(\omega,\textbf{Q}_{TE})=-\xi_{TE}^*(\omega,\textbf{Q}_{TE})$$
$$\xi_{TO}(\omega,\textbf{Q}_{TO})=\xi_{TO}^*(\omega,\textbf{Q}_{TO})$$
$$\zeta_{TE}(\omega,\textbf{Q}_{TE})=-\zeta_{TE}^*(\omega,\textbf{Q}_{TE})$$
$$\zeta_{TO}(\omega,\textbf{Q}_{TO})=\zeta_{TO}^*(\omega,\textbf{Q}_{TO})$$
$$\mu_{TE}(\omega,\textbf{Q}_{TE})=\mu_{TE}^*(\omega,\textbf{Q}_{TE})$$
$$\mu_{TO}(\omega,\textbf{Q}_{TO})=-\mu_{TO}^*(\omega,\textbf{Q}_{TO})$$

But these are too many, because we are also constrained by the Onsager-Casimir relations. Applying those, we are left with 6 relations,
$$\epsilon_{TE} = \epsilon_{TE}^T = \epsilon_{r}$$
$$\epsilon_{TO} = -\epsilon_{TO}^T = \epsilon_{n}$$
$$\mu_{TE} = \mu_{TE}^T = \mu_{r}$$
$$\mu_{TO} = \mu_{TO}^T = \mu_{n}$$
$$\xi_{TE} = -\zeta_{TE}^T = -\frac{j}{c}\kappa$$
$$\xi_{TO} = \zeta_{TO}^T = \frac{1}{c}\chi$$

The new material tensors introduced in the last two relations are the reciprocal chirality tensor and the nonreciprocal Tellegen tensor. The imaginary unit $j$ only appears with the chirality tensor, an important consequence of the fact that chirality is an effect of spatial dispersion and vanishes when $\omega \approx 0$. After some manipulations and using the Condon model(included in the references), we can get the time inversion transformations for the material tensors
$$\epsilon_{r}(\omega, \textbf{Q}_{TE}) = \epsilon_{r}^*(\omega, \textbf{Q}_{TE})$$
$$\epsilon_{n}(\omega, \textbf{Q}_{TO}) = -\epsilon_{n}^*(\omega, \textbf{Q}_{TO})$$
$$\mu_{r}(\omega, \textbf{Q}_{TE}) = \mu_{r}^*(\omega, \textbf{Q}_{TE})$$
$$\mu_{n}(\omega, \textbf{Q}_{TO}) = -\mu_{n}^*(\omega, \textbf{Q}_{TO})$$
$$\kappa(\omega, \textbf{Q}_{TE}) = \kappa^*(\omega, \textbf{Q}_{TE})$$
$$\chi(\omega, \textbf{Q}_{TO}) = \chi^*(\omega, \textbf{Q}_{TO})$$

Having derived the time reversal properties of the medium, we now consider the parity inversion properties. First we utilise the linearity of the material tensors to figure out which of them are true tensors(gets a - sign under parity) and which of them are pseudo-tensors(stays the same under parity).
$\epsilon_{r}(\omega,\textbf{Q}_{TE}),\epsilon_{n}(\omega,\textbf{Q}_{TO}) = \epsilon_n(\omega,Q_{TO})\textbf{Q}_{TO}\times I,\mu_{n}(\omega,\textbf{Q}_{TO}) =\mu_{n}(\omega,Q_{TO})\textbf{Q}_{TO}\times I ,\mu_{r}(\omega,\textbf{Q}_{TE})$ are all true tensors, while $\kappa_{s}(\omega,\textbf{Q}_{TE}),\kappa_{a}(\omega,\textbf{Q}_{TO}) = \kappa_a(\omega,Q_{TO})\textbf{Q}_{TO}\times I,\chi_{a}(\omega,\textbf{Q}_{TO}) =\chi_{a}(\omega,Q_{TO})\textbf{Q}_{TO}\times I ,\chi_{s}(\omega,\textbf{Q}_{TE})$ are all pseudo tensors.

## VI. Nonreciprocity in non-linear systems

It is clear from the discussion on the Lorentz reciprocity conditions that non-linearity may be our way to break the condition. This is often done to manipulate the transition characteristics of systems to develop nonreciprocal components.

The general principle behind such devices is the fact that the field distribution at the source and observer positions can be significantly different. We generally position nonlinear objects where the field strength is high for one direction and weak for the other in order to utilise the non-linearity of the system.


image.png

We follow this explanation with a simple example using fig 17. The figure shows the coupling of two antennas, the right one having a non-linear object that is sensitive to the current flowing in the circuit(a fuse would be a simple example). To simplify the mathematics and make it more intuitive, we place the antennae far apart and make them in such a way that the radii $r$ are much smaller than the free space wavelength $\lambda$, $r<<\lambda$. We now analyse our setup.

Our nonlinear object acts as an open switch when the current is high and a closed switch when the current is low. When the right loop is receiving, the free space path loss $(\frac{\lambda}{4\pi r})^2$ leads to a small current flowing in the circuit even if the current in the left loop is high. However, when the circuit is in the transmitting regime, since the current is high, the nonlinear object is shorted and no power is transferred. Thus one could simply design a nonreciprocal link between antennae using nonlinearity.

In the above example, we had the free space path loss attenuating the energy and assuring field asymmetry. What if we wanted to design a nonreciprocal system **without** power spread? We can try to create such a system.

We consider an infinite(in the transverse direction) dielectric slab as shown in the figure below

image.png

The top and the bottom of the figure represent the behaviour of the magnitude of the electric field inside the slab using a colour gradient. Due to non-identical homogeneous permittivities and thickness, the setup is spatially asymmetric, Now;
1. We setup standing waves from either side, one by one, by illuminating the slab.
2. We find the ratio of the magnitudes of the electric fields(forward/backward) within the slab
3. Put the nonlinear component where the ratio is highest.

This way of constructing our system makes it so that the response to the electric fields is vastly different. Nonlinearity with spatial asymmetry has helped us break the Lorentz reciprocity.

The abovementioned explanations were based on assumption that the forward and backward incident waves are not concurrently present, clearly an *non*practical assumption. We give another example of the problem.

image.png

Fig 19 illustrates a dielectric slab with a non homogeneous response to electric fields described by a permittivity $\epsilon(\textbf{r})$ for weak excitations and by $\epsilon(\textbf{r})+ \Delta\epsilon(\textbf{r},E)$ for strong excitations. The system is obviously reciprocal for weak electrical fields{fig 19 (a) and (b)}. For strong electric fields, the response function $\Delta\epsilon(\textbf{r},E)$ will introduce different behaviours for forward and backward transmission{fig 19 (c) and (d)}. In the case that we illuminate the slab with both a low power and a high power wave, the low power wave will also experience the high transmission as shown in fig 19(c).

We now figure out the effective relative permittivity for a low power wave when it is incident backwards WITH a high power wave incident in the forward direction.

According to non-linear optics
$$\nabla\times\nabla\times \tilde{E} + \mu_0\epsilon_0\epsilon(\textbf{r})\frac{\partial^2\tilde{E}}{\partial t^2} + \mu_0\frac{\partial^2\tilde{P_{NL}}}{\partial t^2} = 0$$
where $\tilde{E}$ is the time-harmonic electric field, $\epsilon$ is the frequency dependent permittivity in the linear regime, and $\tilde{P_NL}$ denotes the nonlinear polarization density vector that,in the general case, depends on the electric field as

$$\tilde{P_{NL}} = \epsilon_0\chi^{(2)}(\textbf{r})\tilde{E}^2 + \epsilon_0\chi^{(3)}(\textbf{r})\tilde{E}^3 + \cdots$$

where $\chi^{(n)}(\textbf{r})$ is the nth order component of the electric susceptibility of the medium. We neglect the components other than the third order, under the guise of looking at the optical Kerr nonlinearity. So, $\tilde{P_{NL}} = \epsilon_0\chi^{(3)}(\textbf{r})\tilde{E}^3$.

Now, $\tilde{E} = E_{total} = E_hexp(j\omega_ht) + E_lexp(j\omega_lt)$, with $|E_h|>>|E_l|$. Substitute this into the expression for $\tilde{P_{NL}}$ and get;

$$\tilde{P_{NL}} = \tilde{P_0} + \tilde{P_{\omega_l}} = \tilde{P_0} + 6\epsilon\chi^{(3)}(\textbf{r})|E_h|^2E_le^{j\omega_lt}$$

where $\tilde{P_{\omega_l}}$ is the polarization density due to the low-power
incident wave and $\tilde{P_0}$ is the rest of the polarization terms
oscillating at different frequencies. We can introduce $\chi_{\omega_1} \approx 6\chi^{(3)}(\textbf{r})|E_h|^2$

With all this in mind, we rewrite our initial differential equation in the frequency domain as a linearised equation for the low power incident wave

$$\nabla\times\nabla\times {E} - k_1^2[\epsilon(\textbf{r}) + 6\chi^{(3)}(\textbf{r})|E_h|^2]E_l = 0$$

with $k_1 = \frac{\omega_1}{c}$. We get that the effective permittivity $\epsilon_{eff}(\textbf{r}) = \epsilon(\textbf{r}) +  6\chi^{(3)}(\textbf{r})|E_h|^2$. The scalar and time independent nature of $\epsilon_{eff}$ leads to the transmission of the low power wave being reciprocal. The backward propagation will not be blocked, precluding the system described from operating as an ideal isolator.

There is an additional constraint to the nonreciprocity via nonlinearity due to the second law of thermodynamics. If an isolator could operate from both sides simultaneously, the radiative thermal power transferred between the ports would not be 0, despite being at the same temperature. There is another limitation in the form of the fact that there is a tradeoff between transmission forward and the level of input intensity for which large isolation can be obtained.

# VII. Non - Reciprocity in Linear Time-Variant Systems

In this section we look at another method to **break reciprocity** is to **break the time invariance** of the system. <br><br>
Let us first derive reciprocity relations in systems whose properties vary harmonically in time in such a way that, at all points in space, the permittivity alternates with the same phase. Consider a case when permittivity of the material is modulated according to the following time-symmetric modulation function: <br>
<br>
$$\varepsilon(\omega, \mathbf{r}, t)=\varepsilon_{\mathrm{st}}(\omega, \mathbf{r})+M(\mathbf{r}) \cos (\Omega t+\phi)$$ 
<br>
Where, $\widetilde{\varepsilon}_{\mathrm{st}}$ is the static permittivity,
$M(\mathbf{r})$ is the modulation strength function,
$\Omega$ is the modulation frequency, and
$\phi$ is an arbitrary phase.
<br>
<br>
Invoking the frequency-domain wave equation for time-varying material:
<br><br>
$$\nabla \times \frac{1}{\mu(\omega)} \nabla \times \mathbf{E}-\omega^2\left[\varepsilon_{\mathrm{st}}(\omega) \mathbf{E}+\mathbf{P}_{\mathrm{M}}\right]=-j \omega \mathbf{J}$$
<br>
where, $\mathbf{P}_{\mathrm{M}}(\omega)$ is the additional polarization density. The expression for $\mathbf{P}_{\mathrm{M}}(\omega)$ is found by applying the modulation function. We then get, <br>
<br>
$$\mathbf{P}_{\mathrm{M}}(\omega)=\frac{M}{2}\left[e^{j \phi} \mathbf{E}(\omega-\Omega)+e^{-j \phi} \mathbf{E}(\omega+\Omega)\right]$$
<br>
The modulation induces some **sideband harmonics** $\mathbf{E}_{\mathrm{n}}$ of frequencies $\omega_{\mathrm{n}} = \omega_{\mathrm{0}} + n\Omega$ where $n$ is an integer. We get $\mathbf{E}_{\mathrm{n}}$ by substituting the equation of Polarization density found into the wave equation and matching frequency components with temporal variations $e^{j\omega t}$

$$\hat{\Theta}_n \mathbf{E}_n-\omega_n^2 \frac{M}{2}\left[e^{j \phi} \mathbf{E}_{n-1}+e^{-j \phi} \mathbf{E}_{n+1}\right]=-j \omega_n \mathbf{J}_n$$

where $\hat{\Theta}_n$ is defined by the action: $$\hat{\Theta}_n \mathbf{E}_n=\nabla \times \frac{1}{\mu_n} \nabla \times \mathbf{E}_n-\varepsilon_{\mathrm{st}, n} \omega_n^2 \mathbf{E}_n$$

Corresponding to the multi-band electric field, we assume a current source with multiple side bands, with $\mathbf{J}_n$ denoting the source at frequencies $\omega_n=\omega_0+n \Omega$. It is shown that writing the equation for the side-band Electric fields $\mathbf{E}_{\mathrm{n}}$ in the matrix form and expressing electric field vector as $\mathbf{E}_n=\overline{\bar{f}}_{n m} \mathbf{J}_m$, we always obtain asymmetric matrix $\overline{\bar{f}}$. However, by dividing both sides of the equation for the side-band Electric fields $\mathbf{E}_{\mathrm{n}}$ by $\omega_n^2$ and using replacement $\mathbf{J}_m^{\prime}=\mathbf{J}_m / \omega_m$, we obtain relation
<br><br>
$$
\mathbf{E}_n(\mathbf{r})=\overline{\bar{F}}_{n m}(\mathbf{r}) \mathbf{J}_m^{\prime}(\mathbf{r})
$$
<br>
where $\overline{\bar{F}}$ is a symmetric matrix if $\phi=0$ is chosen. Formally solving this equation, the radiated field harmonics by given current source $J$ read as
<br><br>
$$
\mathbf{E}_n(\mathbf{r})=\int_V \overline{\bar{G}}_{n m}\left(\mathbf{r}, \mathbf{r}^{\prime}\right) \mathbf{J}_m^{\prime}\left(\mathbf{r}^{\prime}\right) d V^{\prime}
$$
<br>
Since $\overline{\bar{F}}$ is symmetric, the system's corresponding dyadic Green's function of time-varying material is also symmetric
<br><br>
$$
\overline{\bar{G}}\left(\mathbf{r}, \mathbf{r}^{\prime}\right)=\overline{\bar{G}}^T\left(\mathbf{r}, \mathbf{r}^{\prime}\right)
$$
<br>
where $\mathbf{r}^{\prime}$ denotes the coordinates of points inside source volume $V$. Comparing the above two equations with the Onsager reciprocity equations for time-invariant electro-magnetic field given, we see that **electromagnetic radiation in time-varying material with global modulation satisfies the Onsager reciprocity conditions**, and is always reciprocal. Repeating derivations analogous to to the case where we have derived the Onsager-Casimir relations, we obtain the following equation:
<br><br>
$$
\int_{V_A} \sum_n \frac{\mathbf{E}_n^{(B)} \cdot \mathbf{J}_n^{(A)}}{\omega_n} d V_A=\int_{V_B} \sum_n \frac{\mathbf{E}_n^{(A)} \cdot \mathbf{J}_n^{(B)}}{\omega_n} d V_B
$$
<br>
The above equation related the fields and current harmonics generated via interaction between a pair of two sources A and B. This relation is the electromagnetic reciprocity theorem for time varying systems. This relation as was in the case of the previous cases **imposes restrictions on the properties of the system**. It connects energies of photons if we divide both sides of the relation by the
reduced Planck constant $\hbar$. Thus, the fractions inside the volume integrals have dimensions of number of photons per second per unit area. <br><br>

**Hence we have seen that the reciprocity conditions hold true for global symmetric time-harmonic modulation of permittivity.**
<br><br>
Now let's break it! Take a case where the perturbation is not globally symmetric. If modulation is not global and can be represented by two modulation functions
$$
\bar{\varepsilon}(\omega, \mathbf{r}, t)=\bar{\varepsilon}_{st}(\omega, \mathbf{r})+M_1(\mathbf{r})\cos(\Omega t+\phi_1)+M_2(\mathbf{r})\cos(\Omega t+\phi_2)
$$
<br>
with $\phi_1 \neq \phi_2$ and $M_1(\mathbf{r}) \neq \alpha M_2(\mathbf{r})$ ( $\alpha$ is an arbitrary real constant), then we cannot write in the form $\mathbf{E}_n(\mathbf{r})=\bar{F}_{n m}(\mathbf{r}) \mathbf{J}_m^{\prime}(\mathbf{r})$, with $\bar{F}$ being symmetric matrix (time translation affects both $\phi_1$ and $\phi_2$ and does not make $\bar{F}$ symmetric). Therefore, Green's function is asymmetric, and this time-modulated system will break reciprocity relation. Nevertheless, one can write the Onsager-Casimir relation that imposes constraint even on the time-modulated nonreciprocal systems
<br><br>
$$
\int_{V_A} \sum_n \frac{\mathbf{E}_n^{(B)}(\Omega) \cdot \mathbf{J}_n^{(A)}}{\omega_n} d V_A=\int_{V_B} \sum_n \frac{\mathbf{E}_n^{(A)}(-\Omega) \cdot \mathbf{J}_n^{(B)}}{\omega_n} d V_B
$$
<br>

Now let us tackle the approaches to create non-reciprocal devices based on time modulation. <br>

We look at **Travelling Wave Modulation**. In such cases the modulation of the medium can encourage the transmission of a certain mode in a certain direction, analagous to the idea of pushing and pulling to increase the amplitude of a swing. Consider an inhomogeneous isotropic dielectric waveguide with permittivity 
<br><br>
$$\tilde{\varepsilon}(x, y, z, t)=\varepsilon_{\mathrm{st}}(x, y)+M(x, y) \cos \left(\omega_m t-k_m z\right)$$
<br>
where $\varepsilon_{\mathrm{st}}(x, y)$ denotes the static permittivity, $M(x, y)$ s the modulation amplitude distribution, $\omega_m$ represents the modulation angular frequency, $k_{\mathrm{m}}$ is the modulation wavenumber, and finally, $z$ is the axis of propagation.
Inside the waveguide two modes of travel for the signal is supported, namely $\omega_1$ and $\omega_2$ with their corresponding $k_1$ and $k_2$. In the forward diraction, if the conditions $\omega_m = \omega_1 - \omega_2$ and the corresponding $k_m = k_1 - k_2$ are satisfied then the **modulation encourages the transition from one mode to another**. This does not happen in the reverse direction where the wave modes are $-k_1$ and $-k_2$, hence giving us non-reciprocal behaviour.

Reciprocity means that if we swap the locations of any two sources, the resulting field at point $\mathbf{P}$ should be the same. Dispersion means that the phase velocity of the wave depends on its frequency, so we need to sum over a range of frequencies to get an accurate result.

To make things even more challenging, let's say that the sources are not all at the same frequency. Instead, each source emits a wave with a frequency distribution given by a complex function $f_i(\omega)$, where $\omega$ is the angular frequency. Furthermore, the amplitude of each source is proportional to a function $g_i(\omega)$, which depends on its frequency as well.

Find an expression for the total electric field at point $\mathbf{P}$, in terms of the source locations, frequencies, amplitudes, and dispersion properties. This expression should satisfy the reciprocity principle, and should be valid for arbitrary distributions of sources and frequencies.

The total electric field at point P due to the $N$ sources can be expressed as a superposition of the electric fields generated by each source, as follows:
