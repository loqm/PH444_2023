## Abstract
Transformation optics has become a valuable tool for designing and understanding electromagnetic systems without the need for a ray approximation. While most studies have focused on manipulating the trajectories of light rays, this paper explores the use of transformation optics to manipulate electromagnetic fields. 

Three examples are presented: 
1. an electromagnetic cavity for deep subwavelength confinement of light
2. the modification of Cherenkov light emitted in a transformation-optical material from simple geometric considerations
3. the enhancement of optical forces through a folded coordinate transformation of a pair of waveguides

These examples demonstrate the potential for an entirely new spectrum of devices that can be created using transformation optics.

## 1) Introduction

Transformation optics takes advantage of form-invariance of Maxwell's equations under coordinate transformations.
The equivalence between non-trivial geometries and optical media is derived from the study of propagation of light in exotic gravitational fields. Here we are going to use the form invariance in analogy with the techniques of optical conformal mapping.

### (a) Equivalence between geometries and media
Form-invariance of the Maxwell equations under 3D coordinate transformations
We start with the Maxwell equations in a medium:

$$\begin{align}
  \textbf{D}& =\epsilon_0\epsilon\textbf{E},&
  \textbf{B}& =\mu_0\mu\textbf{H}
\end{align}\tag{1}$$

from which, we get the following:

$$\nabla\cdot\mathbf{E} = \frac{\rho}{\epsilon_0\boldsymbol{\epsilon}}$$
$$\nabla\times\textbf{E} = -\frac{\partial(\mu_0\boldsymbol{\mu}\mathbf{H})}{\partial t}$$
$$\nabla\cdot(\mu_0\boldsymbol{\mu}\mathbf{H}) = 0$$
$$\nabla\times\mathbf{H} = -\frac{\partial(\epsilon_0\boldsymbol{\epsilon}\mathbf{E})}{\partial t} + \mathfrak{J}$$

where $\rho$ and $\mathfrak{J}$ denote free charges and currents in the medium.



Now we use the Jacobian $\mathcal{J}$ to transform the equations to an arbitrary coordinate system. 


$$\mathcal{J}_{ij} = \frac{\partial x'_i}{\partial x_j}$$

where $x'_i$ is the new coordinate element and $x_j$ is the old coordinate element.

For a 3D system,

$$\mathcal{J} = \begin{pmatrix}
\mathcal{J}_{11} & \mathcal{J}_{12} & \mathcal{J}_{13}\\
\mathcal{J}_{21} & \mathcal{J}_{22} & \mathcal{J}_{23}\\
\mathcal{J}_{31} & \mathcal{J}_{32} & \mathcal{J}_{33}
\end{pmatrix}$$

We replace $\nabla$ with $\nabla'$, the fields and material properties are transformed under the Jacobian:
$$\mathbf{E}' = (\boldsymbol{\mathcal{J}}^T)^{-1}\mathbf{E}$$
$$\implies E_{i} = \mathcal{J}_{ji}E'_j$$
$$\mathbf{H}' = (\boldsymbol{\mathcal{J}}^T)^{-1}\mathbf{E}$$
$$\implies H_{i} = \mathcal{J}_{ji}H'_j$$
$$\boldsymbol{\epsilon}' = \frac{\boldsymbol{\mathcal{J}}\boldsymbol{\epsilon}\boldsymbol{\mathcal{J}}^T}{det\boldsymbol{\mathcal{J}}}$$
$$\boldsymbol{\mu}' = \frac{\boldsymbol{\mathcal{J}}\boldsymbol{\mu}\boldsymbol{\mathcal{J}}^T}{det\boldsymbol{\mathcal{J}}}$$
$$\boldsymbol{\mathfrak{J}}' = \frac{\boldsymbol{\mathcal{J}}\boldsymbol{\mathfrak{J}}}{det\boldsymbol{\mathcal{J}}}$$
$$\rho' = \frac{\rho}{det\boldsymbol{\mathcal{J}}}$$
We also have
$$\frac{\partial}{\partial x^i} = \mathcal{J}_{ji}\frac{\partial}{\partial x'^j}$$

In the Cartesian coordinate system, the last equation in the set of Maxwell equations is written as:
$$[ijk]\frac{\partial H_k}{\partial x^j} = -\frac{\partial(\epsilon_0\epsilon^{ij}E_j)}{\partial t}+\mathfrak{J}^i$$

After applying the Jacobian, we get:

$$[ijk]\mathcal{J}_{ai}\frac{\partial}{\partial x'^a}\mathcal{J}_{bj}H'_b= -\frac{\partial}{\partial t}(\epsilon_0\epsilon^{ki}\mathcal{J}_{ai}E'_a)+\mathfrak{J}^k$$

Here, the $\mathcal{J}_{ai}\frac{\partial}{\partial x'^a} = \frac{\partial}{\partial x^i}$ term falls on both $\mathcal{J}_{bj}$ and $H'_b$ terms, but we can eliminate the $\mathcal{J}_{bj}$ because of the permutation symbol, $[ijk]$.

$[ijk]\frac{\partial}{\partial x^i}\mathcal{J}_{bj} = 0$ because $\frac{\partial}{\partial x^i}\mathcal{J}_{bj} = \frac{\partial}{\partial x^j}\mathcal{J}_{bi}$.

Multiply both sides by the Jacobian $\mathcal{J}_{ck}$ to get:
$$[ijk]\mathcal{J}_{ck}\mathcal{J}_{bj}\mathcal{J}_{ai}\frac{\partial}{\partial x'^a}H'_b= -\epsilon_0\mathcal{J}_{ck}\epsilon^{ki}\mathcal{J}_{ai}\frac{\partial}{\partial t}E'_a+\mathcal{J}_{ck}\mathfrak{J}^k$$

We know that
$$[ijk]\mathcal{J}_{ai}\mathcal{J}_{bj}\mathcal{J}_{ck} = [ijk]det\boldsymbol{\mathcal{J}}$$
by definition of the determinant.

This gives us
$$[ijk]\frac{\partial}{\partial x'^a}H'_b = -\epsilon_0\frac{\mathcal{J}_{ak}\epsilon^{ki}\mathcal{J}_{ai}}{det\boldsymbol{\mathcal{J}}} \frac{\partial E'_a}{\partial t} + \frac{\mathcal{J}_{ck}\mathfrak{J}^k}{det\boldsymbol{\mathcal{J}}}$$

In vector form:
$$\nabla'\times\textbf{H}' = -\frac{\partial }{\partial t}(\epsilon_0\boldsymbol{\epsilon}'\textbf{E}') + \boldsymbol{\mathfrak{J}}'$$

where

$$\boldsymbol{\epsilon}' = \frac{\boldsymbol{\mathcal{J}\epsilon\mathcal{J}}^T}{det\boldsymbol{\mathcal{J}}}$$

and

$$\boldsymbol{\mathfrak{J}}' = \frac{\boldsymbol{\mathcal{J}\mathfrak{J}}}{det\boldsymbol{\mathcal{J}}}$$

In a similar argument, we obtain:
$$\nabla'\times\textbf{E}' = -\frac{\partial }{\partial t}(\mu_0\boldsymbol{\mu}'\textbf{H}')$$

where

$$\boldsymbol{\mu}' = \frac{\boldsymbol{\mathcal{J}\mu\mathcal{J}}^T}{det\boldsymbol{\mathcal{J}}}$$

As for the remaining divergence equations, we get:

$$\rho = \frac{\partial}{\partial x^i}(\epsilon_0\epsilon^{ij}E_j)$$

Applying the Jacobian:
$$\rho = \mathcal{J}_{ai}\frac{\partial}{\partial x'^a}(\epsilon_0\epsilon^{ij}\mathcal{J}_{bj}E'_b)$$

We make use of the transformed permittivity after replacing $\nabla$ with $\nabla'$
$$\implies \epsilon_0\mathcal{J}_{ai}\frac{\partial}{\partial x'^a}(det\boldsymbol{\mathcal{J}}) \mathcal{J}_{ic}^{-1}\epsilon'^{cb}E'_b$$

$$\implies \epsilon_0(det\boldsymbol{\mathcal{J}})\frac{\partial}{\partial x'^a}\epsilon'^{ab}E'_b + \frac{\partial}{\partial x^i}\mathcal{J}_{ic}^{-1}det\boldsymbol{\mathcal{J}}\epsilon'^{cb}E'_b$$

$$\implies \rho = \epsilon_0(det\boldsymbol{\mathcal{J}})\frac{\partial}{\partial x'^a}\epsilon'^{ab}E'_b $$

We made use of:
$$\frac{\partial}{\partial x^i}\mathcal{J}_{ic}^{-1}det\boldsymbol{\mathcal{J}} = [abc][inm]\frac{\partial}{\partial x^i}\frac{\mathcal{J}_{an}\mathcal{J}_{bm}}{2} = 0$$

This gives us:
$$\nabla'\cdot(\epsilon_0\boldsymbol{\epsilon}'\textbf{E}') = \rho'$$
where
$$\rho' = \frac{\rho}{det\boldsymbol{\mathcal{J}}}$$

Similarly for the divergence of $\textbf{H}'$, we get:
$$\nabla'\cdot(\mu_0\boldsymbol{\mu}'\textbf{H}') = 0$$

The above transformations are from the Cartesian coordinate system to an arbitrary coordinate system where $x'^i$ is a coordinate element. 

We compare the Maxwell equations in both the coordinate systems and observe that the Maxwell equations are form-invariant under arbitrary spatial coordinate transforms. The effects of a coordinate transformation can be absorbed by the material properties of the medium.

Optical components can be designed by using geometric formalism to obtain a better understanding of the propagation of light through complex dielectrics.

The design process is as follows:

1. We start off with vacuum or a simple dielectric, where we understand the propagation of light to a large extent.

2. We transform this electromagnetic problem to a new and deformed coordinate system which is identified by the metric tensor $g_{ij}$

3. The equivalence relations allow us to calculate the material properties $\epsilon$ and $\mu$. In some cases, we can also calculate $\rho$ and $J$ when the transformation reproduces the same electromagnetic fields as the ones from the curved electromagnetic space in the original coordinate system.

Transformation optics is basically a generalization of the optical conformal mapping (OCM) technique. In other words, this design method is similar to the OCM technique in some ways:

1. The equivalence relations are valid for the full-wave Maxwell equations. Unlike OCM, transformation optics successfully manipulates the wavelike behavior of light via geometrical concepts. Such manipulations are incompatible with the geometrical concepts used in the eikonal approximation. OCM transformations transform only the Helmholtz equation with form invariance. Transformation optics overcomes the shortcomings that occur in media with inhomogeneous permittivity and permeability by transforming the full Maxwell equations. One of the shortcomings include using a short-wavelength approximation to transform the wave equation for the electromagnetic fields in the Helmholtz form.

2. Transformation optics introduces anisotropy in the material implementation of a general coordinate transformation. OCM makes use of isotropy.

3. It is possible to execute local transformations of the coordinate system in transformation optics. In a finite region, we can manipulate light by using a non-trivial coordinate transformation that approaches unity at the boundaries of the region. We can create finite components that smoothly manipulates light without causing reflections or refractions at the boundaries. 



## 2) Transforming space to confine light

Light confinement is important for both theoretical studies of quantum mechanics of light and advanced information processing systems. Optical cavities are traditionally used to achieve light confinement. Dielectric microcavities consist of a dielectric material  with high-index of refraction $η_{in}$ surrounded by a lower-index of refraction $\eta_{out}$ (air), which traps light inside due to internal reflection at the boundaries. However, when the dimensions of the resonator are reduced to the same magnitude as the wavelength of confined light, diffraction effects should be taken into account. Partial confinement occurs, and some energy is lost to the surrounding environment. The eigenmodes of these optical cavities are characterized by a discrete set of complex eigenfrequencies, where the real part ($\omega'$) is proportional to the inverse of the wavelength of the confined light and the imaginary part ($\omega''$) is a measure of the temporal confinement of the wave inside the cavity.




### Cylindrical Cloak


In the following subsections, we review certain cylindrical systems to design optical cavities within the framework of transformation optics. But before we look at the examples, we derive the dispersion relations of a cavity based on radial coordinate transform. 

We consider an infinite cylindrical cloak that is bound by radii $\rho = R_1$ and $\rho = R_2$. Regions (I) and (III) are vacuum and region (II) contains the medium that performs the coordinate transformation. Without loss of generality, we will consider the time-harmonic solutions that are polarized along the z-axis: $E(\textbf{r}, t) = E(\rho, \phi) exp(− i\omega t)\textbf{1}_z$.

We will look for the eigenmodes of this system by solving Maxwell’s equations inside each region and matching the solutions using the proper boundary conditions.  Inside the vacuum regions, the solutions of Maxwell’s equations are well known: the angular variation of the electric
field is a sum of complex exponentials, characterized by indices $\nu_{I}$ and $\nu_{III}$ whereas the radial dependence obeys the Bessel equation. The radial dependence of the electric field can therefore
be written as a sum of Bessel functions $J_\nu$ and $Y_\nu$ or as a sum of Hankel functions $H_\nu^{(1)}$ and $H_\nu^{(2)}$. In the inner region (I), we decompose the field using the Bessel functions where we can reject the
Bessel function $Y_\nu$, because it has a singularity at the origin. In the surrounding vacuum region
(III), we decompose using the Hankel functions, where we can drop the Hankel function $H_\nu^{(2)}$, which represents an incoming wave: 
$$E_I (\rho,\phi) = A_{\nu_I} J_{\nu_I}(k_0 \rho)$$ 
and 
$$E_{III} (\rho,\phi) = C_{\nu_{III}} J_{\nu_{III}}(k_0 \rho)$$
where $A_{\nu_I}$ and $C_{\nu_{III}}$  are complex integration constants and $k_0 = \omega/c$ represents the vacuum wave number. In order to calculate the solutions in the transformation-optical region (II), we need to insert the constitutive parameters for the material in Maxwell’s equations.

In this section, we consider the case of an arbitrary radial transformation, leaving the azimuthal angle and the z-axis unchanged: (ρ,φ,z) is transformed into (ρ′, φ′, z′) such that ρ′ = f (ρ), φ′ = φ and z′ = z. From the results derived in Section 1, we can write
$$\epsilon^\rho_\rho = \mu^\rho_\rho = \frac{f(\rho)}{\rho f'(\rho)}$$
$$\epsilon^\phi_\phi = \mu^\phi_\phi = \frac{\rho f'(\rho)}{f(\rho)}$$
$$\epsilon^z_z = \mu^z_z = \frac{f(\rho)f'(\rho)}{\rho }$$

where the prime denotes differentiation with respect to ρ. The derivation for the above equations is given in Appendix A. Using the constitutive equations $B^i = μ_0 μ^i_j H^j$ and $D^i = ε_0 ε^i_j E^j$ , we can insert these
parameters in Maxwell’s equations and combine them into the following equation for the electric field in region (II):
$$ \frac{f(\rho)}{f'(\rho)} \frac{\partial}{\partial \rho} \left( \frac{f(\rho)}{f'(\rho)} \frac{\partial E_{II}}{\partial \rho}\right) + \frac{\partial^2 E_{II}}{\partial \phi^2} + k^2_0 f^2(\rho) E_{II} = 0$$
The solutions of the above equation will of course keep their harmonic azimuthal character. As to the radial part of this equation, it can be reduced to the same Bessel equation as in regions (I) and (III), but in the variable $ρ′ = f (ρ)$. As a result, the solutions inside the transformation-optical region (II) are given by
$$E_{II} (\rho,\phi) = [F_{\nu_{II}} J_{\nu_{II}}(k_0 f(\rho)) + G_{\nu_{II}}Y_{\nu_{II}}(k_0 f(\rho))] exp(i \nu_{II} \phi)$$
where, once again, $F_{\nu_{II}}$ and $G_{\nu_{II}}$ are complex integration constants.

This solution needs to be matched to the solutions of the electromagnetic field in regions (I)
and (III) using the Maxwell boundary conditions. The dispersion relation is thus retrieved by imposing the continuity of the tangential components of the electric ($E^z$) and magnetic fields ($H^\phi$) at the boundaries between the regions. Because of the symmetry in the angular direction, the azimuthal mode number $m_i$ remains the same in each region ($m_i = m$) and, as a result, we find the following set of equations that govern the eigenmodes of the cavity:

$$ A_m J_m(k_0 R_1) = F_m J_m (k_0 f(R_1)) + G_m Y_m (k_0 f(R_1)) $$

$$ A_m J'_m(k_0 R_1)  = F_m \frac{f(R_1)}{R_1} J'_m (k_0 f(R_1)) + G_m \frac{f(R_1)}{R_1}  Y'_m (k_0 f(R_1)) $$

$$ F_m J_m (k_0 f(R_2)) + G_m Y_m (k_0 f(R_2))  = C_m H_m^{(1)} (k_0 R_2)$$

$$ F_m \frac{f(R_2)}{R_2} J'_m (k_0 f(R_2)) + G_m \frac{f(R_2)}{R_2} Y'_m (k_0 f(R_2))  = C_m H_m^{'(1)} (k_0 R_2) $$

where $A_m, F_m, G_m$ and $C_m$ are complex integration constants. It is important to note that this relation is valid for any cavity of the type shown in figure 3, implementing a radial coordinate transformation ρ′ =f(ρ) between $R_1$ and $R_2$. In the following sections, we will evaluate this relation to discuss the confined modes of two cavities, whose functionality is based on a radial coordinate transformation.

### Making Electromagnetic space larger

A cavity ideally confines electromagnetic energy in small region of space for a very long time. This can be interpreted in terms of coordinate transformations, by mapping a large domain of electromagnetic space onto a much smaller region in physical space. An example of such a transformation is a hyperbolic map, which grows to infinity at a finite point. Therefore, we can consider a coordinate transformation, in which f : $[R_1, R_2] → [R_1, ∞] : ρ → ρ′$, where
$$ f(R_1) = R_1$$
and 
$$f(R_2) = \infty$$ 

The matching at the inner boundary enables a smooth transition of the waves. As there cannot be anything ‘beyond infinity’, the above equations ensures that electromagnetic energy cannot escape the device. In the figure, where we consider the transformation given by 
$$f(ρ)=R_1(R_1 −R_2)/(ρ−R_2)$$
the Cartesian coordinate lines in physical space become denser as we approach the outer radius $R_2$. 
We find that the hyperbolic map does not confine any electromagnetic modes.
We find that, independent of the azimuthal mode number $m$, the dispersion relation can only
be satisfied if $k_0 = 0$, i.e. the static solution. This result fits with the intuitive idea that in this
configuration an electromagnetic wave travels an infinitely long time to reach the outer boundary
of the cavity. Therefore, no standing wave can be created in the cavity: waves are not reflected at
the outer boundary, $ρ = R_2$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define physical coordinates
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)

# Define radial coordinate transformation
R1 = 0.5
R2 = 1
def f(rho):
    return R1*(R1-R2)/(rho-R2)

# Define transformed coordinates
rho = X**2 + Y**2
phi = np.arctan2(Y, X)
rho_prime = f(rho)
X_prime = rho_prime * np.cos(phi)
Y_prime = rho_prime * np.sin(phi)

# Plot physical coordinates
plt.figure(figsize=(4, 4))
plt.scatter(X, Y, s=1, c='black')

# Plot transformed coordinates
plt.scatter(X_prime, Y_prime, c='blue', alpha=0.5)

# Plot innermost circle
theta = np.linspace(0, 2*np.pi,5)
x_inner = R2 * np.cos(theta)
y_inner = R2 * np.sin(theta)
plt.plot(x_inner, y_inner, c='black')

plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f_1(rho,R1,R2):
    return R1*(R1-R2)/(rho - R2)

def f_1_prime(rho,R1,R2):
    return -R1*(R1-R2)/((rho - R2)**2)

def e_rho(rho,R1,R2):
    return f_1(rho,R1,R2)/(rho*f_1_prime(rho,R1,R2))

def e_phi(rho,R1,R2):
    return 1/e_rho(rho,R1,R2)

def e_z(rho,R1,R2):
    return (f_1(rho,R1,R2)*f_1_prime(rho,R1,R2))/(rho)

R1 = 1
R2 = 2
rho = np.linspace(R1,R2,50)

y1 = e_rho(rho,R1,R2)
y2 = e_phi(rho,R1,R2)
y3 = e_z(rho,R1,R2)

plt.plot(rho/R2,y1,label='epsilon_rho, mu_rho')
plt.plot(rho/R2,y2,label='epsilon_phi, mu_phi')
plt.plot(rho/R2,y3,label='epsilon_z, mu_z')
plt.legend()
plt.ylim((-1,10))
plt.show()
```

### Folding the electromagnetic space

An invisibility cloak turns a finite region in space invisible by guiding light around it. For an
ideal cloak, it is impossible for light of any wavelength to enter the cloaked region. A perfect invisibility cloak is characterized by a continuous radial coordinate transformation $f : [R_1 , R_2 ] → [0, R_2] : ρ → ρ′$ for which the inner boundary in the physical space is mapped onto the origin in
electromagnetic space, whereas the outer boundary is mapped onto itself:
$$ f(R_1) = 0$$
and 
$$f(R_2) = R_2$$ 

We can, however, construct a cavity from this cloaking perspective and design a device that
cloaks away the volume surrounding the device instead of the volume inside the device. Such a device should smoothly guide the electromagnetic waves in the cavity so that they never penetrate the outer boundary. The effect of such a transformation is shown in the figure. As we want to cloak away region (III), we will use a radial coordinate transformation that maps the physical coordinates $(ρ, φ, z)$ onto the electromagnetic coordinates $(ρ′, φ′, z′)$. To achieve perfect cloaking of region (III) from the viewpoint of region (I), the radial transformation function has to satisfy the following boundary requirements:
$$ f(R_1) = R_1$$
and 
$$f(R_2) = 0$$

To calculate the modes of this cavity, we need to evaluate the above four equations, where we now have to insert $f(R_1)=R_1$ and $f(R_2)=0$. In this case, we find

$$ A_m J_m(k_0 R_1) = F_m J_m (k_0R_1) + G_m Y_m (k_0 R_1) $$

$$ A_m J'_m(k_0 R_1)  = F_m  J'_m (k_0 R_1) + G_m   Y'_m (k_0 R_1) $$

$$ F_m \lim_{x \rightarrow 0} J_m (k_0 x) + G_m \lim_{x \rightarrow 0} Y_m (k_0 x)  = C_m H_m^{(1)} (k_0 R_2)$$

$$ F_m \lim_{x \rightarrow 0}  \left[ \frac{x}{R_2} J'_m (k_0 x) \right] +  G_m \lim_{x \rightarrow 0} \left[ \frac{x}{R_2} Y'_m (k_0 x) \right] = C_m H_m^{'(1)} (k_0 R_2) $$

These limits should be handled carefully, as they contain infinities and indefinite expressions. For azimuthal mode numbers $m  \neq 0$, we can evaluate these limits unambiguously ($\lim_{x → 0} Y_m(k_0 x) = - ∞$ and $\lim_{x → 0} [x/R_2 Y'_m(k_0 x)] = + \infty$) because of which the coefficient $G_m$ needs to be zero. The previous set of equations then transforms into
$$ A_m J_m(k_0 R_1) = F_m J_m(k_0 R_1)$$
and 
$$ A_m J'_m(k_0 R_1) = F_m J'_m(k_0 R_1)$$

Surprisingly, there is no quantization of the wavevector $k_0$. In other words, the cavity supports a continuous spectrum of modes, even wavelengths that are much larger than the size of the cavity can be confined. Furthermore, the quality factor of these modes is infinite: there is no tunnelling of electromagnetic energy into the surrounding vacuum.


```python
import math
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm

colormap = cm.get_cmap("rainbow")


def plot_grid(
    x1: float,
    x2: float,
    x3: float,
    x4: float,
    y1: float,
    y2: float,
    y3:float,
    y4:float,
    n_lines: int,
    line_points: int,
    map_func,
):
    """
    Plot a transformation of a regular grid.

    :param xmin: Minimum x value
    :param xmax: Maximum x value
    :param ymin: Minimum y value
    :param ymax: Maximum y value
    :param n_lines: Number of lines per axis
    :param line_points: Number of points per line
    :param map_func: Function to map the grid points to new coordinates
    """
    # List for gathering the lines into.
    lines = []

    # Iterate over horizontal lines.
    for y in np.linspace(y1, y2, n_lines):
        lines.append([map_func(x, y) for x in np.linspace(x1, x4, line_points)])
        
    for y in np.linspace(y3, y4, n_lines):
        lines.append([map_func(x, y) for x in np.linspace(x1, x4, line_points)])

    # Iterate over vertical lines.
    for x in np.linspace(x1, x2, n_lines):
        lines.append([map_func(x, y) for y in np.linspace(y1, y4, line_points)])
        
    for x in np.linspace(x3, x4, n_lines):
        lines.append([map_func(x, y) for y in np.linspace(y1, y4, line_points)])

    # Plot all the lines.
    for i, line in enumerate(lines):
        p = i / (len(lines) - 1)  # Normalize to 0-1.
        # Transpose the list of points for passing to plot.
        xs, ys = zip(*line)
        # Get the line color from the colormap.
        plt.plot(xs, ys, color=colormap(p))


# Define some mapping functions.

def cloak(x: float, y: float):
    R1 = 1
    R2 = 2
    rho = np.sqrt(x**2 + y**2)
    cos_phi = (x/rho)
    sin_phi = (y/rho)
    f_rho = R1*np.sqrt(R2**2 - rho**2)/(R2**2 - R1**2)
    return f_rho*cos_phi, f_rho*sin_phi

def cavity(x: float, y: float):
    R = 0.5
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    f_rho = R
    return f_rho*np.cos(phi), f_rho*np.sin(phi)


# Set up the plot surface...
fig = plt.figure(figsize=(10, 5))

# Plot the 2D view
ax1 = fig.add_subplot(1, 2, 1)
plot_grid(-3, 3, -3, 3, -3, 3, -3, 3, 40, 80, cloak)
plot_grid(-3, 3, -3, 3, -3, 3, -3, 3, 40, 80, cavity)
ax1.set
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f_1(rho,R1,R2):
    return R1*(np.sqrt(R2**2 - rho**2))/(np.sqrt(R2**2 - R1**2))

def f_1_prime(rho,R1,R2):
    return -R1*rho/(np.sqrt(R2**2 - R1**2) * (np.sqrt(R2**2 - rho**2)))

def e_rho(rho,R1,R2):
    return f_1(rho,R1,R2)/(rho*f_1_prime(rho,R1,R2))

def e_phi(rho,R1,R2):
    return 1/e_rho(rho,R1,R2)

def e_z(rho,R1,R2):
    return (f_1(rho,R1,R2)*f_1_prime(rho,R1,R2))/(rho)

R1 = 1
R2 = 2
rho = np.linspace(R1,R2,50)

y1 = e_rho(rho,R1,R2)
y2 = e_phi(rho,R1,R2)
y3 = e_z(rho,R1,R2)

plt.plot(rho/R2,y1,label='epsilon_rho, mu_rho')
plt.plot(rho/R2,y2,label='epsilon_phi, mu_phi')
plt.plot(rho/R2,y3,label='epsilon_z, mu_z')
plt.legend()
plt.ylim((-10,0))
plt.show()
```

# 3) Cherenkov Radiation
Let us consider a simple scaling coordinate transform along the coordinate axis. Thus $$\begin{matrix}
                &x' = f(x), & y'=g(y),  & z'= h(z) \\
  \text{where } & f'(x) = F & g'(y) = G & h'(z)= H & \text{are constants}
\end{matrix}$$
where we take the unprimed coordinates in electromagnetic space, and the primed coordinates in material space.
We thus obtain the following information about the transformation : $$\begin{align}
  \Lambda_{i}^{i'} &=  \begin{bmatrix} F & 0 & 0 \\ 0 & G & 0 \\ 0 & 0 & H \end{bmatrix} & 
  \Lambda_{i'}^{i} &=  \begin{bmatrix} F^{-1} & 0 & 0 \\ 0 & G^{-1} & 0 \\ 0 & 0 & H^{-1} \end{bmatrix} \\
  g_{ij} &=  \begin{bmatrix} F^2 & 0 & 0 \\ 0 & G^2 & 0 \\ 0 & 0 & H^2 \end{bmatrix} & 
  g^{ij} &=  \begin{bmatrix} F^{-2} & 0 & 0 \\ 0 & G^{-2} & 0 \\ 0 & 0 & H^{-2} \end{bmatrix} \\
  \implies \sqrt{g} = \left|\mathcal{J}\right| &= FGH
\end{align}$$
We take a simple non-magnetic dielectric for the electromagnetic space, where $\mu^i_j = \mu_0 \delta^i_j$ and $\epsilon^i_j = \epsilon_0 \epsilon_b \delta^i_j$, where the refractive index is given as $n_b = \epsilon_b^2$. Thus the response properties of the material space must be $$\begin{align}
  \overline{\epsilon}\,^{ij} &= \sqrt{g} \,g^{ik} \epsilon_k^{i} = \epsilon_0 \epsilon_b \begin{bmatrix} \frac{GH}{F} & 0 & 0 \\ 0 & \frac{HF}{G} & 0 \\ 0 & 0 & \frac{FG}{H} \end{bmatrix} & 
  \overline{\mu}\,^{ij} &= \mu_0 \begin{bmatrix} \frac{GH}{F} & 0 & 0 \\ 0 & \frac{HF}{G} & 0 \\ 0 & 0 & \frac{FG}{H} \end{bmatrix} \\
  \overline{\eta}_{ij} &= \frac{1}{\mu_0} \begin{bmatrix} \frac{F}{GH} & 0 & 0 \\ 0 & \frac{G}{HF} & 0 \\ 0 & 0 & \frac{H}{FG} \end{bmatrix}
\end{align}$$
where $\eta_{ij}$ is the reluctance tensor, inverse of the permeability tensor.


## Material Space solution
Let us solve the problem of Cherenkov radiation for the problem of an anisotropic material. We will consider potentials in the material in using the generalized Lorentz gauge : $$\begin{align}
  B^i &= \varepsilon^{ijk}\nabla_j A_k  & \implies \nabla \cdot B &= 0\\
  E_i &= - \nabla_i V - \partial_t A_i & \implies \nabla \times E &= - \partial_t B \\
  \text{Gauge condition : } \frac{\epsilon_0}{c^2} \partial_t V &= - \nabla_i(\epsilon^{ij}A_j)
\end{align}$$
We thus obtain from the inhomogenous equations, $$\begin{align} 
  \rho &= \nabla_i(D^i) = \nabla_i ( \overline{\epsilon}\,^{ij} E_j ) \\
  &= \nabla_i\left( \vphantom{\sum}-\overline{\epsilon}\,^{ij}\nabla_j V - \overline{\epsilon}\,^{ij}\partial_t A_j \right) = - \nabla_i (\overline{\epsilon}\,^{ij} \nabla_j V) - \partial_t \left( \nabla_i \overline{\epsilon}\,^{ij} A_j  \right) \\
  \implies \rho &= - \nabla \cdot( \overline{\epsilon} \cdot\nabla V ) + \frac{\epsilon_0}{c^2} \partial_t^2 V \\[1.5em]
  J^i &=  \epsilon^{ijk} \nabla_j H_k -  \partial_t D \\
  &= \epsilon^{ijk} \nabla_j \left( \overline{\eta}_{kl} B^l \right) -  \partial_t \left( \overline{\epsilon}\,^{ij} E_j \right) \\
  &= \epsilon^{ijk} \nabla_j \left( \vphantom{\sum} \overline{\eta}_{kl} \varepsilon^{lpq}\nabla_p A_q \right) + \overline{\epsilon}\,^{ij} \partial_t \left( \nabla_j V + \partial_t A_j  \right) \\
  &= \epsilon^{ijk} \nabla_j \left( \vphantom{\sum} \overline{\eta}_{kl} \varepsilon^{lpq}\nabla_p A_q \right) + \overline{\epsilon}\,^{ij} \partial^2_t A_j + \overline{\epsilon}\,^{ij} \nabla_j  \left(  \partial_t V \right) \\
  &= \epsilon^{ijk} \nabla_j \left( \vphantom{\sum} \overline{\eta}_{kl} \varepsilon^{lpq}\nabla_p A_q \right) + \overline{\epsilon}\,^{ij} \partial^2_t A_j - \frac{c^2 \overline{\epsilon}\,^{ij}}{\epsilon_0} \nabla_j  \left(  \nabla_k ( \overline{\epsilon}\,^{kl} A_l ) \right) \\
  \implies \mathbf{J} &= \nabla \times \left( \vphantom{\sum} \overline{\eta} \cdot \nabla \times \mathbf{A} \right) -  \frac{c^2\overline{\epsilon}}{\epsilon_0} \cdot \nabla \left( \vphantom{\sum} \nabla \cdot( \overline{\epsilon} \cdot \mathbf{A} )\right)  + \overline{\epsilon} \cdot \partial_t^2 \mathbf{A}
\end{align}$$
In the case of Cherenkov radiation, $\rho = q \delta(\mathbf{r} - \mathbf{v}t), \mathbf{J} = q \mathbf{v} \delta(\mathbf{r} - \mathbf{v}t)$. Let us take the Fourier transform with respect to space i.e. $\int d\mathbf{r} e^{-\iota \mathbf{k} \cdot \mathbf{r}} \left( \circ\right)$. Note that the material properties are constant throughout the material, and the first derivatives of $\overline{\epsilon}, \overline{\eta}$ and thus the first derivatives of $V$ and $\mathbf{A}$  do not feature in the equations. Thus all derivatives $\nabla_i$ can be replaced with $\iota k_i$.  $$\begin{align}
  \implies q e^{-\iota \mathbf{v} \cdot \mathbf{k} t}  &= \frac{\epsilon_0}{c^2} \partial_t^2 V_\mathbf{k} + \mathbf{k} \cdot (\overline{\epsilon} \cdot \mathbf{k} V_\mathbf{k}) \\
  \implies \mathbf{v} e^{-\iota \mathbf{v} \cdot \mathbf{k} t} &= -\mathbf{k} \times \left( \vphantom{\sum} \overline{\eta} \cdot \mathbf{k} \times \mathbf{A}_\mathbf{k} \right) + \frac{c^2\overline{\epsilon}}{\epsilon_0} \cdot \mathbf{k} \left( \vphantom{\sum} \mathbf{k} \cdot( \overline{\epsilon} \cdot \mathbf{A}_\mathbf{k} )\right)  + \overline{\epsilon} \cdot \partial_t^2 \mathbf{A}_\mathbf{k}
\end{align}$$
We can consider $V_\mathbf{k}$ and $\mathbf{A}_\mathbf{k}$ to be an SHM (or 3 couples SHMs), being driven at the frequency of the LHS terms, $\omega = \mathbf{v} \cdot \mathbf{k}$. However, there is another relation between $\omega$ and $\mathbf{k}$, the dispersion relation of the medium. We need to consider a source free material for that. $$\begin{align}
  & \nabla \times \mathbf{H} - \partial_t \mathbf{D} = 0 \equiv \nabla \times ( \overline{\eta} \mathbf{B}) - \partial_t (\overline{\epsilon} \mathbf{E}) = 0\\
  \implies & \nabla \times ( \overline{\eta} (-\partial_t \mathbf{B})) + \partial_t^2 (\overline{\epsilon} \mathbf{E}) = 0 \\
  \implies & \nabla \times \left( \vphantom{\sum} \overline{\eta} \cdot \nabla \times \mathbf{E}\right) + \partial_t^2 ( \overline{\epsilon} \mathbf{E}) = 0
\end{align}$$
We can solve this linear system on $\mathbf{E}$ after taking the Fourier transform with respect to space and time i.e. $\int dt e^{-\iota \omega t} \int d\mathbf{r} e^{-\iota \mathbf{k} \cdot \mathbf{r}} \left( \circ\right)$, where we would obtian solutions only if the matrix is singular. $$\begin{align}
  M \mathbf{E} &= \mathbf{k} \times (\overline{\eta} (\mathbf{k} \times \mathbf{E})) + \omega^2 \overline{\epsilon} \mathbf{E} \\
  \implies M &= [\mathbf{k}]_\times \cdot \overline{\eta} \cdot[\mathbf{k}]_\times + \omega^2 \overline{\epsilon} \\
  &= \begin{bmatrix} 0 & - k_z & k_y \\ k_z & 0 & -k_x \\ -k_y & k_x & 0
  \end{bmatrix} \cdot \frac{1}{\mu_0} \begin{bmatrix} \frac{F}{GH} & 0 & 0 \\ 0 & \frac{G}{HF} & 0 \\ 0 & 0 & \frac{H}{FG} \end{bmatrix} \begin{bmatrix} 0 & - k_z & k_y \\ k_z & 0 & -k_x \\ -k_y & k_x & 0
  \end{bmatrix} + \omega^2 \overline{\epsilon} \\
  \implies \mu_0 M &= \begin{bmatrix} 0 & - k_z & k_y \\ k_z & 0 & -k_x \\ -k_y & k_x & 0
  \end{bmatrix} \begin{bmatrix} 0 & - \frac{F}{GH}k_z & \frac{F}{GH}k_y \\ \frac{G}{HF}k_z & 0 & -\frac{G}{HF}k_x \\ -\frac{H}{FG}k_y & \frac{H}{FG}k_x & 0
  \end{bmatrix} + \omega^2 \mu_0 \overline{\epsilon} \\
  &= \begin{bmatrix} 
  - \frac{H}{FG} k_y^2 - \frac{G}{HF} k_z^2 & \frac{H}{FG} k_x k_y & \frac{G}{HF} k_x k_z \\ 
  \frac{H}{FG} k_y k_x & - \frac{F}{GH} k_z^2 - \frac{H}{FG} k_x^2 & \frac{F}{GH} k_y k_z \\ 
  \frac{G}{HF} k_z k_x & \frac{F}{GH} k_z k_y & - \frac{G}{HF} k_x^2  -  \frac{F}{GH} k_y^2
  \end{bmatrix} \\
  &+ \frac{\omega^2}{c^2} \epsilon_b \begin{bmatrix} \frac{GH}{F} & 0 & 0 \\ 0 & \frac{HF}{G} & 0 \\ 0 & 0 & \frac{FG}{H} \end{bmatrix} \\
  \implies \mu_0 FGH M &= \begin{bmatrix} 
  - H^2 k_y^2 - G^2 k_z^2 & H^2 k_x k_y & G^2 k_x k_z \\ 
  H^2 k_y k_x & - F^2 k_z^2 - H^2 k_x^2 & F^2 k_y k_z \\ 
  G^2 k_z k_x & F^2 k_z k_y & - G^2 k_x^2 - F^2 k_y^2
  \end{bmatrix} \\
  &+ \frac{\omega^2}{c^2} \epsilon_b \begin{bmatrix} (GH)^2 & 0 & 0 \\ 0 & (HF)^2 & 0 \\ 0 & 0 & (FG)^2 \end{bmatrix} \\
\end{align}$$
Let us redefine $\frac{\omega}{c} = k_0$, the vaccum wavevector length, and apply the tranformation $\left( R_1 \to \frac{1}{(GH)^2} R_1, R_2 \to \frac{1}{(HF)^2} R_2, R_3 \to \frac{1}{(FG)^2} R_3 \right)$ to $\mu_0 FGHM$, giving us $M'$. $$\require{cancel} \begin{align}
  M' &= \begin{bmatrix} 
  k_0^2 \epsilon_b - \frac{k_y^2}{G^2} - \frac{k_z^2}{H^2} & \frac{F}{G} \frac{k_x}{F} \frac{k_y}{G} & \frac{F}{H} \frac{k_x}{F} \frac{k_z}{H} \\ 
  \frac{G}{F} \frac{k_y}{G} \frac{k_x}{F} & k_0^2 \epsilon_b - \frac{k_z^2}{H^2} - \frac{k_x^2}{F^2} & \frac{G}{H} \frac{k_y}{G} \frac{k_z}{H} \\ 
  \frac{H}{F} \frac{k_z}{H} \frac{k_x}{F} & \frac{H}{F} \frac{k_z}{H} \frac{k_y}{G} & k_0^2 \epsilon_b - \frac{k_x^2}{F^2} - \frac{k_y^2}{G^2}
  \end{bmatrix} \\
  \text{Let } \mathbf{p} &= \begin{pmatrix} p_x & p_y & p_z\end{pmatrix} = \begin{pmatrix} \frac{k_x}{F} & \frac{k_y}{G} & \frac{k_z}{H} \end{pmatrix} \\
  \implies \det M' &= \left(k_0^2 \epsilon_b^2 - p_y^2 - p_z^2\right) \left(k_0^2 \epsilon_b^2 - p_z^2 - p_x^2\right) \left(k_0^2 \epsilon_b^2 - p_x^2 - p_y^2\right) + 2 p_x^2 p_y^2 p_z^2 \\
  &-\left( \vphantom{\sum}\left( k_0^2 \epsilon_b - p_y^2 - p_z^2\right) p_y^2 p_z^2 + \left( k_0^2 \epsilon_b - p_z^2 - p_x^2\right) p_z^2 p_x^2 + \left( k_0^2 \epsilon_b - p_x^2 - p_y^2\right) p_x^2 p_y^2 \right) \\
  &= \left(k_0^2 \epsilon_b - p^2\right) \left[ \vphantom{\sum^{\sum^\sum}} \cancel{\frac{2 p_x^2 p_y^2 p_z^2}{k_0^2 \epsilon_b - p^2}} - \left( p_y^2 p_z^2 + p_z^2 p_x^2 + p_x^2 p_y^2 + \cancel{\frac{3 p_x^2 p_y^2 p_z^2}{k_0^2 \epsilon_b - p^2} }\right) \right. \\
  &  \phantom{=}\left. \left(k_0^2 \epsilon_b^2 - p_z^2 - p_x^2\right) \left(k_0^2 \epsilon_b^2 - p_x^2 - p_y^2\right) + p_x^2 \left(k_0^2 \epsilon_b^2 - p_x^2 - p_y^2\right) + p_x^2 p_y^2 + \cancel{\frac{ p_x^2 p_y^2 p_z^2}{k_0^2 \epsilon_b - p^2}} \vphantom{\sum^{\sum^\sum}} \right] \\
  &= \left(k_0^2 \epsilon_b - p^2\right) \left[ \vphantom{\sum} \left( k_0^2 \epsilon_b\right)^2 - \left(p^2 + p_x^2\right) k_0^2 \epsilon_b + p_x^4 + p_y^2 p_x^2 + p_z^2 p_x^2 + p_y^2 p_z^2 + \right. \\
  & \phantom{=} \left. p_x^2 k_0^2 \epsilon_b- p_x^4 - p_y^2 p_x^2 + p_x^2 p_y^2 - p_y^2 p_z^2 - p_z^2 p_x^2 - p_x^2 p_y^2 \vphantom{\sum} \right] \\
  &= \left(k_0^2 \epsilon_b - p^2\right)^2 \cdot k_0^2 \epsilon_b
\end{align}$$
We thus observe that the dispersion relation either allows static solutions ($\omega = 0$) or it follows the relation $$\frac{\omega^2}{c^2} \epsilon_b = \frac{k_x^2}{F^2} + \frac{k_y^2}{G^2} + \frac{k_z^2}{H^2}$$
Remembering our determination earlier that $\omega = \mathbf{k} \cdot \mathbf{v}$, let us take the particle to be travelling along the x-direction so that we may engineer our material for specific direction of incoming particles. $$\omega = k_x v = \beta c k_x \implies k^2_x \left(\beta^2 \epsilon_b - \frac{1}{F^2}\right) = \frac{k_y^2}{G^2} + \frac{k_z^2}{H^2}$$
In the x-y plane, $k_x = 0$, and thus angle of the wavevector is $$\tan (\alpha_{My}) = \frac{k_y}{k_x} = \sqrt{G^2\left(\beta^2 \epsilon_b - \frac{1}{F^2}\right)} = \frac{G}{F} \sqrt{F^2 \beta^2 n_b^2 - 1} = \frac{G}{F} \tan\left(\vphantom{\sum}\sec^{-1}\left(F \beta n_b\right)\right)$$
Note that the half-angle made by the cone in x-y plane, $\theta_{My}$, is complementary to $\alpha_{My}$, and is given as $$\theta_{My} = \tan^{-1} \left( \frac{F}{G} \tan \left(\vphantom{\sum} \csc^{-1}(F \beta n_b)\right)\right)$$
Similarly, the wavevector angle and the cone half-angle in the x-z plane is given as $$\begin{align}
  \alpha_{Mz} &= \tan^{-1} \left( \frac{H}{F} \tan \left(\vphantom{\sum} \sec^{-1}(F \beta n_b)\right)\right) \\
  \theta_{Mz} &= \tan^{-1} \left( \frac{F}{H} \tan \left(\vphantom{\sum} \csc^{-1}(F \beta n_b)\right)\right)
\end{align}$$


## Electromagnetic Space Solution
We can solve the EM space solution where the medium is a non-magnetic, isotropic $\epsilon_b$ dielectric. Here we observe the charge with magnitude $q_{EM} = q FGH$ travelling at a velocity $\mathbf{v}_{EM} = \lambda^{i'}_j (v_{M})^j = \begin{pmatrix} Fv_x & 0 & 0 \end{pmatrix}$
In such a case, the dispersion relation is simply $$\frac{\omega^2}{c^2} \epsilon_b = k^2 = k_x^2 + k_y^2 + k_z^2$$
and thus other condition is $\beta v_x k_x = \omega$, providing us with the angle of the wavevector and the uniform half-angle of the cone as $$\alpha_{EM} = \tan^{-1} \left( \sqrt{F^2 \beta^2 \epsilon_b - 1 } \right) = \sec^{-1} (F \beta n_b) \hspace{3em}\theta_{EM} = \csc^{-1} (F \beta n_b)$$
This can also be observed from the following wavefront propogation diagram. 
<p style="background-color:black" align=center>  </p>
When transforming back to the material space, the vectors in space get scaled as $\begin{pmatrix} x & y & z \end{pmatrix} \to \begin{pmatrix} \frac{x}{F} & \frac{y}{G} & \frac{z}{H} \end{pmatrix}$. However one-forms like the wave vector get scaled as $\begin{pmatrix} k_x & k_y & k_z \end{pmatrix} \to \begin{pmatrix} F k_x & G k_y & H k_z \end{pmatrix}$. Thus the tangent of $\alpha_{EM}$, equivalent to $\frac{k_{y,z}}{k_x}$, gets scaled as $$\begin{align}
  \text{x-y plane : } \tan(\alpha_{EM}) = \frac{k_y}{k_x} &\to \frac{Gk_y}{F k_x} = \tan(\alpha_{My}) = \frac{G}{F} \tan(\alpha_{EM}) \\
  \text{x-z plane : } \tan(\alpha_{EM}) = \frac{k_z}{k_x} &\to \frac{Hk_z}{F k_x} = \tan(\alpha_{Mz}) = \frac{H}{F} \tan(\alpha_{EM})
\end{align}$$
In contrast and correspondence, the tangent of $\theta_{EM}$, equivalent to ratio of displacements $\frac{\delta(y,z)}{\delta x}$, gets scaled as $$\begin{align}
  \text{x-y plane : } \tan(\theta_{EM}) = \frac{\delta y}{\delta x} &\to \frac{\frac{1}{G} \delta y}{ \frac{1}{F} \delta x} = \tan(\theta_{My}) = \frac{F}{G} \tan(\theta_{EM}) \\
  \text{x-z plane : } \tan(\theta_{EM}) = \frac{\delta z}{\delta x} &\to \frac{\frac{1}{H} \delta z}{ \frac{1}{F} \delta x} = \tan(\theta_{Mz}) = \frac{F}{H} \tan(\theta_{EM})
\end{align}$$
So for minimal effort, we obtain the following cone angles for the Cherenkov radiation produced in the material space. $$\begin{align} 
  \alpha_{Mz} &= \tan^{-1} \left( \frac{G}{F} \tan \left(\vphantom{\sum} \sec^{-1}(F \beta n_b)\right)\right) \\
  \alpha_{Mz} &= \tan^{-1} \left( \frac{H}{F} \tan \left(\vphantom{\sum} \sec^{-1}(F \beta n_b)\right)\right) \\[1em]
  \theta_{Mz} &= \tan^{-1} \left( \frac{F}{G} \tan \left(\vphantom{\sum} \csc^{-1}(F \beta n_b)\right)\right) \\
  \theta_{Mz} &= \tan^{-1} \left( \frac{F}{H} \tan \left(\vphantom{\sum} \csc^{-1}(F \beta n_b)\right)\right)
\end{align}$$


## Discussion of results
We thus observe that the cone angle decreases for increasing $G$ and $F$. It is trivial for the case of $G$ as it compresses the longitudinal directions, however for $F$, there is also a change in the velocity. Thus we see that there is cut-off value, below which Cherenkov radiation is not emitted. We conclude that we can induce Cherenkov radiation from any particle with a suitable transformation which increases it's speed sufficiently, which is equivalent to selecting a suitable medium with a large enough refractive index that Cherenkov radiation is produced.

# 4) Transforming space to enhance optical gradient forces
In this section, we generalize the scope of transformation optics to include the study of
momentum transfer in complex metamaterial structures. At macroscopic scales, this momentum
is normally too small to have any significant effect, but in nanoscale devices the transfer of
linear momentum between light and matter and the associated optical forces start to play an
increasingly important role. Depending on whether the transferred momentum is parallel or perpendicular to the direction of propagation, the optical forces can be divided into scattering and gradient forces.

## (a) Optical gradient forces in coupled waveguides

### (i) Method

We now consider optical gradient forces between coupled dielectric waveguides. To this aim,
we calculate the forces that act on two slab waveguides, as shown in figure 9. To evaluate the
forces that act on the rightmost waveguide, we need to evaluate the fields in regions I, II and
III. The
waveguide modes are found by solving Maxwell’s equations in combination with appropriate
boundary conditions. However, in this section, we are looking for the propagating (travelling) modes of the
waveguide system and, therefore, we use the ansatz ${\bf E} = E(x) e^{[ i(βy − ωt)]}{\bf \hat{z}}$ .

The mode profile E(x) is a solution of a Helmholtz equation in each region (I, II and III):

\begin{align}
\frac{\partial^2 E^2}{∂ x^2} + (n_i^2 k_o^2 - \beta^2) = 0 \tag({4.1}
\end{align}



where $n_i = n$ inside the dielectric core of the waveguide and $n_i = 1$ in the vacuum surroundings.
The requirement that the field vanishes at infinity implies that $k_o^2 <\beta^2$ (Exponential Solution) and $n_i^2k_o^2 >\beta^2$ (Sinusoidal soluion inside dielectric)

A general solution of the electromagnetic fields in this symmetric waveguide set-up can always
be decomposed as a sum of even and odd solutions. Therefore, we can restrict our analysis to the
even or odd modes. In the case of even modes, we obtain the following solutions for the fields in
each region

image.png


\begin{align}
E_I (x) &= A cosh(k_1 x) \tag{4.2} \\
E_{II} (x) &= B cos(k_2 x) + C sin(k_2 x) \tag{4.3} \\
E_{III} (x) &= D e^{(−k_1 x)} \tag{4.4} \\
E_{IV} (x) &= B cos(k_2 x) − C sin(k_2 x) \tag{4.5} \\
E_{V} (x) &= D e^{(k_1 x)} \tag{4.6} \\
\end{align}

where $k_1^2 = \beta^2-k_o^2$ and $k_2^2 = n_2^2 k_0^2 − \beta^2$. The odd mode solution is similar, apart from the
hyperbolic cosine being replaced by a hyperbolic sine in region I and the signs in front of B, C
and D being inverted in regions IV and V.

The dispersion relation of this coupled-waveguides system can be obtained using the
requirement that the tangential components of both the electric and magnetic fields are
continuous at each boundary. This eigenvalue problem determines the solutions β as a function of the frequency $\omega$

First we calculate the magnetic fields using $∇ \times {\bf E} = - \frac{∂ {\bf B}}{∂t}$

\begin{align}
\implies {\bf \hat{x}} \frac{∂ E_z}{∂y} - {\bf \hat{y}} \frac{∂ E_z}{∂x} = - \frac{∂ {\bf B}}{∂t} \\
\end{align}

Since we are only concerned with the propagating modes the time derivative of ${\bf B}$ is equal to $-i\omega {\bf B}$. Plugging the value of $E_z$ we get

\begin{align}
B_x = \frac{\beta}{\omega} E_z \\
B_y = \frac{i \partial}{\omega ∂x} E_z 
\end{align}

For finding the boundary conditions we just need to equate the tangential components of $\bf{E}$ (i.e. $E_z$) and $\bf{H}$ (i.e $H_y$) on both sides of a interface

The even mode solution for magnetic field H_y can be written as

\begin{align}
H_{y I} (x) &= \left(\frac{ik_1}{\omega \mu_1}\right)A sinh(k_1 x)  \\
H_{y II} (x) &= \left(\frac{ik_2}{\omega \mu_2}\right)(-B sin(k_2 x) + C cos(k_2 x)) \\
H_{y III} (x) &= \left(\frac{ik_1}{\omega \mu_1}\right)(-D e^{(−k_1 x)})  \\
H_{y IV} (x) &= \left(\frac{ik_2}{\omega \mu_2}\right)(-B sin(k_2 x) − C cos(k_2 x)) \\

### (ii) Results 
We evaluate equation (4.13) to calculate the optical gradient force as a function of the parameters
of the waveguides. In particular, we investigate the dependence on the gap distance, the
waveguide thickness and the refractive index of the core. Clearly, the optical force is proportional
to the amount of optical power that is coupled into the waveguide system. Therefore, we
normalize the field amplitude in the following simulations to an input power of $1 mW \mu m^{-1}$ .

```python


from scipy.optimize import minimize, dual_annealing
import numpy as np
import matplotlib.pyplot as plt

def gap(x,n,d,t,k0):
    k1 = x[0]
    k2 = abs(np.sqrt((n**2 - 1)*k0**2 - k1**2))
    num = k1*k2*2*(1+np.tanh(k1*d/2))
    denom = k2**2 - (k1**2)*np.tanh(k1*d/2)
    return (np.tan(k2*t) - (num/denom))**2

def gap2(beta,n,d,t,k0):
    k1 = abs(np.sqrt(beta[0]**2-k0**2))
    k2 = abs(np.sqrt((n**2)*k0**2 - beta[0]**2))
    num = k2*(np.tan(k2*t))-k1
    denom = (k1)*(np.tan(k2*t))+k2
    return  abs((k1/k2)*np.tanh(k1*d/2)-num/denom)

def power(k1,n,d,t,k0):
    
    c0 = 3*1e17
    k2 = abs(np.sqrt((n**2 - 1)*k0**2 - k1**2))
    beta = np.sqrt(k1**2 + k0**2)
    gamma_L = k1
    gamma_S = gamma_L*(n**2)
    w = t
    n_eff = beta/k0
    a = d/2
    
    #Kh = k2
    
    term_1 = a/2 + (np.sinh(2*a*gamma_L)/(4*gamma_L))
    term_2_1 = (np.sin(2*k2*w)/(4*k2))*(1/2 + (np.cosh(2*gamma_L*a))/(2))
    term_2_2 = (w/2)*(1/2 + (np.cosh(2*gamma_L*a))/(2))
    term_2_3 = (gamma_S/(2*k2**2))*np.sinh(2*gamma_L*a)*(np.sin(k2*w))**2
    term_2_4 = -((gamma_S**2)/(4*k2**3))*np.sin(2*k2*w)*(-1/2 + (np.cosh(2*gamma_L*a))/(2))
    term_2_5 = ((w*gamma_S**2)/(2*k2**2))*(-1/2 + (np.cosh(2*gamma_L*a))/(2))
    term_2 = term_2_1 + term_2_2 + term_2_3 + term_2_4 + term_2_5
    
    term_3_1 = np.cos(k2*w)*(1/2 + (np.cosh(2*gamma_L*a))/(2))
    term_3_2 = (gamma_L/k2)*np.sin(k2*w)*(-1/2 + (np.cosh(2*gamma_L*a))/(2))
    
    term_3 = (1/(2*gamma_L))*((term_3_1 + term_3_2)**2)
    
    return (2*n_eff/c0)*(term_1+term_2+term_3)
    
    
    
    
    
    
# n = 2.5
# t = 600 #nm
# d = np.linspace(150,1000,20) #nm
# k0 = 2*np.pi/1550 # nano meter ^ -1

# sols = []
# print(k0)
# for d_i in d:
#     sol = minimize(gap, [0.0001], args = (n,d_i,t,k0),method='Nelder-Mead')#,bounds=[[0,np.inf]])
#     print(sol.x)
#     sols.append(sol.x[0]**2/power(abs(sol.x[0]),n,d_i,t,k0))

# plt.plot(d,sols)

n = 2.5
t = 600 #nm
d = np.linspace(150,1000,20) #nm
k0 = 2*np.pi/1550 # nano meter ^ -1

sols = []
#print(k0)
for d_i in d:
    sol = minimize(gap2, [0.008], args = (n,d_i,t,k0),method='Nelder-Mead',bounds=[[0.006,0.0085]])#,bounds=[[0,np.inf]])
    #print(sol.x)
    #print(sol.fun)
    #sols.append(sol.x[0]**2/power(abs(sol.x[0]),n,d_i,t,k0))
    k1 = abs(np.sqrt(sol.x[0]**2-k0**2))
    sols.append(-k1**2/(power(abs(k1),n,d_i,t,k0)*1e9))
plt.plot(d,sols)
plt.xlabel("gap (nm)")
plt.ylabel("force (pN $\mu m^{-1} mW^{-1}$")

```

### (iii) Discussion

The magnitude of the gradient forces between two slab waveguides depends strongly on the
geometric parameters and refractive indices of the waveguides (figures 10–12). In general, the
magnitude increases as the interaction between the evanescent tails and the waveguides becomes
stronger. The interaction strength is largest when the separation distance and the waveguide
thickness are small. Therefore, realistic set-ups, that aim for measurable displacements, typically
use very thin silicon waveguides (w = ±300 nm) that are positioned very close to each other
(gap = ±100 nm). The displacement of the waveguides can further be enhanced by
reduction of the mass of the waveguides and enlargement of the waveguide length.
Nevertheless, the resulting displacements remain too small to be useful for all-optical
actuation. Therefore, several techniques have been developed to enhance the gradient forces
inside integrated systems. Typically, the approach consists of enhancing the optical fields at
certain positions. Hence, these methods allow for amplification of the optical force per milliwatt
of input power.

## (c) Enhancing optical gradient forces

image.png

We now review our mechanism to enhance optical gradient forces using the method of
transformation optics. The idea originates from the observation that the optical gradient
force between two waveguides decays exponentially with the distance between the waveguides.
Transformation optics allows to engineer the interwaveguide distance perceived by light. We
achieve this result by designing a medium that annihilates the optical space between two objects
by implementing a folded coordinate transformation. We study two waveguides separated
by such an annihilating optical medium. For symmetry purposes and practical feasibility, the
metamaterial cladding can be attached to the interior boundaries of both waveguides.

To annihilate a distance that is $\alpha$ times larger than the physical thickness t of the metamaterial
cladding, we need a transformation that maps the physical coordinate x to the electromagnetic
coordinate f (x) = −$\alpha$x. This can be understood from figure below, from which it is clear that the
two physical points x1 and x3 are mapped on the same point x∗ in electromagnetic space.
Consequently, the electromagnetic fields are identical in points x1 and x3 . The position of
x1 depends on the transformation inside the metamaterial slab: x1 = x2 − $\alpha$(x3 − x2 ) = x2 − $\alpha$t.
Inserting the function f (x) in the equivalence relation of transformation optics results in the
following materials parameters: $\epsilon^x_x$ ,  $\mu^x_x$  = −1/$\alpha$, $\epsilon^y_y$ ,  $\mu^y_y$ = −$\alpha$, and $\epsilon^z_z$ ,  $\mu^z_z$ = −$\alpha$, where we used
the linear transformation f (x) = −$\alpha$x.

## (d) Discussion

Our study shows that transformation optics allows for the design of materials which can
significantly enhance the optical gradient force between two slab waveguides—the prototype
system for the study of optical gradient forces. When the space-annihilating waveguide cladding
is implemented by a double-negative metamaterial, the force amplification is limited by
dissipative loss.

Transformation optics sheds new light on the propagation of light in complex media. Based on the
equivalence between curved geometries and inhomogeneous electromagnetic materials, it allows
for the investigation of the propagation of light using the tools of differential geometry. In this

components, with accuracy beyond the geometrical approximation

### Solution:

a) It follows that the Jacobian matrix is 

$$\mathcal{J} = \begin{pmatrix}
\frac{\partial \rho}{\partial \rho'} & \frac{1}{\rho'} \frac{\partial \rho}{\partial \phi'} & \ \frac{\partial \rho}{\partial z'}\\
\rho \frac{\partial \phi}{\partial \rho'} & \frac{\rho}{\rho'} \frac{\partial \phi}{\partial \phi'} & \ \rho \frac{\partial \phi}{\partial z'}\\
\frac{\partial z}{\partial \rho'} & \frac{1}{\rho'} \frac{\partial z}{\partial \phi'} & \ \frac{\partial z}{\partial z'}
\end{pmatrix}$$
For our transformation, we thus get
$$\implies \mathcal{J} = \begin{pmatrix}
\frac{1}{f'(\rho)} & 0 & 0\\
0 & \frac{\rho}{f(\rho)} & 0\\
0 & 0 & 1
\end{pmatrix}$$
We can see that $f'(\rho)$ can be written as 
$$ f'(\rho) = \frac{R_2 - R_1}{R_2}$$
Thus we get the jacobian as
$$ \mathcal{J} = \begin{pmatrix}
\frac{R_2 }{R_2- R_1} & 0 & 0\\
0 & \frac{\rho R_2}{R_1R_2 + (R_2 - R_1) \rho} & 0\\
0 & 0 & 1
\end{pmatrix}$$
We assume the permitivitty and permeability tensors in the transformed coordinate system are identity matrices.
Thus the permitivitty is given by

$$ \boldsymbol \epsilon  = \frac{\mathcal{J}  \mathcal{J}^T}{\det(\mathcal{J})}$$
$$\implies \boldsymbol \epsilon  = \frac{(R_2- R_1)(R_1R_2 + (R_2 - R_1) \rho)}{ \rho R_2^2} \begin{pmatrix}
\left (\frac{R_2 }{R_2- R_1} \right)^2 & 0 & 0\\
0 & \left( \frac{\rho R_2}{R_1R_2 + (R_2 - R_1) \rho} \right)^2 & 0\\
0 & 0 & 1
\end{pmatrix}$$

$$\implies \boldsymbol \epsilon  = \boldsymbol \mu = \frac{\left( \frac{R_1R_2}{R_2- R_1}\right) + \rho}{ \rho R_2^2} \begin{pmatrix}
\left (\frac{R_2 }{R_2- R_1} \right)^2 & 0 & 0\\
0 & \left( \frac{\rho R_2}{R_1R_2 + (R_2 - R_1) \rho} \right)^2 & 0\\
0 & 0 & 1
\end{pmatrix}$$

b) Since, our jacobian for the transformation will be diagonal, we can simply write the diagonal elements of the jacobian as the differential of the old coordinates with respect to the new coordinates.

$$\implies \mathcal{J} = \begin{pmatrix}
\frac{R_2 }{R_2- R_1} & 0 & 0\\
0 & \frac{1}{\frac{R_1}{r} + \frac{R_2 }{R_2- R_1}} & 0\\
0 & 0 & \frac{1}{\frac{R_1}{r} + \frac{R_2 }{R_2- R_1}}
\end{pmatrix}$$
Thus our permitivitty and permeability tensors in the radial coordinates would be 
$$ \boldsymbol \epsilon  = \boldsymbol \mu = \frac{R_2 - R_1}{R_2} \times \left( \frac{R_1}{r} + \frac{R_2 }{R_2- R_1}\right)^2 \begin{pmatrix}
\left( \frac{R_2 }{R_2- R_1} \right)^2 & 0 & 0\\
0 & \left( \frac{1}{\frac{R_1}{r} + \frac{R_2 }{R_2- R_1}} \right)^2 & 0\\
0 & 0 & \left( \frac{1}{\frac{R_1}{r} + \frac{R_2 }{R_2- R_1}} \right)^2
\end{pmatrix}$$

We now calculate the cartesian permitivitty and permeability tensor. The jacobian for transformation from spherical coordinates to cartesian coordinates is given by
$$ \mathcal{J'} = \begin{pmatrix}
\sin \theta \cos \phi & r \cos \theta \cos \phi  & -r \sin \theta \sin \phi\\
\sin \theta \sin \phi & r \cos \theta \sin \phi & r \sin \theta \cos \phi\\
\cos \theta & -r \sin \theta & 0
\end{pmatrix}$$

Using

$$\implies \boldsymbol \epsilon'  = \boldsymbol \mu' = \frac{\mathcal{J'}  \boldsymbol ϵ \mathcal{J'}^T}{\det(\mathcal{J'})}$$

Substituting the above relations gives us the permitivitty and permeability in the cartesian coordinates

## Appendix A

#### Derivation of permittivity and permeability tensors in the radially transformed coordinates

We assume the permittivity and permeability tensors in the transformed coordinates to be identity matrices. 

$$ ϵ^̃ = \begin{pmatrix}
1 & 0 & 0\\
0 & 1 & 0\\
0 & 0 & 1
\end{pmatrix} \; \text{and} \; \mu^̃ = \begin{pmatrix}
1 & 0 & 0\\
0 & 1 & 0\\
0 & 0 & 1
\end{pmatrix}$$

We derive the permittivity tensor in the cylindrical coordinates, the same method follows for the permeability tensor. We have 
$$ \boldsymbol \epsilon  = \frac{\mathcal{J}  \boldsymbol ϵ^̃ \mathcal{J}^T}{det(\mathcal{J})}$$

where $\mathcal{J}$ is the Jacobian matrix obtained while transforming primed coordinates to unprimed cylindrical coordinates.

 $$\mathcal{J} = \frac{\partial(\rho,\phi,z)}{\partial(\rho',\phi',z')}$$ 
 $$\implies \mathcal{J} = \begin{pmatrix}
\frac{1}{f'(\rho)} & 0 & 0\\
0 & \frac{\rho}{f(\rho)} & 0\\
0 & 0 & 1
\end{pmatrix}$$
Substituting in the expression for $\boldsymbol \epsilon$, we get
$$\boldsymbol \epsilon = \frac{f(\rho) f'(\rho)}{\rho} \begin{pmatrix}
\frac{1}{f'(\rho)} & 0 & 0\\
0 & \frac{\rho}{f(\rho)} & 0\\
0 & 0 & 1
\end{pmatrix} \begin{pmatrix}
\frac{1}{f'(\rho)} & 0 & 0\\
0 & \frac{\rho}{f(\rho)} & 0\\
0 & 0 & 1
\end{pmatrix}$$

$$\implies \boldsymbol \epsilon = \frac{f(\rho) f'(\rho)}{\rho}  \begin{pmatrix}
(\frac{1}{f'(\rho)})^2 & 0 & 0\\
0 & (\frac{\rho}{f(\rho)})^2 & 0\\
0 & 0 & 1
\end{pmatrix}$$


$$\implies \boldsymbol \epsilon = \begin{pmatrix}
\frac{f(\rho)}{\rho f'(\rho)} & 0 & 0\\
0 & \frac{\rho f'(\rho)}{f(\rho)} & 0\\
0 & 0 & \frac{f(\rho) f'(\rho)}{\rho}
\end{pmatrix}$$
This gives the desired result.
