# What is Multipole Expansion?
Multipole expansions are very frequently used in the study of electromagnetic and gravitational fields, where the fields at distant points are given in terms of sources in a small region. The multipole expansion with angles is often combined with an expansion in radius. Such a combination gives an expansion describing a function throughout three-dimensional space.

# Multipole Expansion of the Electromagnetic Fields

The Maxwell equations in a source-free region of empty space with time dependence of $e^{-iwt}$ :

$$\nabla \times E = ikZ_{0}H \tag{1}$$ $$\nabla \times H = -ikE/Z \tag{2}$$ $$\nabla \times E = 0$$ $$\nabla \cdot H = 0$$

where $k = w/c$

From equation (1) and (2)
- If __E__ is eliminated :
  $$(\nabla^{2}+k^{2})H=0$$ 
  $$\nabla \cdot H=0$$
  $\,$
  $$E=\frac{iZ_{0}}{k}\nabla \times H \tag{3}$$
- If __H__ is eliminated :
  $$(\nabla^{2}+k^{2})E=0$$ 
  $$\nabla\cdot E=0$$
  $\,$
  $$H=-\frac{i}{kZ_{0}}\nabla\times E\tag{4}$$

In equation (3) & (4), each cartesian component of __H__ and __E__ satisfies the Helmholtz wave equation [$(\nabla ^{2}+ k^{2})\psi(x,\omega)=0$] and each such component can be written as an expansion of the general form: $$\psi(x)=\sum_{l,m}[A_{lm}^{(1)}h_{l}^{(1)}(kr)+A_{lm}^{(2)}h_{l}^{(2)}(kr)]Y_{lm}(\theta , \phi)\tag{5}$$ but we will follow a different and easier path to find a pure multipole field of order $(l,m)$.
Let's define a magnetic multipole field of order $(l,m)$ with the following conditions:
$$r\cdot H_{lm}^{(M)}=\frac{l(l+1)}{k}g_{l}(kr)Y_{lm}(\theta , \phi)\tag{6}$$ $\;$ $$r\cdot E_{lm}^{(M)}=0$$ 
  where $$g_{l}(kr)=A_{lm}^{(1)}h_{l}^{(1)}(kr)+A_{lm}^{(2)}h_{l}^{(2)}(kr)$$

from eq.(4), $Z_{0}kr\cdot H = \frac{1}{i}r\cdot (\nabla\times E) = \frac{1}{i}(r\times\nabla)\cdot E = L\cdot E$, thus combining this equation with eq.(6), we conclude that the electric field of the magnetic multipole must satisfy: $$L\cdot E_{lm}^{(M)}(r, \theta , \phi)=l(l+1)Z_{0}g_{l}(kr)Y_{lm}(\theta , \phi)\tag{7}$$

To determine purely transverse electric field from eq.(7):
1. __L__ acts only on the angular variables $(\theta , \phi)$ i.e. radial dependence of $E_{lm}^{(M)}$ must be given by $g_{l}(kr)$.
2. __L__ acting on $Y_{lm}$ transforms the $m$ but $l$ is unchanged. Thus components of $E_{lm}^{(M)}$ can be said to be linear combinations of $Y_{lm}$ with different $m$ values and a common $l$ with eq.(7). Also, components should be prepared beforehand for these changes in $m$ done by __L__. This gives [$E_{+/-}\propto L_{+/-}Y_{lm}$], thus electric field should be: $$E_{lm}^{(M)}=Z_{0}g_{l}(kr)LY_{lm}(\theta , \phi)\tag{8}$$ also $$H_{lm}^{(M)}=- \frac{i}{kZ_{0}} \nabla \times E_{lm}^{(M)} $$ 

Eq.(8) together with eq.(4) gives us the electromagnetic fields of a _magnetic multipole_ of order $(l,m)$. The multipole fields are also called _transverse electric (TE)_ rather than magnetic for eq.(8) being transverse to the radius vector.

Similarly, the fields of an _electric_ or _Transverse magnetic (TM) multipole_ of order $(l,m)$ can be specified by the conditions, $$r\cdot E_{lm}^{(E)}=-Z_{0}\frac{l(l+1)}{k}f_{l}(kr)Y_{lm}(\theta , \phi)\tag{9}$$ $\;$ $$r\cdot H_{lm}^{(E)}=0$$ 

thus the electric multipole fields are:$$H_{lm}^{(E)}=f_{l}(kr)LY_{lm}(\theta , \phi)\tag{10}$$ together with _equation(3)_.

The two sets of multipole field (8) and (10) can be shown to form a complete set of vector solutions to the maxwell equations in a source free region. By combinng these two fields we can write the general solution to Maxwell equations:
$$H = \sum_{l,m}[a_{E}(l,m)f_{l}(kr)X_{lm}-\frac{i}{k}a_{M}(l,m)\nabla\times g_{l}(kr)X_{lm}]$$ $\,$ $$E = Z_{0}\sum_{l,m}[\frac{i}{k}a_{E}(l,m)\nabla\times f_{l}(kr)X_{lm}+a_{M}(l,m)g_{l}(kr)X_{lm}]\tag{11}$$
Here, 
  $$X_{lm}(\theta,\phi)=\frac{1}{\sqrt{l(l+1)}}LY_{lm}(\theta,\phi)\tag{Normalized form}$$
  $$a_{M}(l,m)g_{l}(kr)=\frac{k}{\sqrt{l(l+1)}}\int Y_{lm}^{*}r\cdot Hd\Omega$$
  $$Z_{0}a_{E}(l,m)f_{l}(kr)=-\frac{k}{\sqrt{l(l+1)}}\int Y_{lm}^{*}r\cdot Ed\Omega$$
$a_{E}(l,m)$ and $a_{M}(l,m)$ specify the ammounts of electric and magnetic multipole fields and are determined by the sources and boundary conditions.

# Properties of Multipole Fields; Energy and Angular Momentum of Multipole Radiation

In the __near zone__ (kr << 1), $$f_{l}(kr)\propto [n_{l}\rightarrow -\frac{(2l-1)!!}{x^{l+1}}(1-\frac{x^{2}}{2(l-2l)}+...)]$$
thus the limiting behaviour of the magnetic field for an electric multipole is $$H_{lm}^{E}\rightarrow -\frac{k}{l}L\frac{Y_{lm}}{r^{l+1}}$$ $\,$
$$\Rightarrow E_{lm}^{(E)}\rightarrow -\frac{i}{l}Z_{0}\nabla\times L(\frac{Y_{lm}}{r^{l+1}})$$
By applying the operator identity $i\nabla\times L=r\nabla^{2}-\nabla(1+r\frac{\partial}{\partial r})\tag{12}$ in both the equations we notice that $H_{lm}^{(E)}$ is smaller in magnitude than $E_{lm}^{(E)}/Z_{0}$ by a factor $kr$. Thus in near zone, magnetic field of electric multipole << electric field (always).
For magnetic multipole fields, __E__ and __H__ are interchanged, $E^{(E)}\rightarrow -Z_{0}H^{(M)}$, $\,$ $H^{(E)}\rightarrow E^{(M)}/Z_{0}$ 

In the __far or radiation zone__ (kr >> 1), the multipole fields depends on the boundry conditions imposed. Taking an example of outgoing waves, $$f_{l}(kr)\propto [h_{l}^{(1)}(kr)\rightarrow (-i)^{l+1}\frac{e^{ix}}{kr}]\tag{Spherical Hankel Function}$$
From the asymptotic form, in the radiation zone, magnetic induction for electric multipole is $$H_{lm}^{(E)}\rightarrow (-i)^{l+1}\frac{e^{ikr}}{kr}LY_{lm}$$ $\,$ $$\Rightarrow E_{lm}^{(E)}=-Z_{0}(-i)^{l+1}\frac{e^{ikr}}{kr}[n\times LY_{lm}-\frac{1}{k}(r\nabla^{2}-\nabla)Y_{lm}]\tag{but kr >> 1}$$ $\,$ $$\Rightarrow E_{lm}^{(E)}=Z_{0}H_{lm}^{(E)}\times n \tag{13}$$
where $n=(\textbf{r}/r)$ is a unit vector in the radial direction.
For magnetic multipoles the same relation holds because the Poynting vector is directed radially outward for both types of multipole.

The multipole fields of a radiating source can be used to calculate the energy and angular momentum carried off by the radiation. Lets consider a linear superposition of electric multipoles with different $m$, but same $l$. For harmonically varying fields, the time-averaged energy density is $$u=\frac{\epsilon_{0}}{4}(E\cdot E^{*}+Z_{0}^{2}H\cdot H^{*})$$ Consequently the energy in a spherical shell between $r$ and $(r+dr)$ (kr >> 1) is $$\frac{dU}{dr}=\frac{\mu_0}{2k^{2}}\sum_{m}|a_{E}(l,m)|^{2}$$
In genral, $\sum_m \rightarrow \sum_{l,m}$ and $|a_{E}|^2 \rightarrow |a_{E}|^2+|a_{M}|^{2}$, total energy = incoherent sum over all multipoles.

Time averaged angular momentum density: $$m=\frac{1}{2c^{2}}Re[r\times (E\times H^{*})]$$ $\,$ $$\Rightarrow dM=\frac{\mu_{0}}{2wk^{2}}Re\sum_{m,m'}a_{E}^{*}(l,m')a_{E}(l,m)\int Y_{lm}^{*}LY_{lm}d\Omega$$

The $x$ and $y$ cartisan components of $dM/dr$ i.e. $M_{x}$ & $M_{y}$, vanishes for a single $m$ value and $\frac{dM_{z}}{dr}=\frac{m}{w}\frac{dU}{dr}$. The obvious quantum interpretation: 
- The radiation from a multipole of order $(l,m)$ carries off $m\hbar$ units of $z$ component of angular momentum per photon of energy $\hbar w$.

Classical result of the ratio of square of anular momentum to the square of the energy for a pure (l,m) multipole is $$\frac{M^{(c)^2}}{U^{2}}=\frac{|M_{z}|^{2}}{U^{2}}=\frac{m^{2}}{w^{2}}\tag{14}$$  
In case of quantum mechanics, $M_{z}^{2}\rightarrow M_{x}^{2}+M_{y}^{2}+M_{z}^{2}$ and $m^{2}\rightarrow l(l+1)$

For (l,m) multipole field containing N photons with both the limiting cases (quantum and classical): $$\frac{[M^{(q)}(N)]^{2}}{[U(N)]^{2}}=\frac{N^{2}m^{2}+Nl(l+1)-m^{2}}{N^{2}w^{2}}\tag{15}$$

The quantum-mechanical interpretation of the radiated angular momentum per photon for multipole fields contains the selection rules for multipole transitions between quantum states.
- With the states $(J,M)$ and $(J',M')$, possible multipole transitions have $(l,m)$ such that $|J-J'|\leq l\leq J+J'$ and $m=M-M'$
- it is necessary to state whether the parities of the initial and final states are the same or different. The parity of the initial state is equal to the product of the parities of the final state and the multipole field.
- We know that parity of $Y_{lm}$ is $(-1)^{l}$, Thus, for the fields of an _electric multipole_ of order $(l,m)$ is $(-1)^{l}$. For _magnetic multipole_, it is $(-1)^{l+1}$.

If the parities of two states are same, we see that parity conservation restricts the possibilities, so that only magnetic dipole and electric quadruple transitions occur. If the states differ in parity, then electric dipole and magnetic quadrupole radiation can be emitted or absorbed.

# Angular Distribution of Multipole radiation#

For a localized source distribution, the fields in radiation zone are given by :  

$\begin{array}{l}{{\mathbf{H}\rightarrow\frac{e^{i k r-i\omega t}}{k r}\sum_{l,m}\left(-i\right)^{l+1}[a_{E}(l,\,m){\bf X}_{l m}\,+\,a_{M}(l,\,m){\bf n}\ \times\,{\bf X}_{l m}]}}\\ {{\mathbf{E}\rightarrow Z_{0}{\bf H}\times{\bf\hat{n}}}}\end{array}$


Where the coefficients $a_{M}(l,\,m)$ and $a_{E}(l,\,m)$ will be discussed later

The time averaged power radiated per unity solid angle is given by :

${\frac{d P}{d\Omega}}={\frac{Z_{0}}{2k^{2}}}\left|\sum_{l,m}\left(-i\right)^{l+1}\left[a_{E}(l,\,m){\bf X}_{l m}\,\times\,\mathbf{n}\,+\,a_{M}(l,\,m){\bf X}_{l m}\right]\right|^{2}$

Where ${\bf X}_{l m}(\theta,\,\phi)\,=\,\frac{1}{\sqrt{l(l\,+\,1)}}\;{\bf L}\:Y_{l m}(\theta,\,\phi)$

Electric and magnetic multipoles have same angular dependence but have polarizations at right angles to each other. So the multipole order
can be determined by the measurement of the angular distribution of the radiated power, 
but the character of the radiation can be determined only by the polarization measurement

For a pure multipole of order (l,m) the angular distribution becomes :
$\frac{d P(l,\,m)}{d\Omega}=\frac{Z_{0}}{2k^{2}}\left|a(l,\,m)\right|^{2}\left|{\bf X}_{l m}\right|^{2}$

Using the properties of ${\bf X}_{l m}$, the distribution can be further reduced to :

$\frac{d P(l,\,m)}{d\Omega}=\frac{Z_{0}\,|a(l,\,m)|^{2}}{2k^{2}l(l+1)}[\begin{array}{c}{{\frac{1}{2}(l-m)(l+m+1)\mid Y_{l,m+1}|^{2}}}\\ {{+\frac{1}{2}(l+m)(l-m+1)\mid Y_{l,m-1}|^{2}+m^{2}\mid Y_{l m}|^{2}}}]\end{array}$

using the sum rule for the spherical harmonic function we can write :

$\sum_{m=-l}^{l}\:|X_{l m}(\theta,\:\phi)|^{2}=\frac{2l+1}{4\pi}$

By this the radiation distribution will be isotropic from a source that consists of a set of multipoles of order l, independent of m with 
coefficients a(l,m) superimposed incoherently. 

The total power radiated by a pure multipole of order (l,m) is given by the integral of the angular distribution over all angles :

$P(l,\,m)\,=\,\frac{Z_{0}}{2k^{2}}\,|a(l,\,m)|^{2}$

For a general source, the angular distribution is given by the : 

${\frac{d P}{d\Omega}}={\frac{Z_{0}}{2k^{2}}}\left|\sum_{l,m}\left(-i\right)^{l+1}\left[a_{E}(l,\,m){\bf X}_{l m}\,\times\,\mathbf{n}\,+\,a_{M}(l,\,m){\bf X}_{l m}\right]\right|^{2}$

On integerating this over all angles we get :

$P=\frac{Z_{0}}{2k^{2}}\sum_{l,m}\left[|a_{E}(l,\,m)|^{2}\,+\,|a_{M}(l,\,m)|^{2}\right]$

Showing that the total power radiated is just an incoherent sum of contributions from the different multipoles.

```python
#@title
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt

x = smp.symbols('x')
l, m = 1, 0

Plm = (-1)**m/(2**l)/smp.factorial(l) * (1- x**2)**(m/2) * smp.diff( (x**2 - 1)**l, x, l+m)
Ylm = smp.sqrt( (2.0*l +1)/4.0/smp.pi * smp.factorial(l-m)/smp.factorial(l+m)) * Plm

if (m+1 <= l) :
  Plma = (-1)**(m+1)/(2**l)/smp.factorial(l) * (1- x**2)**((m+1)/2) * smp.diff( (x**2 - 1)**l, x, l+m+1)
  Ylma = smp.sqrt( (2.0*l +1)/4.0/smp.pi * smp.factorial(l-m-1)/smp.factorial(l+m+1)) * Plma
else :
  Ylma = 0

if (m-1 >= 0) :
  Plms = (-1)**(m-1)/(2**l)/smp.factorial(l) * (1- x**2)**((m-1)/2) * smp.diff( (x**2 - 1)**l, x, l+m-1)
  Ylms = smp.sqrt( (2.0*l +1)/4.0/smp.pi * smp.factorial(l-m+1)/smp.factorial(l+m-1)) * Plms
else :
  Ylms = 0

F = 1/l/(l+1) * ( smp.Rational(1,2)*(l-m)*(l+m+1)* (Ylma)**2 + smp.Rational(1,2)*(l-m+1)*(l+m)* (Ylms)**2 + (m**2)* (Ylm)**2)
F.simplify()
```

# Sources of Multipole Radiation, Multipole moments#

Assume that there exists localized and well behaved distributions of charge, current and intrinsic magnetization, we also assume that the time 
dependence can be analyzed into its fourier components, and we only take harmonically varying sources, the charge, current and intrinsic magnetization
are respectively assumed to have a distribution :

$\rho(\mathbf{x})e^{-i\omega t},\qquad\mathbf{J}(\mathbf{x})e^{-i\omega t},\qquad\displaystyle\operatorname{}\mathcal{M}(\mathbf{x})e^{-i\omega t}$

The basic maxwell equations for the electric field (E), and H'

$\begin{array}{r}{\mathbf{\nabla}\cdot\mathbf{H}^{\prime}\,=\,0,}\\{\mathbf{\nabla}\cdot\mathbf{E}\,=\,\rho/\epsilon_{o},}\end{array}$

$\begin{array}{l}{{\nabla\times{\bf E}\,-\,i k Z_{0}{\bf H}^{\prime}\,=\,0}}\\ {{\nabla\times{\bf H}^{\prime}\,+\,i k E/Z_{0}\,=\,{\bf J}\,+\,\nabla\,\times\,\mathcal{M}}}\end{array}$

Helmholz wave equations : 

$(\nabla^{2}\ +\ k^{2}){\bf H}^{\prime}\ =\ -\nabla\times\ ({\bf J}\ +\nabla\times\mathcal{M})$

$(\nabla^{2}\:+\:k^{2})\mathbf{E}^{\prime}\:=\:-i Z_{0}k\nabla\:\times\:\left(\mathcal{M}+\:\frac{1}{k^{2}}\:\nabla\:\times\:\mathbf{J}\right)$

$i\omega\rho\,=\,\nabla\,\cdot\,{\bf J}\,$

$\mathbf{E}^{\prime}\,=\,\mathbf{E}\,+\,{\frac{i}{\omega\epsilon_{0}}}\,\mathbf{J}$

Multipole coefficients are determined by r.H' and r.E' rather than the individual fields, so it is sufficient to consider the wave equations
for them.

For any vector field A, $\mathbf{r}\cdot(\nabla\times\mathbf{A})=(\mathbf{r}\times\nabla)\cdot\mathbf{A}=i\mathbf{L}\cdot\mathbf{A}$ 

The inhomogeneous wave equations can be written as :

$\begin{array}{c}{{(\nabla^{2}\,+\,k^{2}){\bf r}\cdot{\bf H}^{\prime}\,=\,-i{\bf L}\cdot({\bf J}\,+\,\nabla\times\,{\cal M})}}\\ {{(\nabla^{2}\,+\,k^{2}){\bf r}\cdot{\bf E^{\prime}}\,=\,Z_{v}k{\bf L}\cdot\left({\it M}\,+\,\frac{1}{k^{2}}\,\nabla\,\times\,{\bf J}\right)}}\end{array}$

With the boundary conditions of outgoing waves at infinity we have :

${\textbf{r}}\cdot{\mathbf{H}}^{\prime}({\mathbf{x}})={\frac{i}{4\pi}}\int{\frac{e^{i k|{\mathbf{x}}-{\mathbf{x}}^{\prime}|}}{|{\mathbf{x}}-{\mathbf{x}}^{\prime}|}}\;{\mathbf{L}}^{\prime}\cdot\left[{\mathbf{J}}({\mathbf{x}}^{\prime})\;+\;{\mathbf{\nabla}}^{\prime}\times{\mathcal{M}}({\mathbf{x}}^{\prime})\right]\,d^{3}x^{\prime}$

${\bf r}\cdot{\bf E}^{\prime}({\bf x})\,=\,-\,\frac{Z_{0}k}{4\pi}\int\frac{e^{i k|{\bf x}-{\bf x}^{\prime}|}}{|{\bf x}\,-\,{\bf x}^{\prime}|}\,{\bf L}^{\prime}\cdot\left[\mathcal{M}({\bf x}^{\prime})\,+\,\frac{1}{k^{2}}\,\nabla^{\prime}\times\,{\bf J}({\bf x}^{\prime})\right]d^{3}x^{\prime}$

The multipole coefficients can be calculated by :

$\begin{array}{c}{{a_{M}(l,\,m)g_{l}(k r)=\displaystyle\frac{k}{\sqrt{l(l+1)}}\int Y_{\,\,\,l m}^{\ast}\mathbf{r}\cdot{\bf H}\,\,d\Omega}}\\ {{\displaystyle\,}}\\ {{\displaystyle{\cal Z}_{0}a_{E}(l,\,m)f_{l}(k r)=\displaystyle-\frac{k}{\sqrt{l(l+1)}}\int\ Y_{\,\,\,l m}^{\ast}\mathbf{r}\cdot{\bf E}\,\,d\Omega}}\end{array}$

Also, 

$\begin{array}{l c l}{{\mathbf{r}\cdot\mathbf{M}_{l m}^{(M)}={\frac{I({\mathit{l}}+1)}{k}}\,g_{l}(k r)Y_{l m}(\theta,\;\phi)}}\\ {{\mathbf{r}\cdot\mathbf{E}_{l m}^{(M)}=0}}&{}\\ {{\mathbf{r}\cdot\mathbf{E}_{l m}^{(M)}=A_{l}^{(1)}h_{l}^{(1)}(k r)\:+\:A_{l}^{(2)}h_{l}^{(2)}(k r)}}\end{array}$

The requirement of outgoing waves at infinity makes $A_{l}^{(2)}=0$, so we choose $f_{l}(k r)\;=\;g_{l}(k r)\;=\;h_{l}^{(1)}(k r)$
as the representation of E and H outside the sources.

For the green function we assume a spherical wave representation and that the point x is outside the spherical surface completely 
enclosing the sphere. Then in the spherical wave projection can be written as :

${\frac{1}{4\pi}}\int\,d\Omega\,\,\,Y_{l\,m}^{\ast}(\theta,\,\phi)\,{\frac{e^{i k|\mathbf{x}-\mathbf{x}^{\prime}|}}{|\mathbf{x}-\mathbf{x}^{\prime}|}}=i k\,\,h_{l}^{(1)}(k r)j_{l}(k r^{\prime})Y_{l\,m}^{\ast}(\theta^{\prime},\,\phi^{\prime})$

Using this in the calculation of the coefficients :

$a_{E}(l,\,m)=\frac{i k^{3}}{\sqrt{l(l+1)}}\int j_{l}(k r)Y_{l m}^{*}{\bf L}\cdot\left({\mathcal{M}}+\frac{1}{k^{2}}\,\nabla\times{\bf J}\right)\,d^{3}x$

$a_{M}(l,\,m)\,=\,\frac{-k^{2}}{\sqrt{l(l\,+\,1)}}\int j_{l}(k r)Y_{l m}^{\ast}{\bf L}\cdot({\bf J}\,+\,{\bf \nabla}\times\,\mathcal{M})\ d^{3}x$

These can be further made into more usefull forms by the identities:

$\begin{array}{c}{{{\bf L}\cdot{\bf A}=i{\bf \nabla}\cdot({\bf r}\times{\bf A})}}\\ {{{\bf L}\cdot({\bf \nabla}\times{\bf A})=i\nabla^{2}({\bf r}\cdot{\bf A})-\frac i r\frac\partial{\partial{\partial r}}\,(r^{2}\nabla\cdot{\bf A})}}\end{array}$

Choose $ {\bf A = {\bf \mathcal{M}}}$ in the 1st eqn and ${\bf A} = {\bf J}$ in the 2nd eqn and rewrite :

$a_{E}(l,\,m)\,=\,-\frac{k^{3}}{\sqrt{l(l\,+\,1)}}\int\,j_{l}(k r)Y_{\,\,l m}^{*}\Biggl[\nabla\cdot(\mathrm{r}\,\times\,\mathcal{M})+\,\frac{1}{k^{2}}\,\nabla^{2}({\bf r}\cdot{\bf J})\,-\,\frac{i c}{k}\,\frac{1}{r}\,\frac{\partial}{\partial r}\,\left(r^{2}\rho\right)\Bigg]\,d^{3}x$

where we have used contunuity equation for the 1st term, green's theorem for the 2nd term and a radial integeration by parts on the 
third term.

$a_{E}(l,\,m)=\frac{k^{2}}{i\sqrt{l(l+1)}}\int\,Y_{l\,m}^{\ast}\Bigg\{\left.\begin{array}{c}{{c\rho\,\frac{\partial}{\partial r}\left[r j_{i}(k r)\right]+i k({\bf r\cdot J})j_{i}(k r)}}\\ {{-\,i k\nabla\cdot({\bf r}\times M)j_{i}(k r)}}\end{array}\right\}d^{3}x$

And the magnetic multipole coefficient :

$a_{M}(l,\,m)=\frac{k^{2}}{i\sqrt{l(l+1)}}\int\,Y_{l m}^{*}\{ \nabla\cdot\left(\mathbf{r}\,\times\,\mathbf{J}\right)j_{l}(k r)\,+\,\nabla\cdot{\mathcal{M}}\,{\frac{\partial}{\partial r}}\left[r j_{l}(k r)\right]-\ k^{2}({\bf r}\cdot\mathcal{M})j_{I}(k r)$ \} $d^{3}x$

In many fields of physics the source dimensions are very small compared to a wavelength. Then for such cases, we can keep only
the lowest powers in kr for terms involving $ \rho , J and {\mathcal{M}}$

$a_{E}(l,\,m)\simeq\frac{c k^{l+2}}{i(2l+1)!!}\left(\frac{l+1}{l}\right)^{1/2}(Q_{l m}+\:Q_{l m}^{\prime})$

Where, multipole moments are

${\cal Q}_{l m}\,=\,\int_{}^{}r^{l}Y_{\,l m}^{\ast}\rho\,\,d^{3}x$

$Q_{l m}^{\prime}={\frac{-i k}{(l+1)c}}\int r^{l}Y_{l m}^{\ast}\nabla\cdot(\mathbf{r}\times{\mathcal{M}})\ d^{3}x\,.$

The moment $Q_{l m}^{\prime}$ is an induced multipole moment due to magnetization. Following long wavelength approximation:

$a_{M}(l,\,m)\simeq\frac{i k^{l+2}}{(2l+1)!!}\left(\frac{l+1}{l}\right)^{1/2}(M_{l m}\,+\,M_{l m}^{\prime})$

where the magnetic moments, 

$M_{l m}\,=\,-\,\frac{1}{l\,+\,1}\,\int\,r^{l}Y_{l\,m}^{\ast}\,\nabla\,\cdot\,({\bf r}\,\times\,{\bf J})\,\,d^{3}x$

$M_{\,\,l m}^{\prime}\,=\,-\int\,r^{l}Y_{\,\,I m}^{*}\,\nabla\cdot\,\mathcal{M}\,d^{3}x$

# Multipole Radiation in atoms and nuclei#

The transistion probability for the emission of the photon of energy $\hbar \omega$ is given by the radiated power divided by $\hbar \omega$. Using the previous calculated results :

$P(l,\,m)\,=\,\frac{Z_{0}}{2k^{2}}\,|a(l,\,m)|^{2}$

And the above calculated coefficients $a_{M}(l,\,m)$ and $a_{E}(l,\,m)$

We can write the transistion probability $\Gamma_{\scriptscriptstyle F}(l,\,m)$ for an electric multipole :

$\Gamma_{\scriptscriptstyle F}(l,\,m)=\frac{\omega Z_{\scriptscriptstyle0}k^{2l}}{2\hbar[(2l+1)!!]^{2}}\left(\frac{l+1}{l}\right)|Q_{l m}+\;Q_{\scriptscriptstyle l m}^{\prime}|^{2}$

As for a magnetic multipole, $Q_{l m}+\,Q_{l m}^{\,\prime}\to(1/c)[{\cal M}_{l m}+{\cal M}_{l m}^{\,\prime}].$

Effective multipole moments can be estimated as to order of magnitude. Effective magnetization can be written as $|\mathcal{M}|\,=\,O(e\hbar/m R^{3})$

Where e = effective charge, m = effective mass of the system, R = effective size of the system and $e\hbar/m R^{3}$ is the effective magnetic moment of the constituents.

Then the estimates of the multipole coefficients are :

$|Q_{l m}|=O(e R^{l});\ \ \ \ \ |Q_{l m}^{\prime}|=O\biggl(\frac{\hbar\omega}{m c^{2}}\,e R^{l}\biggr)$

$\frac{1}{c}\left|M_{l m}+M_{l m}^{\prime}\right|=O\left(\frac{e\hbar}{m c}\,R^{l-1}\right)$

With these order of magnitude estimates we can extract some qualitative features of the atomic and nuclear radiative transistions

In atoms electrons are radiating. The size of the system can be estimated via $R=O(a_{0}/Z_{\mathrm{eff}})$, where $a_{0}$ = bohr radius and $Z_{\mathrm{eff}}$ = unity for valence electron transistions and z for K,L shell x-ray transistions. From the above results the relative size of the magnetic multipole moment and the electric multipole moment is $|M|/c|_{}Q|=O(\hbar/m c R)=O(Z_{\mathrm{eff}}/137)$ For the same transistion energy the transistion probabilities will be in the ratio :

$\frac{\Gamma_{M}(l)}{\Gamma_{E}(l)}=\,{\cal O}\left(\frac{Z_{\mathrm{eff}}^{2}}{(137)^{2}}\right)$

Also the relative size of the transistion probabilities for multipoles differing by one unit (ignoring order of unity) :

$\frac{\Gamma_{E,M}(l+1)}{\Gamma_{E,M}(l)}=~{\cal O}(k^{2}R^{2})$

For atomic transistions, in which the angular momentum selection rule permits several multipoles, the lowest multipole dominates. The ratio of succesive El multipole obeys $\frac{\Gamma_{M}(l)}{\Gamma_{E}(l)}=\,{\cal O}\left(\frac{Z_{\mathrm{eff}}^{2}}{(137)^{2}}\right)$ in atoms.

The dominant transistions are E1, high angular momentum states de-excite by a lot of E1 transistions in atoms.

For high energy nuclear transistions succesive multipoles of the same type are not as suppressed as in atoms. For lower energies the suppression rate with multipole order is dramatic. The nuclear estimates for magnetic to electric transition rates of the same order, but for an increase in order of electric multipole transistion to magnetic transistion, the estimates are :

$\frac{\Gamma_{M}(l)}{\Gamma_{E}(l)}=\,O(0.2\,A^{-2/3})$ 

$\frac{\Gamma_{E}(l+1)}{\Gamma_{M}(l)}=\;O\biggl(\frac{(\hbar\omega[\mathrm{MeV}])^{2}\;A^{4/3}}{4000}\biggr)$

Here we have taken the effective magnetization to be roughly $3\ e\hbar/m_{\mathrm{N}}R^{3}$, with g = 3 to account for the magnetic moment of the nucleons.

#                               **Tutorial Problem**

**Q.** (a) Let $\Phi(R, \theta, \phi)$ be specified values of the electrostatic potential on the surface of a sphere.
Using the general form of  an exterior, spherical multipole expansion, show what $\Phi(r)$ will look like for $r > R$.                                      
(b) The eight octants of a spherical shell are maintained at alternating electrostatic potentials $\pm V$. Using the results of part (a), find the asymptotic (r → ∞) form of the potential produced
by the shell.
                                                                              

**A.** (a) The general form of an exterior, spherical multipole expansion is given by:

$$\Phi(r, \Omega) = \sum_{l=0}^{\infty} \sum_{m=-l}^{l} \frac{A_{lm}} {r^{l+1}} Y_{lm}(\Omega)\ \ \ \ \ \ \ \ \ \ \ r>R$$     

where, $$A_{lm} = \frac{4\pi}{2l+1} \int \rho(r') r'^l Y_{lm}^*(\Omega ') dV'$$

At the surface:
$$\Phi(R, \Omega) = \sum_{l=0}^{\infty} \sum_{m=-l}^{l} \frac{A_{lm}} {R^{l+1}} Y_{lm}(\Omega)$$

Multiplying both sides by $Y_{l'm'}^*(\Omega ')$ and integrating :

$$\int \Phi(R, \Omega ') Y_{l'm'}^*(\Omega ') d \Omega ' =  \sum_{l=0}^{\infty} \sum_{m=-l}^{l} \frac{A_{lm}} {R^{l+1}} \int Y_{lm}(\Omega ') Y_{l'm'}^*(\Omega ') d \Omega '$$ 

Spherical harmonics are orthonormal:

=> $$\int Y_{lm}(\Omega ') Y_{l'm'}^*(\Omega ') d \Omega ' = \delta _{ll'} \delta_{mm'}$$

=>  $$A_{lm} = R^{l+1}\int \Phi(R, \Omega ') Y_{l'm'}^*(\Omega ') d \Omega '$$

Thus:
$$\Phi(r) = \sum_{l=0}^{\infty} \sum_{m=-l}^{l} (\frac {R}{r})^{l+1} Y_{lm}(\Omega) \int \Phi(R, \Omega ') Y_{l'm'}^*(\Omega ') d \Omega '\ \ \ \ \ \ \ \ \ r>R$$

(b) The octants have alternating $\pm V$ potential => potential changes sign every time $\phi$ is an integer multiple of $\frac{\pi}{2}$. => $\ \ m= \pm 2$, whic in turn implies $l \geq 2$.                                                   
Similarly  potential changes sign every time $\theta$ is an integer multiple of $\frac{\pi}{2}$.                             
$Y_{22} \propto \sin^2 \theta$ does not satisfy this requirement. However, $Y_{32} \propto \sin^2 \theta \cos \theta \propto \cos \theta - \cos 3\theta$ does. For the sake of asymptotic form of the potential, we only need to keep the lowest value of $l$ necessary. 
=> potential is a linear combination of $Y_{3 \pm 2}$ and must be equal to $\pm V$ at $r=R$.                                                                     
=> $$\Phi(r) = V (\frac{R}{r})^4 2\sqrt {\frac {2\pi}{105}}(Y_{32} + Y_{3-2}) \\ =  V (\frac{R}{r})^4 \sin^2 \theta \cos \theta \cos 2\phi \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ r \rightarrow \infty$$
