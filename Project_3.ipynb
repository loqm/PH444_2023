## **Introduction to Axions**



Dark matter is a field of enormous interest for both theorists and experimentalists. Despite it's compelling existential evidence, we don't yet know what it is, except for the fact that, it doesn't interact with matter at observable scales. There are several proposals suggesting what could be a constiuent of dark matter and axion is one of them.  


Axions are originally proposed to solve strong-CP problem in QCD. Experimentally it is long known that CP is a good symmetry at the level of strong interactions. CP symmetry is, if you interchange all particles with their respective anti-particles and perform a parity transformation on all co-ordinates, the laws of physics should remain invariant.However, the lagrangian of QCD has a free parameter $\theta$ in it, and for CP symmetry to exist, $\theta$ should be zero. Infact, experimentally the value of $\theta$ is constrained to be less than $10^{-12}$. But, why should a free parameter be constrained, why should it have such a small value, or If it is zero, why should it specifically be zero? These questions lead to the strong CP problem. 

There are several proposed solutions to solve the strong CP problem, but none of them have been verified through experiments. The most popular solution to the problem is through Peccei-Quinn Mechanism. It introduces a new symmetry which is now represented by, $U(1)_{PQ}$ invariance, so that the value of $\theta$ naturally reduces to 0. This symmetry is spontaneously broken below a temperature or energy, giving rise to a boson called Axion. 

*   Axion is an extremely light particle and is highly non-interactive with known matter.
*   In the early stages of the universe, the spontaneous symmetry breaking of $U(1)_{PQ}$ symmetry could have led to the production of Axions in enormous quantities. 

These two peculiar properties of the Axion made it a promising candidate to explain DM. Production of axions is considered in the early universe and it is found that as the universe is cooling down, at a particular scale of temperature, axions are produced copiously. QCD effects produce an effective periodic potential in which the axion field evolves. The oscillations of the axion field about the minima of such potential produces such huge number of axions. This is called the misalignment mechanism and the time evolution of the axion field produced through this mechanism is expected to be given by,
$$a(\mathbf{x},t)=a_0\cos(\omega_at-\mathbf{x\cdot k_D}) \tag{1.1}$$

where the frequency $\omega_a$, is approximately equal to the axion mass $m_a$, (i.e; $\omega_a=m_a$).

Axion couplings(analogous to interations) are in general very weak. But if a huge density of axions is acting in coherence, the coherence can enhance the effects. Axion couplings to fermion cannot lead to coherence, and axion coupling to gravitons is too weak even in presence of coherence. This makes axion photon couplings and it's effects the main source to investigate the presence of axions.








## **Modified Maxwell Equations**

Similar to the paper, throughout this report, we consider $c=\mu_0=\epsilon_0=1$, for the sake of convience.
We now define few terms which will be used to derive the modified maxwell equations through lagrange formalism.

Curvature in space-time is assumed to be Identically zero and lowering and raising of indices is done using using the minkowski space-time metric $\eta^{\mu\nu}$ given by,

$$\eta^{\mu\nu}=\text{diag}(1,-1,-1,-1)$$

${J^e}_{\mu}$ is the covariant four vector electron current given by,

$${J^e}_{\mu}=(\rho,-{J^e}_x,-{J^e}_y,-{J^e}_z)$$

$A^\mu$ represents the contravariant four vector potential given by,

$$A^\mu=(\Phi,A^x,A^y,A^z)$$

$F^{\mu\nu}$ is the contravariant field strength tensor given by,

$$F^{\mu\nu}=\partial^{\mu}A^{\nu}-\partial^{\nu}A^{\mu}$$

where,

$$\partial^{\mu}=(\partial^t,-\partial^x,-\partial^y,-\partial^z)$$

We assume lorentz guage,

$$\vec{E}=-\nabla\Phi-\frac{\partial\vec{A}}{\partial t}$$
$$\vec{B}=\nabla\times\vec{A}$$

Therefore, $F^{\mu\nu}$ in matrix representation is given by,

$$F^{\mu\nu}=\begin{bmatrix}0 && -E_x && -E_y && -E_z\\E_x && 0 && -B_z && B_y\\E_y && B_z && 0 && -B_x\\E_z && -B_y && B_x && 0\\
\end{bmatrix}$$

The covariant version is given by,

$$F_{\mu\nu}=\begin{bmatrix}0 && E_x && E_y && E_z\\-E_x && 0 && -B_z && B_y\\-E_y && B_z && 0 && -B_x\\-E_z && -B_y && B_x && 0\\
\end{bmatrix}$$

The dual field strenth tensor, $\tilde{F}_{\mu\nu}=\frac{1}{2}\epsilon_{\mu\nu\sigma\rho}F^{\sigma\rho}$, in matrix representation is given by,

$$\tilde{F}_{\mu\nu}=\begin{bmatrix}0 && +B_x && +B_y && +B_z\\-B_x&& 0 && E_z && -E_y\\-B_y && -E_z && 0 && E_x\\-B_z && E_y && -E_x && 0\ \
\end{bmatrix}$$

The Lagrangian for axion electrodynamics is given by,

$$L_{aEM}={J^e}_{\mu}A^{\mu}-\frac{1}{4}F^{\mu\nu}F_{\mu\nu}-\frac{1}{4}g_{a\gamma\gamma}aF^{\mu\nu}\tilde{F}_{\mu\nu}+\frac{1}{2}(\partial^{\mu}a)(\partial_{\mu}a)-U(a)$$

where $g_{a\gamma\gamma}$ gives the axion-photon coupling strength and is treated to be a constant, $a$ is the scalar axion field and $U(a)$ is the effective axion potential. In the above lagrangian the first two terms come from the lagrangian of Electromagnetic fields, the last two terms come from the lagrangian of axion field in presence of a potential. Whereas, the middle term is the interaction term between axion and photon. The Euler-Lagrange equations of this Lagrangian give us the modified Maxwell equations, and an equation describing the evolution of the axion field.

The Euler-Lagrange equation w.r.t. four vector potential is given by,

$$\frac{\partial L}{\partial A^{\alpha}}-\partial^{\beta}\left(\frac{\partial L}{\partial(\partial^{\beta}A^{\alpha})}\right)=0$$

Note that,

$$\frac{\partial L}{\partial A^{\alpha}}={J^e}_{\mu}\frac{\partial A^{\mu}}{\partial A^{\alpha}}={J^e}_{\mu}{\delta^{\mu}}_{\alpha}={J^e}_{\alpha}$$

and,

$$\frac{\partial F^{\mu\nu}}{\partial(\partial^{\beta}A^{\alpha})}=\frac{\partial (\partial^{\mu}A^{\nu})}{\partial(\partial^{\beta}A^{\alpha})}-\frac{\partial(\partial^{\nu}A^{\mu})}{\partial(\partial^{\beta}A^{\alpha})}={\delta_{\beta}}^{\mu}{\delta_{\alpha}}^{\nu}-{\delta_{\beta}}^{\nu}{\delta_{\alpha}}^{\mu}$$

Therefore, the Euler Lagrange Equation turns out to be,

$${J^e}_{\alpha}-\frac{1}{4}\partial^{\beta}\left(({\delta_{\beta}}^{\mu}{\delta_{\alpha}}^{\nu}-{\delta_{\beta}}^{\nu}{\delta_{\alpha}}^{\mu})F_{\mu\nu}+F^{\mu\nu}(\delta_{\beta\mu}\delta_{\alpha\nu}-\delta_{\beta\nu}\delta_{\alpha\mu})\right)-\frac{1}{8}g_{a\gamma\gamma}\epsilon_{\mu\nu\kappa\sigma}\partial^{\beta}[a\left(({\delta_{\beta}}^{\mu}{\delta_{\alpha}}^{\nu}-{\delta_{\beta}}^{\nu}{\delta_{\alpha}}^{\mu})F^{\kappa\sigma}+({\delta_{\beta}}^{\kappa}{\delta_{\alpha}}^{\sigma}-{\delta_{\beta}}^{\sigma}{\delta_{\alpha}}^{\kappa})F^{\mu\nu}\right)]=0$$

Using anti-symmetric property of $F^{\beta\alpha}$,

$${J^e}_{\alpha}-\partial^{\beta}F_{\beta\alpha}-g_{a\gamma\gamma}\partial^{\beta}(a\tilde{F}_{\beta\alpha})=0$$

Resorting out of the indices notation and calculating individual equations,

For $\alpha=1$,

$$\rho-\frac{\partial E_x}{\partial x}-\frac{\partial E_y}{\partial y}-\frac{\partial E_z}{\partial z}-g_{a\gamma\gamma}\left(\frac{\partial(aB_x)}{\partial x}+\frac{\partial(aB_y)}{\partial y}+\frac{\partial(aB_z)}{\partial z}\right)=0$$

For $\alpha=2$,

$$-J_x-\frac{\partial E_x}{\partial t}+\frac{\partial B_z}{\partial y}-\frac{\partial B_y}{\partial z}+g_{a\gamma\gamma}\left(-\frac{\partial(aB_x)}{\partial t}-\frac{\partial (aE_z)}{\partial y}+\frac{\partial (aE_y)}{\partial z}\right)=0$$

Similarly,

$$-J_y-\frac{\partial E_y}{\partial t}-\frac{\partial B_z}{\partial x}+\frac{\partial B_x}{\partial z}+g_{a\gamma\gamma}\left(-\frac{\partial(aB_y)}{\partial t}+\frac{\partial (aE_z)}{\partial x}-\frac{\partial (aE_x)}{\partial z}\right)=0$$

$$-J_z-\frac{\partial E_z}{\partial t}+\frac{\partial B_y}{\partial x}-\frac{\partial B_x}{\partial y}+g_{a\gamma\gamma}\left(-\frac{\partial (aB_z)}{\partial t}-\frac{\partial (aE_y)}{\partial x} +\frac{\partial (aE_x)}{\partial y}\right)=0$$

Combining these equations,

\begin{equation}\nabla\cdot E=\rho-g_{a\gamma\gamma}\nabla\cdot(aB) \tag{1.2}\end{equation}

\begin{equation}\nabla\times B=J+\frac{\partial E}{\partial t}+g_{a\gamma\gamma}\left[\nabla\times(aE)+\frac{\partial (aB)}{\partial t}\right] \tag{1.3}\end{equation}

Using the facts that divergence of curl and curl of gradient are vanishing and the Lorentz guage definitions of $E,B$ will lead to the following two Idenitites,

$$\nabla\cdot B=0 \tag{1.4}$$
$$\nabla\times E=-\frac{\partial B}{\partial t} \tag{1.5}$$

Using equations $(1.4),(1.5)$, the equations $(1.2),(1.3)$ can be simplified to,

\begin{equation}\nabla\cdot E=\rho-g_{a\gamma\gamma}B\cdot\nabla a \tag{1.6}\end{equation}

\begin{equation}\nabla\times B=J+\frac{\partial E}{\partial t}-g_{a\gamma\gamma}\left[E\times\nabla a-B\frac{\partial a}{\partial t}\right] \tag{1.7}\end{equation}

The Euler-Lagrange Equation w.r.t axion field is given by,

$$\frac{\partial L}{\partial a}-\partial^{\beta}\left(\frac{\partial L}{\partial(\partial^{\beta}a)}\right)=0$$

Substituting the Lagrangian and simplifying it leads to,

$$\partial^{\beta}\partial_{\beta}a+\frac{\partial U(a)}{\partial a}+\frac{1}{4}g_{a\gamma\gamma}F^{\mu\nu}\tilde{F}_{\mu\nu}=0$$

As, $F^{\mu\nu}\tilde{F}_{\mu\nu}=-4E\cdot B$,

$$\frac{\partial^2 a}{\partial t^2}-\nabla^2a=-\frac{\partial U(a)}{\partial a}+g_{a\gamma\gamma}E\cdot B \tag{1.8}$$

The five equations $(1.4),(1.5),(1.6),(1.7),(1.8)$ together govern the evolution of EM and axion fields. It should be noted that if the axion field is considered to be zero, the equations go back to maxwell equations. However, in the rest of the report, we wouldn't consider the changes in $a(\vec{r},t)$ due to E,B. As it will only lead to corrections of $O(g_{a\gamma\gamma}^2)$ in $E,B$ fields which we're going to neglect.

## **Experimental Methods**

There are three major classes of experiments being done to investigate the existence of axion through photon-axion coupling,


#### **Light Shining through wall**



In this experiment a laser beam is passed through a strong magnetic field, which will stimulate a conversion of small fraction of photons to axions. A material barrier is introduced to block the laser, but axions pass through matter without any effect. On the other side of the barrier, a strong magentic field is placed to stimulate the conversion of a small fraction of these axions back to photons. But the sensitivity needed to perform this experiment and observe axions is quite high, as the probability of photons converting to axions and axions converting back to photons is extremely small. It has been suggested that matched optical resonators in both axion production and photon regeneration regions will result in very high gains.

#### **Helioscopes**

This class of experiments search for axions coming from sun, hence the name heliscopes. It is found that axions could be produced from blackbody photons in the solar core in the presence of strong electromagnetic fields in the plasma. As axions are weakly interacting they can easily escape sun's core and reach earth. This class of experiments aim to convert these axions to photons and observe them. Typically these experiments involve a strong magnetic field equipped with a mechanical system, so that the field can be re-oriented to follow the sun and thus maximizing flux. At the other end of the magnetic field, an X-ray detector is placed, and all external light is blocked from entering the experimental set-up.

####**Haloscopes**

This class of experiments is based on observing the axion dark matter in the milkyway halo, hence the name Haloscope. As we believe dark matter is abundant and is passing through us every second, these experiments don't need any source for axion production. In the haloscope method resonance cavities have been under usage since long.



**Axion Mass, Photon wavelength and Detector size**

In general every haloscope experiment works by creating a strong static magentic field and looking for small oscillating fields at frequency of the axion field, $\omega_a$. We've seen in equation $(1)$ in Introduction, that the axion field frequency is same as it's mass, $\omega_a=m_a$. Compton wavelength of a particle is the wavelenth of a photon whose energy is equal to the rest mass energy of the particle. Compton wavelength is given by,

$$\lambda_a=\frac{h}{cm_a}$$

The above equation is dimensionally correct and no constants are assumed to be one. As axions are approximated to be cold and non-relativistic, we assume that the actual photon wavelenght will be extremely close to the compton wave length. We know that in a cavity, the behaviour of EM fields is extremely dependant on the wavelength and the dimensions of the cavity. Therefore, the size of the detector has direct impact on the range of axion masses being investigated.

**Resonant Cavities**

This experimental method relies on the resonant conversion of DM axions into photons in presence of strong magentic field and observing the generated photons using sensors or looking for small AC fields. As the method of probing is resonance, it'll only detect those axion fields whose frequency is close to the cavity frequency,

$$\omega_a\approx\omega_{cavity}$$

One can play around with the cavity resonance frequency, but it's still in a limited range. Typically cavities have frequencies around $~1000MHz$, and the axion masses being probed will be around,

$$m_a\sim 10^{-5}-10^{-6}eV$$

In these experiments size of the detector should be of the order of the compton wavelength. But if we need to investigate more smaller axion masses, say $~10^{-9}eV$, we need much larger cavities that aren't practically feasible.




**Broadband Approach for low masses**

To overcome the limitations of the resonant cavity approach recently, several experiments have been proposed to search for lower axion masses and can cover a wide range of orders, $10^{-14}-10^{-6}eV$. These are now being called broadband approaches. Note that, in these experiments the compton wavelengths are much larger than the detector.

In most of these experiments the proposal is to measure the oscillating magnetic flux using sensitive magnetometers. These sensetive magnetometers are proposed to be superconductors. There are some proposals on axion interferometry. And there is also a proposal to measure the electric field.

## **Field Expansions**



####**Introduction**


The approach presented in the paper is to Taylor expand the Electric field ($\textbf E$) and magnetic field ($\textbf B$) in power of $g_{a\gamma\gamma}$ (the coupling between the axion and photon). Thus coverting the Modified Maxwell equations into wave equations grouped on the basis of order of $g_{a\gamma\gamma}$.The detector's field are driven by a static current $\textbf J_e(x)$ and charge distribution $\rho_e(\textbf x)$.An example of such a detector could be a power supply that produces a current to drive a magnet or generate an electric field. In the paper, only the vacuum behaviour of ($\textbf E$) and ($\textbf B$) is calculated in response to the  sources.

As mentioned in the previous section, the coupling factor $\textbf g_{a \gamma \gamma}$ is expected to be very small. Also, it is to be noted that $ J_a^{\mu} << J_e^{\mu}$.Hence we can Taylor expand the fields as follows: 

$$ \textbf E(x,t) = \textbf E_0(x,t) +  \textbf E_1(x,t) + O(\ g_{a \gamma \gamma}^2)  \tag{2.1.1}  $$
$$ \textbf B(x,t) = \textbf B_0(x,t) +  \textbf B_1(x,t) + O(\ g_{a \gamma \gamma}^2) \tag{2.1.1}  $$.


Here, $E_0$ and $B_0$ are independent of $g_{a \gamma \gamma}$ while $E_1$ and $B_1$ are proportional to $g_{a \gamma \gamma}$. We now substitute these expression into the modified maxwell's equation , take the time derivative of equation(1.7) and group equations of same power of $g_{a \gamma \gamma}$.Also, we will neglect all terms of $O(g_{a \gamma \gamma}^2)$ or more. The equation is given by:

\begin{equation}\nabla\times B=J+\frac{\partial E}{\partial t}-g_{a\gamma\gamma}\left[E\times\nabla a-B\frac{\partial a}{\partial t}\right] \end{equation}.

As is common, we will assume that the spatial gradients
of the axion field are negligible, ∇a ≈ 0. This is because the
de Broglie wavelength is about 3 orders of magnitude larger
than the oscillation wavelength $(λ_D ≈ 10^3λ_a)$, and thus
spatial gradient terms are suppressed.
Taking a time derivative of the above equation,we get:

$$ \nabla \times(\frac{\partial \textbf B}{\partial t}) = \frac{\partial^2 \textbf E}{\partial t^2} +  \frac{\partial \textbf J}{\partial t} + g_{a \gamma \gamma}(\frac{\partial }{\partial t}(\frac{\partial a}{\partial t}\textbf B)) \tag{2.2}$$.

The current and charge density $\textbf J_e(x)$ and $\rho_e(x)$ are assumed to be constant in time.Hence the above equation simplifies to:

$$ \nabla \times(\frac{\partial \textbf B}{\partial t}) =\frac{\partial^2 \textbf E}{\partial t^2} +  g_{a \gamma \gamma}(\frac{\partial }{\partial t}(\frac{\partial a}{\partial t}\textbf B)) \tag{2.3}$$.

Using equation(1.5):
$$ \nabla \times \textbf E = -\frac{\partial \textbf B}{\partial t} $$

we arrive at:
$$ - \nabla \times(\nabla \times \textbf E) = \frac{\partial^2 \textbf E}{\partial t^2} +  g_{a \gamma \gamma}(\frac{\partial }{\partial t}(\frac{\partial a}{\partial t}\textbf B)) \tag{2.4}$$.

We can now use the vector identity,
$$ -\nabla \times (\nabla \times \textbf E) = \nabla^2 \textbf E - \nabla(\nabla. \textbf E) = \nabla^2 \textbf E - \nabla  \rho_e \tag{2.5} $$

Substituting in the above equation:

$$ \nabla^2 \textbf E = \nabla \rho_e + \frac{\partial^2 \textbf E}{\partial t^2} + g_{a \gamma \gamma}(\frac{\partial }{\partial t}(\frac{\partial a}{\partial t}\textbf B)) \tag{2.6}  $$

The third term of the expression can be  expanded as:
$$ g_{a \gamma \gamma}(\frac{\partial }{\partial t}(\frac{\partial a}{\partial t}\textbf B)) = g_{a \gamma \gamma}(\frac{\partial^2a}{\partial t^2} \textbf B   + \frac{\partial a}{\partial t}\frac{\partial \textbf B}{\partial t} )    \tag{2.7} $$

We neglect the second term of the RHS.This becomes clear later on when we derive the wave equation for $\textbf B_0$. The term $ g_{a \gamma \gamma}\frac{\partial \textbf B_0}{\partial t}\frac{\partial a}{\partial t}$ would have been first order in $g_{a \gamma \gamma}$ but $\textbf B_0$ itself turns out to be static due to the time independent nature of $\textbf J_e(x)$, $\frac{\partial \textbf B_0}{\partial t}$ is hence 0. The other term would be of $O(g_{a \gamma \gamma}^2)$ and is thus neglected. The equation becomes:

$$ \nabla^2 \textbf E = \nabla \rho_e + \frac{\partial^2 \textbf E}{\partial t^2} +  g_{a \gamma \gamma}(\frac{\partial^2a}{\partial t^2}) \textbf B \tag{2.8} $$

We now expand the electric and magnetic field terms in accordance with equation(2.1).

$$ \nabla^2 (\textbf E_0 +\textbf E_1) = \nabla \rho_e + \frac{\partial^2 (\textbf E_0 +\textbf E_1)}{\partial t^2} +  g_{a \gamma \gamma}(\frac{\partial^2a}{\partial t^2}) (\textbf B_0 +\textbf B_1 )$$

This leads to:

$$ \nabla^2 (\textbf E_0) + \nabla^2(\textbf E_1) = \nabla \rho_e + \frac{\partial^2 (\textbf E_0)}{\partial t^2}+\frac{\partial^2 (\textbf E_1)}{\partial t^2} +  g_{a \gamma \gamma}(\frac{\partial^2a}{\partial t^2}) (\textbf B_0 ) + g_{a \gamma \gamma}(\frac{\partial^2a}{\partial t^2}) (\textbf B_1 ) \tag{2.9} $$

Using the fact that  $E_0$ and $B_0$ are independent of $g_{a \gamma \gamma}$ while $E_1$ and $B_1$ are proportional to $g_{a \gamma \gamma}$, we split the above equation on the basis of same power of dependence on $g_{a \gamma \gamma}$. Hence we arrive at:




$$ \nabla^2 \textbf E_o(x,t) = \frac{\partial^2 \textbf E_0(x,t)}{∂t^2} + \nabla \rho_e(x), \tag{2.10.1} $$

$$ \nabla^2 \textbf E_1(x,t) = \frac{\partial^2 \textbf E_1(x,t)}{∂t^2} + g_{a \gamma \gamma}\frac{\partial^2 a}{∂t^2} \textbf B_0(x,t) \tag{2.10.2} $$




 In a similar way, we can use  equation(1.5) to arrive at the wave equations for $\textbf B$ in the following way :

$$ \nabla \times \textbf E = -\frac{\partial \textbf B}{\partial t} $$

Taking the time derivative of the above equation:

$$ \nabla \times \frac{\partial \textbf E}{\partial t} = -\frac{\partial^2 \textbf B}{\partial t^2} \tag{2.11} $$

By utilizing equation (1.7). We can write:

$$  \frac{\partial \textbf E}{\partial t} = \nabla \times \textbf B - \textbf J + g_{a \gamma \gamma}(\textbf E \times \nabla a -\frac{\partial a}{\partial t} \textbf B) $$

As before, we neglect the $\nabla a$ term. Substituting the above equation:

$$ \nabla \times (\nabla \times \textbf B ) - \nabla \times \textbf J + g_{a \gamma \gamma}(\frac{\partial a}{\partial t})(\nabla \times \textbf B) = -\frac{\partial^2 \textbf B}{\partial t^2} \tag{2.12} $$

Using the identity and the fact that $\nabla.\textbf B = 0$ :

$$ \nabla \times (\nabla \times \textbf B) = -\nabla^2 \textbf B +   \nabla(\nabla.\textbf B) = - \nabla^2\textbf B \tag{2.13} $$

Hence, the equation simplifies to:

$$ -\nabla^2 \textbf B - \nabla \times \textbf J + g_{a \gamma \gamma}(\frac{\partial a}{\partial t})(\nabla \times \textbf B) = -\frac{\partial^2 \textbf B}{\partial t^2} \tag{2.14}  $$.

Expanding the electric and magnetic field in accordance with equation (2.1):

$$ -\nabla^2 (\textbf B_0 + \textbf B_1) - \nabla \times \textbf J + g_{a \gamma \gamma}(\frac{\partial a}{\partial t})(\nabla \times (\textbf B_0 + \textbf B_1)) = -\frac{\partial^2 (\textbf B_0 + \textbf B_1)}{\partial t^2}  $$.

$$ -\nabla^2 (\textbf B_0 ) + -\nabla^2(\textbf B_1) - \nabla \times \textbf J + g_{a \gamma \gamma}(\frac{\partial a}{\partial t})(\nabla \times \textbf B_1)  + g_{a \gamma \gamma}(\frac{\partial a}{\partial t})(\nabla \times \textbf B_0 ) = -\frac{\partial^2 \textbf B_0}{\partial t^2} + -\frac{\partial^2 \textbf B_1}{\partial t^2} \tag{2.15}  $$.

Using the fact that  $E_0$ and $B_0$ are independent of $g_{a \gamma \gamma}$ while $E_1$ and $B_1$ are proportional to $g_{a \gamma \gamma}$, we split the above equation on the basis of same power of dependence on $g_{a \gamma \gamma}$. Hence we arrive at:







$$ -\nabla^2 \textbf B_o(x,t) + \nabla \times \textbf J_e(x) = -\frac{\partial^2 \textbf B_0(x,t)}{∂t^2}  \tag{2.16.1}  $$

$$ -\nabla^2 \textbf B_1(x,t) + g_{a \gamma \gamma}\frac{\partial a}{∂t} \textbf B_0(x,t) = -\frac{\partial^2 \textbf B_1(x,t)}{∂t^2} \tag{2.16.2}   $$




Hence we arrive at the following 4 wave equations to solve:

$$ \nabla^2 \textbf E_o(x,t) -  \frac{\partial^2 \textbf E_0(x,t)}{∂t^2} =  \nabla \rho_e(x), \tag{2.10.1}  $$


$$ \nabla^2 \textbf B_o(x,t) -\frac{\partial^2 \textbf B_0(x,t)}{∂t^2} =  - \nabla \times \textbf  J_e(x) \tag{2.16.1}   $$

$$ \nabla^2 \textbf E_1(x,t) - \frac{\partial^2 \textbf E_1(x,t)}{∂t^2} = g_{a \gamma \gamma}\frac{\partial^2 a}{∂t^2} \textbf B_0(x,t) \tag{2.10.2}  $$


$$ \nabla^2 \textbf B_1(x,t) -\frac{\partial^2 \textbf B_1(x,t)}{∂t^2}   = g_{a \gamma \gamma}\frac{\partial a}{∂t} \textbf B_0(x,t)  \tag{2.16.2}  $$

We must note the wave equation for $\textbf E_0$ and  $\textbf B_0$. Since the source term is time independent($\textbf J_e(x)$ and $\rho_e(x)$), these two fields are static.Our assumption while stating eqation (2.7) thus holds good.Going forward we use these expansion to examine the
effect of the axion field on particular choices of geometry.

```python
# In this cell we compute the expressions of E and B and using the scipy for support for hankel and bessel functions

j_1 = jv(1,rho_1)     # defining bessel functions and hankel function required to plot the magnetic field
H_1 = hankel1(1,rho_2)
h = hankel1(1,omega_a*R)
j = jv(1,omega_a*R)

j_0 = jv(0,rho_1)  # defining bessel functions and hankel function required to plot the electric field
H_0 = hankel1(0,rho_2)

by_1 = -np.pi*omega_a*h*j_1   # magnetic field
by_2 = -np.pi*omega_a*j*H_1

ey_1 = 1j*np.pi*omega_a*h*j_0 -2/R   # electric field
ey_2 = 1j*np.pi*omega_a*j*H_0
```

```python
# In this cell we plot the real and imaginary parts of both E and B

plt.plot(rho_by_R_1, np.real(by_1),color= 'b', linestyle = "dashed", label = "$Re[B_{\phi}]$") #B for reals
plt.plot(rho_by_R_2, np.real(by_2), color='b', linestyle = "dashed" )

plt.plot(rho_by_R_1, np.real(ey_1),color= 'r',label = "$Re[E_{z}]$") #E for reals
plt.plot(rho_by_R_2, np.real(ey_2), color='r' )

plt.plot(rho_by_R_1, np.imag(by_1), 'b',label = "$Im[B_{\phi}]$")  # B for imag
plt.plot(rho_by_R_2, np.imag(by_2), 'b')

plt.plot(rho_by_R_1, np.imag(ey_1), 'r', linestyle = "dashed",label = "$Im[E_{z}]$")  # E for imag
plt.plot(rho_by_R_2, np.imag(ey_2), 'r', linestyle = "dashed")
plt.legend()
plt.axvline(x=omega_a, color='black')
```

## **Axion Dark Matter in the Infinite Solenoid**



We assume that we have an infinite solenoid of radius $R$ pointing in the $\textbf{$\hat{z}$}$ direction. We also assume that current cannot flow along the length of the solenoid and is confined to flow in the $\hat{\phi}$ direction in cylindrical coordinates, as in a set of densely packed current carrying loops. Further we assume that $ρ_e=0$, such that $\textbf{$E_0$}=0$. In cylindrical co-ordinates, we have the current density along the walls: $${\vec{J_e}} = B_0δ(\rho-R)\hat{\phi}\tag{3.1}$$
From Maxwell's unmodified equations, we have:
$$\vec{∇}⋅\vec{B_0}=0\tag{3.2}$$
$$\vec{∇}\times \vec{B_0}=\vec{J_e}\tag{3.3}$$
Solving, we get:
$$\frac{∂B_{0,z}}{∂ρ}=-B_0δ(\rho-R)\tag{3.4}$$
Thus, we get:
$$\vec{B_0}=\begin{cases}
    B_0\hat{z}, & \text{$\rho<R$}\\
    0, & \text{$\rho>R$}
\end{cases}\tag{3.5}$$
Now, the wave equations for $E_0$ and $B_0$ reproduce the results derived above. The wave equations for $E_1$ and $B_1$ reduce to:
$$\vec{∇}^2\vec{E_1} - \frac{\partial^2\vec{E_1}}{\partial t^2}=\begin{cases}
    g_{a\gamma\gamma}\frac{\partial^2a}{\partial t^2}B_0\hat{z}, & \text{$rho<R$}\\
    0, & \text{$\rho>R$}
\end{cases}\tag{3.6a}$$


$$\vec{∇}^2\vec{B_1} - \frac{\partial^2\vec{B_1}}{\partial t^2}=-g_{a\gamma\gamma}\frac{∂a}{∂t}B_0δ(\rho-R)\hat{\phi}\tag{3.6b}$$
Evidently, only $E_z$ and $B_{\phi}$ have nonzero solutions as the other terms lack any source on the RHS. Now, we know that the axion field is easily decomposable into frequency modes, so we move into the frequency space. We leverage azimuthal symmetry and drop the transient terms to obtain:
$$E_{1z}(\rho,t)=\psi_E(\rho)e^{i\omega_a t}\tag{3.7a}$$
$$B_{1ϕ}(\rho,t)=\psi_B(\rho)e^{i\omega_a t}\tag{3.7b}$$
(because $a=a_0e^{i\omega_at}$)
## The B field solution

The laplacian of a vector in cylindrical coordinates is:
$$\vec{∇}^2\vec{v} = (\frac{∂^2v_{\rho}}{\partial \rho^2} + \frac{1}{\rho^2}\frac{∂^2v_{\rho}}{\partial \phi^2} + \frac{∂^2v_{\rho}}{\partial z^2} + \frac{1}{\rho}\frac{∂v_{\rho}}{\partial \rho} - \frac{2}{\rho^2}\frac{∂v_{\phi}}{\partial \phi} - \frac{v_{\rho}}{\rho^2}) \hat{\rho}\\ + (\frac{∂^2v_{\phi}}{\partial \rho^2} + \frac{1}{\rho^2}\frac{∂^2v_{\phi}}{\partial \phi^2} + \frac{∂^2v_{\phi}}{\partial z^2} + \frac{1}{\rho}\frac{∂v_{\phi}}{\partial \rho} + \frac{2}{\rho^2}\frac{∂v_{\rho}}{\partial \phi} - \frac{v_{\phi}}{\rho^2})\hat{\phi}\\ + (\frac{∂^2v_z}{\partial \rho^2} + \frac{1}{\rho^2}\frac{∂^2v_z}{\partial \phi^2} + \frac{1}{\rho}\frac{∂v_{z}}{\partial \rho} + \frac{\partial^2v_z}{\partial z^2}) \hat{z}\tag{3.8}$$

We substitute eqn.(3.7b) into eqn.(3.6b), exploiting the fact that $B_z$ and $B_{\rho}$ are zero, along with all the $\frac{\partial}{∂ϕ}$ and $\frac{\partial}{∂z}$ terms due to azimuthal symmetry. After changing variables to $ρ' =\omega_a\rho$, we get the Bessel equation with a boundary condition at $\rho = R$ :
$$ (∂^2_{ρ'}+\frac{1}{\rho'} ∂_{ρ'}+(1-\frac{1}{\rho'^2}))ψ_B=-ig_{a\gamma\gamma}a_0B_0\delta(\rho'-\omega_aR)\tag{3.9}$$
Now, the solutions to this equation are Bessel functions of order 1, with boundary conditions at $\rho'=0$ and $\rho'=\omega_aR$:
$$ψ_B(\rho')=\begin{cases}
    a_BJ_1(\rho')\, & \text{$\rho'<\omega_aR$}\\
    b_BH^+_1(\rho'), & \text{$\rho'>\omega_aR$}
\end{cases}\tag{3.10}$$
Naturally, for $\rho'<\omega_aR$, the Neumann function component would blow up at the origin, so we only retain the Bessel function component. For $\rho'>\omega_aR$, we choose to work with Hankel functions instead because of their neat asymptotic form:
$$H^{(1)}_1(z) \sim \sqrt{\frac{2}{\pi z}}exp(i (z-\frac{3\pi}{4}))\tag{3.11a}$$
$$H^{(2)}_1(z) \sim \sqrt{\frac{2}{\pi z}}exp(-i (z-\frac{3\pi}{4}))\tag{3.11b}$$
When taken along with $e^{i\omega_at}$, these represent inward and outward travelling magnetic fields respectively. We discard the inward flowing field since that implies power flowing into the system from infinity. Note that the author uses $H^+_1$ and $H^-_1$ for $H^{(2)}_1$ and $H^{(1)}_1$ respectively, which is the convention we follow going ahead.
To find the full solution, we firstly exploit the continuity of $B_{||}$ across the boundary:
$$a_BJ_1(\omega_aR) - b_BH^+_1(\omega_aR)=0\tag{3.12a}$$
and a step discontinuity in $\frac{\partial\psi_B}{\partial \rho'}$ because of the presence of the $\delta$ function. We multiply eqn. (3.9) with $d\rho'$ and integrate from $\omega_aR-ϵ$ to $\omega_aR+ϵ$ (where $\epsilon → 0$) to get:
$$(b_B \partial_{\rho'}H^+_1(\rho')-a_B \partial_{\rho'}J_1(\rho'))|_{\rho'=\omega_aR}=-ig_{a\gamma\gamma}a_0B_0\tag{3.12b}$$
We calculate $a_B$ explicitly from the eqn. (3.12a) and substitute it into eqn. (3.12b) to get:
$$\frac{b_B}{J_1}(J_1\frac{\partial H^+_1}{∂ \rho'}-\frac{\partial J_1}{∂ \rho'}H^+_1)=-ig_{a\gamma\gamma}a_0B_0\tag{3.13}$$
We now use the standard identity of Wronskian of Bessel functions:
$$W(J_1,H^+_1)=J_1\frac{\partial H^+_1}{∂ \rho'}-\frac{\partial J_1}{∂ \rho'}H^+_1=\frac{2i}{\pi\rho'}\tag{3.14}$$
Solving for $a_B$ and $b_B$, we get:
$$a_B=-\frac{\pi}{2}g_{a\gamma\gamma}\omega_aRa_0B_0H^+_1(\omega_aR)\tag{3.15a}$$
$$b_B=-\frac{\pi}{2}g_{a\gamma\gamma}\omega_aRa_0B_0J_1(\omega_aR)\tag{3.15b}$$
We have thus calculated the $\vec{B}$ field.

## The E field solution

Proceeding similarly as before, we get another Bessel equation:
$$(∂^2_{ρ'}+\frac{1}{\rho'} ∂_{ρ'}+1)ψ_E= \begin{cases}
  -g_{a\gamma\gamma}a_0B_0, & {\rho'<\omega_aR}\\
    0, & {\rho'>\omega_aR}
\end{cases}\tag{3.16}$$
which has solutions:
$$ψ_E(\rho')=\begin{cases}
    a_EJ_0(\rho')-g_{a\gamma\gamma}a_0B_0\, & \text{$\rho'<\omega_aR$}\\
    b_EH^+_0(\rho'), & \text{$\rho'>\omega_aR$}
\end{cases}\tag{3.17}$$
The boundary conditions require the continuity od $E_z$ and its derivative across the boundary. The former can be derived by integrating $\vec{\nabla} × \vec{E}$ around a small contour just inside and outside the solenoid.
$$a_EJ_0(\omega_aR)-g_{a\gamma\gamma}a_0B_0=b_EH^+_0(\omega_aR)\tag{3.18a}$$
$$a_EJ_1(\omega_aR)=b_EH^+_1(\omega_aR)\tag{3.18b}$$
We use the Bessel function identities $2∂_{\rho'}Ω_{\nu}=\Omega_{\nu-1}-\Omega_{\nu+1}$ and $\Omega_{-\nu}=(-1)^{\nu}\Omega_{\nu}$ for $\Omega_{\nu}⊂ [J_{\nu},H^+_{\nu}]$ to get:
$$a_E=\frac{i\pi}{2}g_{a\gamma\gamma}\omega_aRa_0B_0H^+_1(\omega_aR)\tag{3.19a}$$
$$b_E=\frac{i\pi}{2}g_{a\gamma\gamma}\omega_aRa_0B_0J_1(\omega_aR)\tag{3.19b}$$
We have thus calculated the $\vec{E}$ field. Putting these together with the $\vec{B}$ field solutions:
$$\begin{pmatrix}
a_E\\b_E\\a_B\\b_B
\end{pmatrix} = \frac{\pi g_{a\gamma\gamma}\omega_aRa_0B_0}{2}\begin{pmatrix}
iH^+_1(\omega_aR)\\iJ_1(\omega_aR)\\-H^+_1(\omega_aR)\\J_1(\omega_aR)
\end{pmatrix}\tag{3.20}$$

# The long-wavelength limit

The varaiable $\rho'$ is nothing but $\omega_a\rho=2\pi\rho/\lambda_a$. Naturally, this wavelength is the relvant lengthscale of the problem. As we saw in the graphs, the behaviour for $R<<\lambda_a$ is very different from $R∼λ_a$ or $R>>\lambda_a$. In this long wavelength limit, we can take the asymptotic forms of the Bessel functions ($\rho'\rightarrow 0$) to see the behaviour of the field near the solenoid. 
$$ψ_B(\rho')=\begin{cases}
    \frac{a_B}{2}\rho'\, & \text{$\rho'<\omega_aR$}\\
    -i\frac{2b_B}{\pi\rho'}, & \text{$\rho'>\omega_aR$}
\end{cases}\tag{3.21}$$
with the coefficients given by:
$$a_B=ig_{a\gamma\gamma}a_0B_0\tag{3.22a}$$
$$b_B=-\frac{\pi}{4}g_{a\gamma\gamma}a_0B_0\omega^2_aR^2\tag{3.22b}$$
Inserting these into eqn.(3.21) and converting back to $\rho$, we get:
$$ψ_B(\rho)=\begin{cases}
    \frac{i}{2}\omega_ag_{a\gamma\gamma}a_0B_0\rho, & \text{$\rho<R$}\\
    \frac{i}{2}\omega_ag_{a\gamma\gamma}a_0B_0\frac{R^2}{\rho}, & \text{$\rho>R$}
\end{cases}\tag{3.23}$$
The factor of i indicating a $\pi/2$ phase shift is expected as the $\vec{B}$ field is driven by $\frac{\partial a}{\partial t}$. We plug this back into eqn.(3.7b) to get the full solution for the axion-induced $\vec{B}$ to first order in $g_{a\gamma\gamma}$ and in the limit of $\rho,R<<\lambda_a$:
$$\vec{B_1}(\vec{x},t)=\begin{cases}
    \frac{1}{2}g_{a\gamma\gamma}\frac{∂a}{\partial t}\rho\hat{\phi}, & \text{$\rho<R$}\\
    \frac{1}{2}g_{a\gamma\gamma}\frac{∂a}{\partial t}\frac{R^2}{\rho}\hat{\phi}, & \text{$\rho>R$}
\end{cases}\tag{3.24}$$
Note that we can sum the axion frequency modes $\omega_a$ to convert $i\omega_aa_0e^{i\omega_at}$ back into $\frac{∂a}{\partial t}$ to  ake the solution true for arbitrary $a(t)$.
Similarly, the electric field behaviour in the long wavelength limit will be:
$$ψ_E(\rho')=\begin{cases}
    a_E(1-\frac{\rho'^2}{4}) - g_{a\gamma\gamma}a_0B_0, & \text{$\rho'<\omega_aR$}\\
    b_E\frac{2i}{\pi}\gamma'(\rho'), & \text{$\rho'>\omega_aR$}
\end{cases}\tag{3.25}$$
where we have defined the function $\gamma'(x)=ln(x/2)+\gamma-i\pi/2$, where $\gamma ≈ 0.5772....$ is the Euler-Mascheroni constant. The coefficients are given by:
$$a_E=g_{a\gamma\gamma}a_0B_0-\frac{g_{a\gamma\gamma}a_0B_0\omega^2_aR^2}{2}(\gamma'(\omega_aR)-\frac{1}{2})\tag{3.26a}$$
$$b_E=\frac{i\pi}{4}g_{a\gamma\gamma}a_0B_0\omega^2_aR^2\tag{3.26b}$$
Expanding this, and dropping terms of order $(\omega_aR)^2(\omega_a\rho)^2$, we get:
$$ψ_E(\rho)=-\frac{1}{2}g_{a\gamma\gamma}a_0B_0(\omega_aR)^2\times\begin{cases}
    (\gamma'(\omega_aR)-\frac{1}{2})+\frac{\rho^2}{2R^2}, & \text{$\rho<R$}\\
   \gamma'(\omega_a\rho), & \text{$\rho>R$}
\end{cases}\tag{3.27}$$
Thus, to the first order in $g_{a\gamma\gamma}$ and for $\rho,R<<\lambda_a$, electric fields are suppressed by $(\frac{R}{\lambda_a})^2ln(\frac{R}{\lambda_a})<<\frac{R}{\lambda_a}$, which can be observed in the graph.
Now, this is in direct contrast to the results obtained in Ref. [25], which searched for an axion-induced electric
field in the long-oscillation-wavelength limit inside the
solenoid. They claimed that the electric field would be measurable as they calculated it to be $\vec{E}=-g_{a\gamma\gamma}a\vec{B_0}$. Though this term does appear as the 'particular soultion' term in eqn. (3.18a), it gets cancelled in the large wavelength limit by the 'homogenous solution' term. Thus, there has been some mistake in Ref. [25].

## **Demonstrating the MQS approximation for a generic detector**




Magneto quasi static (MQS) approximation basically states that the dispacement current (dependent of the time derivative of electric field) is negligible. 
In this section, this fact is generalised for a broad class of detectors that
MQS approximation is always valid in the large - **λa** limit and
that the suppression of the electric field is a generic quality.
In the following argument, we
will make two assumptions:

1. Our detector is composed of a collection of time-independent charges and currents, $\mathbf{\rho_e}$ and $\mathbf{J_e}$.
2. Our detector fits into some box with a diagonal size **L**.

The exact form of the box in the second assumption doesn't matter much. It simply sets a standard size for our detector. Initially, we won't make any suppositions about how the currents and charges are arranged within the box.

Initially, we modify the right side of the wave equations in section 2 by replacing $\mathbf{E_0}$ and $\mathbf{B_0}$ with terms involving $\mathbf{\rho_e}$ and $\mathbf{J_e}$. This is necessary because $\mathbf{E_0}$ and $\mathbf{B_0}$ can exist beyond the box, but the second assumption limits charges and currents to be contained within the box, implying that they are zero on the surface.

At this point, it is clear that the primary fields $\mathbf{E_0}$ and $\mathbf{B_0}$(which are solutions to the axion-free equations) will be
independent of time. 


\begin{align}
\nabla^{2}\mathbf{E_0(x)} &= \nabla \rho _e (\mathbf{x}) \tag{4.1} \\
 \nabla^{2}\mathbf{B_0(x)} &= - \nabla \times \mathbf{J_e} (\mathbf{x}) \tag{4.2}\\
\end{align}

We can then Fourier decompose $\mathbf{E_1(x)}$
and $\mathbf{B_1(x)}$ in the frequency domain,

$$ \mathbf{E_1(x,t)} = \mathbf{E_1(x)} e^{iw_at} $$
$$ \mathbf{B_1(x,t)} = \mathbf{B_1(x)} e^{iw_at} $$

and write the following time independent wave equations:
\begin{align}
 \nabla^{2}\mathbf{E_1(x)} + w_a^{2}\mathbf{E_1(x)} &= -g_{a\gamma \gamma}w_a ^{2}a_0 \mathbf{B_0(x)} \tag{4.3}\\
\nabla^{2}\mathbf{B_1(x)} + w_a^{2}\mathbf{B_1(x)} &= ig_{a\gamma \gamma}w_a a_0 \nabla\times\mathbf{B_0(x)} \tag{4.4}
\end{align}

Rewriting the rhs of equation $(4.4)$ in terms of $\mathbf{J_e(x)}$
using equation $(4.2)$ and 

$$\nabla \times (\nabla \times \mathbf{B_0}) = \nabla(\nabla \cdot \mathbf{B_0}) - \nabla^{2} \mathbf{B_0} = - \nabla^{2}\mathbf{B_0} $$
putting this in equation $(4.2)$ to get -
$$ \nabla \times (\nabla \times \mathbf{B_0(x)}) =  \nabla \times \mathbf{J_e} (\mathbf{x})$$

this implies
$$ \nabla \times \mathbf{B_0(x)} =  \mathbf{J_e(x)}$$

this gives 
\begin{align}
 \nabla^{2}\mathbf{B_1(x)} + w_a^{2}\mathbf{B_1(x)} = ig_{a\gamma \gamma}w_a a_0 \mathbf{Je(x)} \tag{4.5} 
\end{align}

Focusing on equation $(4.3)$, we can split  $ \mathbf{E_1(x)} $
into $ \mathbf{E_1(x)} = \mathbf{E_1'(x)} - g_{a \gamma \gamma} a_0 \mathbf{B_0 (x)} $, and get an equation
for $\mathbf{E_1'(x)}$


$$ \nabla^{2}\mathbf{E_1'(x)} + w_a^{2}\mathbf{E_1'(x)} - g_{a \gamma \gamma}a_0 \nabla^{2}\mathbf{B_0(x)} - g_{a \gamma \gamma}a_0 w_a^{2} \mathbf{B_0(x)}   = -g_{a\gamma \gamma}w_a ^{2}a_0 \mathbf{B_0(x)}$$


\begin{align}
\nabla^{2}\mathbf{E_1'(x)} + w_a^{2}\mathbf{E_1'(x)} &= g_{a \gamma \gamma}a_0 \nabla^{2}\mathbf{B_0(x)} \\
&= g_{a\gamma \gamma} a_0 \nabla \times \mathbf{J_e(x)}\tag{4.6}
\end{align}

Now, solving the equations for $\mathbf{E_0(x)}$ and $\mathbf{B_0(x)}$ using the Laplace's equation form gives us

$$\mathbf{E_0(x)} = -\frac{1}{4 \pi} \int \frac{\nabla \rho_e (\mathbf{x})}{| \mathbf{x} - \mathbf{x'}|} d^3 \mathbf{x'} $$

$$\mathbf{B_0(x)} = \frac{1}{4 \pi} \int \frac{\nabla \times \mathbf{J}_e (\mathbf{x'})}{| \mathbf{x} - \mathbf{x'}|} d^3 \mathbf{x'} $$


Solving the equations $(4.5)$ and $(4.6)$ for $ \mathbf{E_1'(x)}$ and $\mathbf{B_1(x)}$ in the time independent wave equation form (Helmholtz equation) to get
\begin{align}
\mathbf{B_1(x)} &= \frac{i g_{a \gamma \gamma } w_a a_0}{4 \pi} \int \frac{e^{iw_a| \mathbf{x} - \mathbf{x'}|}}{| \mathbf{x} - \mathbf{x'}|} \mathbf{J}_e (\mathbf{x'}) d^3 \mathbf{x'} \tag{4.7}\\
\mathbf{E_1'(x)} &= \frac{g_{a \gamma \gamma }a_0}{4 \pi} \int \frac{e^{iw_a| \mathbf{x} - \mathbf{x'}|} }{| \mathbf{x} - \mathbf{x'}|} ∇\times \mathbf{J}_e (\mathbf{x'}) d^3 \mathbf{x'} 
\end{align}

Including the offset term $- g_{a \gamma \gamma} a_0 \mathbf{B_0 (x)} $ to get equation for $\mathbf{E_1(x)}$

$$\mathbf{E_1(x)} +  g_{a \gamma \gamma} a_0 \mathbf{B_0 (x)} = \frac{g_{a \gamma \gamma }a_0}{4 \pi} \int \frac{e^{iw_a| \mathbf{x} - \mathbf{x'}|} }{| \mathbf{x} - \mathbf{x'}|} ∇\times \mathbf{J}_e (\mathbf{x'}) d^3 \mathbf{x'} $$


and substituting for $\mathbf{B_0(x)}$ gives 
\begin{align}
\mathbf{E_1(x)} = \frac{g_{a \gamma \gamma }a_0}{4 \pi} \int \frac{e^{iw_a| \mathbf{x} - \mathbf{x'}|} -1}{| \mathbf{x} - \mathbf{x'}|} ∇\times \mathbf{J}_e (\mathbf{x'}) d^3 \mathbf{x'} \tag{4.8}
\end{align}

Up to now, we have only used the first assumption that
our charges and currents are constant in time. We use the
second assumption to examine what happens in the limit of
$L << \lambda_a$. Notice that our solutions are completely in terms
of charges and currents, which are completely contained
within our box of size L, as opposed to fields, which can
extend outside of the box.
If both $\mathbf{x}$ and $\mathbf{x_0}$ are within our box then $|\mathbf{x} − \mathbf{x_0}| \le L$.
And now we examine the behavior of the axion-induced
electric fields by Taylor expanding equation $(4.8)$ in the limit of
$ w_a L<< 1$, and keeping terms of first order:

\begin{align}
\mathbf{E_1(x)} &\approx \frac{g_{a \gamma \gamma }a_0}{4 \pi} \int \frac{ iw_a| \mathbf{x} - \mathbf{x'}| + O((w_aL)^2)}{| \mathbf{x} - \mathbf{x'}|} ∇\times \mathbf{J}_e (\mathbf{x'}) d^3 \mathbf{x'} \\
&=\frac{g_{a \gamma \gamma }w_a a_0}{4 \pi} \int  ∇\times \mathbf{J}_e (\mathbf{x'}) d^3 \mathbf{x'}  + O((w_aL)^2)\\
&=\frac{g_{a \gamma \gamma }w_a a_0}{4 \pi} \int_S \hat{n} \times \mathbf{J}_e (\mathbf{x'}) d A'  + O((w_aL)^2)\\
&=O((w_aL)^2)
\end{align}

Where the surface integral vanishes due to the fact that our
current is contained within S and so is equal to zero at the
surface.

We can use the same method for equation $(4.7)$, but the only difference from equation $(4.8)$ is a -1 in the numerator. However, the main term remains the same and is not weakened by $ \lambda_a$. This conclusion is applicable in general and is not reliant on the specifics of our detector. 

Our assumptions were limited to the constancy of currents and charges that produce the primary fields, as well as a detector of size $|\mathbf{x} − \mathbf{x_0}| \le L$. Based on these assumptions, we have established that electric fields produced by axions are constantly weakened. Therefore, it can be concluded that in the presence of an oscillating axion field with a large $\lambda_a$, the MQS approximation still holds true.

An interesting thing worth noting is that in this calculation we have neglected terms proportional to $\nabla a $ as they
are suppressed by factors of$\frac{\lambda_a}{\lambda_D} \approx 10^{−3}$. However, when
$\frac{L}{\lambda_a} \lesssim 10^{−3}$, it is possible for electric fields generated by
the $\nabla a \cdot \mathbf{B}$ term in equation $(1.6)$ to dominate over the electric
fields generated by the $ \frac{\partial a}{\partial t}\mathbf{B} $ term in equation $(1.7)$.

## **Approach based on polarization**

An alternate approach to the one underlined in section 3 of the paper is to incorporate the axion induced effects as a sort of vacuum polarization. This is similar to the way material polarization is incorporated into Maxwell's equations. A number of references have used this approach before but reached incorrect physical conclusions. The analogy between material and axion induced polarizations is a subtle one, and correct boundary conditions must be imposed.

We have the maxwell's equations,

$$
\begin{align}
\nabla \cdot \vec{D} &= \rho_f + g_{a \gamma γ} \vec{B} \cdot \nabla a \tag{5.1} \\
\nabla \cdot \vec{B} &= 0 \tag{5.2}\\
\nabla \times \vec{E} &= - \frac{∂ \vec{B}}{∂t} \tag{5.3}\\
\nabla \times \vec{H} &= \vec{J_f} + \frac{∂ D}{∂ t} - g_{a \gamma γ} \left(\vec{E} \times \nabla a + \frac{∂a}{∂t} \vec{B} \right) \tag{5.4}
\end{align}
$$


Here, $ρ_f$ and $\vec{J_f}$ are the free charge density and current density respectively.

We define: 
$\vec{D_a} = \vec{D} - g_{a \gamma γ} a \vec{B} \tag{5.6} $ 

$\hspace{4em} \vec{H_a} = \vec{H} + g_{a γ γ} a \vec{E} \tag{5.7} $



Substituting into the inhomogenous equation for $\vec{D}$,

$$
\begin{align}
\nabla \cdot \vec{D} - g_{a \gamma \gamma} \vec{B} \cdot \nabla a &= \rho_f \\
\implies \nabla \cdot \left( \vec{D_a} - g_{a \gamma \gamma} \vec{B} a \right) &= \rho_f \hspace{2em} \left[ \because \nabla \cdot \vec{B} = 0 \right] \\
\implies  \nabla \cdot \vec{D_a} &= \rho_f
\end{align}
$$

Similarly, from the inhomogenous equation for $\vec{H}$,

$$
\begin{align}
\nabla \times \vec{H} &= \vec{J_e} + \vec{J_f} + \frac{\partial D}{\partial t} - g_{a \gamma γ} \left(\vec{E} \times \nabla a + \frac{∂a}{∂t} \vec{B} \right) \\
\implies \nabla \times \vec{H} + g_{a \gamma γ} \left(\vec{E} \times \nabla a \right) &= \vec{J_f} + \frac{\partial (D_a + g_{a \gamma γ} a \vec{B})}{\partial t} - g_{a \gamma γ} \frac{\partial a}{\partial t} \vec{B} \\
\implies \nabla \times \vec{H} + g_{a \gamma γ} \left(\vec{E} \times \nabla a \right) &= \vec{J_f} + \frac{\partial D_a }{\partial t} + g_{a \gamma \gamma} a \frac{∂ B}{∂ t} \\
\implies \nabla \times \vec{H} + g_{a \gamma γ} \left(\vec{E} \times \nabla a - a \frac{∂ B}{∂ t} \right) &= \vec{J_f} + \frac{\partial D_a }{\partial t} \\
\implies \nabla \times \vec{H} + g_{a \gamma γ} \left(\vec{E} \times \nabla a + a \nabla \times \vec{E} \right) &= \vec{J_f} + \frac{\partial D_a }{\partial t}  \hspace{2em} \left[ \because \nabla \times \vec{E} = - \frac{\partial B}{\partial t} \right]\\
\implies \nabla \times \vec{H} + \nabla \times (a g_{a \gamma γ} \vec{E}) &= \vec{J_f} + \frac{\partial D_a }{\partial t} \\
\implies \nabla \times \left( \vec{H} + a g_{a \gamma γ} \vec{E} \right) &= \vec{J_f} + \frac{\partial D_a }{\partial t} \\
\implies \nabla \times \vec{H_a} &= \vec{J_f} + \frac{\partial D_a }{\partial t} 
\end{align}
$$

Therefore, we can recast the Maxwell's equations similar to their previous form, **without the axion modification terms**.

$$
\begin{align}
\nabla \cdot \vec{D_a} &= \rho_f \tag{5.8} \\
\nabla \cdot \vec{B} &= 0 \tag{5.9} \\
\nabla \times \vec{E} &= - \frac{∂ \vec{B}}{∂t} \tag{5.10} \\
\nabla \times \vec{H_a} &= \vec{J_f} + \frac{\partial D_a }{\partial t} \tag{5.11} 
\end{align}
$$

Since this is exactly in the form as the unmodified Maxwell's equations, we can cast these into the form of a field tensor, $F_a^{\mu ν}$, such that,

$$
F_a^{\mu \nu} = \begin{bmatrix}
0 & {H_a}_3  & -{H_a}_2 & -i {D_a}_1 \\
-{H_a}_3 & 0 & {H_a}_1 & -i {D_a}_2 \\
{H_a}_2 & -{H_a}_1 & 0 & -i {D_a}_3 \\
i {D_a}_1 & i {D_a}_2 & i {D_a}_3 & 0 
\end{bmatrix}
$$

The elements of this field tensor would satisfy the differential equation $ \partial_{\nu} F_a^{\mu \nu} = \frac{1}{c} J_f^\mu$ , where $J_f^\mu$ and $\partial_\nu$ are 4-vectors whose elements are as follows:  

$$
\partial_\nu = \begin{bmatrix} \frac{\partial}{\partial x_1}\\
\frac{\partial}{\partial x_2} \\
\frac{\partial}{\partial x_3}\\
\frac{\partial}{\partial x_4} 
\end{bmatrix} = 
\begin{bmatrix}
\frac{\partial}{\partial x}\\
\frac{\partial}{\partial y} \\
\frac{\partial}{\partial z}\\
\frac{\partial}{\partial (- i t)}
\end{bmatrix} \\
$$
$$ \\ \\
J_f^\mu = \begin{bmatrix}
{J_f}_1 \\
{J_f}_2 \\
{J_f}_3 \\
i \rho_f
\end{bmatrix}
$$

Let's express the field tensor now as $F_a^{\mu \nu} = F^{\mu \nu} - P_a^{\mu \nu}$, where $F^{\mu \nu}$ is the field tensor solution to the original set of Maxwell's equations, with the axion terms as it is, and $P_a^{\mu \nu}$ is a polarization tensor that takes into account the effects of the axion induced polarization. Next we find the elements of the $P_a^{\mu \nu}$ tensor.

We have

$$\begin{align}
∂_\nu F_a^{\mu \nu} &= J_f^\mu \tag{5.12} \\
\implies ∂_\nu (F^{\mu \nu} - P_a^{\mu \nu}) &= J_f^\mu \\
\end{align}
$$
Since, $$∂_{\nu} F^{\mu \nu} = {J_f}^{\mu} + J_a^\mu \tag{5.13} $$
where $J_f^\mu$ and $J_a^\mu$ are the 4-vector current density contributions of free charges and the axion induced effects. The elements of $J_a^\mu$ are given by:

$$
J_a^\mu = \left( - g_{a \gamma γ} \left(\vec{E} \times \nabla a + \frac{∂a}{∂t} \vec{B} \right) , g_{a \gamma γ} \vec{B}\nabla \cdot a \right) ^ T
\tag{5.14}$$

Therefore, on substituting, we get,

$$
\begin{align}
∂_{\nu} F^{\mu \nu} - ∂_{\nu} P_a^{\mu \nu} &= J_f^μ \\
{J_f}^{\mu} + J_a^\mu - ∂_{\nu} P_a^{\mu \nu} &= J_f^μ \\
∂_{\nu} P_a^{\mu \nu} &= J_a^\mu \tag{5.15} 
\end{align}
$$

The expression for $J_a^{\mu \nu}$ can be rewritten as follows:

$$
\begin{align}
J_a^\mu &= \left( - g_{a \gamma γ} \left(\vec{E} \times \nabla a + \frac{∂a}{∂t} \vec{B} \right) , g_{a \gamma γ} \vec{B} \cdot \nabla a \right) ^ T \\
J_a^\mu &= \left( - g_{a \gamma γ} \left(\vec{E} \times \nabla a + \frac{∂\left(a \vec{B}\right)}{∂t} - a \frac{∂ B}{∂ t}\right) , g_{a \gamma γ} \vec{B} \cdot \nabla a \right) ^ T \\
J_a^\mu &= \left( - g_{a \gamma γ} \left(\vec{E} \times \nabla a + \frac{∂\left(a \vec{B}\right)}{∂t} + a \nabla \times \vec{E}\right) , g_{a \gamma γ} \vec{B} \cdot \nabla a \right) ^ T \hspace{1.5em} \left[ \because \nabla \times \vec{E} = - \frac{∂ B}{∂ t}\right] \\
J_a^\mu &= \left( - g_{a \gamma γ} \left(\nabla \times \left( a \vec{E} \right)  + \frac{∂\left(a \vec{B}\right)}{∂t} \right) , g_{a \gamma γ} \nabla \cdot \left(a \vec{B} \right) \right) ^ T \hspace{1.5em} \left[ \because \nabla \cdot \vec{B} = 0 \right] \tag{5.16} 
\end{align}
$$

Therefore, elements of $J_a^\mu$ are: 

$$
\begin{align}
J_a^{\mu} &= -g_{a γ γ} \left( \epsilon_{\mu \nu \lambda} a ∂_ν E_λ + \frac{∂(a B_\mu)}{∂t} \right) \hspace{1.5em} \left[\text{if } \mu = 1, 2,3\right] \\
J_a^{\mu} &= g_{a γ γ} \nabla \cdot \left(a \vec{B} \right) \hspace{10em} \left[\text{if } \mu = 4 \right]
\end{align}
$$

Comparing the elements of $J_a^\mu$ with $∂_{\nu} P_a^{\mu \nu} = J_a^\mu$, we can compare the corresponding derivatives on both sides of the equation and obtain the elements of $P_a^{\mu \nu}$ to be:

$$
\begin{align}
P_a^{\mu \nu} &= g_{a γ γ} a \tilde{F}^{\mu \nu} \\
&= g_{a γ γ} a \begin{bmatrix}
0 & -B_x & -B_y & -B_z \\
B_x & 0 & E_z & -E_y \\
B_y & -E_z & 0 & E_x \\
B_z & E_y & -E_x & 0 \\
\end{bmatrix}
\end{align}
$$

This is completely analogous to the approach where we take into account material polarization by introducing $\vec{J}_{bound}$ and $\rho_{bound}$ , i.e. the bound current density and the bound charge density, respectively.  There, we transform the field tensor solution as $F^{\mu \nu} → G^{\mu \nu} = F^{\mu \nu} - P^{\mu \nu}$ , where $P^{\mu \nu}$ is the polarization tensor.

$$
P^{\mu \nu} = \begin{bmatrix}
0 & P_x & P_y & P_z \\
-P_x & 0 & M_z & -M_y \\
-P_y & -M_z & 0 & M_x \\
-P_z & M_y & -M_x & 0  
\end{bmatrix} \tag{5.17} 
$$

Therefore, comparing the elements of both the tensors, we conclude that the expressions for axion induced polarization and magnetization are:

$$
\begin{align}
\vec{P}_a &= - g_{a γ γ} a \vec{B} \tag{5.18} \\ 
\vec{M}_a &= g_{a γ γ} a \vec{E} \tag{5.19} 
\end{align}
$$

**Conclusions**

From here, we can compute an expression for axion induced bound charge density.

$$
\begin{align}
\nabla \cdot \vec{P}_a &= - g_{a γ γ} \vec{B} ∇a \\ 
& ∝ O(g_{a γ γ} v_{dm})
\end{align}
$$

Thus in the limit of small spatial gradients (which is what we have assumed) axion induced bound charge density is suppressed. Therefore, we can **neglect axion electric polarization**.

Therefore it is the time varying nature of the $\vec{P}_a$ that generates a time varying magnetic field, which in turn generates a time varying electric field. 

Further, an axion $a$ without any spatial or time derivatives can have **no physical manifestations**. 

## **Conclusion**

The paper started with modified maxwell equations, solved them completely in the context of a solenoid. It is shown that the electric fields are always supressed in the long wave length limit. Later the same is found to be true for all bounded geometries, not just for solenoid. The entire analysis could be done in a completely equivalent Polarisation approach. But great care must be taken, as application of boundary conditions is quite subtle and wrong physical intuitions can lead to potentially wrong results. The effects of $\nabla a$ and $\nabla \times a$ have been neglected throughout the paper. It would be interesting to do the analysis in presence of those terms and is relevant in searching for axion winds during cosmoslogical events.

\begin{align}
        \nabla\cdot E&=\rho-g_{a\gamma\gamma}B\cdot\nabla a \\
        \nabla\times E&=-\frac{\partial B}{\partial t}\\
        \nabla\cdot B&=0\\
        \nabla\times B&=J+\frac{\partial E}{\partial t}-g_{a\gamma\gamma}\left[E\times\nabla a-B\frac{\partial a}{\partial t}\right]\\
\end{align}

Where, $a$ is the axion field and is given by,

$$a(\vec{r},t)=a_0e^{ i\omega_a t}$$

Typically such weak modifications to existing laws will be dealt as a perturbation. Here, the strength of the interaction is given by $g_{a\gamma\gamma}$ and it is known to be extremely small. Electric and Magnetic fields can be expressed as a perturbation series,

\begin{align}
E(\vec{r},t)&=E_{0}(\vec{r},t)+g_{a\gamma\gamma}E_{1}(\vec{r},t)+g_{a\gamma\gamma}^2E_{2}(\vec{r},t)+...\\
B(\vec{r},t)&=B_{0}(\vec{r},t)+g_{a\gamma\gamma}B_{1}(\vec{r},t)+g_{a\gamma\gamma}^2B_{2}(\vec{r},t)+...\\
\end{align}

**(a)**  Substitute the perturbative expansions in modified maxwell equations. Seperate the zeroth order and first order maxwell equations.

**(b)** Consider an unbounded system with a constant zeroth order magnetic field and a vanishing zeroth order electric field, everywhere. 

$$B_0(\vec{x},t)=\vec{P}\qquad \forall \vec{x}\in\mathbb{R},\forall t$$
$$E_0(\vec{x},t)=0\qquad \forall \vec{x}\in\mathbb{R},\forall t$$

Find the first order electric field $E_1(\vec{r},t)$ without assuming any ansatz. Note that as Boundary conditions and Initial conditions  aren't provided, you can just solve for any particular solution.

**(c)** A rectangular cavity with relative permeability $\mu_r=1$ is introduced in this constant magnetic field. The medium inside the rectangular cavity is vacuum. The resonance wavelengths of the rectangular cavity are given by,

$$k_{mnq}=\sqrt{\left(\frac{m\pi}{l}\right)^2+\left(\frac{n\pi}{b}\right)^2+\left(\frac{q\pi}{z}\right)^2} \qquad \forall {m,n,q}\in\mathbb{Z}$$

Observing resonance can confirm the existence of axions. #### **Soln.Q.1.**

###### **(a)**



Substituting the perturbative expansions in the modified maxwell equations, retaining only terms of zeroth and first order in $g_{a\gamma\gamma}$ and re-arranging then,

\begin{align}
        \nabla\cdot E_0-\rho&=-g_{a\gamma\gamma}(\nabla\cdot E_1+B_0\cdot\nabla a) \\
        \nabla\times E_0+\frac{\partial B_0}{\partial t}&=-g_{a\gamma\gamma}\left(\nabla\times E_1+\frac{\partial B_1}{\partial t}\right)\\
        \nabla\cdot B_0&=-g_{a\gamma\gamma}\nabla\cdot B_1\\
        \nabla\times B_0-J-\frac{\partial E_0}{\partial t}&=g_{a\gamma\gamma}\left(\frac{\partial E_1}{\partial t}-\nabla\times B_1-\left[E_0\times\nabla a-B_0\frac{\partial a}{\partial t}\right]\right)\\
\end{align}

The R.H.S of all 4 equations are dependent on $g_{a\gamma\gamma}$, whereas the L.H.S aren't. This is satisfied for all small yet arbitrary $g_{a\gamma\gamma}$, iff both L.H.S and R.H.S are individually zero.

Therefore the maxwell equations governing zeroth order fields are,

\begin{align}
        \nabla\cdot E_0&=\rho\\
        \nabla\times E_0&=-\frac{\partial B_0}{\partial t}\\
        \nabla\cdot B_0&=0\\
        \nabla\times B_0&=J+\frac{\partial E_0}{\partial t}\\
\end{align}

Note, that these are the normal maxwell equations. The maxwell equations governing first order fields are,

\begin{align}
        \nabla\cdot E_1&=-B_0\cdot\nabla a \\
        \nabla\times E_1&=-\frac{\partial B_1}{\partial t}\\
        \nabla\cdot B_1&=0\\
        \nabla\times B_1&=\frac{\partial E_1}{\partial t}-\left[E_0\times\nabla a-B_0\frac{\partial a}{\partial t}\right]\\
\end{align}

Note that the, zeroth order fields together with axion fields are acting like a source to the first order fields.

 

###### **(b)**

Given, $B_0(x,t)=\vec{P}$ and $E_0(x,t)=0$.

The maxwell equations governing the first order fields turn out to be,

\begin{align}
        \nabla\cdot E_1&=0 \tag{6.1}\\
        \nabla\times E_1&=-\frac{\partial B_1}{\partial t} \tag{6.2}\\
        \nabla\cdot B_1&=0 \tag{6.3}\\
        \nabla\times B_1&=\frac{\partial E_1}{\partial t}+\vec{P}\frac{\partial a}{\partial t}\tag{6.4}\\
\end{align}

Taking curl of equation $(6.2)$ and using the identity,

$$\nabla\times(\nabla\times A)=\nabla(\nabla.A)-\nabla^2A$$

leads to,

$$\nabla(\nabla.E_1)-\nabla^2E_1=-\frac{\partial^2 E_1}{\partial t^2}-\vec{P}\frac{\partial^2 a}{\partial t^2}-\frac{\partial \vec{P}}{\partial t}\frac{\partial a}{\partial t}$$

Using equation $(6.1)$, given information like $a=a_0e^{i\omega_a t}$, and the fact that $\vec{P}$ is a constant vector leads to,

$$\nabla^2E_1(\vec{r},t)-\frac{\partial^2 E_1}{\partial t^2}(\vec{r},t)=-a_0\omega_a^2\vec{P}e^{i\omega_a t} \tag{6.5}$$


At this point, it might be tempting to use the single term solution of the wave equation. But, the source terms aren't vanishing at infinities, so better not to proceed that way and end up with an unsolvable diverging integral. 

One can easily consider an ansatz, but let's take a bit more rigorous approach in which green's functions are typically derived.

Fourier transforming the equation $(6.5)$ in time domain, 

$$\nabla^2E_1(\vec{r},\omega)+\omega^2E_1(\vec{r},\omega)=a_0\omega_a^2\vec{P}\delta(\omega-\omega_a) \tag{6.6}$$

Fourier transforming the equation $(6.6)$ in space domain,

$$-k^2E_1(\vec{k},\omega)+\omega^2E_1(\vec{k},\omega)=a_0\omega_a^2\vec{P}\delta(\omega-\omega_a)\delta(\vec{k}) \tag{6.7}$$

$$E_1(\vec{k},\omega)=a_0\omega_a^2\vec{P}\frac{\delta(\omega-\omega_a)\delta(\vec{k})}{\omega^2-k^2}$$

Performing an inverse fourier transform in both space and time domains is trivial due to the dirac deltas in integrand,

$$E_1(\vec{r},t)=a_0\omega_a^2\vec{P}\int_{0}^{\infty}d\omega e^{i\omega t}\int_{\text{k-space}}d^3ke^{i\vec{k}\cdot\vec{r}}\frac{\delta(\omega-\omega_a)\delta(\vec{k})}{\omega^2-k^2}$$

Solving both the integrals gives us,

$$E_1(\vec{r},t)=a_0\vec{P}e^{i\omega_a t}$$

Please note that, this solution is a particular solution and the actual one is upto homogenous wave equation solutions. We can determine the unique solution only if boundary conditions are provided.



###### **(c)**

Given that haloscope experiments try matching least possible non-zero resonant frequency, they try for $(1,0,0)$ mode, where $1$ will be along the longest dimension.

In this cavity's case, $z>l,b$. Therefore, $(m,n,q)=(0,0,1)$.

$$\omega_{mno}=ck_{mno}=\frac{c\pi}{z}$$

$E_1(\vec{r},t)$ obtained in part $(b)$ is,

$$E_1(\vec{r},t)=a_0\vec{P}e^{i\omega_a t}$$

Therefore the frequency is, 

$$\omega=\omega_a=\frac{m_ac^2}{\hbar}$$

Haloscope experiments try matching the resonant frequency $\omega_{mno}$ and $\omega_a$.

$$\frac{c\pi}{z}=\frac{m_ac^2}{\hbar}$$

$$z=\frac{\hbar c\pi}{m_ac^2}$$

$$z\in\left(\frac{6.58\times10^{-16}\times3\times10^{8}\times3.14}{10^{-6}}\text{ m},\frac{6.58\times10^{-16}\times3\times10^{8}\times3.14}{10^{-14}}\text{ m}\right)$$

$$z\in(0.62\text{ m},6.2\times10^{7}\text{ m})$$

You might have wondered if it's a mistake, because one end of the length scale is impractical. No, you didn't. This resonance technique is not suitable for probing axions of such low masses, and that is why other methods have been suggested in the recent years.
