###Introduction
<p align="justify">
The conservation of momentum is a fundamental principle in physics that states that the total momentum of a closed system remains constant unless acted upon by an external force. In recent years, this principle has been extended to photonic *time-varying media , which are materials whose properties change over time.* These media are of great interest in a variety of fields, including photonics, plasmonics, and metamaterials, due to their unique properties, such as negative refraction and superluminal propagation.                            
<p align="justify">
In this tutorial review, we explore the conservation of momentum in time-varying environments. We show that the Minkowski momentum is an unchanging value using three different approaches that increase in complexity:

1.   By using the boundary conditions for Maxwell equations at a temporal boundary, we can demonstrate conservation of momentum.

2.   We can also check for constants of motion and derive conservation laws to confirm the conservation of momentum. 

3.   By applying temporal and spatial translations within the framework of the Lagrangian theory of the electromagnetic field is another way to show that the Minkowski momentum remains constant over time.




<p align="justify">
Overall, we provide multiple methods to demonstrate that momentum is conserved in environments that change over time.

<p align="justify">
When light travels through a medium that varies with time, it can experience a change in momentum due to the interaction between the light and the medium. This change in momentum is related to the Poynting vector, which describes the direction and magnitude of the energy flow of electromagnetic waves. The Poynting vector is given by:

$$
S = E \times H
$$
<p align="justify">
where E is the electric field and H is the magnetic field. The direction of the Poynting vector is perpendicular to both the electric and magnetic fields and represents the direction of energy flow.
<p align="justify">
In a time-varying medium, the Poynting vector is not conserved, and the momentum of the electromagnetic wave can change due to the interaction with the medium. This change in momentum can be described using either the Abraham momentum or the Minkowski momentum.

The Abraham momentum is defined as:


$$
P_{A}(t)=\int d^{3} r P_{A}(r, t)=\mu_{0} \varepsilon_{0} \int d^{3} r 
E(r, t) \times H(r, t)
$$


 The Abraham momentum takes into account the contribution of the electric field to the momentum of the wave.

The Minkowski momentum is defined as:

$$
P_{M}(t)=\int d^{3} r P_{M}(r, t)=\mu_{0} \varepsilon_{0} \int d^{3} r 
D(r, t) \times B(r, t)
$$

The Minkowski momentum takes into account the contribution of both the electric and magnetic fields to the momentum of the wave.
<p align="justify">
The relationship between the Abraham momentum and the Minkowski momentum can be described using the Abraham-Minkowski controversy. The controversy arises because the two momentum definitions predict different results for the momentum transfer in certain situations, such as in the case of a dielectric slab reflecting light. The controversy was resolved by the consensus that both momenta are valid and lead to the same result in the case of a closed system.

## Conservation of Minkowski momentum

###Method 1: Boundary conditions at temporal boundary

Our starting point is Maxwell curl equations in time-varying media, which, in the absence of charges and currents, can be written as follows

$$
\begin{gathered}
\nabla \times \vec{E}(\vec{r}, t)=-\partial_{t} \vec{B}(\vec{r}, t) \qquad (3) \\
\nabla \times \vec{H}(\vec{r}, t)=\partial_{t} \vec{D}(\vec{r}, t) \qquad (4)
\end{gathered}
$$
<br>
For the sake of simplicity, we assume homogeneous and instantaneous time-varying media, with constitutive relations

$$
\vec{D}(\vec{r}, t)=\varepsilon(t) \vec{E}(\vec{r}, t) \qquad (5)
$$



$$
\vec{B}(\vec{r}, t)=\mu(t) \vec{H}(\vec{r}, t) \qquad (6)
$$
<p align="justify">
A more complete description of time-varying media would include the impact of dispersion and loss [17, 36]. However, a system with dissipation does not necessarily conserve quantities even in the presence of symmetries. In addition, the assumption of instantaneous media is widespread in the field of temporal metamaterials [2].

Integrating Maxwell equations (3)-(4) accross a temporal boundary taking place at $t_{0}$, where material parameters suddenly change from $\varepsilon\left(t_{0}^{-}\right), \mu\left(t_{0}^{-}\right)$to $\varepsilon\left(t_{0}^{+}\right), \mu\left(t_{0}^{+}\right)$, gives
<br></br>
$$
\begin{gathered}
\int_{t_{0}^{-}}^{t_{0}^{+}} d t \nabla \times \vec{H}(\vec{r}, t)=\int_{t_{0}^{-}}^{t_{0}^{+}} d t \partial_{t} \vec{D}(\vec{r}, t)=\vec{D}\left(\vec{r}, t_{0}^{+}\right)-\vec{D}\left(\vec{r}, t_{0}^{-}\right) \qquad (7)\\
-\int_{t_{0}^{-}}^{t_{0}^{+}} d t \nabla \times \vec{E}(\vec{r}, t)=\int_{t_{0}^{-}}^{t_{0}^{+}} d t \partial_{t} \vec{B}(\vec{r}, t)=\vec{B}\left(\vec{r}, t_{0}^{+}\right)-\vec{B}\left(\vec{r}, t_{0}^{-}\right) \qquad (8)
\end{gathered}
$$
</br>
<p align="justify">
Therefore, we find that $\vec{D}(\vec{r}, t)$ and $\vec{B}(\vec{r}, t)$ must be continuous accross changes of the constitutive parameters, for finite $\vec{E}(\vec{r}, t)$ and $\vec{H}(\vec{r}, t)$ fields. This property is well-known since early works on time-varying media [16]. As a consequence, this reasoning confirms that the Minkowski momentum - uniquely defined as a function of $\vec{D}$ and $\vec{B}$ fields via Eq. (2) - is a continuous quantity across a temporal boundary, suggesting that is should be a conserved quantity in time-varying media. However, this approach fails at providing any insight on the associated conservation law and/or how it can be related to invariance under spatial translations. Moreover, it does not clarify the (non) conservation of Abraham momentum.

### Method 2: Constants of motion and conservation laws
<p align = "justify">
In this section, we address the conservation of momentum in time-varying media by direcly testing if a given quantity is a constant of motion. To this end, one can take the time derivative of the quantity under question and check if it is zero, in which case it shall be a constant of motion/conserved quantity. Before addressing the momentum, it is instructive to analyze the energy of the electromagnetic field, which in time-varying media can be written as</p>
<br></br>
$$
U(t)=\int d^{3} r u(\vec{r}, t) \hspace{5em} (9)       
$$

with energy density
<br></br>
$$
u(\vec{r}, t)=\frac{1}{2}\left[\varepsilon(t) E^{2}(\vec{r}, t)+\mu(t) H^{2}(\vec{r}, t)\right] \hspace{5em} (10)
$$

Taking the time derivative of the energy in the equation (9) and, substituting equation (10), leads to the following expression
<br></br>
$$
\frac{d U}{d t}=\int \frac{1}{2}\left[\frac{d \varepsilon(t)}{d t} E^{2}(\vec{r}, t)+2E\varepsilon(t)\frac{d E(\vec{r},t)}{d t} +\frac{d \mu(t)}{d t} H^{2}(\vec{r}, t)+2H\mu(t)\frac{d H(\vec{r},t)}{d t} \right] d^{3} r \qquad (10a)
$$

Then  From Maxwell equations (5)-(6),

$$
\vec{D}(\vec{r}, t)=\varepsilon(t) \vec{E}(\vec{r}, t)
$$

Then taking derivative on both side with respect to time

$$
\partial_{t} \vec{D}(\vec{r}, t) = \frac{d \varepsilon(t)}{d t} E(\vec{r}, t)+\varepsilon(t)\frac{d E(\vec{r},t)}{d t}
$$

and 

$$
\vec{B}(\vec{r}, t)=\mu(t) \vec{H}(\vec{r}, t)
$$

then taking derivative on both side with respect to time<br></br>
$$
\partial_{t} \vec{B}(\vec{r}, t) = \frac{d \mu(t)}{d t} H(\vec{r}, t)+\mu(t)\frac{d H(\vec{r},t)}{d t}
$$

So, we get, $\quad \varepsilon(t)\frac{d E(\vec{r},t)}{d t} = \partial_{t} \vec{D}(\vec{r}, t)-\frac{d \varepsilon(t)}{d t} E(\vec{r}, t)$

Similarly we get, $\, \mu(t)\frac{d H(\vec{r},t)}{d t} = \partial_{t} \vec{B}(\vec{r}, t)-\frac{d \mu(t)}{d t} H(\vec{r}, t)\,$
<br></br>

Now we substitute these equations in above equation (10a) as,<br></br>
$$
\frac{d U}{d t}=\int \frac{1}{2}\left[\frac{d \varepsilon(t)}{d t} E^{2}(\vec{r}, t)+2E(\vec{r}, t)\left(\partial_{t} \vec{D}(\vec{r}, t)-\frac{d \varepsilon(t)}{d t} E(\vec{r}, t)\right) +\frac{d \mu(t)}{d t} H^{2}(\vec{r}, t)+2H(\vec{r}, t)\left(\partial_{t} \vec{B}(\vec{r}, t)-\frac{d \mu(t)}{d t} H(\vec{r}, t)\right) \right] d^{3} r \qquad (10b)
$$

Further simplifying,<br></br>
$$
\frac{d U}{d t}=\int \frac{1}{2}\left[-\left(\frac{d \varepsilon(t)}{d t} E^{2}(\vec{r}, t)+\frac{d \mu(t)}{d t} H^{2}(\vec{r}, t)\right)+2E(\vec{r}, t)\partial_{t} \vec{D}+2H(\vec{r}, t)\partial_{t} \vec{B}\right]d^{3} r
$$
<br></br>
Then substituting the Maxwell equations for $ \, \partial_{t} \vec{D}(\vec{r}, t)=\nabla \times \vec{H}(\vec{r}, t)\,$ and $\, -\partial_{t} \vec{B}(\vec{r}, t)=\nabla \times \vec{E}(\vec{r}, t)\, $ then we get as,<br></br>
$$
\frac{d U}{d t}=\int \left[-\frac{1}{2}\left(\frac{d \varepsilon(t)}{d t} E^{2}(\vec{r}, t)+\frac{d \mu(t)}{d t} H^{2}(\vec{r}, t)\right)+E(\vec{r}, t)\cdot\left(\nabla \times \vec{H}(\vec{r}, t)\right)-H(\vec{r}, t)\cdot\left(\nabla \times \vec{E}(\vec{r}, t)\right)\right]d^{3} r \qquad (10c)
$$ 

Now we recall vector calculus formulas<br></br> $\,\nabla \cdot \left(\vec{E}(\vec{r}, t) \times \vec{H}(\vec{r}, t)\right)=\left(\nabla \times \vec{E}(\vec{r}, t) \right) \cdot \vec{H}(\vec{r}, t) - \vec{E}(\vec{r},t) \cdot \left(\nabla \times \vec{H}(\vec{r},t) \right) \,$

Substituting this whole formula in above equation (10c), we get<br></br>
$$
\frac{d U}{d t}=-\int \nabla \cdot \left(\vec{E}(\vec{r}, t) \times \vec{H}(\vec{r}, t)\right) d^{3} r - \int d^{3} r\frac{1}{2}\left(\frac{d \varepsilon(t)}{d t} E^{2}(\vec{r}, t)+\frac{d \mu(t)}{d t} H^{2}(\vec{r}, t)\right) \qquad (10d)
$$

Now we recall the Poynting vector $\, S = \left(\vec{E}(\vec{r}, t) \times \vec{H}(\vec{r}, t)\right)\,$ and using the Gauss Divergence Theorem $\, \int_v \nabla \cdot V d v = \int_S V\cdot d S\,$ then after substitution in equation (10d)  we get<br></br> 
$$
\frac{d U}{d t}=-\int d \vec{S} \cdot S-\frac{1}{2} \int d^{3} r\left[\frac{d \varepsilon(t)}{d t} E^{2}(\vec{r}, t)+\frac{d \mu(t)}{d t} H^{2}(\vec{r}, t)\right] 
$$

Then from (1) we substitute $\,S=\frac{\vec{p_A}}{\mu_{0} \varepsilon_{0}}\,$ and get the final expression as<br></br>
$$
\frac{d U}{d t}=-\frac{1}{\mu_{0} \varepsilon_{0}} \int d \vec{S} \cdot \vec{p}_{A}-\frac{1}{2} \int d^{3} r\left[\frac{d \varepsilon(t)}{d t} E^{2}(\vec{r}, t)+\frac{d \mu(t)}{d t} H^{2}(\vec{r}, t)\right] \hspace{5em} (11)
$$
<p align="justify">
On the one hand, the first term in the r.h.s. of (11) is a surface term proportional to the $\vec{E}$ and $\vec{H}$ fields. This term physically means that the change of energy over time is partly due to energy either leaking out or coming into the system. It can be seen as a flux of either outgoing or incoming Poynting vector field, hence setting down a link with $\vec{P}_{A}$ (Eq. (1)). It confirms the role of the Abraham momentum as the kinetic momentum, associated with energy transport. If the volume is large enough to capture the entirety of the $\vec{E}$ and $\vec{H}$ fields within the time interval of interest, its contribution vanishes. On the other hand, the second term in the r.h.s. of (11) is a volume integral directly linked to the time modulation of the permittivity and permeability, which results in a change of the energy of the system. It represents the energy that must be pumped into or retracted from the system in order to realize the time modulation of the material parameters. In other words, the time variation of the material parameters act as sources or sinks of electromagnetic energy. By contrast, Eq. (11) shows that for a medium with static material properties $d U / d t=0$ and energy would be a conserved quantity.</p>

Eq. (11) can also be casted as a local conservation law as a function of the energy and momentum densities

$$
\frac{d u(\vec{r}, t)}{d t}+\frac{1}{\mu_{0} \varepsilon_{0}} \nabla \cdot \vec{p}_{A}(\vec{r}, t)=-\frac{1}{2}\left[\frac{d \varepsilon(t)}{d t} E^{2}(\vec{r}, t)+\frac{d \mu(t)}{d t} H^{2}(\vec{r}, t)\right] \hspace{5em} (12)
$$

where we clearly identify the source/sink at the r.h.s..

Let us now tackle the conservation of Minkowski momentum that is equation (2) and examine the time variation of Abraham momentum. 

$$
\vec{P}_{M}(t)=\int d^{3} r \vec{p}_{M}(\vec{r}, t)=\int d^{3} r \vec{D}(\vec{r}, t) \times \vec{B}(\vec{r}, t) 
$$

By introducing Maxwell equations and applying a few vector calculus identities, and finding the time derivative of equation (2):

$$
\frac{d \vec{P}_{M}(t)}{d t}=\int d^{3} r \frac{d (\vec{D}(\vec{r}, t) \times \vec{B}(\vec{r}, t))}{d t} 
$$

Now using the identity :

$$
\frac{d(\vec{A} \times \vec{B})}{d t} = \frac{d \vec{A}}{d t} \times \vec{B} + \vec{A} \times \frac{d \vec{B}}{d t}
$$

we get as:

$$
\frac{d \vec{P}_{M}(t)}{d t} = \int d^{3} r \left[\frac{d \vec{D}(\vec{r}, t)}{d t} \times \vec{B}(\vec{r}, t) + \vec{D}(\vec{r}, t) \times \frac{d \vec{B}(\vec{r}, t)}{d t} \right]
$$ 

By using Maxwell's Equation, we get:

$$
\frac{d \vec{P}_{M}(t)}{d t} = \int d^{3} r \left[\left(\nabla \times \vec{H}(\vec{r},t)\right) \times \vec{B}(\vec{r},t) - \vec{D}(\vec{r},t) \times \left(\nabla \times \vec{E}(\vec{r},t)\right) \right]
$$

Again using $\,\vec{D}(\vec{r}, t)=\varepsilon(t) \vec{E}(\vec{r}, t)\,$ and $\, \vec{B}(\vec{r}, t)=\mu(t) \vec{H}(\vec{r}, t)\,$ we get,

$$
\frac{d \vec{P}_{M}(t)}{d t} = \int d^{3} r \, \mu(t)\left[\left(\nabla \times \vec{H}(\vec{r},t)\right) \times \vec{H}(\vec{r},t)\right] - \int d^{3} r \, \varepsilon(t)\left[\vec{E}(\vec{r},t) \times \left(\nabla \times \vec{E}(\vec{r},t)\right)\right]
$$
Then again using the following identities:

$$
(\vec{A} \cdot \nabla)\vec{A} = \frac{1}{2}\nabla|\vec{A}|^{2} - \vec{A} \times (\nabla \times \vec{A}) = \frac{1}{2}\nabla|\vec{A}|^{2} + (\nabla \times \vec{A}) \times \vec{A}
$$

And then converting all volume integral to surface integral we get,

$$
\frac{d \vec{P}_{M}(t)}{d t}=\varepsilon(t)\left[\int d \vec{S} \cdot (E \vec{E}) - \frac{1}{2} \int d \vec{S}(\vec{E} \cdot \vec{E})\right]
$$

$$
+ \mu(t) \left[\int d \vec{S} \cdot (H \vec{H}) - \frac{1}{2} \int d \vec{S}(\vec{H} \cdot \vec{H})\right]
$$

After writing the above equation in components form we get the time derivative of the Minkowski momentum as:

$$
\frac{d \vec{P}_{M}(t)}{d t}=\varepsilon(t)\left[\sum_{p=x, y, z} \vec{u}_{p} \int d \vec{S} \cdot\left(E_{p} \vec{E}\right)-\frac{1}{2} \int d \vec{S}(\vec{E} \cdot \vec{E})\right] 
$$

$$
+\mu(t)\left[\sum_{p=x, y, z} \vec{u}_{p} \int d \vec{S} \cdot\left(H_{p} \vec{H}\right)-\frac{1}{2} \int d \vec{S}(\vec{H} \cdot \vec{H})\right] \hspace{5em} (13)
$$

where $\, \mathbf{u}{x}=\hat{\mathbf{i}}, \mathbf{u}{y}=\hat{\mathbf{j}},$ and $\mathbf{u}_{z}=\hat{\mathbf{k}}\,$ are the unit vectors in the $x$, $y$, and $z$ directions, respectively, and $\int d\mathbf{S}$ denotes the surface integral over a closed surface enclosing the region of interest.

<p align = "justify">
Equation (13) is known as the Minkowski momentum formula, and it describes the rate of change of the total momentum of the electromagnetic field in terms of the electric and magnetic fields. The first term on the right-hand side represents the contribution of the electric field to the momentum, while the second term represents the contribution of the magnetic field. The factor $\varepsilon(t)$ and $\mu(t)$ represent the time-varying permittivity and permeability of the medium, respectively.<br></br>
Note that the Minkowski momentum formula applies to the electromagnetic field only, and does not include the momentum of matter. The total momentum of a system including both the electromagnetic field and matter can be obtained by adding the Minkowski momentum to the mechanical momentum of matter.<br></br>
To further explain the terms in equation (13), let us consider the first term on the right-hand side, which represents the contribution of the electric field to the momentum. This term involves the dot product of the electric field $\vec{E}$ with the surface element $d\mathbf{S}$, which gives the flux of the electric field through the surface. Multiplying this by the corresponding unit vector $\mathbf{u}_p$ and integrating over the surface gives the total momentum contribution in the $p$ direction due to the electric field.<br></br>
The second term on the right-hand side represents the contribution of the electric field to the energy density of the field. This term involves the dot product of the electric field with itself, which gives the squared magnitude of the electric field. Integrating this over the surface gives the total electric field energy. Multiplying this by $-1/2$ and integrating again over time gives the rate of change of the electric field energy.<br></br>
The terms involving the magnetic field can be similarly explained. The dot product of the magnetic field with the surface element gives the magnetic flux through the surface, and multiplying this by the corresponding unit vector and integrating over the surface gives the total momentum contribution in the $p$ direction due to the magnetic field. The dot product of the magnetic field with itself gives the squared magnitude of the magnetic field, and integrating this over the surface and over time gives the rate of change of the magnetic field energy.<br></br>
Now, we find that the time derivative of the Minkowski momentum reduces to surface terms. Once again, if the volume of integration is taken large enough so that all the $\vec{E}$ and $\vec{H}$ fields are confined within its interior, all surface terms vanish. In other words, $d \vec{P}_{M}(t) / d t=0$, proving that the Minkowski momentum is a constant of motion as expected. It is also instructive to note that the above equation can be written in a differential form as a conservation law for the momentum density:</p>

$$
\frac{d \vec{p}_{M}(t)}{d t}=\nabla \cdot \overline{\overline{\vec{T}}}_{M}(\vec{r}, t) \hspace{5em} (14)
$$

where we define the Minkowski stress tensor for time-varying media as

$$
\overline{\overline{\vec{T}}}_{M}(\vec{r}, t)=\varepsilon(t)\left(\vec{E} \otimes \vec{E}-\frac{1}{2} \overline{\overline{\vec{I}}}(\vec{E} \cdot \vec{E})\right)+\mu(t)\left(\vec{H} \otimes \vec{H}-\frac{1}{2} \overline{\overline{\vec{I}}}(\vec{H} \cdot \vec{H})\right) \hspace{3em} (15)
$$

with $\overline{\overline{\vec{I}}}$ being the identity dyadic:

$$
\overline{\overline{\vec{I}}} = \hat{i}\hat{i} + \hat{j}\hat{j} + \hat{k}\hat{k}
$$

Proceeding similarly with the Abraham momentum reveals that in general it is not a conserved quantity:
From the equation (1):

$$
\vec{P}_{A}(t)=\int d^{3} r \vec{p}_{A}(\vec{r}, t)=\mu_{0} \varepsilon_{0} \int d^{3} r \vec{E}(\vec{r}, t) \times \vec{H}(\vec{r}, t)
$$

We can write this as

$$
\vec{P}_{A}(t)=\mu_{0} \varepsilon_{0} \int d^{3} r \left[\frac{\vec{D}(\vec{r}, t)}{\varepsilon(t)} \times \frac{\vec{B}(\vec{r}, t)}{\mu(t)}\right]
$$

Similarly finding the time derivative of above equation by using the identity $\, \frac{d(\vec{A} \times \vec{B})}{d t} = \frac{d \vec{A}}{d t} \times \vec{B} + \vec{A} \times \frac{d \vec{B}}{d t}\,$:

$$
\frac{d \vec{P}_A (t)}{d t} = \mu_{0} \varepsilon_{0} \int d^{3} r \left[\left(\frac{-\vec{D}(\vec{r},t)}{\varepsilon^{2}(t)}\frac{d \varepsilon(t)}{d t} + \frac{1}{\varepsilon(t)}\frac{d \vec{D}(\vec{r},t)}{d t}\right) \times \frac{\vec{B}(\vec{r},t)}{\mu(t)} + \frac{\vec{D}(\vec{r},t)}{\varepsilon(t)} \times \left(-\frac{d \mu(t)}{d t}\frac{\vec{B}(\vec{r},t)}{\mu^{2}(t)} + \frac{d \vec{B}(\vec{r},t)}{d t}\frac{1}{\mu(t)}\right)\right]
$$

Now after arranging the terms:

$$
\frac{d \vec{P}_A (t)}{d t} = \mu_{0} \varepsilon_{0} \int d^{3} r\left[\frac{-\vec{D}(\vec{r},t)}{\varepsilon^{2}(t)}\frac{d \varepsilon(t)}{d t} \times \frac{\vec{B}(\vec{r},t)}{\mu(t)} - \frac{\vec{D}(\vec{r},t)}{\varepsilon(t)} \times \frac{d \mu(t)}{d t}\frac{\vec{B}(\vec{r},t)}{\mu^{2}(t)}\right] 
$$

$$
+ \mu_{0} \varepsilon_{0} \int d^{3} r\left[\frac{1}{\varepsilon(t)}\frac{d \vec{D}(\vec{r},t)}{d t} \times \frac{\vec{B}(\vec{r},t)}{\mu(t)} - \frac{1}{\mu(t)}\frac{d \vec{B}(\vec{r},t)}{d t} \times \frac{\vec{D}(\vec{r},t)}{\varepsilon(t)} \right]
$$

Now further simplifying the first term in above equation we get:

$$
\frac{d \vec{P}_A (t)}{d t} = \mu_{0} \varepsilon_{0} \int d^{3} r \left[\frac{1}{\varepsilon(t)}\frac{d \varepsilon(t)}{d t}\left(-\frac{\vec{D}(\vec{r},t)}{\varepsilon(t)} \times \frac{\vec{B}(\vec{r},t)}{\mu(t)}\right) + \frac{1}{\mu(t)}\frac{d \mu(t)}{d t}\left(-\frac{\vec{D}(\vec{r},t)}{\varepsilon(t)} \times \frac{\vec{B}(\vec{r},t)}{\mu(t)}\right)\right]
$$

$$
+ \mu_{0} \varepsilon_{0} \int d^{3} r\left[\frac{1}{\varepsilon(t)}\frac{d \vec{D}(\vec{r},t)}{d t} \times \frac{\vec{B}(\vec{r},t)}{\mu(t)} - \frac{1}{\mu(t)}\frac{d \vec{B}(\vec{r},t)}{d t} \times \frac{\vec{D}(\vec{r},t)}{\varepsilon(t)} \right]
$$

Again further simplifying:

$$
\frac{d \vec{P}_A (t)}{d t} = \left[-\mu_{0} \varepsilon_{0} \int d^{3} r (\vec{E} \times \vec{H})\right]\left[\frac{1}{\varepsilon(t)}\frac{d \varepsilon(t)}{d t} + \frac{1}{\mu(t)}\frac{d \mu(t)}{d t}\right]
$$

$$
+ \mu_{0} \varepsilon_{0} \int d^{3} r\left[\frac{1}{\varepsilon(t)}\frac{d \vec{D}(\vec{r},t)}{d t} \times \frac{\vec{B}(\vec{r},t)}{\mu(t)} - \frac{1}{\mu(t)}\frac{d \vec{B}(\vec{r},t)}{d t} \times \frac{\vec{D}(\vec{r},t)}{\varepsilon(t)} \right]
$$

Now we can see that $\,
\vec{P}_{A}(t)=\mu_{0} \varepsilon_{0} \int d^{3} r \vec{E}(\vec{r}, t) \times \vec{H}(\vec{r}, t)\,
$ is being used and the second term is the Minkowski term from equation (13), so finally we can write as:

$$
\begin{gathered}
\frac{d \vec{P}_{A}}{d t}=-\left(\frac{1}{\varepsilon(t)} \frac{d \varepsilon(t)}{d t}+\frac{1}{\mu(t)} \frac{d \mu(t)}{d t}\right) \vec{P}_{A}(t) \\
-\frac{\varepsilon_{0} \mu_{0}}{\mu(t)}\left[\frac{1}{2} \int d \vec{S}(\vec{E} \cdot \vec{E})-\sum_{p=x, y, z} \vec{u}_{p} \int d \vec{S} \cdot\left(E_{p} \vec{E}\right)\right] \\
-\frac{\varepsilon_{0} \mu_{0}}{\varepsilon(t)}\left[\frac{1}{2} \int d \vec{S}(\vec{H} \cdot \vec{H})-\sum_{p=x, y, z} \vec{u}_{p} \int d \vec{S} \cdot\left(H_{p} \vec{H}\right)\right]
\end{gathered} \hspace{5em} (16)
$$
<p align = "justify">
Here again, the second and third terms are surface terms that would vanish for a sufficiently large volume. However, the first term illustrates that the Abraham momentum does change in time, following the change in the permittivity and permeability of the medium. Equation (16) can also be compactly written as a local conservation law for the momentum density</p>

$$
\frac{d \vec{p}_{A}}{d t}=\nabla \cdot \overline{\overline{\vec{T}}}_{A}-\left(\frac{1}{\varepsilon(t)} \frac{d \varepsilon(t)}{d t}+\frac{1}{\mu(t)} \frac{d \mu(t)}{d t}\right) \vec{p}_{A}(\vec{r}, t) \hspace{5em} (17)
$$

where we define the Abraham stress tensor in time-varying media, related to the Minkowski stress tensor as follows

$$
\overline{\overline{\vec{T}}}_{A}=\frac{\varepsilon_{0} \mu_{0}}{\mu(t) \varepsilon(t)} \overline{\overline{\vec{T}}}_{M}(\vec{r}, t) \hspace{5em} (18)
$$
<p align = "justify">
In conclusion, testing for constants of motions provides an independent confirmation that the Minkowski momentum is indeed a conserved quantity in time-varying media. In addition, it provides insight in the form of the conservation law that supports its invariance. Furthermore, it shows that the Abraham momentum is not a constant of motion in close connection to energy considerations, and re-emphasizes its role as the kinetic momentum of the electromagnetic field. Nevertheless, writing the conservation law does not clarify the role of the invariance of the system under spatial translations in the conservation of momentum.</p>

### Method 3: Lagrangian approach
<p align = "justify">
In this section, we use the classical lagrangian approach of electromagnetic fields to show that conservation of minkowski momentum as a consequence of invariance under spatial translations is a classical effect. First, we will review the lagrangian description extended to time varying media, then we will see a form of Noether theorem in this formalism and then finally show the quantities associated with temporal and spatial translations for time varying media.</p>

### Langrangian description
Maxwell's equations can be derived from the principle of least or stationary action using the lagrangian:
$$
L(t)=\int d^3 r\mathcal{L}(\vec r,t)
$$
where $\mathcal{L}$ is the lagrangian density given by<br></br>
$$
\mathcal{L}(\vec r,t)=\frac{1}{2}\int d^3r[\epsilon(t)\vec E(\vec r,t)\cdot \vec E(\vec r,t)-\mu(t)\vec H(\vec r,t)\cdot \vec H(\vec r,t) ]
$$
<p align = "justify">
This form of the lagrangian recovers the equations of motion for the electromagnetic field when applied on lagrange's equation (to be shown in the tutorial)
The action is defined as:
$$
S(t_1,t_2)=\int_{t_1}^{t_2}dtL(t)
$$
<p align = "justify">
For the lagrangian description it is convenient to work with the scalar potential $V(\vec r,t)$ and the vector potential $\vec A(\vec r,t)$ instead of the fields. If we use coulomb gauge ($\nabla\cdot \vec A(\vec r,t)$=0) then in the absence of charges, the scalar potential is 0 and both fields can be written as a function of the vector potential:
$$
\vec D(\vec r,t)=-\epsilon(t)\partial_t\vec A(\vec r,t) \\
\vec B(\vec r,t)=\nabla\times \vec A(\vec r,t)\\
$$
Now, using maxwell's equations for each component of the vector potential (p=x,y,z):<br></br>
$$
\begin{align}
\nabla^2A_p(\vec r,t)=\mu(t)\partial_t\{\epsilon(t)\partial_t A_p(\vec r,t)\}
\end{align}$$
In the coulomb gauge, the minkowski momentum can be compactly written as:<br></br>
$$
\vec P_M=-\epsilon(t)\sum_p\int d^3r\partial_t A_p(\vec r,t)\nabla A_p(\vec r,t)
$$
Similarly the lagrangian density now becomes:<br></br>
$$
\mathcal{L}=\frac{1}{2}\sum_p\Big(\epsilon(t)\dot A_p^2(\vec r,t)-\frac{1}{\mu(t)}(\nabla\times\vec A(\vec r,t))^2_p\Big)
$$
Now if we impose $\delta S=0$ we get back the lagrange equations:<br></br>
$$
\frac{\partial \mathcal{L}}{\partial A_p}-\sum_q\partial_q\Big( \frac{\partial \mathcal{L}}{\partial (\partial_q A_p)}\Big)-\frac{d}{dt}\frac{\partial \mathcal{L}}{ \partial \dot A_p}=0
$$
Using the expression for lagrangian density we can get conjugate momentum for each vector component $A_p$:
$$
\Pi_p(\vec r,t)=\frac{\partial \mathcal{L}}{\partial \dot A_p}=\epsilon(t)\dot A_p(\vec r,t)=-D_p(\vec r,t)
$$
<p align = "justify">
At this juncture, it is important to differentiate between momentum associated with spatial translations and conjugate momentum. In a typical lagrangian treatment, for eg. for a free particle with lagrangian $L=\sum_p\frac{1}{2}m\dot r_p^2$ position is the dynamic variable and hence the conjugate momenta $\partial L/\partial \dot r_p=m\dot r_p$ correspond to the components of linear momentum (associated with spatial translations). However, in the above description the vector potential is the dynamical variable. Hence, it is important to distinguish between the two.<br></br>
Finally the hamiltonian is defined as a function of conjugate momentum as follows:<br></br>
$$
H=\sum_p\int d^3r\Pi_p(\vec r,t)\dot A_p(\vec r,t)-L
$$


### Noether's theorem in coulomb gauge
<p align="justify">
In this section, we cast a form of Noether's theorem which allows us to discern the conserved quantities associated with the continuous symmetries of time-varying media. Consider a motion between times $t_1$ and $t_2$ defined by the dynamic variables $\vec A(\vec r,t)$ and an infinitesimally close motion between times $t_1'$ and $t_2'$ described by $A_p'(\vec r,t)$. The variation of the dynamic variables at a given point of time is $dA_p(\vec r,t)=A_p'(\vec r,t)-A_p(\vec r,t)$ and the variation of action can be written as:
$$
dS=S'-S=\int_{t_1'}^{t_2'}dt L(A_p')-\int_{t_1}^{t_2}dt L(A_p)
$$
$$
=\int_{t_1}^{t_2}dt[L(A_p')-L(A(p)]+\int_{t2}^{t2'}dt L(A_p')-\int_{t_1}^{t_1'}dt L(A'p)
$$
To first order the last 2 terms can be approximated by:
$$
\int_{t_2}^{t_2'}dtL(A_p')=L(A_p')|_{t_2}dt_2
$$ and similarly
$$
\int_{t_1}^{t_1'}dtL(A_p')=L(A_p')|_{t_1}dt_1
$$
For infinitesimally close motions, the first term:
$$
\int_{t_1}^{t_2}dt[L(A_p')-L(A(p)]
$$ 

$$
=\int_{t_1}^{t_2}dt\sum_p\int d^3r\Big\{\frac{\partial \mathcal{L}}{\partial A_p(\vec r,t)}dA_p(\vec r,t)+\frac{\partial\mathcal{L}}{\partial \dot A_p(\vec r,t)}d\dot A_p(\vec r,t)+\sum_q\Big(\frac{\partial \mathcal{L}}{\partial(\partial_qA_p(\vec r,t)}\Big)d\partial_qA_p(\vec r,t)\Big\}
$$
Similarly to the derivation of Lagrange's equation, we integrate by parts the second term
with respect to time and the third term with respect to $\vec r$, so the variation of the action
reduces to:<br></br>
$$
\int_{t_1}^{t_2}dt\sum_p\int d^3r\Big\{\frac{\partial \mathcal{L}}{\partial A_p(\vec r,t)}-\frac{d}{dt}\frac{\partial\mathcal{L}}{\partial \dot A_p(\vec r,t)}-\sum_q\partial_q\Big(\frac{\partial \mathcal{L}}{\partial(\partial_qA_p(\vec r,t)}\Big)d\partial_qA_p(\vec r,t)\Big\}A_p(\vec r,t)+\sum_p\int d^3r\frac{\partial \mathcal{L}}{\partial\dot A_p(\vec r,t)}dA_p(\vec r,t)|_{t_1}^{t_2}
$$

The integrand of the first term is a solution to the lagrange equation and reduces to zero. Hence,<br></br>
$$
dS=\sum_p\int d^3r\Big\{\frac{\partial \mathcal{L}}{\partial\dot A_p(\vec r,t)}dA_p(\vec r,t)|_{t_2}+L(A_p)_{t_2}dt_2-\frac{\partial L}{\partial \dot A_p(\vec r,t)}dA_p(\vec r,t)|_{t_1}-L(A_p)|_{t_1}dt_1\Big\}
$$

If a system has continuous symmetry, the corresponding action remains invariant with respect to infinitesimal displacements, i.e., $dS=0$ for any pair of times $t_1$ and $t_2$. Hence we obtain:<br></br>
$$
\frac{\partial \mathcal{L}}{\partial\dot A_p(\vec r,t)}dA_p(\vec r,t)|_{t_2}+L(A_p)_{t_2}dt_2-\frac{\partial L}{\partial \dot A_p(\vec r,t)}dA_p(\vec r,t)|_{t_1}-L(A_p)|_{t_1}dt_1=0
$$
or,
$$
\frac{\partial \mathcal{L}}{\partial\dot A_p(\vec r,t)}dA_p(\vec r,t)+L(A_p)_{t}dt=\text{constant}
$$


#### Using spatial and temporal translations

##Temporal and spatial translations

 
<p align = "justify"> 
First, let us assume that the variation is produced by an infinitesimal temporal displacement $d t$, such that $d t_{2}=d t_{1}=d t$. If the system is invariant with respect to temporal translations we can write $A_{p}^{\prime}(\vec{r}, t)=A_{p}(\vec{r}, t-d t) \simeq A_{p}(\vec{r}, t)-\dot{A}{p}(\vec{r}, t) d t$. Then, we have $d A{p}(\vec{r}, t)=-\dot{A}_{p}(\vec{r}, t) d t$. Substituting this result for dS and factoring out $d t$ we find that the conserved quantity is

$$
\begin{gathered}
\sum_{p} \int d^{3} r\left[-\frac{\partial \mathcal{L}}{\partial \dot{A}{p}(\vec{r}, t)} \dot{A}{p}(\vec{r}, t)+L\left(A_{p}\right)\right]  \\
=-\sum_{p} \int d^{3} r \frac{1}{2} \sum_{p}\left(\varepsilon(t) \dot{A}{p}^{2}(\vec{r}, t)+\frac{1}{\mu(t)}(\nabla \times \vec{A}(\vec{r}, t)){p}^{2}\right)=-H
\end{gathered}
$$
<p align = "justify">
Therefore, it is found that invariance with respect to temporal translations implies that the Hamiltonian must be a conserved quantity. In time-varying media, the system is not invariant under temporal translations, and, consequently, the Hamiltonian manifestly depends on time. As shown in the previous section, taking its time derivative explicitly shows that it is not a constant of motion.
<p align = "justify">
Second, we assume that the variation is produced by an infinitesimal spatial displacement $\boldsymbol{\eta}$. Then, if the system is invariant under spatial translations we must have $A_{p}^{\prime}(\vec{r}, t)=A_{p}(\vec{r}-\boldsymbol{\eta}, t) \simeq A_{p}(\vec{r}, t)-\boldsymbol{\eta} \cdot \nabla A_{p}(\vec{r}, t)$, so that $d A_{p}(\vec{r}, t)=-\boldsymbol{\eta} \cdot \nabla A_{p}(\vec{r}, t)$. Again substituting this result into (34) and factoring out $\boldsymbol{\eta}$ we find that the conserved quantity must be

$$
\mathbf{P}=\sum_{p} \int d^{3} r \frac{\partial \mathcal{L}}{\partial \dot{A}{p}(\vec{r}, t)} \nabla A{p}(\vec{r}, t)=-\sum_{p} \int d^{3} r \varepsilon(t) \dot{A}{p}(\vec{r}, t) \nabla A{p}(\vec{r}, t)
$$
<p align = "justify">
which equals the Minkowski momentum as shown before. Therefore, we finally found that the fact that time-varying media are invariant under spatial translations directly enforces that the Minkowski momentum is a conserved quantity.

can be recast as:

-\nabla^2\phi-\partial_t(\nabla.\vec A)=\frac{\rho}{\epsilon(t)}\text{ ...(1)}\\

b) Show that the Lagrangian density for EM waves is given by
$$
	\mathcal{L}=\frac{1}{2}[\epsilon(t)\vec E\cdot \vec E -\mu(t)\vec B\cdot \vec B]-\rho\phi+\vec j\cdot\vec A
$$
when applied on lagrange equations gives back equations (1) and (2)

Hint: Lagrange equation for a coordinate $\eta_i$ are given by $$

### Solution

The Maxwell's equations are:
$$
\begin{aligned}
\nabla\cdot \vec E=\frac{\rho}{\epsilon(t)}\\
\nabla\cdot \vec B=0\\
\nabla\times \vec E=-\partial_t\vec B\\
\nabla\times \vec B=\mu (t)\vec j+\mu(t)\epsilon(t)\partial_t\vec E
\end{aligned}
$$

From equation 3 we have:
$$
\begin{align}
\nabla\times\vec E+\partial_t(\nabla\times\vec A)=0 \\
\Rightarrow \nabla\times(\vec E+\partial_t \vec A)=0 \\
\end{align}
$$
Setting $\vec E+\partial_t\vec A=-\nabla\phi$
Hence, $\vec E=-\nabla\phi-\partial_t\vec A$
Now from equation (2) we get
$$
\begin{align}
\nabla\cdot(-\nabla\phi-\partial_tA)=\frac{\rho}{\epsilon(t)}\\
\Rightarrow -\nabla^2\phi-\partial_t(\nabla.\vec A)=\frac{\rho}{\epsilon(t)}
\end{align}
$$
Similarly from 4th maxwell equation

$$
\nabla\times(\nabla\times\vec A)=\mu(t)\vec j+\mu(t)\epsilon(t)\partial_t(-\nabla\phi-\partial_t\vec A)
$$

Using the vector identity $\nabla\times(\nabla\times\vec A)=\nabla(\nabla\cdot\vec A)-\nabla^2 \vec A$ and rearranging we obtain:

$$
\mu(t)\epsilon(t)\partial^2_t\vec A-\nabla^2 \vec A+\nabla(\nabla\cdot \vec A+\mu(t)\epsilon(t)\partial_t\phi)=\mu(t)\vec j
$$

Hence,

$$
\begin{aligned}
-\nabla^2\phi-\partial_t(\nabla.\vec A)=\frac{\rho}{\epsilon(t)}\text{ ...(1)}\\
\mu(t)\epsilon(t)\partial^2_t\vec A-\nabla^2 \vec A+\nabla(\nabla\cdot \vec A+\mu(t)\epsilon(t)\partial_t\phi)=\mu(t)j\text{ ...(2)}
\end{aligned}
$$

We will now show that the langrangian density 
$$
	\mathcal{L}=\frac{1}{2}[\epsilon(t)\vec E\cdot \vec E -\mu(t)\vec B\cdot \vec B]-\rho\phi+\vec j\cdot\vec A
$$
gives back the maxwell equations

To see this let us write the lagrange equation
$$
\partial_t(\frac{\partial \mathcal{L}}{\partial \dot\eta_i})+\nabla\cdot\frac{\partial\mathcal{L}}{\partial(\nabla\eta_i)}-\frac{\partial\mathcal{L}}{\partial \eta_i}=0
$$
where $\mathcal{L}$ is a function of the field coordinates $\eta=(A_1,A_2,A_3,\phi)$, their time derivatives and gradients.

Let us write the lagrangian density in terms of $\vec A$ and $\phi$:
$$
\begin{aligned}
\mathcal{L}=\frac{1}{2}[\epsilon(t)(-\nabla\phi-\partial_t\vec A)\cdot(-\nabla\phi-\partial_t\vec A)-\mu(t)(\nabla\times A)\cdot(\nabla\times A)]-\rho\phi+\vec j.\vec A\\
=\frac{1}{2}\Big(\epsilon(t)\Big [||\nabla\phi||^2+||(\partial_t\vec A)||^2+2\nabla\phi\cdot\partial_t\vec A\Big ]-\mu(t)\sum_{i=1}^3\Big[||\nabla A_i||^2-\frac{\partial \vec A}{\partial x_i}\cdot \nabla A_i\Big]\Big)-\rho\phi+\vec j\cdot\vec A\\
\end{aligned}
$$
Writing the lagrange equation for $\eta_i=\phi$ we get
$$
\begin{aligned}
\partial_t(\frac{\partial \mathcal{L}}{\partial \dot\phi})+\nabla\cdot\frac{\partial\mathcal{L}}{\partial(\nabla\phi)}-\frac{\partial\mathcal{L}}{\partial \phi}=0\\
\Rightarrow \partial_t(0)+\nabla\cdot[\frac{1}{2}(2\epsilon(t)\nabla\phi+2\partial_t\vec A)]-\rho=0\\
\Rightarrow \epsilon(t)[\nabla^2\phi+\partial_t(\nabla.A)]-\rho=0\\
\Rightarrow \nabla^2\phi+\partial_t(\nabla.A)-\frac{\rho}{\epsilon(t)}=0
\end{aligned}
$$
Thus recovering equation (1) back
Now, writing the lagrange equation for $\eta_i=A_1$
$$
\begin{aligned}
\partial_t(\frac{\partial \mathcal{L}}{\partial \dot A_1})+\nabla\cdot\frac{\partial\mathcal{L}}{\partial(\nabla A_1)}-\frac{\partial\mathcal{L}}{\partial A_1}=0\\
\Rightarrow \partial_t(\frac{1}{2}\epsilon(t)[2\partial_tA_1+2\nabla\phi])+\nabla\cdot(\mu(t)[2\nabla A_1-\frac{\partial\vec A_1}{\partial x_1}])-j_1=0\\
\Rightarrow \epsilon(t)\partial_t^2A_1+\nabla(\epsilon(t)\nabla\phi+\frac{\partial A_1}{\partial x_1})=j_1\\
\Rightarrow\mu(t)\epsilon(t)\partial_t^2A_1+\nabla(\mu(t)\epsilon(t)\nabla\phi+\frac{\partial A_1}{\partial x_1})=j_1\mu(t)\\
\end{aligned}
$$
which is the 1st component of equation (2)
Following the same procedure for $\eta_2$ and $\eta_3$ yields equation (2)
Hence, we have obtained the maxwell equations from the lagrange equation using the above lagrangian.

## Conclusion
<p align = "justify">
Symmetries play a crucial role in understanding the properties of physical systems, including those in electromagnetism. In electromagnetism, symmetries refer to the invariance of the laws of electromagnetism under transformations of space and time.
<p align = "justify">
Spatial symmetries refer to the invariance of the laws of electromagnetism under transformations of space, such as translation, rotation, and reflection. For example, if we rotate a system of charges or currents, the electromagnetic fields at any point in space will be unaffected. Similarly, if we translate a system of charges or currents, the electromagnetic fields will remain the same. These symmetries make it easier to solve problems and identify conserved quantities, such as the total electric charge or magnetic moment.
<p align = "justify">
Temporal symmetries refer to the invariance of the laws of electromagnetism under transformations of time, such as time reversal. However, in most traditional photonics systems, temporal symmetries are conserved. Time-varying media and temporal metamaterials break temporal symmetries, which provides a fresh perspective on electromagnetic theory. For example, in time-varying media, the electric and magnetic fields can depend explicitly on time, and the electromagnetic waves can propagate with varying velocities.
<p align = "justify">
The connection between spatial and temporal symmetries and the properties of time-varying media can be analyzed through the point of view of Lagrangian mechanics. Lagrangian mechanics is a formalism used to describe the motion of particles and systems, and it is based on the principle of least action. The Lagrangian is a function that describes the difference between the kinetic and potential energies of a system, and it is used to derive the equations of motion for the system. By analyzing the Lagrangian, we can identify the symmetries of the system and their consequences.
<p align = "justify">
The conservation of electromagnetic momentum in time-varying media is a subtle issue, and it is related to the debate between the Abraham and Minkowski forms of the electromagnetic momentum. The Abraham form of momentum is based on the momentum of the electromagnetic fields, while the Minkowski form includes both the momentum of the fields and the momentum of the material medium. In time-varying media, the momentum of the fields can be different from the momentum of the medium, leading to different predictions for the motion of objects in the field.
<p align = "justify">
In summary, the role of symmetries in understanding the properties of physical systems, including time-varying media in electromagnetism, cannot be overstated. By analyzing the symmetries of a system through the point of view of Lagrangian mechanics, we can identify conserved quantities, solve problems more easily, and gain insights into the behavior of physical systems.
