# 1. Dirac's Argument for Magnetic monopole

## Aharanov-Bohm Experiment

slit.png

Consider the situation above, which shows the classic electron double slit experiment with a solenoid in the picture. It is impenetrable and carries a flux $\phi$ into the page. Outside the solenoid, $B = 0$ but $A$ is still present. The Lagrangian of a particle of mass $m$ and charge $e$ in the presence of a magnetic field can be given by 

$$
\mathcal{L}(\boldsymbol{r}, \boldsymbol{v})=\frac{1}{2} m v^2+e \boldsymbol{v} \cdot \boldsymbol{A}
$$

One way to characterise the effect of the solenoid on the electrons is to use Feynman's path integral approach, where each path is weighted by a factor of

$$
\exp\left(\frac{i}{\hbar} \int \mathcal{L}dt\right)
$$

Suppose that without the $A$, at any point $r$ on the screen, 

$$
\psi(\boldsymbol{r})=\psi_{P 1}(\boldsymbol{r})+\psi_{P 2}(\boldsymbol{r})
$$

Now, because of the spatially varying $e \boldsymbol{v} \cdot \boldsymbol{A}$ term in $\mathcal{L}$, the contributions of the two paths P1 and P2 will be modified to yield

$$
\begin{aligned}
\psi(\boldsymbol{r}) & =\psi_{P 1}(\boldsymbol{r}) \exp \left[\frac{i e}{\hbar} \int_{P 1} \boldsymbol{A} \cdot d \boldsymbol{r}\right]+\psi_{P 2}(\boldsymbol{r}) \exp \left[\frac{i e}{\hbar} \int_{P 2} \boldsymbol{A} \cdot d \boldsymbol{r}\right] \\
| \psi(\boldsymbol{r})|^2 & =\left|\exp \left[\frac{i e}{\hbar} \oint_{C=P 1-P 2} \boldsymbol{A} \cdot d \boldsymbol{r}\right] \psi_{P 1}+\psi_{P 2}\right|^2 
\end{aligned}
$$

Thus, the regular phase difference at $r$ is now supplemented by the extra contribution due to the path integral along the closed loop that includes P1 and the reverse of P2. This contribution is **unobservable** if the flux through the loop $\phi = \oint \boldsymbol{A}\cdot d \boldsymbol{l}$ satisfies the condition $ \phi = m\frac{2 \pi \hbar}{e} = m \Phi_0$, where $\Phi_0$ is known as the flux quantum. Consequently, the solenoid itself would be unobservable if this is satisfied, and this specific experiment would not be able to detect it.

## Dirac's argument

Let there be a magnetic monopole $g$, with a particle moving around it. The magnetic field will be given by,

$$
\boldsymbol{B}=\frac{g \boldsymbol{e}_r}{r^2}
$$

The total flux emitted by this monopole will be $4\pi g$. We will now show that this $B$ cannot be described by a non-singular $A$. Consider a sphere around $g$, and a closed loop $C$ around the north pole along one of the latitudes of the sphere. By Stoke's theorem,

$$
\oint_C \boldsymbol{A} \cdot d \boldsymbol{r}=\Phi_{e n c}
$$

As C is lowered and reaches the equator, it expands and $\Phi_{e n c}$ increases. On lowering more, C starts to shrink. As we reach the south pole, C is infinitesimally small but $\Phi_{e n c}$ is now $4 \pi g$. For an infinitesimally small loop to enclose a finite flux, $A$ **must have a singularity at that point**. 

image.png

It can be shown that the position of this singularity is a function of the gauge choice. Hence, the physical cause of this phenomenon must be unobservable. Dirac theorized that an infinitesimally thin *Dirac string* was bringing the $4 \pi g$ flux from infinity and releasing it at the origin in a spherically symmetrical way. By extension, this Dirac string must be unobservable too, and it must be so everywhere. As we saw above, a particle going around it would gain the Aharanov-Bohm phase factor

$$
\exp \left[\frac{i e \Phi}{\hbar}\right]=\exp \left[\frac{4 \pi i e g}{\hbar}\right]
$$

Dirac argued that this must be unobservable, yielding the condition

$$
\frac{4 \pi i e g}{\hbar} = 2\pi m
$$

for $m = 0, \pm 1, \pm 2$. Rearranging, we get

$$
e g=\frac{1}{2} m \hbar
$$

This result is special because it implies the presence of even a single magnetic monopole in the Universe forces the electric charge of particles to be integral multiples of $\frac{1}{2g} m \hbar$.

#2. The Quantum Hall Effect

We enter into a digression to understand topological isnulators(TI), and why they are interesting for us. We'll motivate the magneto-electric effect in 3D TI materials, starting from the Quantum Hall Effect.


image.png

Consider 2D electron gas, acting like a classical Hallbar, with magnetic field $B\hat z$. We consider the conductor to be infinite in x direction (later we will impose the periodic boundary condition in x direction) but having a finite width in the y direction. We have not yet applied $\vec E$. We apply a potential $U(y)$ in the y direction. We'll solve for the eigenstates and eigenvalues for this system, and see how we can relate it to the classical Hall Effect.


The Schrodinger's equation can be written via the minimal coupling Hamiltonian:
\begin{equation}
\left[E_s + \frac{(i\hbar\nabla + eA)^2}{2m} + U(y)\right]\Psi(x,y) = E \Psi(x,y)
\end{equation}

with $E_s$ being the reference ground state energy

We work in the gauge of $\vec A =\hat x By $
Now we can rewrite the Hamiltonian as
\begin{equation}
\left[E_s + \frac{(\hat p_x+eBy)^2}{2m} + \frac{\hat p_y^2}{2m} + U(y)\right]\Psi(x,y) = E \Psi(x,y)
\end{equation}



Since the potential in $x$ direction is essentially that of a free particle, we can use variable separation to write $\Psi(x,y) = \frac{1}{\sqrt{L}}e^{ikx} \chi(y)$

Eliminating $e^{ikx}$, the equation for $\chi(y)$ becomes

\begin{equation}
\left[E_s + \frac{(\hbar k + e B y)^2}{2m} + \frac{p_y^2}{2m} + U(y)\right]\chi(y) = E \chi(y)
\end{equation}

Now, the potential $U(y)$ is applied by having positive voltage on the lateral electrodes, such that the the  potential is higher towards the sides and lower in the center.
 Approximate $U(y) = \frac{1}{2}m\omega_{o}^2 y^2$
 

Now to simply this , define $\omega_c= \frac{|e|B}{m}$, $y_k=\frac{\hbar k}{eB}$, $\omega_{co}^2 =\omega_{c}^2 + \omega_{o}^2$

\begin{equation}
\left[\underbrace{\left(E_s + \frac{1}{2}\frac{m\omega_o^2\omega_c^2}{\omega_{co}^2}y_k^2 \right)}_{\text{constant for a k}}+ \frac{p_y^2}{2m} + \underbrace{\frac{1}{2}m\omega_{co}^2\left(y+\frac{\omega_o^2}{\omega_{co}}y_k\right)^2}_{\frac{1}{2}m\omega_{o}\tilde y^2} \right]\chi(y) = E  \chi(y)  
\end{equation}

This looks exactly like Schrodinger's equation for a quantum harmonic oscillator, with the potential shifetd by $y_k$. So for each k, we will have levels like that of a QHO. The eigenvalues will hence have contributions from the "free electron" part of the wavefunction (in $x$ direction) and the QHO part (along $y$ direction)

\begin{equation}
E_n = E_s +(n + \frac{1}{2}) \hbar \omega_{co} + \frac{\hbar^2}{2m}\frac{\omega_o^2}{\omega_{co}^2}k^2
\end{equation}

 and the $\chi(y)$ are the eigenfunctions of the QHO Hamiltonian, now centred around different y coordinates 

 \begin{equation}
  \chi_{n,k}(y) = u_n(\sqrt{m\omega_o/\hbar} y + \frac{\omega_{c}^2}{\omega_{co}^2}\sqrt{m\omega_o/\hbar} y_k)
 \end{equation}

 where $u_n$ are the eigenfunctions of QHO Hamiltonian

The group velocity at a particular $k$ is given by
\begin{equation}
v(k)= \frac{1}{\hbar}\frac{\partial{E(k)}}{ \partial{k}} = \frac{\hbar}{m}\frac{\omega_{o}^2}{\omega_{co}^2}k
\end{equation}


Each $k$ has transverese modes centred along different y coordinates, dependent on $y_k$. From above equation, we see that
\begin{equation}
y_k = \frac{\hbar k}{e B}  \propto v(k)
\end{equation}

### States with different velocities are **spatially separated**

# Quantizing Conductance

Now suppose there are two contacts, in $x$ direction. They are maintained at fermi levels $\mu_{1}$ and $\mu_{2}$ respectively.

The $-k$ states, originating on right contact are in equilibrium with it, so left right propagating bands upto energy  $\mu_{2}$ are filled. Similarly,$+k$ states are in equilibrium with left contact and right propagating states upto energy $\mu_{1}$ are filled. The currents in bulk cancel out. Say if $\mu_{1}>\mu_{2}$, only those k states that are not compenated by $-k$ states contribute to current. These are the excess parts of the bands that lie between $\mu_{1}$ and $\mu_{2}$ in the band structure. 

QHE2.PNG

Each band acts like a "mode" or "channel", with only these relevant parts contributing to current. Each mode, quantum mechanically has conductance $G = 2 e^2 /h$.
So if M modes lie between $\mu_{1}$ and $\mu_{2}$, the conductance is $2Me^2/h$

```python
e = 1
m = 1
hbar = 1
Es = 1
```

```python
@dataclass
class resultContainer:
    k: np.array
    E: np.array
    minima: float
    wc: float
    wco: float
    wo: float
    n: int
```

```python
def En(n, B, wo, k=np.linspace(-10, 10, 1000)):
    wc = e*B/m
    wco = ( wc**2 + wo**2 )**0.5
    E = Es + (n+0.5)*hbar*wco + hbar**2 * k**2 / (2*m) * wo**2 / wco**2
    return resultContainer(
        k, E, Es + (n+0.5)*hbar*wco, wc, wco, wo, n
    )
```

```python
def get_energies(Nstart, Nend, B, wo, k):
    return [
        En(n, B, wo, k) for n in range(Nstart, Nend)
    ]
```

```python
def plot_and_count(Nstart, Nend, Binit, wo_init, L1_init, L2_init):
    from matplotlib.widgets import Slider
    k = np.linspace(-10, 10, 1000)
    fig, _ax = plt.subplots(ncols=2)
    ax = _ax[0]
    ax1 = _ax[1]
    fig.subplots_adjust(left=0.15, bottom=0.25)
    ax.set_ylim(0.0, 10)
    ax.set_xlim(-10, 10)
    ax1.set_ylim(-25, 25)
    ax1.set_xlim(-1, 1)
    results = get_energies(Nstart, Nend, Binit, wo_init, k)
    lines = [
        ax.plot(
            res.k, res.E, 'blue'
        )[0] for res in results
    ]

    def count_minimas(results, L1, L2):
        count = 0
        for result in results:
            if L1 < result.minima < L2:
                count += 1
        return count


    def calc_intersections(results: list[resultContainer], L1, L2, B):
        intersections = []
        for result in results:
            yk1_sq = ((L1 - Es - (result.n + 0.5) * hbar * result.wco)
                * 2*m * result.wco**2 / (hbar**2 * result.wo**2)
            )
            yk2_sq = ((L2 - Es - (result.n + 0.5) * hbar * result.wco)
                * 2*m * result.wco**2 / (hbar**2 * result.wo**2)
            )
            if yk1_sq > 0:
                k = -yk1_sq**0.5
            if yk2_sq > 0:
                k = yk2_sq**0.5
            intersections.append(
                result.wc**2 / result.wco**(1.5) * k/B
            )
        return intersections

    L1, = ax.plot([-15, 0], [L1_init, L1_init], '--')
    L2, = ax.plot([-0, 15], [L2_init, L2_init], '--')
    counts = ax.text(8, 9, f'$ G = ({count_minimas(results, L1_init, L2_init)}) * 2e^2/h$')

    intersection_lines = [
        ax1.plot(
            [-1, 1], [-100, -100]
        )[0] for i in range(Nstart*2, Nend*2)
    ]
    for line, ii in zip(intersection_lines, calc_intersections(results, L1_init, L2_init, Binit)):
        line.set_ydata([ii, ii])

    ax_B = fig.add_axes([0.25, 0.1, 0.65, 0.03])
    ax_wo = fig.add_axes([0.25, 0.15, 0.65, 0.03])
    B_slider = Slider(
        ax = ax_B,
        label='Magnetic Field',
        valmin=0.,
        valmax=10,
        valinit=Binit
    )
    wo_slider = Slider(
        ax = ax_wo,
        label=r'$\omega_o$',
        valmin=0.,
        valmax=10,
        valinit=Binit
    )
    
    ax_L1 = fig.add_axes([0.04, 0.25, 0.0225, 0.63])
    L1_slider = Slider(
        ax=ax_L1,
        label=r"$\mu_1$",
        valmin=0,
        valmax=10,
        valinit=L1_init,
        orientation="vertical"
    )

    ax_L2 = fig.add_axes([0.08, 0.25, 0.0225, 0.63])
    L2_slider = Slider(
        ax=ax_L2,
        label=r"$\mu_2$",
        valmin=0,
        valmax=10,
        valinit=L2_init,
        orientation="vertical"
    )


    def update(_):
        B = B_slider.val
        wo = wo_slider.val
        results = get_energies(Nstart, Nend, B, wo, k)
        for line, result in zip(lines, results):
            line.set_ydata(result.E)
        L1_v, L2_v = L1_slider.val, L2_slider.val
        L1.set_ydata([L1_v, L1_v])
        L2.set_ydata([L2_v, L2_v])
        if L1_v < L2_v:
            c = count_minimas(results, L1_v, L2_v)
        else:
            c = count_minimas(results, L2_v, L1_v)
        counts.set_text(f'$ G = ({c}) * 2e^2/h$')
        for line in intersection_lines:
            line.set_ydata([-100, -100])
        for line, ii in zip(intersection_lines, calc_intersections(results, L1_v, L2_v, B)):
            line.set_ydata([ii, ii])    
    
    B_slider.on_changed(update)
    wo_slider.on_changed(update)
    L1_slider.on_changed(update)
    L2_slider.on_changed(update)

    return B_slider, wo_slider, L1_slider, L2_slider

if not _RUNNING_COLAB:
  %matplotlib qt5 
  # pip install pyqt5
  _=plot_and_count(0, 100, 1, 1, 2, 5)
else:
  print('+--------------------------------------------+')
  print('Cannot run this on colab...')
  print('Download this file and run it locally')
  print('+--------------------------------------------+')
```

# Where do Topological Materials come into picture?

In the above setup, we saw that the response of the material to a magnetic field was quantized. But it required us to have the lateral volatge $U(y)$ and the nodes being maintained at different chemical potentials. Now there exist materials in which the phenomenon of "edge states", i.e., conduction at the boundaries and no net current within , does not require the presence of magnetic field. 

For example, consider the 1D kitaev chain.
A kitaev chain is a 1D bravais lattice, with a 2 point basis. (2 atoms corresponding to one lattice site). We can model it using a tight binding Hamiltonian for it solve it exactly to get the disersion relation. We can have $+k$ states and $-k$ Bloch functions supported this chain.

kitaev.PNG


 Now consider stacking these chains to get a 2D array. We can model the stacking by selectively choosing to couple the +k states of one chain with the -k state of the other chain, with some coupling parameter. Now this coupling scheme leaves the $+k$ states at one edge and $-k$ states at the other edge uncoupled. This is very similar to the case of QHE. The bulk states, on a net, don't conduct current, but the edge states can conduct!
We have a special type of insulator. This is called the "Chern Insulator". Like in QHE, the conductance is quantized, but instead of being quantized  in multiples of $2e^2/h$, it is now quantized in integer multiples of the **Chern Number**, a property of the band structure of a material.



This is the **quantum anomalous hall effect**. 

It is as if "bulk" supports the existence of the "edge" or "boundary" states. This is the bulk-boundary correspondence. The most trivial anology is that of bar magnets.


*We must appreciate that a phenomenon that couldn't occur in a lower dimension, could occur in a higher dimension, just by putting together lower dimensional elements*


It turns out that when modelling a system by designing its Hamiltonians, we can endow the Hamiltonian with certain symmetries. For example, if we envolve spins in our picture, we arrive at the quantum spin Hall effect, wherein instead of 1 states at each edge, two states a $+k$ and $-k$ states exist, each with different spin.

# $\theta$- Electrodynamics

The surface phenomena of TI materials can be incorporated into our clasical electromagnetic formalism by treating them as bulk materials, with magento-electric polarizabilities. 


## Modifying the Langrangian
The Langrangian density which is normally given as $\mathcal{L} = \frac{1}{2}\epsilon E^2 -\frac{1}{2\mu}B^2$
doesn't have "surface terms".
We want to have terms that when integrated over volume would give surface integrals, so we need a complete derivative(divergence term).
It turns out that to account for topological surface effects, we need to modify:
\begin{equation}
 \mathcal{L}_{\theta}  =\frac{1}{2}\epsilon E^2 -\frac{1}{2\mu}B^2 + 2\alpha\sqrt\frac{\epsilon_0}{\mu_0}\frac{\theta}{2\pi} \vec E \cdot \vec B
\end{equation}

Since a volume intergral would give a term of the form 

$\int dV   \textbf E \cdot \textbf B   = \int dV  \textbf E \cdot (\vec\nabla \times \textbf A) =    \int d V ( -\vec \nabla \cdot (\textbf E \times \textbf B) + (\vec \nabla \times \textbf E )\cdot \textbf B   ) =   \oint_{surface}  (\textbf E \times \textbf B) .d\textbf a$

(In the static case, when $\frac{\partial }{\partial t}\textbf B =0 $)

This motivates the new term. We can arrive at the Maxwell's equations using this Langrangian, but instead we use an alternative approach.

Solving the Lagrangian, we get the following equations for electric and magnetic field:
$$
\boldsymbol{\nabla} \cdot \textbf{E} = \frac{\rho}{\epsilon_0} - 2c\alpha\boldsymbol{\nabla}\left(\frac{\theta}{2\pi}\right)\cdot\textbf{B}
$$
$$
\boldsymbol{\nabla} \times \textbf{B} = \mu_{0}\textbf{J} + \frac{1}{c^2}\frac{\partial{\textbf{E}}}{\partial t} + \frac{2\alpha}{c}\left[\textbf{B}\frac{\partial}{\partial t}\left(\frac{\theta}{2\pi}\right) + \boldsymbol{\nabla}\left(\frac{\theta}{2\pi}\right)\times\textbf{E}\right]
$$

The electric and magnetic fields are modified due to the magneto-electric effect, which we will explore in the subsequent sections.

## The Maxwell's Equations

We write the standard Maxwell's equation in atomic units as follows:

$$
\begin{equation}
\nabla \cdot \textbf{D}= 4\pi\rho 
\end{equation}
$$
$$
\begin{equation}
\boldsymbol{\nabla} \cdot \textbf{B}= 0 
\end{equation}
$$
$$
\begin{equation}
\boldsymbol{\nabla} \times \textbf{E}= - \frac{1}{c} \frac{\partial \textbf{B}}{\partial t} 
\end{equation}
$$
$$
\begin{equation}
\boldsymbol{\nabla} \times \textbf{H}=  \frac{1}{c}\frac{\partial \textbf{D} }{\partial t} 
\end{equation}
$$

Now, we need to modify the constituent definitions of $\textbf H$ and $\textbf B$ to include the magneto-electric terms

$\textbf D = \epsilon \textbf E - \theta \alpha \frac{1}{\pi} \textbf{B}$

and

$\textbf H = \frac{1}{\mu}\textbf{B} + \frac{\theta}{\pi}\alpha\textbf{E}$

where $\alpha = \frac{1}{137}$ is the fine-structure constant and $\theta$ is a phenomenological parameter that arises from Landau-Ginzburg theory. Topological insulators have $\theta = \pi$.

Using the Lagrangian obtained above, Euler-Lagrange variational principles lead to the Maxwell equations as:
$$
\begin{equation}
\boldsymbol{\nabla} \cdot \textbf{D}= 4\pi\rho 
\end{equation}
$$
$$
\begin{equation}
\boldsymbol{\nabla} \times \textbf{E}= - \frac{1}{c} \frac{\partial \textbf{B}}{\partial t} 
\end{equation}
$$
$$
\begin{equation}
\boldsymbol{\nabla} \cdot \textbf{B}= 0
\end{equation}
$$
$$
\begin{equation}
\boldsymbol{\nabla} \times \textbf{H}=  \frac{1}{c}\frac{\partial \textbf{D} }{\partial t} + \frac{4\pi}{c}\textbf{J}
\end{equation}
$$

Substituting the constitutive relation for $\textbf{D}$ in the first maxwell equation above, we get:
$\boldsymbol{\nabla} \cdot (\epsilon\textbf{E}) = 4\pi\rho + \frac{\alpha}{\pi} \boldsymbol{\nabla} \cdot (\theta\textbf{B}) = 4\pi\rho + \frac{\alpha\boldsymbol{\nabla}\theta}{\pi}\textbf{B}$


Next, we substitute the constitutive relations for $\textbf{D}$ and $\textbf{H}$ in the fourth Maxwell equation to obtain:

$$
\boldsymbol{\nabla} \times (\frac{\textbf{B}}{\mu})  - \frac{1}{c} \frac{\partial(\epsilon\textbf{E})}{\partial{t}} = \frac{4\pi}{c}\textbf{J} - \frac{1}{c}\frac{\alpha}{\pi}\frac{\partial (\theta\textbf{B} )}{\partial{t}} - \frac{\alpha}{\pi} \boldsymbol{\nabla} \times (\theta\textbf{E})
$$

We carry out the derivatives in the last two terms and simplify as follows:

\begin{aligned}
- \frac{1}{c}\frac{\alpha}{\pi}\frac{\partial (\theta\textbf{B} )}{\partial{t}} - \frac{\alpha}{\pi} \boldsymbol{\nabla} \times (\theta\textbf{E}) &= -\frac{1}{c}\frac{\alpha}{\pi}\left( \frac{\partial{\theta}}{\partial{t}}\textbf{B} + \theta\frac{\partial{\textbf{B}}}{\partial{t}}\right) - \frac{\alpha}{\pi}\left( \boldsymbol{\nabla} \theta \times \textbf{E} + \theta \boldsymbol{\nabla} \times {\textbf{E}} \right) \\
&= -\frac{\alpha}{\pi}\theta \left(\frac{1}{c}\frac{\partial{\textbf{B}}}{\partial{t}} + \boldsymbol{\nabla} \times \textbf{E}\right) - \frac{\alpha}{\pi}\left(\frac{1}{c}\frac{\partial{\theta}}{\partial{t}}\textbf{B} + \boldsymbol{\nabla}\theta \times \textbf{E} \right)
\end{aligned}

The first term here is zero from Maxwell's second equation <br>
$$
\boldsymbol{\nabla} \times \left(\frac{\textbf{B}}{\mu}\right)  - \frac{1}{c} \frac{\partial(\epsilon\textbf{E})}{\partial{t}} = \frac{4\pi}{c}\textbf{J} - \frac{1}{c}\frac{\alpha}{\pi}\frac{\partial\theta}{\partial{t}} \textbf{B} - \frac{\alpha}{\pi}\boldsymbol{\nabla}\theta \times \textbf{E}
$$

From the above equations, we can define an effective charge $\rho_{\theta}$ and effective current density $\textbf{J}_{\theta}$ as follows:

$$
4\pi\rho_{\theta} = \frac{\alpha \boldsymbol{\nabla} \theta}{\pi} \cdot \textbf{B} \implies \rho_{\theta} = \frac{\alpha}{4\pi^2}\boldsymbol{\nabla}{\theta}\cdot{\textbf{B}}
$$

$$
\frac{4\pi}{c}\textbf{J}_{\theta} = \frac{\alpha}{\pi} \left( -\frac{1}{c}\frac{\partial{\theta}}{\partial t}\textbf{B} - \boldsymbol{\nabla}\theta \times \textbf{E}\right) \implies \textbf{J}_{\theta} = \frac{\alpha}{4\pi^2} \left( -\frac{\partial{\theta}}{\partial t}\textbf{B} - c\boldsymbol{\nabla}\theta \times \textbf{E}\right)
$$

We recover the original Maxwell equations when $\theta = 0$. If $\theta$ is time and coordinates independent then the effective charge and current density defined above go to zero, giving back Maxwell's equations.

When we move from a topologically trivial insulator with $\theta = 0$ to $\theta = \pi$ for time reversal topological insulators, $\theta$ behaves as a piecewise constant function. Therefore the gradient of $\theta$ will be a Dirac delta function at the point of transition, $\boldsymbol{\nabla}{\theta} = \pi\delta(z_0)\textbf{n}$, where $\textbf{n}$ is the unit vector in the $z$ direction. Since $\theta$ is time-independent in this case, we get

$$
\textbf{J}_{\theta} = -\frac{c\alpha}{4\pi}\delta({z_0})\textbf{n} \times \textbf{E}
$$

$$
\rho_{\theta} = \frac{\alpha}{4\pi}\delta(z_0)\textbf{n}\cdot\textbf{B}
$$

# The parameter $\theta$

# Boundary Conditions at interface of a topological insulator

Consider the an interface bewteen two surfaces











For a pill box in the first image , we follow the standard procedure for arriving at the boundary conditions. We are in the static limit with all time derivates set to 0 and no free charge or free current. 


**The Perpendicular Components:**


Since we have no free charge at the interface, we can use, $\boldsymbol\nabla \cdot \textbf D =0$, we integrate over the cylindrical pill box volume and with the height being infinitesimally small. Using Gauss theorem, we can relate the volume integral to surface integral over the circular area to get

$(\textbf D1 - \textbf D2)\cdot \hat  n = 0$ 

This gives, using our definition of $\textbf D$,

\begin{equation}
(\epsilon_1 \textbf E_1 - \epsilon_2 \textbf E_2)\cdot \hat n =  \frac{\alpha}{\pi} (\theta_1 \textbf B_1 -\theta_2 \textbf B_2)
\end{equation}


Similarly, since $\nabla \cdot \textbf B =0$, we can use the same cylindrical pillbox to arrive at 

\begin{equation}
\textbf B_1 \cdot \hat n = \textbf B_2 \cdot \hat n
\end{equation}




**The Parallel components**

Consider the an area perpendicular to the interface. Since we are in the static case, we can use $\nabla \times \textbf E =0$, to perform an integral over the surface pof this area. Using Stoke's theorem, we can relate this to the line integral over the boundary of the surface. In the limit that the side of the area perpendicular to the interface goes to 0, we have 
$$
\begin{equation}
(\textbf E_1 -\textbf E_2)_∥ =0
\end{equation}
$$
Now , the free surface current on the interface is 0 (the  $J_\theta$ that we had definde was just an artefact of our definition of $\textbf H$)

So similarly we can use $\nabla \times \textbf H = 0$ and the definition of $\textbf H$ to arrive at 
$$
\left(\frac{1}{\mu_1} \mathbf{B}1-\frac{1}{\mu_2} \mathbf{B}\mathbf{2}\right)_{\|}=-\frac{1}{\pi} \alpha\left(\theta_1 \mathbf{E}_1-\theta_2 \mathbf{E}_2\right)
$$




We will use these boundary conditions to arrive at the fields of the monopole near the interface.

# Image charge and monopole for a point electric charge outside a sphere
Consider a spherical topological insulator of radius $a$, centered at the origin. A magneto-electric polarization $P_3$ is centered at the origin. A point charge $q$ is placed at $(0,0,d)$, with $d>a$. The dielectric constant and magnetic permeability are $\epsilon_1$, $\mu_1$ inside the sphere, and $\epsilon_2$, $\mu_2$ inside the sphere.


It can be shown that the electric/magnetic fields inside and outside the sphere can be viewd as inducted by a point electric charge/magnetic monopole plus a line of image electric/magnetic charge. Moreover, inside the sphere the total free electric and magnetic charge vanish.

One can show that when the seperation between the electric charge and the surface is much smaller than the radius of the sphere, the electric/magnetic field is dominated by the point electric charge/magnetic monopole and contribution due to the line of image electric/magnetic charge density is vanishingly small.

We derive the electric and magnetic fields produced when a point charge is placed close to a spherical topological insulator.

Screenshot from 2023-03-31 22-54-00.png

As shown in the figure, a Topological Insulator (TI) of radius $a$ is centered at the origin and a point charge ${q_1}$ is placed at $(0,0,d)$ with $d>a$. The medium outside the TI has dielectic constant $\epsilon_1$ and magnetic permeability $\mu_1$ while the TI has dielectric constant $\epsilon_1$, permeability $\mu_2$ and magneto-electric constant $P_3$.

In the regions $0 < r < a$ and $a < r < d$ the curls and divergances of the electric fields are both zero. Therefore the fields can be expressed as the gradient of a scalar potential in the two regions.
$$
        E^{(i)} = -\nabla \Phi_E^{(i)}
$$
$$
        M^{(i)} = -\nabla \Phi_M^{(i)}
$$
where $i=2,1$ refer to the regions $0 < r < a$ and $a < r < d$ respectively.

Notice that the problem has azimuthat symmetry. Thus, the solution can be expanded in terms of Legendre polynomials. The electric and magnetic potentials satisfy the Laplace's equations and can be expressed as follows:

```python
import sympy
from matplotlib import pyplot as plt
import numpy as np
```

```python
q, r, d, a, l, theta = sympy.symbols(r'q, r, d, a, l, \theta')
eps1, eps2, mu1, mu2, alpha = sympy.symbols(r'\epsilon_1, \epsilon_2, \mu_1, \mu_2, \alpha')
Al, Bl, Cl, Dl = sympy.symbols('A_l, B_l, C_l, D_l')
P3 = sympy.symbols('P3')
PhiE1_l, PhiE2_l, PhiM1_l, PhiM2_l = sympy.symbols(r'\Phi_E^{(1)}, \Phi_E^{(2)}, \Phi_M^{(1)}, \Phi_M^{(2)}')
PhiE1_la, PhiE2_la, PhiM1_la, PhiM2_la = sympy.symbols(r'\Phi_E^{(1)}(a), \Phi_E^{(2)}(a), \Phi_M^{(1)}(a), \Phi_M^{(2)}(a)')

def P_l_costheta(l):
    return sympy.legendre(l, sympy.cos(theta))
```

```python
# Electric potentials
phi_1_E_l = (q/eps1 * (
    r**l / d ** (l+1) * P_l_costheta(l)) + 
    Al * (a/r)**(l+1) * P_l_costheta(l))
phi_2_E_l = Bl * (r/a) ** l * P_l_costheta(l)

# magnetic potentials
phi_1_M_l = Cl * (a/r) ** (l+1) * P_l_costheta(l)
phi_2_M_l = Dl * (r/a) ** l * P_l_costheta(l)
```

```python
sympy.Eq(
    PhiE1_l, sympy.Sum(phi_1_E_l, (l, 1, sympy.oo))
)
```

```python
sympy.Eq(
    PhiE2_l, sympy.Sum(phi_2_E_l, (l, 1, sympy.oo))
)
```

```python
sympy.Eq(
    PhiM1_l, sympy.Sum(phi_1_M_l, (l, 1, sympy.oo))
)
```

```python
sympy.Eq(
    PhiM2_l, sympy.Sum(phi_2_M_l, (l, 1, sympy.oo))
)
```

The boundary conditions will be:

```python
dphi1M_dr     = phi_1_M_l.diff(r).subs(r, a)
dphi1M_dtheta = phi_1_M_l.diff(theta).subs(r, a)

dphi2M_dr     = phi_2_M_l.diff(r).subs(r, a)
dphi2M_dtheta = phi_2_M_l.diff(theta).subs(r, a)

dphi1E_dr     = phi_1_E_l.diff(r).subs(r, a)
dphi1E_dtheta = phi_1_E_l.diff(theta).subs(r, a)

dphi2E_dr     = phi_2_E_l.diff(r).subs(r, a)
dphi2E_dtheta = phi_2_E_l.diff(theta).subs(r, a)
```

```python
eq1 = sympy.Eq(
    eps1 * dphi1E_dr,
    eps2 * dphi2E_dr - 2*alpha*P3*eps1 * dphi2M_dr
)
sympy.Eq(
    eps1 * sympy.Derivative(PhiE1_la, r),
    eps2 * sympy.Derivative(PhiE2_la, r) - 2*alpha*P3*eps1 * sympy.Derivative(PhiM2_la, r)
)
```

```python
eq2 = sympy.Eq(
    1/mu1 * dphi1M_dtheta,
    1/mu2 * dphi2M_dtheta + 2*alpha*P3*eps1 * dphi2E_dtheta
)
sympy.Eq(
    1/mu1 * sympy.Derivative(PhiM1_la, theta),
    1/mu2 * sympy.Derivative(PhiM2_la, theta) + 2*alpha*P3*eps1 * sympy.Derivative(PhiE2_la, theta)
)
```

```python
eq3 = sympy.Eq(
    dphi1E_dtheta, dphi2E_dtheta
)
sympy.Eq(
    sympy.Derivative(PhiE1_la, theta), sympy.Derivative(PhiE2_la, theta)
)
```

```python
eq4 = sympy.Eq(
    dphi1M_dr, dphi2M_dr
)
sympy.Eq(
    sympy.Derivative(PhiM1_la, r), sympy.Derivative(PhiM2_la, r)
)
```

We now solve the above boundary conditions to determine the values of $A_l$, $B_l$, $C_l$, $D_l$

```python
result = sympy.solve(
    [ eq1, eq2, eq3, eq4 ],
    [ Al, Bl, Cl, Dl ],
)
```

```python
sympy.Eq(
    Al,
    result[Al]
        .simplify()
        .collect(-4*P3**2*alpha**2*eps1**2*mu2*mu1)
        .collect(l)
        .collect(eps2)
        .collect(eps1)
        .collect(-1)
)
```

```python
sympy.Eq(
    Bl,
    result[Bl]
        .simplify()
        .collect(mu1)
        .collect(mu2)
        .collect(eps1)
        .collect(eps2)
        .collect(l)
        .collect(P3)
        .collect(alpha)
)
```

```python
sympy.Eq(
    Cl,
    result[Cl]
        .simplify()
        .collect(4*P3**2*alpha**2*eps1**2*mu1*mu2)
        .collect(eps1)
        .collect(eps2)
        .collect(l)
)
```

```python
sympy.Eq(
    Dl,
    result[Dl]
        .simplify()
        .collect(4*P3**2*alpha**2*eps1**2*mu1*mu2)
        .collect(eps1)
        .collect(eps2)
        .collect(l)
)
```

## Alternate Method 

We use an alternate method of using direct image method for an infinite sheet. Note that the presence of Dirac's string (at a larger distance) is not to be forgotten!


The electric and magnetic potentials in both the regions according to the image method are given by

$$
\begin{aligned}
& V(x, z>0)=\frac{Q / \epsilon_1}{\left[x^2+(z-a)^2\right]^{1 / 2}}+\frac{q}{\left[x^2+(z+a)^2\right]^{1 / 2}} \\
& V(x, z<0)=\frac{q / \epsilon_1}{\left[x^2+(z-a)^2\right]^{1 / 2}}+\frac{q}{\left[x^2+(z-a)^2\right]^{1 / 2}} \\
& U(x, z>0)=\frac{p_2}{\left[x^2+(z+b)^2\right]^{1 / 2}} \\
& U(x, z<0)=\frac{p_1}{\left[x^2+(z-b)^2\right]^{1 / 2}}
\end{aligned}
$$

Now, we apply boundary conditions to get the following results - 

1. From continuity of $ \mathbf{B}_{\perp}=-\frac{\partial U}{\partial z} $, 

$$
\begin{aligned}
\left(\frac{\partial U(z>0)}{\partial z}\right)_0&=\left(\frac{\partial U(z<0)}{\partial z}\right)_0 \\
p_2 \frac{\left( 0+b\right)}{\left[x^2+\left(0+b\right)^2\right]^{3 / 2}}&=p_1 \frac{\left(0-b\right)}{\left[x^2+\left({0}-b\right)^2\right]^{3 / 2}} \\
p_1 &= -p_2
\end{aligned}
$$

2. From continuity of $ \mathbf{D}_{\perp}$, we have $ \epsilon_1 \mathbf{E}_{1 \perp}=\epsilon_2 \mathbf{E}_{2 \perp}-\alpha B_z $ where $B_z=-\left(\frac{\partial U}{\partial z}\right)_0$ and $\mathbf{E}_{\perp}=-\left(\frac{\partial V}{\partial z}\right)_0$

$$
\begin{aligned}
\epsilon_1\left(\frac{Q}{\epsilon_1} \frac{\left(0-a\right)}{\left[x^2+\left(0-a\right)^2\right]^{3 / 2}}+q \frac{\left(0+a\right)}{\left[x^2+\left(0+a\right)^2\right]^{3 / 2}}\right) &= 
\epsilon_2\left(\frac{Q}{\epsilon_1} \frac{\left({0}-a\right)}{\left[x^2+\left(0-a\right)^2\right]^{3 / 2}}+q \frac{\left(0-a\right)}{\left[x^2+\left(0-a\right)^2\right]^{3 / 2}}\right) 
-\alpha p_1 \frac{\left(0+b\right)}{\left[x^2+\left(0+b\right)^2\right]^{3 / 2}} \\
Q\left(1-\frac{\epsilon_2}{\epsilon_1}\right) \frac{(-a)}{\left[x^2+a^2\right]^{3 / 2}}+\frac{a q}{\left[x^2+a^2\right]^{3 / 2}}\left(\epsilon_1+\epsilon_2\right) &= -\frac{b}{\left[x^2+b^2\right]^{3 / 2}} \alpha p_1
\end{aligned}
$$

  Therefore, $a = b$, which gives


$$
\begin{aligned}
p_1&=-\frac{Q}{\alpha} \frac{\epsilon_2-\epsilon_1}{\epsilon_1}-\frac{q}{\alpha}\left(\epsilon_1+\epsilon_2\right)
\end{aligned}
$$


3. From continuity of $ \mathbf{H_{\parallel}}$, $\frac{1}{\mu_1} \mathbf{B}_{1 \|}=\frac{1}{\mu_2} \mathbf{B}_{2 \|}+\alpha \mathbf{E}_{\|}$, with $p_1 = -p_2$, $\mathbf{B}_{\|}=-\left(\frac{\partial U}{\partial x}\right)_{z=0}$, $\mathbf{E}_{\|}=-\left(\frac{\partial V}{\partial x}\right)_{z=0}$ and $\left(\frac{\partial V(z>0)}{\partial x}\right)_{z=0}=\left(\frac{\partial V(z<0)}{\partial x}\right)_{z=0}$

$$
\begin{aligned}
\frac{1}{\mu_1} \frac{x p_1}{\left[x^2+a^2\right]^{3 / 2}} & =\frac{1}{\mu_2} \frac{x\left(-p_1\right)}{\left[x^2+a^2\right]^{3 / 2}}+\alpha\left(\frac{Q}{\epsilon_1} \frac{x}{\left[x^2+a^2\right]^{3 / 2}}+q \frac{x}{\left[x^2+a^2\right]^{3 / 2}}\right) \\
\rightarrow\left(\frac{1}{\mu_1}+\frac{1}{\mu_2}\right) p_1 & =\alpha\left(\frac{Q}{\epsilon_1}+q\right) \\
& \rightarrow p_1=\frac{\alpha}{\frac{1}{\mu_1}+\frac{1}{\mu_2}}\left(\frac{Q}{\epsilon_1}+q\right)
\end{aligned}
$$

From $2.$ and $3.$ above, we get 

$$
q=\frac{Q}{\epsilon_1} \frac{\left(\frac{1}{\mu_1}+\frac{1}{\mu_2}\right)\left(\epsilon_1-\epsilon_2\right)-\alpha^2}{\left(\frac{1}{\mu_1}+\frac{1}{\mu_2}\right)\left(\epsilon_1+\epsilon_2\right)+\alpha^2} 
$$

We replace this $q$ in Equation $3.$ to obtain 

$$
p_1=\alpha \frac{2 Q}{\left(\frac{1}{\mu_1}+\frac{1}{\mu_2}\right)\left(\epsilon_1+\epsilon_2\right)+\alpha^2}
$$


We have thus found the values of the charge and magnetic monopole induced by the charge $Q$.

# Experimental Proposal

The effect of magnetic monopole can be observed using a magnetic force microscope (MFM). The setup is as follows

Screenshot from 2023-04-02 18-19-33.png

The MFM has a tip at distance $d$ away from the surface of the TI, in presence of an impurity charge $Q$ at distance $D$ away from the surface. The tip is also separated by a distance $r$ horizontally from the impurity. Considering both the tip and impurity as point-like, we obtain an image charge $Q'$, magnetic flux $G'$ for the impurity and image charge $q'$ and magnetic flux $g'$. The image charges for the impurity will be obtained by the same equations used to calculate the fields above. Boundary conditions will change when the tip has magnetic flux g. Thus three forces are applied to the MFM tip, exerted by the image charge and monopole of itself, the impurity and the image charge and monopole of the impurity. The MFM measures the gradient of this force. The tip is used to scan across position $r$ as well as tip voltage $V$. For a given $r$, $f_{min} (r)$ determines the minimal force applied to the tip when scanning the voltage. In the limit $r>>D$, the contribution from conventional charge leads to a $\frac{1}{r^6}$ dependence of $f_{min}(r)$. However a magnetic monopole leads to a $\frac{1}{r^3}$ dependence. Even defects such as fringe magentic fields will lead to a higher order contribution, keeeping the monopole term dominant.

## Tutorial Problem

Find the images of a magnetic charge placed close to the interface of an ordinary and topological insulator. Compare this with the result for an electric charge.

Like images of the electric charge, a magnetic monopole and its images will generate an electric ($V$) and magnetic ($U$) potentials given by 

$$\begin{aligned} U_1(x, z>0) & =\frac{\mu_1 P}{\left[x^2+(z-a)^2\right]^{1 / 2}}+\frac{p_2}{\left[x^2+(z+a)^2\right]^{1 / 2}} \\ U_2(x, z<0) & =\frac{\mu_1 P}{\left[x^2+(z-a)^2\right]^{1 / 2}}+\frac{p_1}{\left[x^2+(z-a)^2\right]^{1 / 2}} \\ V_1(x, z>0) & =\frac{q}{\left[x^2+(z+b)^2\right]^{1 / 2}} \\ V_2(x, z<0) & =\frac{q}{\left[x^2+(z-b)^2\right]^{1 / 2}}\end{aligned}$$

The boundary conditions are 

$$\begin{aligned} \epsilon_1\left(\frac{\partial V_1}{\partial z}\right)_0 & =\epsilon_2\left(\frac{\partial V_2}{\partial z}\right)_0-\alpha\left(\frac{\partial U}{\partial z}\right)_0 \\ \frac{1}{\mu_1}\left(\frac{\partial U_1}{\partial x}\right)_0 & =\frac{1}{\mu_2}\left(\frac{\partial U_2}{\partial x}\right)_0+\alpha\left(\frac{\partial V}{\partial x}\right)_0\end{aligned}$$



Since $\left(\frac{\partial U_1}{\partial z}\right)_0=\left(\frac{\partial U_2}{\partial z}\right)_0$ and $\left(\frac{\partial V_1}{\partial x}\right)_0=\left(\frac{\partial V_2}{\partial x}\right)_0$, we can employ $U$ and $V$ of either region.

From the first condition, we get $a=b$, and subsequently 

$$
q=\frac{\alpha}{\epsilon_1+\epsilon_2}\left(p_1+\mu_1 P\right)
$$

From the second equation, we will get 

$$
P\left(1-\frac{\mu_1}{\mu_2}\right)+\frac{p_2}{\mu_1}-\frac{p_1}{\mu_2}-\alpha q=0
$$

Finally, $\mathbf{B}_{1 \perp}=\mathbf{B}_{2 \perp}$ yields $p_1 = -p_2$. 

Now, by combining the previous two results, we get

$$
p_1=\mu_1 P \frac{\frac{1}{\mu_1}-\frac{1}{\mu_2}-\frac{\alpha^2}{\epsilon_1+\epsilon_2}}{\frac{1}{\mu_1}+\frac{1}{\mu_2}+\frac{\alpha^2}{\epsilon_1+\epsilon_2}}
$$

By inserting this into the equation for $q$, we get,

$$
q=\frac{\alpha}{\epsilon_1+\epsilon_2} \mu_1 P\left(\frac{\frac{1}{\mu_1}-\frac{1}{\mu_2}-\frac{\alpha^2}{\epsilon_1+\epsilon_2}}{\frac{1}{\mu_2}+\frac{1}{\mu_1}+\frac{\alpha^2}{\epsilon_1+\epsilon_2}}+1\right)
$$

Simplifying, we can obtain

$$
\begin{aligned}
&p_1=\mu_1 P \frac{\left(\frac{1}{\mu_1}-\frac{1}{\mu_2}\right)\left(\epsilon_1+\epsilon_2\right)-\alpha^2}{\left(\frac{1}{\mu_1}+\frac{1}{\mu_2}\right)\left(\epsilon_1+\epsilon_2\right)+\alpha^2}\\
&q=\alpha \frac{2 P}{\left(\frac{1}{\mu_1}+\frac{1}{\mu_2}\right)\left(\epsilon_1+\epsilon_2\right)+\alpha^2}
\end{aligned}
$$



The result is a perfect analogy of the image of an electric charge, with the detail that the image $q$ of the charge $Q$, is proportional to $\left(\frac{1}{\mu_1}+\frac{1}{\mu_2}\right)\left(\epsilon_1-\epsilon_2\right)$ while the image $p$ of the monopole $P$, is proportional to $\left(\frac{1}{\mu_1}-\frac{1}{\mu_2}\right)\left(\epsilon_1+\epsilon_2\right)$. Dielectric constants and magnetic permeabilities exchange their role.
