#Green’s function approach to Chern-Simons extended electrodynamics: An effective theory describing topological insulators



# **Abstract**
The article discusses the analysis of boundary effects produced by a Chern-Simons extension to electrodynamics using the Green's function method. The electromagnetic field coupled to a θ term is considered, which is piecewise constant in different regions of space separated by a common interface Σ, called the θ boundary. The method allows solving the CS modified field equations for a given θ boundary with arbitrary configuration of sources. The static fields of a pointlike charge interacting with a planar TI are calculated, and the force between the charge and the θ boundary is computed using two different methods. The infinitely straight current-carrying wire is also analyzed.

# **Introduction**


We are concerned with a simple case of CS theories, to which we will refer as $\theta$ electrodynamics or simply $\theta$ $\mathrm{ED}$, and it amounts to extending Maxwell electrodynamics by a parity-violating term of the form

$$
\Delta \mathcal{L}_{\theta}=\theta\left(\alpha / 4 \pi^{2}\right) \mathbf{E} \cdot \mathbf{B}=-\frac{\theta}{4}\left(\alpha / 4 \pi^{2}\right) F_{\mu \nu} \tilde{F}^{\mu \nu} \tag{1}
$$

where $\tilde{F}^{\mu \nu}=\frac{1}{2} \epsilon^{\mu \nu \alpha \beta} F_{\alpha \beta}$ and $\epsilon^{\mu \nu \alpha \beta}$ is the Levi-Cività symbol.

 \therefore \mathbf{A^μ}|_{z=a^-}^{z=a^+} = 0 $$

 Integrating equation (2) over an infinitesimally small interval above $z=a$. We get,
$$-∂_{z}\mathbf{A^\mu}|_{z=a^-}^{z=a^+} - θϵ^{3μα}_{ν}∂_{α}\mathbf{A}^\nu|_{z=a} = 4\pi ∫_{z=a^-}^{z=a^+}j^\mu dz\\
\therefore ∂_z\mathbf{A^\mu}|_{z=a^-}^{z=a^+} = -\theta ϵ^{3μα}_{ν}∂_{α}\mathbf{A}^\nu|_{z=a}$$

In order to solve Eq. (2) we introduce the Green function $ G^\nu_\sigma (\mathbf{x}, \mathbf{x}')$ and solve Eq. (2) for a point source,

$$
(-\eta^\mu_\nu \nabla^2 - \tilde{\theta} \delta(z-a)\epsilon^{3\mu\alpha}_\nu \partial_\alpha)G^\nu_\sigma (\mathbf{x}, \mathbf{x}') = 4\pi\eta^\mu_\sigma\delta^3(\mathbf{x} -\mathbf{x}')
$$

Multiplying the above equation with $A^\nu(\mathbf{x}')$ and multiplying Eq. (2) with $G^\nu_\sigma(\mathbf{x}, \mathbf{x}')$ and subtracting the resulting pair of expressions we get,

$$
-\eta^\mu_\nu (G^\nu_\sigma \nabla^2 A^\nu - A^\nu \nabla^2 G^\nu_\sigma) - \tilde{\theta} \delta (z'-a) \epsilon^{3\mu\alpha}_\nu (G^\nu_\sigma \partial_\alpha A^\nu - A^\nu \partial_\alpha G^\nu_\sigma)
= 4\pi (j^\mu G^\nu_\sigma - \eta^\mu_\sigma \delta^3 (\mathbf{x} - \mathbf{x}') A^\nu)
$$

integrating the above equation we get

$$
\eta^\mu_\sigma A^\mu (\mathbf{x}) = \underbrace{\frac{1}{4\pi} \int_\nu d^3\mathbf{x}' \eta^\mu_\nu \nabla (G^\nu_\sigma(\mathbf{x}, \mathbf{x}')\nabla A^\nu(\mathbf{x}') - A^\nu(\mathbf{x}') \nabla G^\nu_\sigma(\mathbf{x}, \mathbf{x}')) }_{I} 
+ \underbrace{\frac{\tilde{\theta}}{4\pi} \int_\nu d^3\mathbf{x}' \delta(z'-a) \epsilon^{3\mu\alpha}_\nu (G^\nu_\sigma(\mathbf{x}, \mathbf{x}') \delta_\alpha A^\nu (\mathbf{x}') - A^\nu(\mathbf{x}')\partial_\alpha G^\nu_\sigma(\mathbf{x}, \mathbf{x}'))}_{II}
+ \underbrace{\int_\nu d^3\mathbf{x}' G^\nu_\sigma(\mathbf{x}, \mathbf{x}') j^\nu (\mathbf{x}')}_{III}
$$

Simplifying each of 3 parts,

$I:$ [by divergence theorem, where $\mathbf{n}$ is the normal to the θ boundary]

$ \frac{1}{4\pi}\oint_S dS'\eta^\mu{}_{\nu}\mathbf{n}\cdot(G^{\nu}{}_\sigma(\mathbf{x},\mathbf{x'})\nabla A^\nu(\mathbf{x'})-A^{\nu}(\mathbf{x'})\nabla G^{\nu}{}_\sigma(\mathbf{x},\mathbf{x'}))$ 

$II: \frac{\tilde{θ}}{4\pi}\int_{\Sigma}d^2\mathbf{x'}_{\Sigma}ϵ^{3\mu\alpha}{}_{\nu}[G^{\nu}{}_\sigma(\mathbf{x},\mathbf{x'})\partial_\alpha A^\nu(\mathbf{x'})-A^\nu(\mathbf{x'})\partial_\alpha G^{\nu}{}_\sigma(\mathbf{x},\mathbf{x'})]_{\Sigma}$

$III: \int_{V} d^3\mathbf{x'}G^{\nu}{}_\sigma(\mathbf{x},\mathbf{x'})j^{\nu}(\mathbf{x'})$

here $I$ can be interpreted as the effects of the bounding surface
S, $II$ as the contributions from the θ boundary and $III$ as the contribution from the sources inside the volume $V$. We now take $S$ as a surface at
infinity and apply appropriate BCs for the fields at the θ interface. By inspection is it can be seen that there are four classes of BCs on Σ that
specify a solution. We know look into these four classes.

##Classes of Boundary conditions
#### Class 1: 
Fix the scalar potential $A^0$ and the vector potential parallel to the θ
boundary $\mathbf{n} × A$ , and $ϵ^{3\mu\alpha}{}_{\nu}G^{\nu}{}_\sigma(\mathbf{x},\mathbf{x'})|_{\Sigma}=0$, which resembles the standard Dirichlet BC. Observe that in the original volume integral for $II$ if the above conditions are applied along with the property that volume integral of a derivative is 0 for well behaved fields, we get $II→0$. This can be applied to subsequant classes too.

####Class 2:
Fix the normal component of the magnetic field $B_n$ and the parallel
component of the electric field $n × E$ on $Σ$ and $ϵ^{3\mu\alpha}{}_{\nu}\partial_{\alpha}G^{\nu}{}_\sigma(\mathbf{x},\mathbf{x'})|_{\Sigma}=1/A_\theta$, where $A_\theta=\int d^2\mathbf{x}_Σ$ is the surface area of the θ boundary, that generates a term in the solution involving the average contribution of the potential and hence resembles the standard Neumann BC.

####Class 3:
Fix the scalar potential $A_0$ and the normal component of the magnetic field $B_n$ on $Σ$, and $ϵ^{3 0\alpha}{}_{\nu}G^{\nu}{}_\sigma(\mathbf{x},\mathbf{x'})|_{\Sigma}=0$, $ϵ^{3\mu\alpha}{}_{0}\partial_{\alpha}G^{0}{}_\sigma(\mathbf{x},\mathbf{x'})|_{\Sigma}=1/A_\theta$.

####Class 4:
Specify $\mathbf{n} × A$ and $\mathbf{n} × E$ on $Σ$ and $ϵ^{3\mu\alpha}{}_{0}G^{0}{}_\sigma(\mathbf{x},\mathbf{x'})|_{\Sigma}=0$, $ϵ^{30\alpha}{}_{\nu}\partial_{\alpha}G^{\nu}{}_\sigma(\mathbf{x},\mathbf{x'})|_{\Sigma}=1/A_\theta$

Having set the surface $S$ at infinity along with the required BC (each BC leads to $II→0$ following a similar  logic as given for class 1) we see that the general solution for the 4-potential in the Coulomb gauge is

$$
A^{\mu}(\mathbf{x})=\int d^{3} \mathbf{x}^{\prime} G^{\mu}{ }_{\nu}\left(\mathbf{x}, \mathbf{x}^{\prime}\right) j^{\nu}\left(\mathbf{x}^{\prime}\right) \tag{14} 
$$


###GF solution

We now construct the solution for the following eqn. of the GF,
$$
\left[-\eta_{\nu}^{\mu} \nabla^{2}-\tilde{\theta} \delta(z-a) \epsilon_{\nu}^{3 \mu \alpha}{ }_{\nu} \partial_{\alpha}\right] G^{\nu}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=4 \pi \eta_{\sigma}^{\mu} \delta^{3}\left(\mathbf{x}-\mathbf{x}^{\prime}\right) \tag{13} 
$$

As this GF has translational invariance in the directions parallel to $Σ$ i.e., in the transverse x and y directions, with invariance broken in the z-direction. By virtue of this symmetry we perform the following Fourier transform in the direction parallel to the plane $Σ$, taking the coordinate dependence to be $(\mathbf{x-x'})_{||}=(x-x',y-y')$
$$
G^{\mu}{ }_{\nu}(\mathbf{x,x^{\prime}}) = 4\pi \int \frac{d^2\mathbf{p}}{(2\pi)^2} \, e^{i\mathbf{p\cdot(x-x^{\prime})}_\parallel} g^{\mu}{ }_{\nu}(z,z^{\prime},\mathbf{p}) \tag{15}
$$

where $\mathbf{p}=\left(p_{x}, p_{y}\right)$ is the momentum parallel to the plane $\Sigma$. We suppress the dependence of
the reduced GF $g^{\mu}{ }_{\nu}$ on $\mathbf{p}$ for the meantime.
Substituting 15 in 13 and writing $δ(\mathbf{x-x'})=\delta(z-z')\int \frac{d^2\mathbf{p}}{(2\pi)^2} \, e^{i\mathbf{p\cdot(x-x^{\prime})}_\parallel}$ , we get,
$$
\left[-\nabla^{2} \eta^{\mu}{ }_{\nu}+i \tilde{\theta} \delta(z-a) \epsilon^{3 \mu \alpha}{ }_{\nu} \partial_{\alpha}\right] G^{\nu}{ }_{\sigma}\left(z, z^{\prime}\right)=\eta^{\mu}{ }_{\sigma} \delta\left(z-z^{\prime}\right)\int \frac{d^2\mathbf{p}}{(2\pi)^2} \, e^{i\mathbf{p\cdot(x-x^{\prime})}_\parallel} \tag{16}
$$
Due to the antisymmetry of the Levi-Cività symbol, the partial derivative appearing in the second term does not introduce derivatives with respect to z but only in the transverse coordinates. Also $\nabla^2e^{i\mathbf{p\cdot(x-x^{\prime})}}=(\mathbf{p}^{2}-\partial_{z}^{2}) e^{i\mathbf{p\cdot(x-x^{\prime})}}$ and so we get the full reduced GF equation,

$$
\left[\partial^{2} \eta^{\mu}{ }_{\nu}+i \tilde{\theta} \delta(z-a) \epsilon^{3 \mu \alpha}{ }_{\nu} p_{\alpha}\right] g^{\nu}{ }_{\sigma}\left(z, z^{\prime}\right)=\eta^{\mu}{ }_{\sigma} \delta\left(z-z^{\prime}\right) , \tag{16}
$$


where $\partial^{2}=\mathbf{p}^{2}-\partial_{z}^{2}, p^{\alpha}=(0, \mathbf{p})$ and $\mathbf{p}^{2}=-p^{\alpha} p_{\alpha}$.

We now solve the full reduced GF equation using a method similar to that used for obtaining the GF for the one-dimensional $\delta$-function potential in quantum mechanics, where the free GF is used for integrating the GF equation with $\delta$ interaction. 

Consider a reduced free GF having the form 
$\mathcal{G}_{\nu}^{\mu}\left(z, z^{\prime}\right)=\mathfrak{g}\left(z, z^{\prime}\right) \eta_{\nu}^{\mu}$, 
associated with the operator $\partial^{2}$ previously defined, that solves

$$
\partial^{2} \mathcal{G}_{\nu}^{\mu}\left(z, z^{\prime}\right)=\eta_{\nu}^{\mu} \delta\left(z-z^{\prime}\right) \tag{17}
$$

satisfying the standard boundary conditions at infinity.

###Solving for free GF
Equation 17 can be rewritten as 
$$
\partial^{2} \mathfrak{g}\left(z, z^{\prime}\right)= \delta\left(z-z^{\prime}\right) 
$$
For $z\neq z'$,
$$
\partial^2_z\mathfrak{g}(z,z')=p^2
$$
with the following general solution,
$$
\mathfrak{g}(z,z')=\begin{array}{lr}
        a(z')e^{pz}+a'(z')e^{-pz}, & \text{for } z>z'\\
        b(z')e^{pz}+b'(z')e^{-pz}, & \text{for } z<z'
\end{array}
$$

and $p=|\mathbf{p}|$. , 

For the case of $z = z'$, we integrate over the limits $z'-\epsilon$ to $z'+0$ in the limit $\epsilon→0$.

$$
\int_{z'-\epsilon}^{z'+\epsilon}(p^2-\partial^2_z)\mathfrak{g}(z,z')dz' = \int_{z'-\epsilon}^{z'+\epsilon}\delta(z-z')dz'
$$
In the limit $\epsilon→0$ the $p^2$ component tends to 0. Thus, we get 

$$
\begin{array}{lr}
-(\partial_z\mathfrak{g}|_{z'+\epsilon}-\partial_z\mathfrak{g}|_{z'-\epsilon}) = 1
\\
a(z')pe^{pz'} - b(z')pe^{pz'} =  -1
\\
p(a(z')e^{pz'}-a'(z')e^{-pz'}-b(z')e^{pz'}+b(z')e^{-pz'})=-1
\end{array}
$$
At infinity, we apply the boundary conditions, $a=0$ and $b'=0$. Thus we get,

$$
a'(z')e^{-pz'}+b(z')e^{pz'}=\frac{1}{p}
$$
Thus, imposing continuity of $\mathfrak{g}$ at $z=z^{\prime}$ we find that the solutions are
$$
a'(z') = \frac{e^{pz'}}{2p} \ \text{and} \ b(z') = \frac{e^{-pz'}}{2p} 
$$
Therefore,
$$
\begin{array}{lr}
\text{for } z>z' : 
        \frac{1}{2p}e^{-p(z-z')} \\
\text{for } z<z' : 
        \frac{1}{2p}e^{-p(z'-z)}
\end{array}
= \frac{1}{2p}e^{-p|z-z'|}
$$
The derivative of $\mathfrak{g}$ is discontinuous at $z=z^{\prime}$, i.e., $\left.\partial_{z} \mathfrak{g}\left(z, z^{\prime}\right)\right|_{z=z^{\prime}} ^{z=z^{\prime}}=-1$.

### Reduced GF solution

Recall the full reduced GF equation, 

$$
\partial^{2} g^{\mu}{ }_{\sigma}\left(z, z^{\prime}\right)+i \tilde{\theta} \delta(z-a) \epsilon^{3 \mu \alpha}{ }_{\nu} p_{\alpha} g^{\nu}{ }_{\sigma}\left(z, z^{\prime}\right)=\eta^{\mu}{ }_{\sigma} \delta\left(z-z^{\prime}\right)
$$

also we know $\delta(z-a)(g^{\nu}{ }_{\sigma}(z,z')-g^{\nu}{ }_{\sigma}(a,z'))=0$ if $g^{\nu}{ }_{\sigma}$ is well behaved. Hence,

$$
\partial^{2} g^{\mu}{ }_{\sigma}\left(z, z^{\prime}\right)+i \tilde{\theta} \delta(z-a) \epsilon^{3 \mu \alpha}{ }_{\nu} p_{\alpha} g^{\nu}{ }_{\sigma}\left(a, z^{\prime}\right)=\eta^{\mu}{ }_{\sigma} \delta\left(z-z^{\prime}\right)
$$

also $\partial^2\mathfrak{g}(z,a)=\delta(z-a)$,

$$
\partial^{2} g^{\mu}{ }_{\sigma}\left(z, z^{\prime}\right)+i \tilde{\theta} \partial^2\mathfrak{g}(z,a) \epsilon^{3 \mu \alpha}{ }_{\nu} p_{\alpha} g^{\nu}{ }_{\sigma}\left(a, z^{\prime}\right)=\eta^{\mu}{ }_{\sigma} \delta\left(z-z^{\prime}\right)\\
⇒\partial^{2}\left( g^{\mu}{ }_{\sigma}\left(z, z^{\prime}\right)+i \tilde{\theta} \mathfrak{g}(z,a) \epsilon^{3 \mu \alpha}{ }_{\nu} p_{\alpha} g^{\nu}{ }_{\sigma}\left(a, z^{\prime}\right)\right)=\eta^{\mu}{ }_{\sigma} \delta\left(z-z^{\prime}\right)
$$

comparing with the reduced free GF equation we get,

$$
g^{\mu}{ }_{\sigma}\left(z, z^{\prime}\right)=\eta^{\mu}{ }_{\sigma} \mathfrak{g}\left(z, z^{\prime}\right)-i \tilde{\theta} \epsilon^{3 \mu \alpha}{ }_{\nu} p_{\alpha} \mathfrak{g}(z, a) g_{\sigma}^{\nu}\left(a, z^{\prime}\right) \tag{19}
$$

hence $g^{\mu}{ }_{\sigma}$ is continuous however $\partial_{z} g^{\mu}{ }_{\sigma}$ is discontinuous at $z=a$.

$$
\left.\partial_{z} g^{\mu}{ }_{\sigma}\left(z, z^{\prime}\right)\right|_{z=a^{-}} ^{z=a^{+}} =-\left.i \tilde{\theta} \epsilon^{3 \mu \alpha}{ }_{\nu} p_{\alpha} \partial_{z} \mathfrak{g}(z, a)\right|_{z=a^{-}} ^{z=a^{+}} g^{\nu}{ }_{\sigma}\left(a, z^{\prime}\right) 
$$
$$
\ \ \ \ =i \tilde{\theta} \epsilon^{3 \mu \alpha}{ }_{\nu} p_{\alpha} g^{\nu}{ }_{\sigma}\left(a, z^{\prime}\right) \tag{20}
$$


We now solve for the various components $g^{\mu}{ }_{\sigma}$. Splitting the components into $\mu=0$ and $\mu=j=1,2,3$:

$$
g^{0}{}_{\sigma}\left(z, z^{\prime}\right)=\eta^{0}{}_{\sigma} \mathfrak{g}\left(z, z^{\prime}\right)-i \tilde{\theta} \epsilon^{30 i}{ }_{j} p_{i} \mathfrak{g}(z, a) g^{j}{ }_{\sigma}\left(a, z^{\prime}\right), \tag{21}
$$

$$
g^{j}{ }_{\sigma}\left(z, z^{\prime}\right)=\eta^{j}{ }_{\sigma} \mathfrak{g}\left(z, z^{\prime}\right)-i \tilde{\theta} \epsilon^{3 j i}{ }_{0} p_{i} \mathfrak{g}(z, a) g_{\sigma}^{0}\left(a, z^{\prime}\right) \tag{22} 
$$

Setting $z=a$ in Eq. (22) and substituting into Eq. (21), and as $\epsilon^{30 i}{ }_{j} \epsilon^{3 j k}{ }_{0} p_{k} p_{i}=p^{2}$ we get

$$
g^{0}{}_{\sigma}\left(z, z^{\prime}\right)= \eta^{0}{}_{\sigma} \mathfrak{g}\left(z, z^{\prime}\right)-i \tilde{\theta} \epsilon^{30 i}{ }_{j} p_{i} \eta^{j}{}_{\sigma} \mathfrak{g}(z, a) \mathfrak{g}\left(a, z^{\prime}\right) -\tilde{\theta}^{2} p^{2} \mathfrak{g}(z, a) \mathfrak{g}(a, a) g^{0}{}_{\sigma}\left(a, z^{\prime}\right), \tag{23}
$$

Solving for $g^{0}{}_{\sigma}\left(a, z^{\prime}\right)$ by setting $z=a$ in Eq. (23) and inserting the result back into that equation, we obtain

$$
g^{0}{}_{\sigma}\left(z, z^{\prime}\right)= \eta^{0}{}_{\sigma}\left[\mathfrak{g}\left(z, z^{\prime}\right)+\tilde{\theta} p^{2} \mathfrak{g}(a, a) A\left(z, z^{\prime}\right)\right] +i \epsilon^{30 i}{ }_{\sigma} p_{i} A\left(z, z^{\prime}\right), \tag{24}
$$

where

$$
A\left(z, z^{\prime}\right)=-\tilde{\theta} \frac{\mathfrak{g}(z, a) \mathfrak{g}\left(a, z^{\prime}\right)}{1+p^{2} \tilde{\theta}^{2} \mathfrak{g}^{2}(a, a)}  \tag{25}
$$

For the remaining components we substituting $g_{\sigma}^{0}\left(a, z^{\prime}\right)$ in Eq. (22). The result is

$$
g_{\sigma}^{j}\left(z, z^{\prime}\right)= \eta_{\sigma}^{j} \mathfrak{g}\left(z, z^{\prime}\right)+i \epsilon_{0}^{3 j k}{ }_{0} p_{k}\left[\eta_{\sigma}^{0}-i \tilde{\theta} \epsilon^{30 i}{ }_{\sigma} p_{i} \mathfrak{g}(a, a)\right] A\left(z, z^{\prime}\right) .  \tag{26}
$$

Equations (24) and (26) allow us to write the general solution as

$$
g_{\nu}^{\mu}\left(z, z^{\prime}\right)= \eta_{\nu}^{\mu} \mathfrak{g}\left(z, z^{\prime}\right)+A\left(z, z^{\prime}\right)\left\{\tilde { \theta } \mathfrak { g } ( a , a ) \left[p^{\mu} p_{\nu}\right.\right. \left.\left.+\left(\eta_{\nu}^{\mu}+n^{\mu} n_{\nu}\right) p^{2}\right]+i \epsilon^{\mu}{ }_{\nu}{ }^{\alpha 3} p_{\alpha}\right\},  \tag{27}
$$

where $n_{\mu}=(0,0,0,1)$ is the normal to $\Sigma$.

The reciprocity between the position of the unit charge and the position at which the GF is evaluated $G_{\mu \nu}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=$ $G_{\nu \mu}\left(\mathbf{x}^{\prime}, \mathbf{x}\right)$ is one of its most remarkable properties. From Eq. (15) this condition demands

$$
g_{\mu \nu}\left(z, z^{\prime}, \mathbf{p}\right)=g_{\nu \mu}\left(z^{\prime}, z,-\mathbf{p}\right),  \tag{28}
$$

# **Derivation of GF solutions**

Here we derive Eqs. (29)-(31). In free case $(\tilde{\theta}=0)$, the reduced vacuum GF was calculated above as

$$
\mathfrak{g}\left(z, z^{\prime}\right)=\frac{1}{2 p} e^{-p\left|z-z^{\prime}\right|}
$$

In the coordinate representation, we obtain the full free Green's function by Fourier transforming the above equation to get

$$
\mathcal{G}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=4 \pi \int \frac{d^{2} \mathbf{p}}{(2 \pi)^{2}} e^{i \mathbf{p} \cdot\left(\mathbf{x}-\mathbf{x}^{\prime}\right)} \frac{1}{2 p} e^{-p\left|z-z^{\prime}\right|} .
$$

image.png

Expressing the area element in polar coordinates, $d^{2} \mathbf{p}=p d p d \varphi$ and choosing the $p_{x}$ axis in the direction of the vector $\mathbf{R}=\left(\mathbf{x}-\mathbf{x}^{\prime}\right)_{\|}$, as shown in the figure, we simplify the integral. Noting that $\mathbf{p} \cdot \mathbf{R}=p R \cos \varphi$, we can write

$$\mathcal{G}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=\int_{0}^{\infty} d p e^{-p\left|z-z^{\prime}\right|}\left\{\frac{1}{2 \pi} \int_{0}^{2 \pi} e^{i p R \cos \varphi} d \varphi\right\}$$

In the above integral, consider $\left|z-z^{\prime}\right|=a, b=R$ and $\theta=\phi-\pi / 2, p=x$ Then the integral can be written as
$$
\begin{aligned}
I & =\int_0^{\infty} d x e^{-a x} \frac{1}{2 \pi} \int d \theta e^{i b x \sin \theta} \\
& =\int \frac{d \theta}{2 \pi} \int_0^{\infty} d x e^{-a x+i b x \sin \theta} \\
& =\int \frac{d \theta}{2 \pi} \frac{1}{a-i b \sin \theta}
\end{aligned}
$$
Substitute $z=e^{i \theta}$. The integration is now on a unit circle in a complex plane.

$$I=\frac{1}{2 \pi} \oint\frac{1}{a-\frac{b}{2}(z-1/z)} \frac{dz} {iz} = \frac{-1}{\pi i b} \oint{\frac{b}{z^2-\frac{2 az}{b}-1}}=-\frac{1}{2 \pi i} \cdot \frac{2}{b} \cdot \oint \frac{d z}{\left(z-z_{-}\right)(z-z)}$$ 

where $z_z=\frac{1}{b}\left(a \pm \sqrt{a^2+b^2}\right)$
To evaluate this integral, we perform contour integration and evaluate the residue as
$$
I=\frac{-1}{2 \pi i}  \frac{2}{b} \frac{2 \pi i}{z_{+}-z_{-}}=\frac{1}{\sqrt{a^2+b^2}}
$$
Thus, integral simplifies to

$$
\mathcal{G}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=\frac{1}{\sqrt{R^{2}+\left|z-z^{\prime}\right|^{2}}}=\frac{1}{\left|\mathbf{x}-\mathbf{x}^{\prime}\right|}
$$

which is the vacuum GF in the coordinate representation [32]. IWe use a similar procedure to compute the required integrals for establishing Eqs. (29) and (30). We first consider the component $G_{0}^{0}$. From Eq. (27), we find that the levi cevita term goes to 0 and the final expression simplifies to
$$
g_{0}^{0}\left(z, z^{\prime}\right)=\mathfrak{g}\left(z, z^{\prime}\right)+A\left(z, z^{\prime}\right) p^{2} \tilde{\theta} \mathfrak{g}(a, a)
$$

Now, in equation 25 if we substitute the solution for reduced free green's function, we find that the expression for $A(z,z')$ reduces to 

$$
A\left(z, z^{\prime}\right)=-\frac{\tilde{\theta}}{4+\tilde{\theta}^{2}} p^{-2} e^{-p Z}
$$

with the notation $Z=|z-a|+\left|z^{\prime}-a\right|$. Thus, we simplify the reduced green's function to be 

$$
g_{0}^{0}(z,z') = \mathfrak{g}\left(z, z^{\prime}\right) - \frac{\tilde{\theta}^2}{4+\tilde{\theta}^{2}}\frac{e^{-pZ}}{2p}
$$

We see that the second term is similar to the expression we had for the free field case, the only difference being, the exponent is now $Z$ instead of $|z-z'|$. We perform a similar integration to obtain
$$
G_{0}^{0}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=\frac{1}{\left|\mathbf{x}-\mathbf{x}^{\prime}\right|}-\frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}} \frac{1}{\sqrt{R^{2}+Z^{2}}}
$$

Now we evaluate the components $G_{1}^{0}{ }_{1}$ and $G_{2}^{0}$. The corresponding reduced GF are

$$
g_{i}^{0}\left(z, z^{\prime}\right)=-i \epsilon_{0 i j 3} p^{j} A\left(z, z^{\prime}\right)
$$

with $A\left(z, z^{\prime}\right)$ given by (A7). For convenience, we define the vector

$$\mathbf{I}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=\left(I^{1}, I^{2}\right)=4 \pi \int \frac{d^{2} \mathbf{p}}{(2 \pi)^{2}} e^{i \mathbf{p} \cdot\left(\mathbf{x}-\mathbf{x}^{\prime}\right)}_{\|} \mathbf{p} p^{-2} e^{-p Z}$$,

with $\mathbf{p}=\left(p_{x}, p_{y}\right)$, in terms of which we have

$$
G_{i}^{0}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=i \frac{\tilde{\theta}}{4+\tilde{\theta}^{2}} \epsilon_{0 i j 3} I^{j}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)
$$

The integral is calculated in the same coordinate system as before and can be written in vector form as

$$\mathbf{I}_{p}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=2 \int_{0}^{\infty} d p e^{-p Z}\left\{\frac{1}{2 \pi} \int_{0}^{2 \pi}\left[\begin{array}{c}\cos \varphi \\ \sin \varphi\end{array}\right] e^{i p R \cos \varphi} d \varphi\right\}$$

where the subscript $p$ indicates that the vector $\mathbf{p}$ is written in the particular coordinate system explained above. Now, we represent the above function in terms of Bessel functions and integrated the well-known radial and angular components to obtain

$$
\begin{aligned}
\mathbf{I}_{p}\left(\mathbf{x}, \mathbf{x}^{\prime}\right) & =2 i \hat{\mathbf{R}} \int_{0}^{\infty} J_{1}(p R) e^{-p Z} d p \\
& =\frac{2 i}{R}\left(1-\frac{Z}{\sqrt{R^{2}+Z^{2}}}\right) \hat{\mathbf{R}} .
\end{aligned}
$$

As a consequence of the chosen coordinate system, we find that $I_{2}=0$, in such a way that vector $\mathbf{I}_{p}$ becomes parallel to $\hat{\mathbf{R}}$. This can be generalized to an arbitrary coordinate system as

$$
\mathbf{I}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=2 i \frac{\mathbf{R}}{R^{2}}\left(1-\frac{Z}{\sqrt{R^{2}+Z^{2}}}\right) .
$$

Thus, we find

$$G_{i}^{0}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=-\frac{2 \tilde{\theta}}{4+\tilde{\theta}^{2}} \frac{\epsilon_{0 i j 3} R^{j}}{R^{2}}\left(1-\frac{Z}{\sqrt{R^{2}+Z^{2}}}\right)$$

In order to evaluate the components $G_{j}^{i}$, we first observe that the corresponding reduced GF can be written as

$$
g_{j}^{i}\left(z, z^{\prime}\right)=\eta_{j}^{i} g_{0}^{0}\left(z, z^{\prime}\right)+\tilde{\theta} \mathfrak{g}(a, a) A\left(z, z^{\prime}\right) p^{i} p_{j},
$$

where $g_{0}^{0}$ has been calculated above. Now we need to compute the Fourier transform of the above equation. The first term has already been calculated and we only evaluate the last term. To this end, we introduce the vector

$\mathbf{K}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=\left(K^{1}, K^{2}\right)=4 \pi \int \frac{d^{2} \mathbf{p}}{(2 \pi)^{2}} e^{i \mathbf{p} \cdot\left(\mathbf{x}-\mathbf{x}^{\prime}\right) \|} \frac{\mathbf{p}}{p} p^{-2} e^{-p Z}$,

from which the required integral will be calculated by taking the spatial derivative. The integral can be computed again in the particular coordinate system defined above. In the polar coordinates defined in the $\mathbf{p}$ plane, the integral reads $$\mathbf{K}_{p}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=2 \int_{0}^{\infty} \frac{d p}{p} e^{-p Z}\left\{\frac{1}{2 \pi} \int_{0}^{2 \pi}\left[\begin{array}{c}\cos \varphi \\ \sin \varphi\end{array}\right] e^{i p R \cos \varphi} d \varphi\right\}$$

Note that the braces in this equation enclose an integral representation of the Bessel function $J_{1}(p R)$. The resulting integral is well-known, and the final result is

$$
\begin{aligned}
\mathbf{K}_{p}\left(\mathbf{x}, \mathbf{x}^{\prime}\right) & =2 i \int_{0}^{\infty} \frac{d p}{p} J_{1}(p R) e^{-p Z} \hat{\mathbf{R}} \\
& =2 i \frac{\sqrt{R^{2}+Z^{2}}-Z}{R} \hat{\mathbf{R}},
\end{aligned}
$$
The generalization to an arbitrary coordinate system is then

$$
\mathbf{K}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=2 i \frac{\sqrt{R^{2}+Z^{2}}-Z}{R^{2}} \mathbf{R}
$$

Note that the required integral involves the term $p^i p_j$, which can be generated from (A18) as follows:
$$
i \partial_j K^i\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=4 \pi \int \frac{d^2 \mathbf{p}}{(2 \pi)^2} e^{i \mathbf{p} \cdot\left(\mathbf{x}-\mathbf{x}^{\prime}\right)_{\|}} \frac{p^i p_j}{p} p^{-2} e^{-p Z}
$$
By using the final form of $\mathbf{K}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)$, one can further check the consistency condition $\partial_1 K^2\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=\partial_2 K^1\left(\mathbf{x}, \mathbf{x}^{\prime}\right)$ required by the cross terms involving $p^1 p_2=p^2 p_1=-p_x p_y$. From the previous results, the $G_j^i$ components of the GF matrix in the coordinate representation can be written as
$$
G_j^i\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=\eta_j^i G_0^0\left(\mathbf{x}, \mathbf{x}^{\prime}\right)-\frac{i}{2} \frac{\tilde{\theta}^2}{4+\tilde{\theta}^2} \partial_j K^i\left(\mathbf{x}, \mathbf{x}^{\prime}\right)
$$

# **III. Applications**

# **A. Point like charge near a planar θ boundary**

A pointlike electric charge $q$ is located at a distance $b>0$ from the $z=0$ plane, where we have chosen $a=0$. Also, the region $z<0$ is filled with a topologically nontrivial insulator, whereas the region $z>0$ is the vacuum $\left(\theta_{2}=0\right)$. We choose the coordinates such that $x^{\prime}=y^{\prime}=0$. Therefore, the current density is $j^{\mu}\left(\mathbf{x}^{\prime}\right)=q \eta_{0}^{\mu} \delta\left(x^{\prime}\right) \delta\left(y^{\prime}\right) \delta\left(z^{\prime}-b\right)$. The solution to this problem is 

$$
A^{\mu}(\mathbf{x})=q G_{0}^{\mu}(\mathbf{x}, \mathbf{r}) \tag{33}
$$

where $\mathbf{r}=b \hat{z}$. We first study the electrostatic potential. We use Eqn (29) to solve this. Here, $R=|\left(\mathbf{x}-\mathbf{x}^{\prime}\right)_{\|}| = \sqrt{x^2+y^2}$ and $Z=|z-a|+\left|z^{\prime}-a\right|=|z|+b$. Also $\mathbf{x}' = b\hat{z} = \mathbf{r}$.

\begin{align}
z>0: & \; \; G_{0}^{0}(\mathbf{x}, \mathbf{r})=\frac{1}{|\mathbf{x}-\mathbf{r}|}-\frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}} \frac{1}{\sqrt{x^2+y^2+(b+z)^2}}=\frac{1}{|\mathbf{x}-\mathbf{r}|}-\frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}} \frac{1}{|\mathbf{x}+\mathbf{r}|}, 
\\
 z<0: & \; \; G_{0}^{0}(\mathbf{x}, \mathbf{r})=\frac{1}{|\mathbf{x}-\mathbf{r}|}-\frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}} \frac{1}{\sqrt{x^2+y^2+(b-z)^2}}=\frac{1}{|\mathbf{x}-\mathbf{r}|}-\frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}} \frac{1}{|\mathbf{x}-\mathbf{r}|}
\end{align}

$$
z>0: G_{0}^{0}(\mathbf{x}, \mathbf{r})=\frac{1}{|\mathbf{x}-\mathbf{r}|}-\frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}} \frac{1}{|\mathbf{x}+\mathbf{r}|}, \tag{34} \\
$$
$$
z<0: G_{0}^{0}(\mathbf{x}, \mathbf{r})=\frac{4}{4+\tilde{\theta}^{2}} \frac{1}{|\mathbf{x}-\mathbf{r}|} \tag{35}
$$

For $z>0$ the GF yields the electric potential $A^{0}(\mathbf{x})=$ $q G_{0}^{0}(\mathbf{x}, \mathbf{r})$ which can be interpreted as due to two pointlike electric charges, one of strength $q$ at $\mathbf{r}$, and the other, the image charge, of strength $-q \tilde{\theta}^{2} /\left(4+\tilde{\theta}^{2}\right)$, at the point $-\mathbf{r}$. For $z<0$ only one pointlike electric charge appears, of strength $4 q /\left(4+\tilde{\theta}^{2}\right)$ located at $\mathbf{r}$.

Now we calculate the components $G_{0}^{1}(\mathbf{x}, \mathbf{r})$ and $G_{0}^{2}(\mathbf{x}, \mathbf{r})$ from equation 30. For $G_{0}^{1}(\mathbf{x}, \mathbf{r})$ we get

$$
\begin{aligned}
   G_{i}^{0}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=&-\frac{2 \tilde{\theta}}{4+\tilde{\theta}^{2}} \frac{\epsilon_{0 1 2 3} R^{2}}{R^{2}}\left(1-\frac{Z}{\sqrt{R^{2}+Z^{2}}}\right) \\
   =&-\frac{2 \tilde{\theta}}{4+\tilde{\theta}^{2}} \frac{y}{R^{2}}\left(1-\frac{|z|+b}{\sqrt{(x-x')^2+y^2+(b-|z|)^2}}\right)
\end{aligned} 
$$

Thus we can simplify $G_{0}^{1}(\mathbf{x}, \mathbf{r})$ the regions $z>0$ and $z<0$ and similarly calculate $G_{0}^{2}(\mathbf{x}, \mathbf{r})$ as

$$
\begin{aligned}
& G_{0}^{1}(\mathbf{x}, \mathbf{r})=-\frac{2 \tilde{\theta}}{4+\tilde{\theta}^{2}} \frac{y}{R^{2}}\left\{
\begin{array}{cc}
1-\frac{z+b}{|\mathbf{x}+\mathbf{r}|}, & z>0 \\
1+\frac{z-b}{|\mathbf{x}-\mathbf{r}|}, & z<0
\end{array}\right. 
\end{aligned}
$$

$$
\begin{aligned}
& G_{0}^{2}(\mathbf{x}, \mathbf{r})=-\frac{2 \tilde{\theta}}{4+\tilde{\theta}^{2}} \frac{x}{R^{2}}\left\{
\begin{array}{cc}
1-\frac{z+b}{|\mathbf{x}+\mathbf{r}|}, & z>0 \\
1+\frac{z-b}{|\mathbf{x}-\mathbf{r}|}, & z<0
\end{array}\right. 
\end{aligned}
$$

Note that $G_{0}^{3}(\mathbf{x}, \mathbf{r})=0$ as $\epsilon_{0 3 j 3}=0$. From Eq. (33) we see that two components of the magnetic vector potential are nonzero, $A^{1}(\mathbf{x})=q G_{0}^{1}(\mathbf{x}, \mathbf{r})$ and $A^{2}(\mathbf{x})=q G_{0}^{2}(\mathbf{x}, \mathbf{r})$. 


The magnetic field $\mathbf{B}=\nabla \times \mathbf{A}$ can be calculated by taking the curl of 4-potential. We first calculate the z component which is given by $\partial_x A_y - \partial_y A_x$. In the region $z>0$, this is simplified as

$$
\begin{aligned}
    \frac{4+\tilde{\theta}^2}{2q\tilde{\theta}}B_z = \frac{\partial}{\partial x}\left(\frac{x}{R^2}-\frac{x}{R^2}\frac{z+b}{|\mathbf{x}+\mathbf{r}|}\right) &- \frac{\partial}{\partial y}\left(-\frac{y}{R^2}+\frac{y}{R^2}\frac{z+b}{|\mathbf{x}+\mathbf{r}|}\right)\\
    = \left(1+\frac{z+b}{|\mathbf{x}+\mathbf{r}|}\right)\left(\frac{\partial}{\partial x}\frac{x}{R^2}+\frac{\partial}{\partial y}\frac{y}{R^2}\right) &- \frac{x}{R^2}\frac{\partial}{\partial x}\frac{z+b}{|\mathbf{x}+\mathbf{r}|} - \frac{y}{R^2}\frac{\partial}{\partial y}\frac{z+b}{|\mathbf{x}+\mathbf{r}|} \\
    = \frac{x(z+b)}{R^2}\frac{-x}{|\mathbf{x}+\mathbf{r}|^3}&+\frac{y(z+b)}{R^2}\frac{-y}{|\mathbf{x}+\mathbf{r}|^3}  = -(z+b)\frac{1}{|\mathbf{x}+\mathbf{r}|^3}
\end{aligned}
$$

After calculating all the components, we find that the magnetic field in both regions is given by

$$
\begin{aligned}
& z>0: \mathbf{B}(\mathbf{x})=\frac{2 q \tilde{\theta}}{4+\tilde{\theta}^{2}} \frac{\mathbf{x}+\mathbf{r}}{|\mathbf{x}+\mathbf{r}|^{3}}, \\
& z<0: \mathbf{B}(\mathbf{x})=\frac{-2 q \tilde{\theta}}{4+\tilde{\theta}^{2}} \frac{\mathbf{x}-\mathbf{r}}{|\mathbf{x}-\mathbf{r}|^{3}}
\end{aligned}
$$

Thus, we observe that the magnetic field for $z>0$ is that due to a magnetic monopole of strength $2 q \tilde{\theta} /\left(4+\tilde{\theta}^{2}\right)$ located at $-\mathbf{r}$. For $z<0$ we have a magnetic monopole of strength $-2 q \tilde{\theta} /\left(4+\tilde{\theta}^{2}\right)$ located at $\mathbf{r}$.

The solution shows that, for an electric charge near the planar surface of a topological insulator, both an image magnetic charge and an image electric charge will be induced. The appearance of magnetic monopoles in this solution seems to violate the Maxwell law $\nabla \cdot \mathbf{B}=0$, which remained unaltered in the case of $\theta$ ED. Nevertheless, recalling that $(\mathbf{x} \pm \mathbf{r}) /|\mathbf{x} \pm \mathbf{r}|^{3} \sim$ $\nabla_{x}(1 /|\mathbf{x} \pm \mathbf{r}|)$, we have $\nabla \cdot \mathbf{B} \sim \nabla_{x}^{2}(1 /|\mathbf{x} \pm \mathbf{r}|) \sim \delta(\mathbf{x} \pm$ $\mathbf{r})$ in a region where $\mathbf{x} \neq \pm \mathbf{r}$.

We can also calculate the surface current density by first calculating the electric field which is given by $-\partial_\mu A^0$ in the static field case. Trivially, the electric field is calculated as 
$$
\mathbf{E} = \left\{
\begin{array}{cc}
\frac{q\mathbf{x}}{|\mathbf{x}-\mathbf{r}|^3}-\frac{\tilde{\theta}^2}{4+\tilde{\theta}^{2}}\frac{q\mathbf{x}}{|\mathbf{x}+\mathbf{r}|^3}, & z>0 \\
\frac{4 q}{4+\tilde{\theta}^{2}}\frac{\mathbf{x}}{|\mathbf{x}-\mathbf{r}|^3}, & z<0
\end{array}\right.
$$

On the surface, $z=0$, the electric field takes the value $\frac{4 q}{4+\tilde{\theta}^{2}}\frac{\mathbf{x}}{(R^2+b^2)^{3/2}}$. Physically, the magnetic field is induced by a surface current density

$$
\mathbf{J}=\tilde{\theta} \delta(z) \mathbf{E} \times \mathbf{n}=\frac{4 q \tilde{\theta}}{4+\tilde{\theta}^{2}} \frac{R}{\left(R^{2}+b^{2}\right)^{3 / 2}} \delta(z) \hat{\varphi}
$$

which is circulating around the origin. However, such an induced field has the correct magnetic field dependence expected from a magnetic monopole. This current is nothing but the Hall current.

# **B. Force between a charge and planar boundary**

The interaction energy between a charge-current distribution and a topological insulator is

$$E_{\mathrm{int}}=\frac{1}{2} \int d \mathbf{x} \int d \mathbf{x}^{\prime} j^{\mu}(\mathbf{x})\left[G_{\mu \nu}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)-\eta_{\mu \nu} \mathcal{G}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)\right] j^{\nu}\left(\mathbf{x}^{\prime}\right) \tag{42}$$

where $\mathcal{G}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=1 /\left|\mathbf{x}-\mathbf{x}^{\prime}\right|$ is the GF in vacuum.

The first contribution represents the total energy of a charge-current distribution in the presence of the $\theta$ boundary, including mutual interactions. We evaluate this energy for the case considered in the previous subsection of a pointlike electric charge at position $\mathbf{r}=b \hat{\mathbf{e}}_{z}$.

We know that for $z>0$,
$$G_{0}^{0}(x,r)=\frac{1}{|x-r|}-\frac{\tilde{\theta}^2}{4+\tilde{\theta}^2}\frac{1}{|x+r|}$$

And $j^μ(x)=qδ(x-r)$

The interaction energy is

$$
E_{\text {int }}=\frac{q^{2}}{2}\left[G_{00}(\mathbf{r}, \mathbf{r})-\mathcal{G}(\mathbf{r}, \mathbf{r})\right]=-\frac{q^{2}}{2} \frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}} \frac{1}{2 b} \tag{43}
$$

force on the charge exerted by the $\theta$ boundary is

$$
\mathbf{F}=-\frac{\partial E_{\text {int }}}{\partial b} \hat{\mathbf{e}}_{z}=-\frac{q^{2}}{(2 b)^{2}} \frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}} \hat{\mathbf{e}}_{z} \tag{44}
$$

The force is always attractive. 
This can be interpreted as force between two charges $q$ and $-q \frac{\tilde{\theta}^{2} }{\left(4+\tilde{\theta}^{2}\right)}$ according to Coulomb law

# Alternatively

Force in terms of Stress energy tensor,

$$
\mathbf{F}=-\hat{\mathbf{e}}_{z} \int_{\Sigma^{+}} d S T_{z z}\left(\Sigma^{+}\right) \tag{45}
$$

where the integration is over the surface $\Sigma^{+}$, just outside the $\theta$ interface, at $z=0^{+}$

# **Derivation of stress tensor in the case of $\theta$ electrodynamics**

We neglect the term coming due to magnetic field and using Maxwell's equation,
$$J.E = \frac{1}{\mu} \cdot (\nabla \times \mathbf{B} - \frac{1}{c^{2}}  \frac{\partial \mathbf{E}}{\partial t})\cdot \mathbf{E}
$$

$$J.E = \mathbf{E}.[\frac{c}{4\pi} \cdot \nabla \times \mathbf{H} - \frac{1}{4\pi}\frac{\partial \mathbf{D}}{\partial t}]\tag{i}$$
As $$\mathbf{B}=\mu \mathbf{H} ;\; \mathbf{D}=ϵ\mathbf{E} ;\; c^2=\frac{1}{ϵ_{0}μ_{0}}$$

using $\nabla\cdot(E\times H)=(\nabla\times E)\cdot H-(\nabla\times H)\cdot E$


$$\mathbf{J} \cdot \mathbf{E} = -\nabla \cdot (\frac{c}{4\pi} \mathbf{E} \times \mathbf{H}) - \frac{1}{4\pi}(\mathbf{E} \cdot \frac{\partial}{\partial t}\mathbf{D} + \mathbf{H} \cdot \frac{\partial}{\partial t} \mathbf{B})$$








We consider rate at which momentum is transferred to the charges, or equivalently, the force density, **f**,
$$\mathbf{f} = ρ\mathbf{E} + \frac{1}{c} \mathbf{J} \times \mathbf{B}$$

Relavant pair of operations here are,
$$
ρ\mathbf{E} + \frac{1}{c} \mathbf{J} \times \mathbf{B} = \frac{1}{4\pi}(\nabla \cdot \mathbf{D})\mathbf{E} + \frac{1}{4\pi}(\nabla \times \mathbf{H} - \frac{1}{c} \frac{\partial}{\partial t} \mathbf{D}) \times \mathbf{B} \tag{iii}$$
$$
0 = \frac{1}{4\pi}(\nabla \cdot \mathbf{B})\mathbf{H} + \frac{1}{4\pi}(\nabla \times \mathbf{E} - \frac{1}{c}\frac{∂}{∂t}\mathbf{B}) \times \mathbf{D}\tag{iv}
$$

Adding (iii) and (iv),
$$
ρ\mathbf{E} + \frac{1}{c} \mathbf{J} \times \mathbf{B} = -\frac{∂}{∂t}(\frac{1}{4\pi c}\mathbf{D} \times \mathbf{B}) - \frac{1}{4\pi}[D_i \nabla E_i - \nabla \cdot \mathbf{(DE)}] - \frac{1}{4π}[B_i \nabla H_i - \nabla \cdot \mathbf{(BH)}]
$$


**Rate at which momentum is tranferred to charges :**
$$\begin{aligned}
(\rho \mathbf{E}+\mathbf{J} \times \mathbf{B})_{k}= & -\frac{\partial}{\partial t}\left(\frac{1}{4 \pi} \mathbf{D} \times \mathbf{B}\right)_{k} \\
& -\frac{1}{4 \pi}\left[D_{i} \partial_{k} E_{i}-\partial_{i}\left(D_{i} E_{k}\right)\right] \\
& -\frac{1}{4 \pi}\left[B_{i} \partial_{k} H_{i}-\partial_{i}\left(B_{i} H_{k}\right)\right] .
\end{aligned}\tag{47}$$


From Eq.(8) and Eq(46),
$$\begin{aligned}Energy\;Flux : \mathbf{S}=\frac{1}{4\pi}\mathbf{E} \times \mathbf{B}\\
Energy\; Density : \mathbf{U} = \frac{1}{8\pi}(\mathbf{E^2} + \mathbf{B^2}) \end{aligned}\tag{48}$$

From time derivative in Eq.(47), we identify momentum density, G,
$$ \frac{1}{4\pi} E \times B
$$
If local momentum conservation law holds then the last two terms of Eq.(47) accounts for divergence of a tensor, representing flow of momentum,
$$\frac{1}{4 \pi}[D_{i} \nabla E_{i}-\nabla(D_{i} E_{k}) + B_{i} \nabla H_{i}-\nabla(B_{i} H_{k})] = ∇ ⋅ \mathbf{T}
$$
We identify stress tensor as,
$$T_{i j} =\frac{1}{8 \pi}(\mathbf{E}^{2}+\mathbf{B}^{2}) \delta_{i j}-\frac{1}{4 \pi}(E_{i} E_{j}+B_{i} B_{j}) \tag{49}
$$
The Stress tensor has same form as in vaccum but it is not conserved on $\theta$ boundary because of self induced charge and current densities are arising there.

We can write $T_{zz}$ for Eq.(45) as,
$$T_{zz} = \frac{1}{8\pi} [\mathbf{E_{||}^2} - \mathbf{E_{Z}^2} + \mathbf{B_{||}^2} - \mathbf{B_{Z}^2} ]\tag{51}
$$

$\mathbf{E_z (B_z)}$ denotes the electric (magnetic) field component normal to the surface and $\mathbf{E_{||} (B_{||})}$ is the component
of the electric (magnetic) field parallel to the surface.

Electric Field and Magnetic Field for $z>0$,
$$\mathbf{E(x)} = q\frac{x-r}{|x-r|^3} - q\frac{\theta^2}{4 + \theta^2}\frac{x-r}{|x-r|^3}\tag{52}$$
$$\mathbf{B(x)} = \frac{2q\theta}{4 + \theta^2}\frac{x-r}{|x-r|^3}\tag{53}
$$

Substituting Eq.(52) and (53) in (51) and then substituting the tensor in Eq.(45),
$$
\begin{aligned}
\mathbf{F}= & \frac{1}{4} \frac{q^{2}}{\left(4+\tilde{\theta}^{2}\right)^{2}} \hat{\mathbf{e}}_{z} \int_{0}^{\infty} d R \frac{R}{\left(R^{2}+b^{2}\right)^{3}} \\
& \times\left[16 R^{2}-\left(4+2 \tilde{\theta}^{2}\right)^{2} b^{2}+4 \tilde{\theta}^{2}\left(R^{2}-b^{2}\right)\right] \\
F = & -\frac{q^{2}}{(2 b)^{2}} \frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}} \hat{\mathbf{e}}_{z},
\end{aligned}
$$


# **C.  Infinitely straight current-carrying wire near a planar θ boundary**
Consider an infinitely straight wire parallel to the $x$ axis and carrying a current $I$ in the $+x$ direction. The wire is located in vacuum $\left(\theta_{2}=0\right)$ at a distance $b$ from an infinite topological medium with $\theta_{1} \neq 0$ in the region $z<0$. For simplicity, choose the coordinates such that $y^{\prime}=0$. Therefore, the current density is $j^{\mu}\left(\mathbf{x}^{\prime}\right)=I \eta_{1}^{\mu} \delta\left(y^{\prime}\right) \delta\left(z^{\prime}-b\right)$.

The solution for this problem is 

$$
A^{\mu}(\mathbf{x})=I \int_{-\infty}^{+\infty} G_{1}^{\mu}(\mathbf{x}, \mathbf{r}) d x^{\prime} \tag{54}
$$

where $\mathbf{x}-\mathbf{r}=\left(x-x^{\prime}\right) \hat{\mathbf{e}}_{x}+y \hat{\mathbf{e}}_{y}+(|z|+b) \hat{\mathbf{e}}_{z}$. The nonzero component $A^{0}(\mathbf{x})$ arising from the GF implies that an electric field is induced. The required component of the GF, $G_{1}^{0}$, from Eq. (30), is 

$G_{1}^{0}(\mathbf{x}, \mathbf{r})=-\frac{2 \tilde{\theta}}{4+\tilde{\theta}^{2}} \frac{y}{R^{2}}\left[1-\frac{|z|+b}{\sqrt{R^{2}+(|z|+b)^{2}}}\right] \tag{55}$.

Substituting Eq. (55) into Eq. (54) gives

$$
A^{\mu}(\mathbf{x})=I \int_{-\infty}^{+\infty} -\frac{2 \tilde{\theta}}{4+\tilde{\theta}^{2}} \frac{y}{R^{2}}(1 - \frac{|z|+b}{\sqrt{R^{2}+(|z|+b)^{2}}}) d x^{\prime}
$$

Splitting the integral into two parts, and redefining $x-x' = X$ and $(|z|+b)=Z$, we get
$$
(I): \int^{\infty}_{-\infty} \frac{y}{X^{2}+y^{2}}dX = tan^{-1}(\frac{X}{y})|^{\infty}_{-\infty} = \pi \\
(II): \int^{\infty}_{-\infty} \frac{yZ dX}{(X^{2}+y^{2})\sqrt{X^{2}+y^{2}+Z^{2}}} = yZ(\frac{tan^{-1}\frac{X|Z|}{|y|\sqrt{X^{2}+y^{2}+Z^{2}}}}{y|Z|})|^{\infty}_{-\infty} = 2tan^{-1}(\frac{|z|+b}{y})
$$
$$
Therefore,\ A^{0}(\mathbf{x}) = (-\pi + 2tan^{-1}(\frac{|z|+b}{y}))\frac{2\tilde{\theta}I}{4+\tilde{\theta}^{2}} \\
Therefore,\ \mathbf{E}(\mathbf{x})=-\nabla A^{0}(\mathbf{x})
$$

$\mathbf{E}(\mathbf{x})=\frac{4 \tilde{\theta} I}{4+\tilde{\theta}^{2}}\left[\frac{|z|+b}{y^{2}+(|z|+b)^{2}} \hat{\mathbf{e}}_{y}-\frac{y \operatorname{sign}(z)}{y^{2}+(|z|+b)^{2}} \hat{\mathbf{e}}_{z}\right] \tag{56}$

Similarly, we compute the magnetic field,
$$\mathbf{B}(\mathbf{x})=\nabla \times\left[I \hat{\mathbf{e}}_{i} \int_{-\infty}^{+\infty} G^{i}_1(\mathbf{x}, \mathbf{r}) d x^{\prime}\right]$$
with $i=1,2$, where the corresponding GF are given by Eqs. (31). The result is


The relevant components of GF are given by
$$
G_{1}^{i}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=\eta_{1}^{i} G_{0}^{0}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)-\frac{i}{2} \frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}} \partial^{i} K_{1}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)
$$
where

$$
G_{0}^{0}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=\frac{1}{\left|\mathbf{x}-\mathbf{x}^{\prime}\right|}-\frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}} \frac{1}{\sqrt{R^{2}+Z^{2}}}
$$

$$
{K}_1\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=2 i \frac{\sqrt{R^{2}+Z^{2}}-Z}{R^{2}} (x-x')
$$

where $R^2=(x-x')^2+y^2$, also $Z=|z|+b, and  |\mathbf{x-x'}|=\sqrt{(x-x')^2+y^2+(z-b)^2}$. Integrating $G^i_1$ and using the property $\int \frac{1}{\sqrt{X^2+a^2}} = ln(X+\sqrt{X^2+a^2})$ we get

$$
\begin{aligned}
\mathbf{B}(\mathbf{x})= & 2 I \operatorname{sign}(z)\left[-\frac{|b-z|}{y^{2}+(b-z)^{2}}\right. \\
& \left.+\frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}} \frac{(|z|+b)}{y^{2}+(|z|+b)^{2}}\right] \hat{\mathbf{e}}_{y} \\
& +2 I y\left[\frac{1}{y^{2}+(b-z)^{2}}-\frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}} \frac{1}{y^{2}+(|z|+b)^{2}}\right] \hat{\mathbf{e}}_{z} .
\end{aligned}
$$




For $z>0$ the magnetic field corresponds to the one produced by an image electric current located at $z^{\prime}=-b$, flowing in the opposite direction to the current of the wire, $\mathbf{j}_{e,>}=-2 \tilde{\theta} I /\left(4+\tilde{\theta}^{2}\right) \hat{\mathbf{e}}_{x}$. For $z<0$ we have an electric current located at $z^{\prime}=b$ of the same strength and flowing in the same direction of the current in the wire.


##Tutorial problem
**Pointlike charge near a planar θ boundary at zero potential**

Consider a pointlike electric charge in front of an infinite planar $\theta$ boundary at zero potentials $\left(A^{0}, \mathbf{A}_{\|}\right)=(0, \mathbf{0})$ located at $z=0$. Derive the greens fuctions for this system and interpret the results.

This case falls under the BC-I class of boundary conditions at the $\theta$ interface. We demand the full Green function $\left(G_{\mathrm{I}}\right)^{\mu}{ }_{\nu}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=0$ for $\mathbf{x'}$ at the θ interface. No additional bounding surfaces are considered.

This problem can be solved by the method of images. Following similar steps as in the standard case, the GF for this example, $\left(G_{\mathrm{I}}\right)^{\mu}{ }_{\nu}{ }^{\prime}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)$, can be constructed as

$$
\left(G_{\mathrm{I}}\right)^{\mu}{ }_{\nu}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=G^{\mu}{ }_{\nu}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)+F^{\mu}{ }_{\nu}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)
$$

with the addition of a matrix $F^{\mu}{ }_{\nu}$ which satisfies the homogeneous equation

$$
\left[-\eta^{\mu}{ }_{\nu} \nabla^{2}-\tilde{\theta} \delta(z) \epsilon^{3 \mu \alpha}{ }_{\nu} \partial_{\alpha}\right] F^{\nu}{ }_{\sigma}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=0
$$

This freedom in the definition of the GF allow us to choose appropriately the matrix ${F^{\mu}}^{\nu}$ in such a way that $\left(G_{\mathrm{I}}\right)^{\mu}{ }_{\nu}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=0$ for $\mathbf{x}^{\prime}$ on $\Sigma$. Let us consider a pointlike electric charge located at $z^{\prime}>0$. The required components of the GF matrix for $z>0$ are

$$
\begin{gathered}
G^{0}{ }_{0}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=\frac{1}{\left|\mathbf{x}-\mathbf{x}^{\prime}\right|}-\frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}} \frac{1}{\left|\mathbf{x}-\mathbf{x}^{\prime \prime}\right|}, \\
G^{0}{ }_{i}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=-\frac{2 \tilde{\theta}}{4+\tilde{\theta}^{2}} \frac{\epsilon_{0 i j 3}\left(\mathbf{x}-\mathbf{x}^{\prime}\right)_{\|}^{j}}{R^{2}}\left(1-\frac{z+z^{\prime}}{\left|\mathbf{x}-\mathbf{x}^{\prime \prime}\right|}\right)
\end{gathered}
$$

and for $z<0$ are

$$
\begin{gathered}
G^{0}{ }_{0}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=\frac{4}{4+\tilde{\theta}^{2}} \frac{1}{\left|\mathbf{x}-\mathbf{x}^{\prime}\right|}, \\
G^{0}{ }_{i}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=-\frac{2 \tilde{\theta}}{4+\tilde{\theta}^{2}} \frac{\epsilon_{0 i j 3}\left(\mathbf{x}-\mathbf{x}^{\prime}\right)_{\|}^{j}}{R^{2}}\left(1+\frac{z-z^{\prime}}{\left|\mathbf{x}-\mathbf{x}^{\prime \prime}\right|}\right)
\end{gathered}
$$

where $\mathbf{x}^{\prime}=\left(x^{\prime}, y^{\prime}, z^{\prime}\right)$ and $\mathbf{x}^{\prime \prime}=\left(x^{\prime}, y^{\prime},-z^{\prime}\right)$ indicate the position of the charge and image charge respectively.

Hence we obtain the full GF for $z>0$ as

$$
\begin{aligned}
\left(G_{\mathrm{I}}\right)^{0}{ }_{0}\left(\mathbf{x}, \mathbf{x}^{\prime}\right) & =G^{0}{ }_{0}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)-G^{0}{ }_{0}\left(\mathbf{x}, \mathbf{x}^{\prime \prime}\right) \\
& =\left(1+\frac{\tilde{\theta}^{2}}{4+\tilde{\theta}^{2}}\right)\left(\frac{1}{\left|\mathbf{x}-\mathbf{x}^{\prime}\right|}-\frac{1}{\left|\mathbf{x}-\mathbf{x}^{\prime \prime}\right|}\right) \\
\left(G_{\mathrm{I}}\right)^{0}{ }_{i}\left(\mathbf{x}, \mathbf{x}^{\prime}\right) & =G^{0}{ }_{i}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)-G^{0}{ }_{i}\left(\mathbf{x}, \mathbf{x}^{\prime \prime}\right) \\
& =\frac{2 \tilde{\theta} }{4+\tilde{\theta}^{2}}\frac{\epsilon_{0 i j 3}\left(\mathbf{x}-\mathbf{x}^{\prime}\right){ }_{\|}^{j}}{R^2}\left(\frac{z+z^{\prime}}{\left|\mathbf{x}-\mathbf{x}^{\prime \prime}\right|}-\frac{z-z^{\prime}}{\left|\mathbf{x}-\mathbf{x}^{\prime}\right|}\right)
\end{aligned}
$$

and for $z<0$ as
$$
\begin{aligned}
\left(G_{\mathrm{I}}\right)^{0}{ }_{0}\left(\mathbf{x}, \mathbf{x}^{\prime}\right) & =G^{0}{ }_{0}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)-G^{0}{ }_{0}\left(\mathbf{x}, \mathbf{x}^{\prime \prime}\right) \\
& =\left(\frac{4}{4+\tilde{\theta}^{2}}\right)\left(\frac{1}{\left|\mathbf{x}-\mathbf{x}^{\prime}\right|}-\frac{1}{\left|\mathbf{x}-\mathbf{x}^{\prime \prime}\right|}\right) \\
\left(G_{\mathrm{I}}\right)^{0}{ }_{i}\left(\mathbf{x}, \mathbf{x}^{\prime}\right) & =G^{0}{ }_{i}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)-G^{0}{ }_{i}\left(\mathbf{x}, \mathbf{x}^{\prime \prime}\right) \\
& =\frac{2 \tilde{\theta} }{4+\tilde{\theta}^{2}}\frac{\epsilon_{0 i j 3}\left(\mathbf{x}-\mathbf{x}^{\prime}\right){ }_{\|}^{j}}{R^2}\left(-\frac{z+z^{\prime}}{\left|\mathbf{x}-\mathbf{x}^{\prime \prime}\right|}+\frac{z-z^{\prime}}{\left|\mathbf{x}-\mathbf{x}^{\prime}\right|}\right)
\end{aligned}
$$


which effectively vanish at $z^{\prime}=0$. 

###Interpretation

The electric field in the region $z>0$ is generated by four electric charges: (i) the original charge of unit strength at $z^{\prime}$, (ii) an equal and opposite charge located at the mirrorimage point, (iii) an electric charge of strength $-\tilde{\theta}^{2} /\left(4+\tilde{\theta}^{2}\right)$ located at the mirror-image point and (iv) an electric charge of strength $+\tilde{\theta}^{2} /\left(4+\tilde{\theta}^{2}\right)$ located at $z^{\prime}$. The magnetic field can be interpreted as that produced by two monopoles, one of strength $2 \tilde{\theta} /\left(4+\tilde{\theta}^{2}\right)$ located at the mirror-image point, induced by the $\theta$ boundary, and the other of strength $-2 \tilde{\theta} /\left(4+\tilde{\theta}^{2}\right)$ at $z^{\prime}$, arising from the Boundary condition $\left. \mathbf{A}_{\|}\right|_{\Sigma}=0$.

In the region $z<0$ is due to an electric charge of strength $-4 /\left(4+\tilde{\theta}^{2}\right)$ located at $z^{\prime}$ plus an electric charge of strength $+4 /\left(4+\tilde{\theta}^{2}\right)$ located at $-z^{\prime}$. The magnetic field can be interpreted as being generated by two magnetic monopoles, one of strength $-2 \tilde{\theta} /\left(4+\tilde{\theta}^{2}\right)$ located at $z^{\prime}$ together with its image of strength $+2 \tilde{\theta} /\left(4+\tilde{\theta}^{2}\right)$ located at $-z^{\prime}$.

#Conclusion
We have discussed  a
particular kind of Chern-Simons extension to electrodynamics consisting of Maxwell Lagrangian supplemented by a parity-violating Pontryagin invariant coupled to a scalar field $\theta$, which is piecewise constant in different regions of space separated by a common interface $\Sigma$. The discontinuity of θ at the interface gives rise to various effects such as induced effective charges and currents, magnetic mirror monopoles, and additional Faraday- and Kerr-like rotation of the plane of polarization of electromagnetic waves transversing the interface $\Sigma$. We also focus on the application of the Green's function (GF) method for the static case in θ electrodynamics, specifically for the case of a planar θ interface. The GF is used to compute the electromagnetic fields on either side of the interface and the force between a pointlike charge and a topological insulator. The results we obtained are consistent with previously existing results and can be extended to other geometries determined by the θ boundary. Our results depend on $\tilde{\theta} = \alpha(\theta_1 - \theta_2)$, and they satisfy the quantum-mechanical periodicity condition
$\theta \rightarrow \theta + 2\pi n$ with $\theta = 0,\pi$. The GF method may also be useful for the dynamic case and for computing other effects such as the Casimir effect. Here, we have made efforts in the context of topological insulators.
