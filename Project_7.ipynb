## I. Elementary Approach to a Relativistic Lagrangian 

First we will use a lesser sophisticated method with ordinary coordinates, velocities, and time and generalize from the nonrelativistic domain in a straightforward way. 

From the first postulate of special relativity the action integral must be a Lorentz scalar because the equations of motion are determined by the extremum condition, $\delta A= 0$. 

Also using $d\tau = \frac{dt}{\gamma(u)}$ we get the action integral as 
$$A = \int _{\tau_{1}}^{\tau_{2}} \gamma Ld\tau$$
We thus obtain that $\gamma L$ will be Lorenz invariant. 


The Lagrangian for a free particle can be a function of its velocity & mass, but
not of its position. The only Lorentz invariant of the velocity available is $U_{\alpha} U^{\alpha} = c^{2}$. Thus we conclude that  Lagrangian for a free particle is proportional to $\gamma ^{-1} = \sqrt{1-\beta ^{2}}$ where $\beta = \frac{u}{c}$. Using the Euler-Lagrange equation, we find that 
$$L_{free} = -mc^{2}\gamma ^{-1}$$ yiels the free - particle equation of motion

This motion at constant velocity is the solution of the free-particle eqn of motion.

A slowly moving charged particle is influenced predominantly by the electric field that is derivable from the scalar potential $\phi$. The potential energy of interaction is $V=e\phi$. Since the nonrelativistic Lagrangian is (T-V), the interaction part $L_{int}$, of the relativistic Lagrangian must reduce in the nonrelativistic limit to
$$ L_{int} ‚Üí L^{NR}_{int} = -e\phi$$

 Since $\phi$ is the time
component of the 4-vector potential $A^{\alpha}$, we anticipate that $\gamma L_{int}$ will involve the scalar product of $A^{\alpha}$ with some 4-vector. The only other 4-vectors available are the momentum and position vectors of the particle. Since gamma times the Lagrangian must be translationally invariant as well as Lorentz invariant, it cannot involve the coordinates explicitly. Hence the interaction Lagrangian must be
$$ L_{int} = - \frac{e}{\gamma c} U_{\alpha}A^{\alpha}$$

Combining Free particle and Interaction parts of the lagrangian, we get
$$L = -mc^{2}\sqrt{1 -\frac{u^{2}}{c^{2}}} -e\phi + \frac{e}{c}\mathbf{u¬∑A}$$

The canonical momentum $\mathbf{P}$ conjugate to the position coordinate $\mathbf{x}$ is obtained by the definition:

Substituting $\mathbf{u}$ in $L$ and $H$, we get the Hamiltonian as,

Writing $H=W$, we get the total energy as,

We get the four vector scalar product,

$$ p^\alpha \equiv \bigg(\frac{E}{c}, \mathbf{p}\bigg) = \bigg(\frac{1}{c} (W-e\phi), \mathbf{P}-\frac{e}{c}\mathbf{A}\bigg)$$

## II. Covariant Treatment of the Relativistic Lagrangian 

We take the 4-vector forms the quantities used above.

$$
L_{free} = \frac{mc}{\gamma} \sqrt{ U_\alpha U^\alpha} 
$$

The action integral in this case will be:
$$
A = -mc \int_{{\tau_1}}^{{\tau_2}} \sqrt{U_\alpha U^\alpha} d\tau
$$

We have the equation of constraint,
$$ 
U_\alpha U^\alpha = c^2 
$$ 

or equivalently,

$$ 
U_\alpha \frac{dU^\alpha}{d\tau} = 0 
$$ 
This can be incorporated by the Lagrange multiplier technique, but we pursue a different, equivalent procedure

We have the integrand,
$$ \sqrt{U_\alpha U^\alpha} d\tau = \sqrt{\frac{dx_\alpha}{d\tau} \frac{dx^\alpha}{d\tau}} d\tau = \sqrt{dx_\alpha dx^\alpha} $$

This is nothing but the infinitesimal length element in 4-space. 
We define $s$ as an arbitrary parameter that is a monotonically increasing function of $\tau$, to put into the action integral.

$$
A = -mc \int^{s_2}_{s_1} \sqrt{\frac{dx_\alpha}{ds} \frac{dx^\alpha}{ds}} ds 
$$

The Lagrangian variables are now $x^\alpha$ and "velocity" $\frac{dx^\alpha }{ds}$, though we consider $s$ to be arbitrary here.  

$$
\sqrt{ \frac{dx_\alpha}{ds} \frac{dx^\alpha}{ds}} ds  = c d\tau
$$

Since the Lagrangian Integrand here doesnt depend on $x_\alpha$, we obtain the Eulers-Lagrange Equation as
$$
mc \frac{d}{ds} \bigg( \frac{dx^\alpha / ds}{\sqrt{\frac{dx_\alpha}{d\tau} \frac{dx^\alpha}{d\tau}} } \bigg) = 0
$$
or 
$$
m\frac{d^2x^\alpha}{d\tau^2} = 0
$$

For a charged particle in an external field, we saw the lagrangian to be,
$$
L_{int} = -e\phi + \frac{e}{c}\mathbf{u¬∑A}
$$

Thus, the manifestly covariant form of action integral will be 
$$
A = - \int^{s_2}_{s_1} \bigg[mc \sqrt{\frac{dx_\alpha}{ds} \frac{dx^\alpha}{ds}} +  \frac{e}{c} \frac{dx_\alpha}{ds} A^\alpha (x)\bigg]ds 
$$
We get the Euler's-Lagrange Equatons using Hamilton's principle,
$$
\frac{d}{ds}\bigg[\frac{\partial \tilde{L}}{\partial \big(\frac{dx_\alpha}{ds}\big)}\bigg] - \partial ^\alpha \tilde{L} = 0 
$$

Where the lagrangian is,
$$
\tilde{L} = - \bigg[mc \sqrt{\frac{dx_\alpha}{ds} \frac{dx^\alpha}{ds}} +  \frac{e}{c} \frac{dx_\alpha}{ds} A^\alpha (x)\bigg]            
$$ 

dividing the Euler-Lagrange equation by $c \frac{d\tau}{ds}$ we get 
$$ 
m\frac{d^2x^\alpha}{d\tau^2} + \frac{e}{c} \frac{dA^\alpha(x)}{d\tau} -\frac{e}{c} \frac{dx_\beta}{d\tau} \partial^\alpha A^\beta = 0
$$

Since
$ \frac{dx_\beta}{d\tau} \partial^\beta A^\alpha = \frac{dA^\alpha}{d\tau}$, we can rewrite the above equation as,
$$
m\frac{d^2x^\alpha}{d\tau^2} = \frac{e}{c} (\partial^\alpha A^\beta - \partial^\beta A^\alpha)\frac{dx_\beta}{d\tau}
$$
This is the covariant equation of motion that we introduced in the beginning.

Now, the conjugate momentum 4-vector is defined as 
$$
P^\alpha = -\frac{\partial \tilde{L}}{\partial(\frac{dx_\alpha}{ds})} = mU^\alpha + \frac{e}{c} A^\alpha
$$

The Hamiltonian can be defined by 
$$ \tilde{H} = P_\alpha U^\alpha + \tilde{L}$$

Just like before we eliminate $U^\alpha$ using conjugate 4-momentum expression and obtain,
$$
\tilde{H} = \frac{1}{m} \bigg(P_\alpha - \frac{eA_\alpha}{c} \bigg)\bigg(P^\alpha - \frac{eA^\alpha}{c} \bigg) -c \sqrt{\bigg(P_\alpha - \frac{eA_\alpha}{c} \bigg)\bigg(P^\alpha - \frac{eA^\alpha}{c} \bigg)}

and
$$
\frac{dP^\alpha}{d\tau} = - \frac{\partial \tilde{H}}{\partial x_\alpha } = \frac{e}{mc} \bigg(P_\beta - \frac{eA_\beta}{c} \bigg) \partial ^\alpha A^\beta

where we have used the constraint

## **Particle Drifts in Nonuniform, Static Magnetic Fields**

We seek to understand how particles behave in spatially varying magnetic fields. We consider variations gentle enough such that a perturbation solution to the motion is an adequate approximation. "Gentle enough" generally means that the distance over which $B$ changes appreciably in magnitude or direction is large compared to the gyration radius $a$ of the particle. Then the lowest-order approximation to the motion is a spiraling around the lines of force at a frequency given by the local value of the magnetic induction. In the next approximation, slow changes occur in the orbit which can be described as a drifting of the guiding center.

We first consider a spatial variation where the the gradient is perpendicular to the field at any point. Let the direction of the gradient at a point be the direction of the unit vector $\vec{n}$. Then, $\vec{n}‚ãÖ\vec{B}=0$. The gyration frequency is written as

$$\vec{\omega _\vec{B}}(\vec{x})=\dfrac{e}{\gamma mc}\vec{B}(\vec{x})$$

Expanding the above expression with respect to the origin, upto first order we get,

$$\vec{\omega _\vec{B}}(\vec{x})=\vec{\omega_0}\left[1+\dfrac{1}{B_0}\left(\dfrac{\partial B}{\partial n}\right)_0 \vec{n}\cdot\vec{x}\right]$$

where $\omega_0 = eB_0/\gamma mc$ is the gyration frequency at origin. The direction of the magnetic field is constant, and hence the motion parallel to it remains uniform translation. We only have to solve for the modifications in transverse motion. Let the transverse velocity be $\vec{v}_\perp = \vec{v_0} + \vec{v_1}$, where $\vec{v_0}$ is the uniform field transverse velocity, and $\vec{v_1}$ is the small correction term.  
We have,
$$\dfrac{d\vec{v}_{\perp}}{dt}=\vec{v}_{\perp}\times \vec{\omega _\vec{B}}(\vec{x})$$

Substituting the value of $\vec{\omega _\vec{B}}(\vec{x})$ obtained earlier, and keeping only the first order terms, we get

$$\dfrac{d\vec{v_1}}{dt} \simeq \left [\vec{v_1} + \vec{v_0} (\vec{n}\cdot\vec{x_0})\dfrac{1}{B_0} \left(\dfrac{\partial B}{\partial n}\right)_0 \right ] \times \vec{\omega_0}$$

For a uniform field, the transverse velocity $\vec{v_0}$ and coordinate $\vec{x_0}$ are related as $\vec{v_0}=-\vec{\omega_0} \times \vec{x_0}$. Substituting this in the above equation we get,

$$\dfrac{d\vec{v_1}}{dt} \simeq \left [\vec{v_1} - \vec{\omega_0} \times \vec{x_0} (\vec{n}\cdot\vec{x_0})\dfrac{1}{B_0} \left(\dfrac{\partial B}{\partial n}\right)_0 \right ] \times \vec{\omega_0}$$

The first term here leads to an oscillatory term of the velocity, and the second leads to a non-oscillatory term with a non-zero average value,

$$\vec{v_G} \equiv \langle \vec{v_1} \rangle = \dfrac{1}{B_0} \left(\dfrac{\partial B}{\partial n}\right)_0 \vec{\omega_0} \times \langle
 (\vec{x_0})_\perp (\vec{n}\cdot\vec{x_0}) \rangle$$

Note that the rectangular components of $\vec{x_0}$ oscillate sinusoidally with peak amplitude $a$ and phase difference $90¬∞$. Hence, $\langle (\vec{x_0})_\perp (\vec{n}\cdot\vec{x_0}) \rangle = \dfrac{a^2}{2}\vec{n}$, and the average becomes

$$\vec{v_G} = \dfrac{a^2}{2} \dfrac{1}{B_0} \left(\dfrac{\partial B}{\partial n}\right)_0 \vec{\omega_0} \times \vec{n}$$

Independent of coordinates, this can be written as

$$\dfrac{\vec{v_G}}{\omega_B a} = \dfrac{a}{2B^2} \left ( \vec{B} \times \nabla_\perp \vec{B} \right ) $$

From this equation, it is evident that if $\left | \nabla B /B \right |$ << $1$, the drift velocity is small compared to the orbital velocity.

The first order approximation can be understood as - the particle tends to spiral along a field line, while the field line curves to the side. We can get the same result through a more straightforward approach by directly solving the Lorentz force equations. If we choose cylindrical coordinates with the center of curvature as origin, then the magnetic induction has only the $\phi$ component, with $B_\phi = B_0$. The Lorentz force equations are then given by

$$\gamma m ((\ddot{\rho}-\rho\dot{\phi}^2)\hat{\rho}+(\rho \ddot{\phi} + 2\dot{\rho}\dot{\phi})\hat{\phi} + \ddot{z}\hat{z}) = \dfrac{e}{c}(\dot{\rho}\hat{\rho}+\rho\dot{\phi}\hat{\phi}+\dot{z}\hat{z})\times B_0 \hat{\phi}$$

From this equation, we get the following three equations,

$$\ddot{\rho}-\rho\dot{\phi}^2 = -\omega _B \dot{z}$$
$$\rho \ddot{\phi} + 2\dot{\rho}\dot{\phi} = 0$$
$$\ddot{z} = \omega_B \dot{\rho}$$

If the zeroth order trajectory is a helix with radius $a$ smaller than the radius of curvature $R$, then to the lowest order, $\dot{\phi}\simeq v_\parallel /R$, while $\rho \simeq R$. Thus, the first equation here yields

$$\dot{z} = \dfrac{v_\parallel ^2}{\omega_B R}$$

which is basically the curvature drift derived earlier.

For regions of space with no currents present, from Maxwell's equations we have $\nabla \times \vec{B} = 0$. And for two-dimensional fields as in the right-hand side of the figure above, this implies

$$\dfrac{\nabla _\perp \vec{B}}{B} = -\dfrac{\vec{R}}{R^2}$$

Substituting this into the equation of $\vec{v_G}$, and adding it to $\vec{v_C}$, we obtain the simple expression

$$\vec{v_D}=\dfrac{1}{\omega_B R}(v_\parallel ^2 + \dfrac{1}{2}v_\perp ^2)\dfrac{\vec{R}\times \vec{B}}{RB}$$

where $v_\perp = \omega_B a$ is the transverse velocity of gyration.

## **Adiabatic Invariance of Flux Through Orbit of Particle**

In the previous section we have dealt with the cases in which the magnetic field was perpendicular to the motion of the particle. To complete our general survey of particle motion in magnetic fields, we must consider motion parallel to the lines of force. It turns out that for slowly varying fields a powerful tool is the concept of adiabatic invariants. In celestial mechanics and in the old quantum theory, adiabatic invariants were useful in discussing perturbations on the one hand, and in deciding what quantities were to be quantized on the other. Our discussion will resemble most closely the celestial-mechanical problem, since we are interested in the behavior of a charged particle in slowly varying fields, which can be viewed as small departures from the simple, uniform, static field considered in previous sections.

The concept of adiabatic invariance is introduced by considering the action integrals of a mechanical system. If q and p, are the generalized canonical coordinates and momenta, then, for each coordinate which is periodic, the action integral J; is defined by

$$J_{i} = \oint p_{i}dq_{i}$$

If now the properties of the system are changed in some way (e.g., a change in spring constant or mass of some particle), the question arises as to how the action integrals change. It can be proved* that, if the change in property is slow compared to the relevant periods of motion and is not related to the periods (such a change is called an adiabatic change), the action integrals are invariant. This means that, if we have a certain mechanical system in some state of motion and we make an adiabatic change in some property so that after a long time we end up with a different mechanical system, the final motion of that different system will be such that the action integrals have the same values as in the initial system.

## **Lowest Order Relativistic Correction to the Lagrangian for Intracting Charge Particles: The Darwin Lagrangian**

Till now we discussed the general Lagrangian formalism for a relativistic particle in external electromagnetic fields described by vector and scalar potentials, *A* and $\phi$. 

The lagrangian is supposed to be a function of instantaneous position and velocity of the particles. When the finite velocity of propagation of electromagnetic field is taken into account, this is no longer possible, since the values of the potentials at one particle due to the other particles depend on their state of motion at "retarded" times. Only when the retarded effects can be neglected is a Lagrangian description in terms of instantaneous position and velocity possible. 

Now we will look into the lowest order relativistic corrections can be included, giving an approximate Lagrangian for interacting particles, correct to order of $(v/c)^2$ inclusive.

In general there will be relativistic corrections to both $\phi_{12}$ and $A_{12}$. But int the *Coulomb gauge*, the scalar potential is given correctly to all orders in $v/c$ by the instantaneous Coulomb potential, thus we calculate in this gauge.

If we are just calculating lowest order of relativistic corrections, retarded effects can be neglacted while computing $A_{12}$. The reason is vector potential enters the Lagrangian (12.11 Jackson) in the combination $q_1(v_1/c).A_{12}$. Since $A_{12}$ itself is of the order of $v_2/c$, greater accuracy in calculating $A_{12}$ is unnecessary. Hence 

$$
    A_{12} \approx \frac{1}{c}\int \frac{J_t(x')d^3x'}{|x_1 - x'|}
$$
where $J_t$ is transverse part of the current density due to second particle

$$J(x') = q_2v_2\delta(x' - x_2)$$
$$J(x') = J_l(x') + J_t(x')$$

$$
J_t = \frac{1}{4\pi} \nabla \times \nabla\times\int \frac{J}{|x - x'|}d^3x' \qquad (6.28 \ \text{Jackson})
$$

$$J_l(x') = \epsilon_0 \nabla' \frac{\partial \Phi}{\partial t} \qquad (6.29 \ \text{Jackson})$$

Using

$$\Phi(x') = \frac{1}{4\pi\epsilon_0}\frac{q_2}{|x' - x_2|}$$
and the chain rule for $|x' - x_2| = \sqrt{(x' - x_2).(x' - x_2)}$

$$\frac{\partial\Phi}{\partial t} = - \frac{1}{4\pi\epsilon_0}\frac{q_2}{|x' - x_2|^3}(x' - x_2).\left (-\frac{dx_2}{dt}\right) = {\frac{1}{4\pi\epsilon_0}\frac{q_2}{|x' - x_2|^3}(x' - x_2).v_2}$$

$$J_t(x') = q_2v_2\delta(x' - x_2) - \frac{q_2}{4\pi} \nabla '\left[\frac{(x' - x_2).v_2}{|x' - x_2|^3}\right] \quad \quad \quad \quad (12.79)$$

$$A_{12} \approx \frac{1}{c} \int d^3x' \frac{1}{|x' - x_1|}\left [ q_2v_2\delta(x' - x_2) - \frac{q_2}{4\pi} \nabla '\left[\frac{(x' - x_2).v_2}{|x' - x_2|^3}\right]\right ]$$
$$= \frac{q_2v_2}{rc} - \frac{q_2}{4\pi c}\int d^3x' \frac{1}{|x' - x_1|}\nabla '\left[\frac{(x' - x_2).v_2}{|x' - x_2|^3}\right]$$

Solving the integral

$$I = \int d^3x' \frac{1}{|x' - x_1|}\nabla '\left[\frac{(x' - x_2).v_2}{|x' - x_2|^3}\right]$$

Setting $y = x' - x_2 \quad \& \quad x' - x_1 = x' - x_2 + x_2 - x_1 = y -r$

Now we have 
$$
    I = \int d^3y \frac{1}{|y - r|} \nabla_y \left( \frac{v_2.y}{y^3} \right)
$$
Use integration by part and thrown away the surface at infinity

Now

$$
    I = -\int d^3y   \left( \frac{v_2.y}{y^3} \right) \nabla_y \left (\frac{1}{|y - r|} \right )
$$

The gradient $\nabla_r \sim -\nabla_r$ action on $y - r$ 

$$
    I = \int d^3y   \left( \frac{v_2.y}{y^3} \right) \nabla_r \left (\frac{1}{|y - r|} \right )
$$

Hence now we can get $\nabla_r$ outside of integral

$$
    I = \nabla_r\int d^3y   \left( \frac{v_2.y}{y^3} \right)  \left (\frac{1}{|y - r|} \right ) \
    = {\nabla_r v_2. \int d^3y   \left( \frac{y}{y^3} \right)  \left (\frac{1}{|y - r|} \right )}
$$

Now consider the integral
$$K_i = \int d^3y \left( \frac{y_i}{y^3} \right)\left ( \frac{1}{|y - r|} \right)$$

In term of spherical coordinate $\textbf{y} = (y, \theta, \phi)$ with the $z$ axis along r

$$K_1 = \int_0 ^\infty y^2dy \int _{-1}^1 d \cos \theta \int_0^{2\pi} d\phi \frac{y \sin \theta \cos \phi }{y^3} \frac{1}{\sqrt{y^2 + r^2 -2yr\cos\theta}} = {0}$$

since 

$$\int_0^{2\pi} d\phi \cos \phi = 0$$

Similarly,

$$K_2 ={0} \quad \quad \text{since} \int_0^{2\pi} d \phi \sin \phi= 0$$

Now,

$$K_3 = 2\pi \int_0^\infty dy \int_{-1}^1 d\cos\theta \frac{\cos \theta}{\sqrt{y^2 + r^2 - 2ry\cos\theta}}$$

$$K_3 = -2\pi\int_0^\infty dy \frac{2}{12y^2r^2} \left\{ \left[2(y^2 + r^2) + 2yr\cos\theta\right] \sqrt{y^2 + r^2 - 2yr\cos\theta} \right\}_{-1}^1$$


$$K_3 = \frac{-2\pi}{3r^2}\int_0^\infty dy \frac{1}{y^2} \left\{ \left[(y^2 + r^2) + yr\cos\theta\right] \sqrt{y^2 + r^2 - 2yr\cos\theta} \right\}_{-1}^1$$

$$K_3 = \frac{-2\pi}{3r^2}\int_0^\infty dy \frac{1}{y^2} \left\{ (y^2 + r^2 + yr)\sqrt{y^2 + r^2 - 2yr} - (y^2 + r^2 - yr)\sqrt{y^2 + r^2 + 2yr}  \right\}$$

$$\sqrt{y^2 + r^2 -2yr} = \LARGE \{_{r-y \quad \text{if} \quad r > y} ^{y - r \quad \text{if} \quad y > r}$$

Hence

$$K_3 =     -\frac{2\pi}{3r^2}\left\{ \int_0^rdy\frac{1}{y^2} \left[ (y^2 + r^2 +yr)(r-y) - (y^2 + r^2  - yr)(y+r) \right] + \int_r^\infty dy \frac{1}{y^2} \left[ (y^2 + r^2 + yr)(y -r) - (y^2 + r^2 -yr)(y + r) \right] \right\}$$

$$
    K_3 = -\frac{2\pi}{3r^2} \left\{ \int_0^rdy \frac{2}{y^2}\left[ -(y^2 + r^2)y  + yr^2\right] + \int_r^\infty dy \frac{2}{y^2}\left[ -(y^2 + r^2)r  + y^2r\right] \right\}
$$

$$K_3 = -\frac{2\pi}{3r^2} \left \{ -2\int_0^r dy \ y - 2r^3\int_r^\infty dy \frac{1}{y^2} \right\} = \frac{4\pi}{3r^2} \left[ \frac{r^2}{2} + r^2 \right] = {2\pi}$$

Since $z$ is in the direction of $r$, we can write
$$\vec{K} = \int d^3y \left( \frac{\vec{y}}{y^3} \right) \left( \frac{1}{|\vec{y} - \vec{r}|} \right) = {2\pi\hat{r} = 2\pi\frac{\vec{r}}{r}}$$

Now,

$$
    I = \nabla_r(\vec{v_2}.\vec{K}) = 2\pi \nabla_r \left( \vec{v_2}.\frac{\vec{r}}{r} \right)
$$

Hence, 

$$
    A_{12} \approx \frac{q_2\vec{v_2}}{cr} - \frac{q_2}{4\pi c}I
$$

$$= \frac{q_2}{c}\left[ \frac{\vec{v_2}}{r} - \frac{1}{2}\nabla_r\left( \frac{\vec{v_2}.\vec{r}}{r} \right) \right]$$

Now let's calculate $\nabla_r \left( \frac{\vec{v_2}.\vec{r}}{r} \right)$

$$
    \frac{\partial}{\partial x_i} \left( \frac{\vec{v_2} . \vec{r}}{r} \right) = \frac{\partial}{\partial x_i} \left( \frac{v_{2j}x_j } {r} \right)\\
     = v_{2j} \left [ \frac{\delta_{ij}}{r} - \frac{x_i}{r^2} \frac{\partial r}{\partial x_i} \right] \\
     = \frac{v_{2i}}{r} - (\vec{v_2}.\vec{r})\frac{1}{r^3}x_i\\
$$

or,

$$
\nabla_r \left( \frac{\vec{v_2}.\vec{r}}{r}\right)  = \frac{1}{r} \left[ \vec{v_2} - \hat{r}(\vec{v_2} . \hat{r}) \right]
$$

Hence,

$$
    A_{12} = \frac{q_2}{2cr} \left[ \vec{v_2} + \frac{\vec{r}(\vec{v_2}.\vec{r})}{r^2} \right]
$$

Interaction Lagrangian is,

$$
    L_{\text{int}} = -\frac{q1}{\gamma c}u_{1\alpha}A_{12}^\alpha \\
     = -q_1 \left[ \Phi_{12} - \frac{1}{c}\vec{v_1}.A_{12} \right] \qquad (\gamma \approx 1)
$$

$$
    = -q_1 \left\{ \frac{q_2}{r} - \frac{q_2}{2c^2r} \vec{v_1}. \left[ \vec{v_2} + \frac{\vec{r}(\vec{v_2}.\vec{r})}{r^2} \right] \right\}
$$

$$
    L_{\text{int}}  = {\frac{q_1q_2}{r} \left \{ -1 + \frac{1}{c^2} \left[ \vec{v_1}.\vec{v_2} + \frac{(\vec{v_1}.\vec{r})(\vec{v_2}.\vec{r})}{r^2} \right] \right\}}
$$

From eq 12.7 of Jackson $L_{free} = -mc^2\sqrt{1 - \frac{u^2}{c^2}}$ and expanding it till $2^{nd}$ order

$$L_{free} = m\left \{-c^2 + \frac{1}{2}v^2 + \frac{1}{8}\frac{v^4}{c^2} + ...\right\}$$

Hence, For a system of interacting charged particles, the complete Darwin Lagrangian, correct upto $1/c^2$ is

$$
    L_{\text{Darwin}} = \frac{1}{2} \sum_i m_iv_i^2 + \frac{1}{8c^2}\sum_i m_iv_i^4 -\frac{1}{2} \sum_{i,j}{}^{'} \frac{q_iq_j}{r_{ij}} + \frac{1}{4c^2} \sum_{ij}{}^{'}\frac{q_iq_j}{r_{ij}}[\vec{v_i}.\vec{v_j} + (\vec{v_i}.\hat{r}_{ij} )(\vec{v_j}.\hat{r}_{ij})]
$$
where $r_{ij} = |x_i - x_j|,\ \hat{r}$ is unit varctor in the dircetion of $x_i - x_j$ and the prime on the double summation indicates the omission of the (self-energy) terms, $i = j$.
